# Missing days 
We have day ranging from different time periods for each of thed dataset. But do we have data for each day? Let's find out

```python
def find_total_missing_days(df, df_name):
    daily_date = pd.date_range(start=df.Date.min(),
                               end=df.Date.max(), freq='D'
                              )
    uniq_days = df.Date.nunique()
    
    print(f"#### {df_name}")
    print(f"Total unique days we have data: {uniq_days}")
    print(f"Total days missing: {len(daily_date) - uniq_days}")
    print("")
    
find_total_missing_days(df = Aquifer_Doganella, df_name='Aquifer_Doganella')
find_total_missing_days(df = Aquifer_Auser, df_name='Aquifer_Auser')
find_total_missing_days(df = Water_Spring_Amiata, df_name='Water_Spring_Amiata')
find_total_missing_days(df = Lake_Bilancino, df_name='Lake_Bilancino')
find_total_missing_days(df = Water_Spring_Madonna_di_Canneto, df_name='Water_Spring_Madonna_di_Canneto')
find_total_missing_days(df = Aquifer_Luco, df_name='Aquifer_Luco')
find_total_missing_days(df = Aquifer_Petrignano, df_name='Aquifer_Petrignano')
find_total_missing_days(df = Water_Spring_Lupa, df_name='Water_Spring_Lupa')
find_total_missing_days(df = River_Arno, df_name='River_Arno')
```

Now that we saw there couple of days missing in each of the dataset. Let's take one step further to see when they are having those missing dates, is it random or it follows a pattern.

```python
def total_missing_days_pattern(df, df_name):

    daily_date = pd.date_range(start=df.Date.min(),
                               end=df.Date.max(), 
                               freq='D'
                              )
    daily_data = pd.DataFrame({'Date' : daily_date})

    temp = df[['Date']].copy()
    temp['missing'] = 0
        
    final = daily_data.merge(temp, on = ['Date'], how='left')
    final.fillna(1, inplace=True)
    
    plt.figure(figsize=(20, 5))
    sns.scatterplot(final.Date, final.missing, hue=final.missing)
    plt.title(df_name)
    
total_missing_days_pattern(df = Aquifer_Doganella, df_name='Aquifer_Doganella')
total_missing_days_pattern(df = Aquifer_Auser, df_name='Aquifer_Auser')
total_missing_days_pattern(df = Water_Spring_Amiata, df_name='Water_Spring_Amiata')
total_missing_days_pattern(df = Lake_Bilancino, df_name='Lake_Bilancino')
total_missing_days_pattern(df = Water_Spring_Madonna_di_Canneto, df_name='Water_Spring_Madonna_di_Canneto')
total_missing_days_pattern(df = Aquifer_Luco, df_name='Aquifer_Luco')
total_missing_days_pattern(df = Aquifer_Petrignano, df_name='Aquifer_Petrignano')
total_missing_days_pattern(df = Water_Spring_Lupa, df_name='Water_Spring_Lupa')
total_missing_days_pattern(df = River_Arno, df_name='River_Arno')
```

Looks like almost all of them have missing values in the year 2020. Additionally Aquifer Petrignano has missing values in 2006, Lake_Bilancino in 2002, Aquifer_Auser in 1998


# Let's understand the total null values for each feature in each of the dataset

```python
def Null_analysis(df, title):
    temp = df.isnull().sum()

    plt.figure(figsize=(15,5))
    g = sns.barplot(temp.index, temp.values)
    plt.xticks(rotation=90)
    plt.ylim(0,(temp.values.max()+1000))
    plt.title(title)
    
    for p in g.patches:
        g.annotate('{:.0f}\n{:.2f}%'.format(p.get_height(), (p.get_height()/df.shape[0]) ), 
                   (p.get_x()+0.4, p.get_height()+10),
                    ha='center', va='bottom',
                    color= 'black')
    
    plt.show()
    
    
    
Null_analysis(df = Aquifer_Doganella, title='Aquifer_Doganella')
Null_analysis(df = Aquifer_Auser, title='Aquifer_Auser')
Null_analysis(df = Water_Spring_Amiata, title='Water_Spring_Amiata')
Null_analysis(df = Lake_Bilancino, title='Lake_Bilancino')
Null_analysis(df = Water_Spring_Madonna_di_Canneto, title='Water_Spring_Madonna_di_Canneto')
Null_analysis(df = Aquifer_Luco, title='Aquifer_Luco')
Null_analysis(df = Aquifer_Petrignano, title='Aquifer_Petrignano')
Null_analysis(df = Water_Spring_Lupa, title='Water_Spring_Lupa')
Null_analysis(df = River_Arno, title='River_Arno')
```

# Temperature distribution
Let's understand the ditribution of temperature for each of the temperature columns in each of the dataset

```python
def temperature_distribution(df, df_name):
    temperature_columns = df.columns[df.columns.str.contains("Temp")]

    
    f, axes = plt.subplots(1, len(temperature_columns), figsize=(25,5))
    if len(temperature_columns) == 1:
        axes = [axes]
    for i, temp_col in enumerate(temperature_columns):
        sns.distplot(df[temp_col], kde=False, ax=axes[i])
        axes[i].set_title(df_name)
       
    
temperature_distribution(df = Aquifer_Doganella, df_name="Aquifer_Doganella")
temperature_distribution(df = Aquifer_Auser, df_name="Aquifer_Auser")
temperature_distribution(df = Water_Spring_Amiata, df_name="Water_Spring_Amiata")
temperature_distribution(df = Lake_Bilancino, df_name="Lake_Bilancino")
temperature_distribution(df = Water_Spring_Madonna_di_Canneto, df_name="Water_Spring_Madonna_di_Canneto")
temperature_distribution(df = Aquifer_Luco, df_name="Aquifer_Luco")
temperature_distribution(df = Aquifer_Petrignano, df_name="Aquifer_Petrignano")
# temperature_distribution(df = Water_Spring_Lupa, df_name="Water_Spring_Lupa")
temperature_distribution(df = River_Arno, df_name="River_Arno")
```

# Average temperature 
Lets check the monthly average temperature

```python
def avg_monthly_temperature(df, df_name):
    temperature_columns = df.columns[df.columns.str.contains("Temp")]
    
    #f, axes = plt.subplots(1, len(temperature_columns), figsize=(25,5))
    
    if len(temperature_columns)>3:
        f, axes = plt.subplots( int(np.ceil(len(temperature_columns)/3)), 3, figsize=(22.5, 15))
        axes = axes.flatten()[0:len(temperature_columns)]
    else:
        f, axes = plt.subplots(1, len(temperature_columns), figsize=(22.5,5))

    if len(temperature_columns) == 1:
        axes = [axes]
        
        
    for i, temp_col in enumerate(temperature_columns):
        
        temp = df.groupby(['month']).agg(col_name = (temp_col,'mean')).reset_index()
        temp = temp.rename(columns={"col_name":temp_col+'_avg_temp'})
    
    
        #plt.figure(figsize=(15,5))
        sns.scatterplot(temp.month, temp[temp_col+'_avg_temp'], 
                        size=temp[temp_col+'_avg_temp'], 
                        hue=temp[temp_col+'_avg_temp'], 
                        size_norm=None,
                        legend=None, 
                        ax=axes[i])
        axes[i].set_title(df_name + " - "+ temp_col)
        axes[i].set_xlabel("Month")
        axes[i].set_ylabel("Avg. temperature")
        plt.xticks(temp.month.unique())
    
    
avg_monthly_temperature(df=Aquifer_Doganella, df_name="Aquifer_Doganella")
avg_monthly_temperature(df = Aquifer_Doganella, df_name="Aquifer_Doganella")
avg_monthly_temperature(df = Aquifer_Auser, df_name="Aquifer_Auser")
avg_monthly_temperature(df = Water_Spring_Amiata, df_name="Water_Spring_Amiata")
avg_monthly_temperature(df = Lake_Bilancino, df_name="Lake_Bilancino")
avg_monthly_temperature(df = Water_Spring_Madonna_di_Canneto, df_name="Water_Spring_Madonna_di_Canneto")
avg_monthly_temperature(df = Aquifer_Luco, df_name="Aquifer_Luco")
avg_monthly_temperature(df = Aquifer_Petrignano, df_name="Aquifer_Petrignano")
avg_monthly_temperature(df = Water_Spring_Lupa, df_name="Water_Spring_Lupa")
avg_monthly_temperature(df = River_Arno, df_name="River_Arno")
```

```python
def avg_yearly_monthly_temperature(df, df_name):
    temperature_columns = df.columns[df.columns.str.contains("Temp")]
    
    # f, axes = plt.subplots(1, len(rain_columns), figsize=(25,5))
    if len(temperature_columns)>3:
        f, axes = plt.subplots( int(np.ceil(len(temperature_columns)/3)), 3, figsize=(22.5, 15))
        axes = axes.flatten()[0:len(temperature_columns)]
    else:
        f, axes = plt.subplots(1, len(temperature_columns), figsize=(22.5,5))

    if len(temperature_columns) == 1:
        axes = [axes]

    for i, temp_col in enumerate(temperature_columns):
        temp = df.groupby(['year', 'month']).agg(col_name = (temp_col,'mean')).reset_index()
        temp['Date'] = temp['year'].astype('str') + "-" + temp['month'].astype('str')
        temp["Date"] = pd.to_datetime(temp["Date"])
        
        temp = temp.rename(columns={"col_name":temp_col+'_avg_temp'})
        sns.lineplot(temp.Date, temp[temp_col+'_avg_temp'], 
                     legend=None, 
                     ax=axes[i])
        sns.scatterplot(temp.Date, temp[temp_col+'_avg_temp'], 
                        size=temp[temp_col+'_avg_temp'], 
                        hue=temp[temp_col+'_avg_temp'], 
                        size_norm=None,
                        legend=None, 
                        ax=axes[i])

        #axes[i].set_xticks(temp.Date)
        axes[i].tick_params(axis='x', rotation=90)
        #axes[i].set_xlabel("Month")
        axes[i].set_ylabel("")
        axes[i].set_ylabel("Avg Temperature")
        axes[i].set_title(df_name + " - \n"+ temp_col)
        
    
avg_yearly_monthly_temperature(df = Aquifer_Doganella, df_name="Aquifer_Doganella")
avg_yearly_monthly_temperature(df = Aquifer_Auser, df_name="Aquifer_Auser")
avg_yearly_monthly_temperature(df = Water_Spring_Amiata, df_name="Water_Spring_Amiata")
avg_yearly_monthly_temperature(df = Lake_Bilancino, df_name="Lake_Bilancino")
# avg_yearly_monthly_temperature(df = Water_Spring_Madonna_di_Canneto, df_name="Water_Spring_Madonna_di_Canneto")
avg_yearly_monthly_temperature(df = Aquifer_Luco, df_name="Aquifer_Luco")
avg_yearly_monthly_temperature(df = Aquifer_Petrignano, df_name="Aquifer_Petrignano")
avg_yearly_monthly_temperature(df = Water_Spring_Lupa, df_name="Water_Spring_Lupa")
avg_yearly_monthly_temperature(df = River_Arno, df_name="River_Arno")
```

```python
def avg_yearly_monthly_rainfall(df, df_name):
    rain_columns = df.columns[df.columns.str.contains("Rainfall")]
    
    # f, axes = plt.subplots(1, len(rain_columns), figsize=(25,5))
    if len(rain_columns)>3:
        f, axes = plt.subplots( int(np.ceil(len(rain_columns)/3)), 3, figsize=(22.5, 15))
        axes = axes.flatten()[0:len(rain_columns)]
    else:
        f, axes = plt.subplots(1, len(rain_columns), figsize=(22.5,5))

    if len(rain_columns) == 1:
        axes = [axes]

    for i, temp_col in enumerate(rain_columns):
        temp = df.groupby(['year', 'month']).agg(col_name = (temp_col,'mean')).reset_index()
        temp['Date'] = temp['year'].astype('str') + "-" + temp['month'].astype('str')
        temp["Date"] = pd.to_datetime(temp["Date"])
        
        temp = temp.rename(columns={"col_name":temp_col+'_avg_rain'})
        sns.lineplot(temp.Date, temp[temp_col+'_avg_rain'], 
                     legend=None, 
                     ax=axes[i])
        sns.scatterplot(temp.Date, temp[temp_col+'_avg_rain'], 
                        size=temp[temp_col+'_avg_rain'], 
                        hue=temp[temp_col+'_avg_rain'], 
                        size_norm=None,
                        legend=None, 
                        ax=axes[i])

        #axes[i].set_xticks(temp.Date)
        axes[i].tick_params(axis='x', rotation=90)
        #axes[i].set_xlabel("Month")
        axes[i].set_ylabel("")
        axes[i].set_ylabel("Avg Rainfall")
        axes[i].set_title(df_name + " - \n"+ temp_col)
        
    
avg_yearly_monthly_rainfall(df = Aquifer_Doganella, df_name="Aquifer_Doganella")
avg_yearly_monthly_rainfall(df = Aquifer_Auser, df_name="Aquifer_Auser")
avg_yearly_monthly_rainfall(df = Water_Spring_Amiata, df_name="Water_Spring_Amiata")
avg_yearly_monthly_rainfall(df = Lake_Bilancino, df_name="Lake_Bilancino")
# avg_yearly_monthly_rainfall(df = Water_Spring_Madonna_di_Canneto, df_name="Water_Spring_Madonna_di_Canneto")
avg_yearly_monthly_rainfall(df = Aquifer_Luco, df_name="Aquifer_Luco")
avg_yearly_monthly_rainfall(df = Aquifer_Petrignano, df_name="Aquifer_Petrignano")
avg_yearly_monthly_rainfall(df = Water_Spring_Lupa, df_name="Water_Spring_Lupa")
avg_yearly_monthly_rainfall(df = River_Arno, df_name="River_Arno")
```
