```python
## Change cat and num list according to your dataframe
def cutevizz(data,num,cat):
    from IPython.core.display import HTML
    def multi_table(table_list):
        ''' Acceps a list of IpyTable objects and returns a table which contains each IpyTable in a cell
        '''
        return HTML(
            '<table><tr style="background-color:white;">' + 
            ''.join(['<td>' + table._repr_html_() + '</td>' for table in table_list]) +
            '</tr></table>')
    def libraries():
        global np,pd,plt,sns
        import numpy as np
        import pandas as pd
        import matplotlib.pyplot as plt
        import seaborn as sns
        %matplotlib inline
        plt.style.use('seaborn-darkgrid')

    def load(data):
        global df
        df=pd.read_csv(data)

    ## Categorical variable analysis using countplot and multitable
    def countplota(column,heading):
        fig, ax = plt.subplots(1, 2, figsize=(15,6))
        x=df[column].value_counts().index
        y=df[column].value_counts().values.tolist()
        data = df.groupby(column).size()
        pal = sns.color_palette("magma", len(data))
        rank = data.argsort().argsort() 
        sns.barplot(x=x,y=y,palette=np.array(pal[::-1])[rank],ax = ax[0])
        for p in ax[0].patches:
                ax[0].annotate('{:.2f}'.format(p.get_height()), (p.get_x()+0.3, p.get_height()),
                            ha='center', va='bottom',
                            color= 'black',fontsize=15)
                my_circle=plt.Circle( (0,0), 0.4, color='#FFFDE7') # Adding circle at the centre
                p=plt.gcf()
                p.gca().add_artist(my_circle)
        _, _, autotexts= ax[1].pie(y, labels = x, colors = pal, autopct='%1.1f%%', textprops={'fontsize': 15},
                explode=[0.03 for i in df[column].value_counts().index])
        for autotext in autotexts:
            autotext.set_color('white')
        plt.legend(bbox_to_anchor=(1, 1))
        plt.suptitle (heading,weight = 'bold')
        plt.show()

    def countplotb(column,heading):
        x=df[column].value_counts().index
        y=df[column].value_counts().values.tolist()
        data = df.groupby(column).size()
        sns.set(rc={"axes.facecolor":"#FFFDE7",'figure.facecolor':'#FFFDE7',
                    'xtick.labelsize':15,'ytick.labelsize':15,
                    'axes.labelsize':20,'figure.figsize':(10.0, 6.0)})
        pal = sns.color_palette("magma", len(data))
        rank = data.argsort().argsort() 
        ax = sns.barplot(x=x,y=y,palette=np.array(pal[::-1])[rank])
        for p in ax.patches:
                ax.annotate('{:.2f}'.format(p.get_height()), (p.get_x()+0.3, p.get_height()),
                            ha='center', va='bottom',
                            color= 'black',fontsize=15)
        plt.title(heading,weight = 'bold',size=18)
        plt.show()

    def scatterplot():
        for i in range(len(num)):
            for j in range(len(num)):
                if num[j] != num[i]:
                    plt.figure(figsize=(13,6))
                    sns.scatterplot(x=df[num[i]] , y=df[num[j]])
                    h = num[i] +" vs "+num[j]
                    plt.title(h,weight = 'bold').set_fontsize('22')
                    plt.show()
                    print("-----------------------------------------------------------------------------------------------")

        for i in range(len(num)):
            for j in range(len(num)):
                if num[j] != num[i]:
                    for k in cat:
                        plt.figure(figsize=(13,6))
                        sns.scatterplot(x=df[num[i]] , y=df[num[j]],hue = df[k])
                        plt.legend(bbox_to_anchor=(1, 1),title=k)
                        h = num[i] +" vs "+num[j]
                        plt.title(h,weight = 'bold').set_fontsize('21')
                        plt.show()
                        print("-------------------------------------------------------------------------------------------")

    def barplot():
        for i in range(len(cat)):
            for j in range(len(num)):
                for k in range(len(cat)):
                    if cat[i] != cat[k]:
                        plt.figure(figsize=(11,6))
                        ax = sns.barplot(x=df[cat[i]] , y=df[num[j]],hue = df[cat[k]],palette = 'magma',ci = None)
                        for p in ax.patches:
                            ax.annotate('{:.2f}'.format(p.get_height()), (p.get_x()+0.125, p.get_height()),
                            ha='center', va='bottom',
                            color= 'black',fontsize=12)
                        plt.legend(bbox_to_anchor=(1, 1),title=cat[k])
                        sns.despine()
                        h = cat[i]+" vs "+num[j]+" with "+cat[k]+" as Hue"
                        plt.title(h,weight = 'bold').set_fontsize('21')
                        plt.show()
                        print("-------------------------------------------------------------------------------------------")

    def catplot():    
        for i in range(len(cat)):
            for j in range(len(num)):
                for k in range(len(cat)):
                    if cat[i] != cat[k]:
                        sns.set(rc={'xtick.labelsize':15,'ytick.labelsize':15,'axes.labelsize':15 , "axes.facecolor":"#FFFDE7" , "axes.grid":False})
                        ax = sns.catplot(x=cat[i] , y=num[j],col = cat[k],data=df,palette = 'magma',ci = None,kind="bar",height=6 , aspect=.5 )
                        for axes in ax.axes.flat:
                            _ = axes.set_xticklabels(axes.get_xticklabels(), rotation=90)
                        plt.tight_layout()
                        sns.despine()
                        plt.show()
                        print("-------------------------------------------------------------------------------------------")

    def histplota():
        for i in range(len(num)):
                plt.figure(figsize=(13,6))
                ax = sns.histplot(x = df[num[i]], palette = 'magma', alpha = 0.8, multiple = 'stack')
                Quantiles = np.quantile(df[num[i]], [0, 0.25, 0.50, 0.75, 1])
                for q in Quantiles: ax.axvline(x = q, linewidth = 0.5, color = 'r')
                h = "Distribution of "+num[i]
                plt.title(h,weight = 'bold').set_fontsize('21')
                plt.show()
                print("-------------------------------------------------------------------------------------------")

    def histplotb():
        for i in range(len(num)):
            for j in cat:
                plt.figure(figsize=(13,6))
                ax = sns.histplot(x = df[num[i]],hue = df[j],palette = 'magma', alpha = 0.8, multiple = 'stack')
                Quantiles = np.quantile(df[num[i]], [0, 0.25, 0.50, 0.75, 1])
                for q in Quantiles: ax.axvline(x = q, linewidth = 0.5, color = 'r')
                legend = ax.get_legend()
                handles = legend.legendHandles
                legend.remove()
                ax.legend(handles, df[j].value_counts().keys().tolist(), title = j, loc = 'upper right',bbox_to_anchor=(1.15, 1))
                plt.tight_layout()
                h = "Distribution of "+num[i]+" with "+j+" as Hue"
                plt.title(h,weight = 'bold').set_fontsize('21')
                plt.show()
                print("-------------------------------------------------------------------------------------------")

    def violinplota():
        for i in range(len(num)):
            for k in cat:
                plt.figure(figsize=(13,6))
                sns.violinplot(y=df[num[i]] , x = df[k] , palette="magma" , scale="count" , bw=.2)
                h = "Distribution of "+num[i]+" with "+k+" as Hue"
                plt.title(h,weight = 'bold').set_fontsize('21')
                plt.show()
                print("-------------------------------------------------------------------------------------------")

    def violinplotb():
        for i in range(len(num)):
            for j in cat:
                for k in cat:
                    if j != k:
                        plt.figure(figsize=(13,6))
                        sns.violinplot(x = df[j],y=df[num[i]] ,hue = df[k], palette="magma")
                        plt.legend(bbox_to_anchor=(1.15, 1),title=k)
                        h = j+" vs "+num[i]+" with "+k+" as Hue"
                        plt.title(h,weight = 'bold').set_fontsize('21')
                        plt.show()
                        print("-------------------------------------------------------------------------------------------")

    def violinplotc():
        catnew = []
        for i in cat:
            if len(df[i].value_counts().keys().tolist()) == 2:
                catnew.append(i)
        for i in range(len(num)):
            for j in cat:
                for k in catnew:
                    if j != k:
                        plt.figure(figsize=(13,6))
                        sns.violinplot(x = df[j],y=df[num[i]] ,hue = df[k], palette="magma",split=True , inner="stick")
                        plt.legend(bbox_to_anchor=(1.15, 1),title=k)
                        h = j+" vs "+num[i]+" with "+k+" as Hue"
                        plt.title(h,weight = 'bold').set_fontsize('21')
                        plt.show()
                        print("-------------------------------------------------------------------------------------------")

    def striplot():
        for i in range(len(num)):
            for j in cat:
                for k in cat:
                    if j != k:
                        params = dict(x = df[j] ,y = df[num[i]] ,hue=df[k],dodge=True)
                        ax = sns.stripplot(**params , size=8,jitter=0.35,palette=['#33FF66','#FF6600'],edgecolor='black',linewidth=1)
                        sns.boxplot(**params ,palette=['#BDBDBD','#E0E0E0'],linewidth=6)
                        plt.legend(bbox_to_anchor=(1.15, 1),title=k)
                        h = j+" vs "+num[i]+" with "+k+" as Hue"
                        plt.title(h,weight = 'bold').set_fontsize('21')
                        plt.show()
                        print("-------------------------------------------------------------------------------------------")

    def jointplot():
        for i in range(len(num)):
            for j in range(len(num)):
                if num[i] != num[j]:
                    sns.jointplot(x=df[num[i]], y=df[num[j]], data=df , height = 10 , kind="hex" , color="#fb9224")
                    plt.show()
                    print("---------------------------------------------------------------------------------------------")

    # Diagonal correlation is always unity & less relevant, shifted variant shows only relevant cases
    def corrMat(df,id=False):

        corr_mat = df.corr().round(2)
        f, ax = plt.subplots(figsize=(12,7))
        mask = np.triu(np.ones_like(corr_mat, dtype=bool))
        mask = mask[1:,:-1]
        corr = corr_mat.iloc[1:,:-1].copy()
        sns.heatmap(corr,mask=mask,vmin=-0.3,vmax=0.3,center=0, 
                    cmap='RdPu_r',square=False,lw=2,annot=True,cbar=False)
        ax.set_title('Shifted Linear Correlation Matrix',weight = 'bold').set_fontsize('21')  
        plt.show()


    libraries()
    load(data)
    print("------------------------"+"\033[1m"+"Distribution of Categorical Vraiables"+"\033[0m"+"------------------------")
    sns.set(rc={"axes.facecolor":"#FFFDE7",'figure.facecolor':'#FFFDE7','xtick.labelsize':15,'ytick.labelsize':15,'axes.labelsize':20})
    for i in cat:
        column = i
        heading = "Distribution of "+i
        if len(df[i].value_counts().keys().tolist()) > 3:
            countplotb(column,str(heading))
        else:
            countplota(column,str(heading))
    print("------------------------"+"\033[1m"+"Scatter Plots of Numerical Features"+"\033[0m"+"------------------------")
    scatterplot()
    print("------------------------"+"\033[1m"+"Bar Plots Analysis"+"\033[0m"+"------------------------")
    barplot()
    print("------------------------"+"\033[1m"+"Cat Plots Analysis"+"\033[0m"+"------------------------")
    catplot()
    print("------------------------"+"\033[1m"+"Histogram Plots Analysis"+"\033[0m"+"------------------------")
    histplota()
    histplotb()
    print("------------------------"+"\033[1m"+"Violin Plots Analysis"+"\033[0m"+"------------------------")
    violinplota()
    violinplotb()
    violinplotc()
    sns.set(rc={"axes.facecolor":"#b0deff","axes.grid":False,'xtick.labelsize':15,'ytick.labelsize':15,'axes.labelsize':20,'figure.figsize':(12.0, 6.0)})
    print("------------------------"+"\033[1m"+"Box/Strip Plots Analysis"+"\033[0m"+"------------------------")
    striplot()
    # Recover default matplotlib settings
    plt.rcParams.update(plt.rcParamsDefault)
    %matplotlib inline
    sns.set(style="white", color_codes=True)
    sns.set(rc={'xtick.labelsize':15,'ytick.labelsize':15,'axes.labelsize':20})
    print("------------------------"+"\033[1m"+"Joint Plots Analysis"+"\033[0m"+"------------------------")
    jointplot()
    sns.set(rc={"axes.facecolor":"#FFFDE7",'figure.facecolor':'#FFFDE7',
                    'xtick.labelsize':15,'ytick.labelsize':15,
                    'axes.labelsize':20,'figure.figsize':(10.0, 6.0)})
    print("------------------------"+"\033[1m"+"Heat Map"+"\033[0m"+"------------------------")
    corrMat(df)
    print("------------------------"+"\033[1m"+"Multi Table Analysis"+"\033[0m"+"------------------------")
    df_nunique = {var: pd.DataFrame(df[var].value_counts()) for var in set(cat)}        
    multi = []
    for i in cat:
        multi.append(df_nunique[i].style\
          .format('{:.2f}')\
          .highlight_max(color = 'lightgreen')\
          .highlight_min(color = 'coral'))
    display(multi_table(multi))

    for i in cat:
        df_groupby = {var: pd.DataFrame(df.groupby([var, i]).size()) 
                      for var in set(cat)}
        multi = []
        for j in cat:
            if j != i: 
                multi.append(df_groupby[j])
        display(multi_table(multi))
```
