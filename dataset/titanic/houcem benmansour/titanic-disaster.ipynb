# check outlier

```python
sns.boxplot(train_df["Age"])
plt.show()
```

```python
#return outlier in list
out=[]
def outliers(train_df):
    q1=train_df.quantile(.25)
    q3=train_df.quantile(.75)
    iqr=q3-q1
    lower_tail=q1-1.5*iqr
    upper_tail=q3+1.5*iqr
    
    for x in train_df:
        if x < lower_tail or x > upper_tail:
            out.append(x)
    print(" The outliers are : ",out)
outliers(train_df["Age"])
```

```python
#remove outlier
sns.boxplot(train_df["Age"])
plt.title("Box Plot before outlier removing")
plt.show()
def drop_outliers(train_df, field_name):
    iqr = 1.5 * (np.percentile(train_df[field_name], 75) - np.percentile(train_df[field_name], 25))
    train_df.drop(train_df[train_df[field_name] > (iqr + np.percentile(train_df[field_name], 75))].index, inplace=True)
    train_df.drop(train_df[train_df[field_name] < (np.percentile(train_df[field_name], 25) - iqr)].index, inplace=True)
drop_outliers(train_df, 'Age')
sns.boxplot(train_df["Age"])
plt.title("Box Plot after outlier removing")
plt.show()
```
