### <span style="color:#011936;">***1.4.1. Business analysis and work initiation | Formulation of the problem***<span>

1. Understand business goals
2. Outline possible approaches and solutions
3. Identify technological limits
4. Think through the risks
5. Rethink and adjust business goals
____
### <span style="color:#011936;">***1.4.1. Business analysis and work initiation | Data evaluation***<span>
1. How many sample object do we have?
2. Is there a markup and what quality is it?
3. Is it possible to makr up all classes?
4. How many classes?
5. Is there an imbalance?
6. What about outliers and omissions?
    
__You must familiarize yourself with the data in advance. Until you see the data, it's a big risk and asking twice as much for the project__
____
### <span style="color:#011936;">***1.4.1. Business analysis and work initiation | Customer and project evaluation***<span>
1. Get to know the customer better
2. Study the experience of other performers working with him
3. Schedule a meeting with all intersted perties from the customer, not only with one of representatives
4. Find out about their budget
____
### <span style="color:#011936;">***1.4.1. Business analysis and work initiation | Selection of metrics***<span>
1. Choose the metrics that will best align with your business goals
2. Explain them to the customer
3. Make sure that the customer understands them. Check again that he understood them EXACTLY
4. If there are a lot of metrics, you need to select a set that will be ised to validate the solutuions

____
### <span style="color:#011936;">***1.4.1. Business analysis and work initiation | Final step***<span>
1. Clearly state the objectives of the study
2. Assess resources for risk
    * Customer Features
    * Uncertainty in requirements
    * Data Uncertainty
3. Formulate a technical task
4. Target metrics

___
___   
### <span style="color:#011936;">***1.4.2. Review and analysis of existing solutions - baseline***<span>
1. Formulation of evaluation criteria
    * Metrics
    * Efficiency

    * Support

## <span style="color:#011936;">***1.5. The concept of reproducibility***<span>
<a id="1.5"></a>

<span style="color:#465362;">**Definition**: Reproducibility is the ability of independent investigators to draw the same conclusions from an experiment by following the documentation shared by the original investigators. Hence, reproducibility requires that another, independent team of investigators have to conduct the same experimen<span>  
    
<span style="color:#465362;">**Repeatablility**: Repeatability (same team, same experimental setup): The measurement can be obtained with stated precision by the same team using the same measurement procedure, the same measuring system, under the same operating conditions, in the same location on multiple trials.<span>

<span style="color:#465362;">**Reproducibility**:Reproducibility (different team, different experimental setup): The measurement can be obtained with stated precision by a different team, a different measuring system, in a different location on multiple trials.<span>

<span style="color:#465362;">**Replicability**: Replicability (different team, same experimental setup): The measurement can be obtained with stated precision by a different team using the same measurement procedure, the same measuring system, under the same operating conditions, in the same or a different location on multiple trials.<span>



### <span style="color:#011936;">***1.5.1 Factors that reduce reproducibility***<span>

1. First place - problems with dependencies in libraries and dependencies in dependencies. Some of the repositories
were used requirements.txt, part setup.py. 
2. The second place is the order of execution. The notebook is saved without clearing the code, the order is not
saved and some variables are declared or initialized after use.
3. Third place - there is no necessary data, for example, absolute paths are specified or there is no data in the repository at all.
___
* Uncontrolled randomness in data or algorithms (40%) (Random Seed) 
* Source data changed manually, not using scripts
* Dependence of output and results on time functions (13%)
* Differences in display on graphs (incorrect use of matplotlib including) (52%)
* External data is not available (3%)
* Differences in the output of floating-point numbers (3%)
* Inconsistent order of traversal of dictionaries and other containers in python (4%)
* Differences in the execution environment (27%)
