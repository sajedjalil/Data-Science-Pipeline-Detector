 # Importing necessary library

```python
import  pandas as pd # for data preprocessing 
import  seaborn as sns  # only for loading dataset
import  plotly.express as px # data visulization
```

```python
df = sns.load_dataset('titanic') # loading dataset
```

```python
df.head(4) 
```

# step 1 : start with simple visualization

* <h2  style="text-align: left;font-size: 20px; font-family: Comic Sans MS; color:darkgslategray;">let's start with simple visualization , we will plot histogram of age column
*  <h2  style="text-align: left;font-size: 20px; font-family: Comic Sans MS; color:darkgslategray;">before ploting histogram we will convert age column datatype to intger 

```python
df.info()
```

```python
df.dropna(inplace=True) # dropping null values 
df['age'] = df['age'].astype(int)# converting float to integers
df.info()
```

```python
#simple histogram
fig= px.histogram(df, x='age')
fig.show()
```

# step 2 : adjust layout 

```python
fig= px.histogram(df, x='age',height=500) # try changing height with 200 ,700 ,1200 and see the results
fig.show()
```

```python
fig= px.histogram(df, x='age',width=700)#  try changing width  with 100 ,500 ,1000 and see the results
fig.show()
```

* <h2  style="text-align: left;font-size: 20px; font-family: Comic Sans MS; color:darkgslategray;">using height and width parameters you can adjust the layout  </h2>

```python
fig =px.histogram(df,x= 'age',width=900,height= 500)
fig.show()
```

#  step 3 : change template 

* <h2  style="text-align: left;font-size: 18px; font-family: Comic Sans MS; color:darkgslategray;">chagne template to simple white this templates removes all the clutter from your graph and gives nice clean look to it


```python

fig= px.histogram(df, x='age',height=500,width=900,template='simple_white')
fig.show()
```

* <h2  style="text-align: left;font-size: 20px; font-family: Comic Sans MS; color:darkgslategray;">other templates

```python
# plotly_dark
fig= px.histogram(df, x='age',height=500,width=900,template='plotly_dark')
fig.show()
```

```python
# seaborn
fig= px.histogram(df, x='age',height=500,width=900,template='seaborn')
fig.show()
```

```python
#ggplot2
fig= px.histogram(df, x='age',height=500,width=900,template='ggplot2')
fig.show()
```

# step 4 : add heading and customized xaxis and yaxis title

```python

fig= px.histogram(df, x='age',height=500,width=900,template='simple_white')
fig.update_layout(title={'text':'histogram of Passenger Traveling by Age'}) # adding heading
fig.show()
```

* <h2  style="text-align: left;font-size: 20px; font-family: Comic Sans MS; color:darkgslategray;">we had added the title but , now I want to change it's position towords right

```python

fig= px.histogram(df, x='age',height=500,width=900,template='simple_white')
fig.update_layout(title={'text':'histogram of Passenger Traveling by Age'},
                 title_x= 0.2) # changing the position of title 
fig.show()
```

<h3 style="padding: 20px;text-align: left;font-size: 16px; border-radius:5px; background-color:powderblue; font-family: Comic Sans MS; color:black;word-spacing: 4px;line-height: 1.3 ;">
    Note: you have to imagine line between 0 to 1 horizontally, 0 means towords left and 1 means towords right so if you want your title position towards left then use 0.1,0.2 and for right use 0.8,0.9 in title_x parameter.

* <h2  style="text-align: left;font-size: 20px; font-family: Comic Sans MS; color:darkgslategray;">adding customized font , color and size to the title

```python

fig= px.histogram(df, x='age',height=500,width=900,template='simple_white')

###########################################################################################
fig.update_layout(title={'text':'Histogram of Passenger Traveling by Age','font':{'size':25}}
                    ,title_font_family="Times New Roman",
    title_font_color="darkgrey",              
                 title_x=0.2)
###########################################################################################
fig.show()
```

<h3 style="padding: 20px;text-align: left;font-size: 16px; border-radius:5px; background-color:powderblue; font-family: Comic Sans MS; color:black;word-spacing: 4px;line-height: 1.3 ;">
Note : I had used light grey color for title beacause I want my users attention towords graph not title 

* <h2  style="text-align: left;font-size: 20px; font-family: Comic Sans MS; color:darkgslategray;">adding customized font , color and size to x axis and y axis labels 

```python


fig= px.histogram(df, x='age',height=500,width=900,template='simple_white')
fig.update_layout(title={'text':'Histogram of Passenger Traveling by Age','font':{'size':25}}
                    ,title_font_family="Times New Roman",
    title_font_color="darkgrey",
    
                 
                 title_x=0.2)
#############################################################################
fig.update_layout(
    font_family='classic-roman',
                  font_color= 'grey',
                  yaxis_title={'text': " count", 'font': {'size':18}},
                xaxis_title={'text': " Age", 'font': {'size':18}}
)
#############################################################################

fig.show()
```

# step 5: using custom color

```python
fig= px.histogram(df, x='age',height=500,width=900,template='simple_white',
                  color_discrete_sequence=['salmon']) # adding custom color





fig.update_layout(title={'text':'Histogram of Passenger Traveling by Age','font':{'size':25}}
                    ,title_font_family="Times New Roman",
    title_font_color="darkgrey",
    
                 
                 title_x=0.2)

fig.update_layout(
    font_family='classic-roman',
                  font_color= 'grey',
                  yaxis_title={'text': " count", 'font': {'size':18}},
                xaxis_title={'text': " Age", 'font': {'size':18}}
)
fig.show()
```

<h3 style="padding: 20px;text-align: left;font-size: 16px; border-radius:5px; background-color:powderblue; font-family: Comic Sans MS; color:black;word-spacing: 4px;line-height: 1.3 ;">Note : if you are using some other graphs like bar and if you did not find color_discrete_sequence then you can use 
color_discrete_scale and color_discrete_map as an alternative 

* <h2  style="text-align: left;font-size: 20px; font-family: Comic Sans MS; color:darkgslategray;">we are adding catergorical column in our histogram to make it more informative.

```python
fig= px.histogram(df, x='age',height=500,width=900,template='simple_white',
                  color='sex', # adding categorical column 
                  color_discrete_sequence=['purple','pink'])




fig.update_layout(title={'text':'Histogram of Passenger Traveling by Age','font':{'size':25}}
                    ,title_font_family="Times New Roman",
    title_font_color="darkgrey",
    
                 
                 title_x=0.2)

fig.update_layout(
    font_family='classic-roman',
                  font_color= 'grey',
                  yaxis_title={'text': " count", 'font': {'size':18}},
                xaxis_title={'text': " Age", 'font': {'size':18}}
)
fig.show()
```

* <h2  style="text-align: left;font-size: 20px; font-family: Comic Sans MS; color:darkgslategray;line-height: 1.5 ;">we can remove legend from above graph to make it look clean and use more descriptive title as an alternative

```python
fig= px.histogram(df, x='age',height=500,width=900,template='simple_white',color='sex',color_discrete_sequence=['purple','pink'])




fig.update_layout(title={'text':'Sex Distribution By Age','font':{'size':25}}
                    ,title_font_family="Times New Roman",
    title_font_color="darkgrey",
    
                 
                 title_x=0.3)

fig.update_layout(
    font_family='classic-roman',
                  font_color= 'grey',
                  yaxis_title={'text': " count", 'font': {'size':18}},
                xaxis_title={'text': " Age", 'font': {'size':18}}
)
######################################################################

fig.update_layout(showlegend=False)

#######################################################################
fig.show()
```

* <h2  style="text-align: left;font-size: 20px; font-family: Comic Sans MS; color:darkgslategray;line-height: 1.5 ;">customizing the bin size , I want bin size as 10 to make this graph more informative for our viewers

```python
fig= px.histogram(df, x='age',height=500,width=900,template='simple_white',color='sex',color_discrete_sequence=['purple','pink'])




fig.update_layout(title={'text':'Sex Distribution By Age','font':{'size':25}}
                    ,title_font_family="Times New Roman",
    title_font_color="darkgrey",
    
                 
                 title_x=0.3)

fig.update_layout(
    font_family='classic-roman',
                  font_color= 'grey',
                  yaxis_title={'text': " count", 'font': {'size':18}},
                xaxis_title={'text': " Age", 'font': {'size':18}}
)
fig.update_layout(showlegend=False)

################################################################################
fig.update_traces(xbins=dict( # bins used for histogram
        start=1,
        end=90,
        size=10
    ))
################################################################################
fig.show()
```

# step 6 : adding text 

```python
fig =px.histogram(df,x= 'age',width=900,height= 500,template='simple_white',color='sex'
                 ,color_discrete_sequence=['pink','mediumpurple'],
                 
                 
                 )  

fig.update_layout(title={'text':'Sex Distribution by Age','font':{'size':22}} 
                  ,title_x= 0.3  # title_x : this parameter is used for changing the title position horizontally
                    
     ,title_font_family="Times New Roman",
    title_font_color="darkgrey"
    
    
                )
fig.update_layout( font_family='classic-roman',
                  font_color= 'grey',
                
                  yaxis_title={'text': " count", 'font': {'size':19},  'standoff': 10},
                  xaxis_title={'text': " Age", 'font': {'size':19}}
                 )
fig.update_layout(showlegend=False)
fig.update_traces(xbins=dict( # bins used for histogram
        start=1,
        end=90,
        size=10
    ))

#####################################################################################
fig.add_annotation(x=40, y=48,
            text="most number of passenger on titanic are between age 30 to 40",
            showarrow=True,   
                   textangle=0,arrowhead= 1,arrowside='end'
                   
            ,font=dict(
            family="Courier New, monospace",
            size=16,
            color="darkblue"
            )
                  )
######################################################################################

fig.show()
```
