### How much does a DS earn?

This is a big question!, it varies demographically alot. As we can See salary for an average DS in US is 3 times more than salary of an average DS in India, this is due to finantial and geopolitical reasons more than skillset or experiance. If you are a DS planning to move to forgin country use this map and choose wisely!!

```python
def salary(x):
    if x == '$0 ($USD)':
        return 0
    if x =='$1-$99':
        return 99
    if x== '$100-$999':
        return 999
    if x=='$1000-$9,999':
        return 9999
    if x=='$10,000-$99,999':
        return 99999
    if x== '$100,000 or more ($USD)':
        return 100000

df = pd.read_csv('https://raw.githubusercontent.com/plotly/datasets/master/2014_world_gdp_with_codes.csv')
code_list = []

Kaggle2020ds['Q3'] = Kaggle2020ds['Q3'].str.replace('United States of America','United States')
Kaggle2020ds['Q3'] = Kaggle2020ds['Q3'].replace('United Kingdom of Great Britain and Northern Ireland','United Kingdom')
Kaggle2020ds['Q3'] = Kaggle2020ds['Q3'].replace('South Korea',"Korea, South")
Kaggle2020ds['Q3'] = Kaggle2020ds['Q3'].replace('Republic of Korea','Korea, North')
Kaggle2020ds['Q3'] = Kaggle2020ds['Q3'].replace('Viet Nam','Vietnam')
Kaggle2020ds['Q3'] = Kaggle2020ds['Q3'].replace('Iran, Islamic Republic of...','Iran')


Kaggle2020ds['salary'] = Kaggle2020ds['Q25'].apply(lambda x: salary(x))

mapdata = Kaggle2020ds.groupby('Q3').mean()
#mapdata
for country in mapdata.index:
    try:
        code_list.append(df[df['COUNTRY']==country]['CODE'].values[0])
    except:
        #print(country)
        code_list.append(country)
mapdata['Code'] = code_list

mapdata['Avg_Code_Exp'] = mapdata['codeExp']
df = px.data.gapminder().query("year==2007")
fig = px.choropleth(mapdata, locations="Code",
                    color="salary", 
                    hover_name="Code",
                    color_continuous_scale=px.colors.sequential.Plasma)
fig.show()
```
