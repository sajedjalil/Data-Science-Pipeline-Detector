## Survey groups 
Survey participants are divided into 4 groups. 
1. People working on NLP
2. People working on Machine Vision
3. People working as Data Scientist
4. All Survey participants

Here is the methodology for dividing survey participants into subgroups - 

>### 1. People working on NLP
People working on NLP have been selected based on the answer given to question **Q27 - Which of the following natural language processing (NLP) methods do you use on a regular basis?  (Select all that apply)**. People who have selected any of the NLP methods (including other) are considered as people working on NLP. So these people have selected any of the following options - 
* **Q27_Part_1**- Word embeddings/vectors (GLoVe, fastText, word2vec)
* **Q27_Part_2** - Encoder-decorder models (seq2seq, vanilla transformers)
* **Q27_Part_3** - Contextualized embeddings (ELMo, CoVe)
* **Q27_Part_4** - Transformer language models (GPT-2, BERT, XLnet, etc)
* **Q27_Part_6** - Other

>### 2. People working on Machine Vision
People working on Machine Vision have been selected based on answer given to question **Q26 Which categories of computer vision methods do you use on a regular basis? (Select all that apply)**. People who have selected any of the Machine vision methods (including other) are considered as people working on Machine Vision. So these people have selected any of the following options - 
* **Q26_Part_1** - General purpose image/video tools (PIL, cv2, skimage, etc) 
* **Q26_Part_2** - Image segmentation methods (U-Net, Mask R-CNN, etc)
* **Q26_Part_3** - Object detection methods (YOLOv3, RetinaNet, etc)
* **Q26_Part_4** - Image classification and other general purpose networks (VGG, Inception, ResNet, ResNeXt, NASNet, EfficientNet, etc)
* **Q26_Part_5** -Generative Networks (GAN, VAE, etc)
* **Q26_Part_7** - Other

>### 3. People working as Data Scientist
People working as Data Scientist are selected based on answer given to question **Q5 Select the title most similar to your current role (or most recent title if retired)**. People belong to this group if **Data Scientist** is selected as the answer to Q5. 

>### 4. All Survey participants
This set consists of everyone who participated in the survey.

Let's check size of the groups and how these group interact with each other-
* 12.95% people work on NLP methods 
* 21.75% work on Machine Vision methods 
* 20.71% people are working as Data Scientist.
