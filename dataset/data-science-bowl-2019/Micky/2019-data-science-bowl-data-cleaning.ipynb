## Train Heatmap

```python
import seaborn as sns
import matplotlib.pyplot as plt
def chk_corr(df):
    corrs = train.corr()
    plt.figure(figsize = (7,7))
    # Heatmap of correlations
    sns.heatmap(corrs, cmap = plt.cm.RdYlBu_r, vmin = -0.25, annot = True, vmax = 0.6)
    plt.title('Correlation Heatmap');
```

```python
chk_corr(train)
```
