<a id="4.2"></a>
<font size="+2" color="indigo"><b>4.2 Random</b></font><br>

Now let us create a new column with random dates greater than the created date.

```python
def random_future_date(start):
    end = pd.to_datetime("2022-02-25 09:23:34",format='%Y-%m-%d %H:%M:%S')  # We will set a random date
    random_date = start + (end - start) * random.random()
    return random_date.strftime("%Y-%m-%d %H:%M:%S")
```

```python
df['future_date']=df['CreationDate'].apply(lambda x:random_future_date(x))
df[['CreationDate','future_date']].sample(3)
```

<a id="4.3"></a>
<font size="+2" color="indigo"><b>4.3 Difference - Year</b></font><br>

```python
df['diff_year']=round((pd.to_datetime(df['future_date'])-pd.to_datetime(df['CreationDate'])).dt.days/365).astype(int)
df[['CreationDate','future_date','diff_year']].sample(3)
```

**Note :** For instance,1 year 11 months will be considered as 1 only.

<a id="4.4"></a>
<font size="+2" color="indigo"><b>4.4 Difference - Month</b></font><br>

```python
def months(d1,d2):
    d1=pd.to_datetime(d1)
    d2=pd.to_datetime(d2)
    return (d1.year-d2.year)*12 + (d1.month-d2.month) 
```

```python
df['diff_month']=df.apply(lambda x:months(x.future_date,x.CreationDate),axis=1)
df[['CreationDate','future_date','diff_month']].sample(3)
```

**Note:** Above calculation is not based on days.It is based on month numbers.

<a id="4.5"></a>
<font size="+2" color="indigo"><b>4.5 Difference - Days</b></font><br>

```python
df['diff_days']=(pd.to_datetime(df['future_date'])-pd.to_datetime(df['CreationDate'])).dt.days
df[['CreationDate','future_date','diff_days']].sample(3)
```

<a id="4.6"></a>
<font size="+2" color="indigo"><b>4.6 Difference - Hours </b></font><br>

```python
df['diff_hours']=(df['diff_days']*24)+(((pd.to_datetime(df['future_date'])-pd.to_datetime(df['CreationDate'])).dt.seconds)// 3600).astype(int)
df[['CreationDate','future_date','diff_hours']].sample(3)
```

<a id="4.7"></a>
<font size="+2" color="indigo"><b>4.7 Difference - Minutes </b></font><br>

<a id="4.12"></a>
<font size="+2" color="indigo"><b>4.12 Weekend or not</b></font><br>

```python
def weekend(date):
    date=pd.to_datetime(date)
    if((date.strftime('%w')=='0')|(date.strftime('%w')=='6')):
        return "weekend"
    else:
        return "working day"
```

```python
df['weekend_or_not']=df['CreationDate'].apply(lambda x:weekend(x))
df[['CreationDate','weekday','weekend_or_not']].sample(3)
```

<a id="4.13"></a>
<font size="+2" color="indigo"><b>4.13 Leap year or not</b></font><br>

```python
def leap(date):
    date=pd.to_datetime(date)
    if(date.year%4==0):
        return "leap"
    else:
        return "non leap"
```

```python
df['leap']=df['CreationDate'].apply(lambda x:leap(x))
df[['CreationDate','year','leap']].sample(3)
```

<font size="+2" color="indigo"><b>5.8 Gantt Chart</b></font><br>

Gantt chart is useful to show activities (tasks or events) displayed against time.Below we will see how long did the kagglers take time to become Grandmasters.Below code is from my own kernel [Become Grandmaster](https://www.kaggle.com/raenish/become-grandmaster) where i have explored more on Grandmasters principles and strategies.

```python
user=pd.read_csv('../input/meta-kaggle/Users.csv')
user_achievement=pd.read_csv('../input/meta-kaggle/UserAchievements.csv')

#Merging User and Userachievement dataset to get TierAchievement date field
only_notebook_gm=user_achievement[(user_achievement['AchievementType']=="Scripts")&(user_achievement['Tier']==4)][['UserId','TierAchievementDate']]
only_notebook_gm=only_notebook_gm.rename(columns={'UserId':'Id'})
only_notebook_gm_df=pd.merge(only_notebook_gm,user,on="Id",how="left")
```

```python
# Scaling difference between register date and tier achievement date for all grandmasters
only_notebook_gm_df['diff_days']=(pd.to_datetime(only_notebook_gm_df['TierAchievementDate'])-pd.to_datetime(only_notebook_gm_df['RegisterDate'])).dt.days 
scaler = MinMaxScaler()
only_notebook_gm_df['scaled_days'] = scaler.fit_transform(only_notebook_gm_df[['diff_days']])*100
only_notebook_gm_df.head(2)
```

```python
def plot_gantt_no_grad(data,height,title):
    df=[]
    for i in data.index:
        df.append(dict(Task=data['DisplayName'][i],Start=data['RegisterDate'][i],Finish=data['TierAchievementDate'][i],
                   Complete=data['scaled_days'][i]))
    df=pd.DataFrame(df)
    df['Start']=pd.to_datetime(df['Start'], format='%m/%d/%Y')
    df['Finish']=pd.to_datetime(df['Finish'], format='%m/%d/%Y')
    df=df.sort_values('Complete',ascending=False)
    fig = ff.create_gantt(df,showgrid_x=True, show_colorbar=True,showgrid_y=True,height=height,title=title)
    fig.update_layout(title_x=0.5)
    fig.show()
```

```python
plot_gantt_no_grad(only_notebook_gm_df,600,"Simple Gantt Chart - Time Taken by each Grandmaster")
```

```python
def plot_gantt_gradient(data,color,height,title):
    df=[]
    for i in data.index:
        df.append(dict(Task=data['DisplayName'][i],Start=data['RegisterDate'][i],Finish=data['TierAchievementDate'][i],
                   Complete=data['scaled_days'][i]))
    df=pd.DataFrame(df)
    df['Start']=pd.to_datetime(df['Start'], format='%m/%d/%Y')
    df['Finish']=pd.to_datetime(df['Finish'], format='%m/%d/%Y')
    df=df.sort_values('Complete',ascending=False)
    fig = ff.create_gantt(df,colors=color, index_col='Complete',showgrid_x=True, show_colorbar=True,showgrid_y=True,height=height,title=title)
    fig.update_layout(title_x=0.5)
    fig.show()
```

```python
plot_gantt_gradient(only_notebook_gm_df,"Blues",600,"Gradient Color Gantt Chart - Time Taken by each Grandmaster")
```
