### Kludgy maximum candies calculation

```python
# Simulate very brute-forcely the optimal selecting of bandits
# to see what expected optimal rewards is, and how low a probability is taken.
# Set up 100 bandits with probs from 0.01 to 1.0:
bandit_probs = (np.array(range(100)) + 1.0)/100.0

# Or, can use the actual values from a simulation further down:
# After running the simulation and getting the initial probs
# return here and uncomment this line:
##bandit_probs = np.sort(np.array(initial_probs)/100.0)
# Optimal value and min probability look very similar to idealized case.

# Keep track of things...
bandit_counts = np.zeros(100)
current_probs = 1.0 * bandit_probs
pulls = []
pulled_probs = []
score = [0]
for ip in range(4000):
    pulls = pulls + [ip]
    # choose the highest prob one
    ihigh = np.argmax(current_probs)
    # add it to probs chosen, count bandit and adjust prob
    score = score + [score[-1] + current_probs[ihigh]]
    pulled_probs = pulled_probs + [current_probs[ihigh]]
    bandit_counts[ihigh] += 1
    current_probs[ihigh] *= 0.97
score = np.array(score[1:])
pulls = np.array(pulls)
bandit_counts = np.array(bandit_counts)

# Some results:
ave_prob_pulled = sum(pulled_probs)/len(pulled_probs)
print("Expected Total rewards: {:.2f}".format(4000*ave_prob_pulled))
print("Average probability of the pulls = {:.4f}".format(ave_prob_pulled))
print("Minimum probability pulled: {:.4f}".format(min(pulled_probs)))
print("\nNumber of times each bandit was pulled (combined player pulls):")
print(bandit_counts)

# These plots are for one of the two optimal players
# Make a Fraz plot: probability of selected bandit vs step
plt.plot(pulls/2, pulled_probs)
plt.xlabel("Game step (each step = 2 pulls)")
plt.ylabel("Bandit probability when selected")
plt.ylim(0.,1.0)
plt.show()

# Expected score vs step
plt.plot(pulls/2, score/2)
plt.xlabel("Game step (each step = 2 pulls)")
plt.ylabel("Expected score (one player)")
plt.show()

# Show the counts vs bin
plt.scatter(bandit_probs,bandit_counts/2,s=2)
plt.xlabel("Bandit Initial Probability")
plt.ylabel("Number of Bandit's Pulls (per player)")
plt.show()

```

### Estimating p_i from observations

```python
# Copy and play with Xin Cui's numerical calculation of the posterior
# mean and std of p_i - it includes the "discount factor" (0.97**m).
# https://www.kaggle.com/xincuimath/weighted-expected-mean-and-std-estimation 
# Added optional plotting of curves.

def get_expected_mean_std(weights, rewards, plotit=False):

    def pdf(p, weights, rewards, normalization=1):
        s = 1
        for weight, reward in zip(weights, rewards):
            if reward == 1:
                s *= (weight * p)
            else:
                s *= (1 - weight * p)
        return s / normalization

    normalization = integrate.quad(
        lambda x: pdf(x, weights, rewards), 0, 1)[0]
    first_order = integrate.quad(
        lambda x: x * pdf(x, weights, rewards, normalization), 0, 1)[0]
    second_order = integrate.quad(
        lambda x: x * x * pdf(x, weights, rewards, normalization), 0, 1)[0]
    
    # Show a plot of the "Bay-XC" pdf along with the naive beta?
    a = sum(rewards)
    b = len(rewards) - a
    
    if plotit:
        # for use in plotting:
        mean = first_order
        std = (second_order - first_order ** 2) ** 0.5
        prob_values = np.linspace(0, 1, 1000)
        fig, ax = plt.subplots(figsize = (10,6))
        # plot the beta pdf
        ax.plot(prob_values, beta.pdf(prob_values, a+1, b+1),c='gray',
                label=f'Beta: a={a}, b={b}')
        # calc and plot the Bay-XC pdf:
        bayxc_pdf_values = pdf(prob_values, weights, rewards, normalization)
        ax.plot(prob_values, bayxc_pdf_values, c='blue',
                label=f'BayXC: p_intrinsic')
        # show the mean
        peak_height = max(bayxc_pdf_values)
        ax.plot([mean,mean],[0.0,0.95*peak_height],"--",c='lightblue')
        # show +/-1 std
        ax.plot([mean+std,mean+std],[0.0,0.2*peak_height],"--",c='lightblue')
        ax.plot([mean-std,mean-std],[0.0,0.2*peak_height],"--",c='lightblue')
        # show + UCB_delta, p=5%
        ucb_delta = np.sqrt(-1.0*np.log(0.05)/(2*len(rewards)))
        ax.plot([mean+ucb_delta,mean+ucb_delta],[0.0,0.2*peak_height],"--",
                c='lightgreen')
        # plot the Bay-XC modified by the minimum weight (aka "discount factor")
        ax.plot(min(weights)*prob_values, bayxc_pdf_values/min(weights),
                c='orange', label=f'BayXC(p_i) * DF={int(1000*min(weights))/1000}')
        plt.xlabel('p_intrinsic',size=20)
        plt.ylabel('probability density value',size=16)
        plt.title("Comparing the 'BayXC' distribution (includes DFs) with Beta"+
                  ", "+str(len(rewards))+" trials", size=16)
        ax.legend(prop={'size': 14})
        plt.show()
    
    return first_order, (second_order - first_order ** 2) ** 0.5
```

```python
# Demonstrate how it works compared with usual beta distribution
trials = 17
p_one = 0.5

# No discounting
discountings = [1 for _ in range(trials)]
rewards = [int(np.random.choice([0, 1], p=(1-p_one, p_one)))
                                       for _ in range(trials)]

# Use the with-weightings formula
(mean, std) = get_expected_mean_std(discountings, rewards, plotit=True)
print("Xin Cui's function gives  mean, std:")
print((mean,std))

# Do the equivalent calculations using beta function:
print("The beta function gives  mean, std:")
print((beta.mean(1 + sum(rewards), trials+1 - sum(rewards)),
        beta.std(1 + sum(rewards), trials+1 - sum(rewards))))

print("--> They are the same when there is no 'discount factor'.")
```

```python
# Now re-make the observations and include the discounting in the "data"
# Put new observed values in the existing variables:

for it in range(trials):
    discountings[it] = 0.97**it
    p_k = p_one * discountings[it]
    rewards[it] = int(np.random.choice([0, 1], p=(1-p_k, p_k)))
    
# Use the with-weightings formula
(mean, std) = get_expected_mean_std(discountings, rewards, plotit=True)
print("Xin Cui's function gives  mean, std:")
print((mean,std))

# Do the equivalent calculations using beta function:
print("The beta function gives  mean, std:")
print((beta.mean(1 + sum(rewards), trials+1 - sum(rewards)),
        beta.std(1 + sum(rewards), trials+1 - sum(rewards))))

print("--> The BayXC (usually) gives a closer estimate" +
              " of the true p_i = {}".format(p_one))
```

```python
%%writefile ucb_half5percent.py

# This is the Santa 2020 UCB agent modified for a constant p_small = 0.05
# and using 1/2 of the delta value.
# Also randomize the starting order of bandits.
import math
import numpy as np

last_bandit = -1
total_reward = 0

sums_of_reward = None
numbers_of_selections = None

def ucb_agent(observation, configuration):    
    global sums_of_reward, numbers_of_selections, last_bandit, total_reward

    if observation.step == 0:
        numbers_of_selections = [0] * configuration["banditCount"]
        sums_of_reward = [0] * configuration["banditCount"]

    if last_bandit > -1:
        reward = observation.reward - total_reward
        sums_of_reward[last_bandit] += reward
        total_reward += reward

    bandit = 0
    max_upper_bound = 0
    for i in range(0, configuration.banditCount):
        if (numbers_of_selections[i] > 0):
            average_reward = sums_of_reward[i] / numbers_of_selections[i]
            ##delta_i = (math.sqrt(2 * math.log(observation.step+1) / 
            ##            numbers_of_selections[i]))
            # Use a fixed 5% to get the upper_bound delta:
            delta_i = math.sqrt( -1.0*math.log(0.05) / 
                                        (2*numbers_of_selections[i]))
            # Use 1/2 of the delta value
            upper_bound = average_reward + 0.5*delta_i
        else:
            # haven't pulled this bandit yet - want to pull it
            upper_bound = 1000.0 + np.random.rand()
        if upper_bound > max_upper_bound and last_bandit != i:
            max_upper_bound = upper_bound
            bandit = i
            last_bandit = bandit

    numbers_of_selections[bandit] += 1

    if bandit is None:
        bandit = 0

    return bandit
```

```python
%%writefile vegas_slots.py

# From https://www.kaggle.com/sirishks/pull-vegas-slot-machines
import numpy as np
import pandas as pd
import random, os, datetime

total_reward = 0
bandit_dict = {}

def set_seed(my_seed=42):
    os.environ['PYTHONHASHSEED'] = str(my_seed)
    random.seed(my_seed)
    np.random.seed(my_seed)

def get_next_bandit():
    best_bandit = 0
    best_bandit_expected = 0
    for bnd in bandit_dict:
        expect = (bandit_dict[bnd]['win'] - bandit_dict[bnd]['loss'] + bandit_dict[bnd]['opp'] - (bandit_dict[bnd]['opp']>0)*1.5) \
                 / (bandit_dict[bnd]['win'] + bandit_dict[bnd]['loss'] + bandit_dict[bnd]['opp'])
        if expect > best_bandit_expected:
            best_bandit_expected = expect
            best_bandit = bnd
    return best_bandit

def multi_armed_probabilities(observation, configuration):
    global total_reward, bandit_dict

    my_pull = random.randrange(configuration['banditCount'])
    if 0 == observation['step']:
        set_seed()
        total_reward = 0
        bandit_dict = {}
        for i in range(configuration['banditCount']):
            bandit_dict[i] = {'win': 1, 'loss': 0, 'opp': 0}
    else:
        last_reward = observation['reward'] - total_reward
        total_reward = observation['reward']
        
        my_idx = observation['agentIndex']
        if 0 < last_reward:
            bandit_dict[observation['lastActions'][my_idx]]['win'] = bandit_dict[observation['lastActions'][my_idx]]['win'] +1
        else:
            bandit_dict[observation['lastActions'][my_idx]]['loss'] = bandit_dict[observation['lastActions'][my_idx]]['loss'] +1
        bandit_dict[observation['lastActions'][1-my_idx]]['opp'] = bandit_dict[observation['lastActions'][1-my_idx]]['opp'] +1
        my_pull = get_next_bandit()
    
    return my_pull
```

```python
%%writefile dfmeansV16_agent.py

import numpy as np
import random   # python module, not same as np.random
import time
from scipy.stats import beta
from scipy import integrate
from matplotlib import pyplot as plt
import pandas as pd

# Routine to calculate discount factor corrected distribution
def get_expected_mean_std(weights, rewards, plotit=False):

    def pdf(p, weights, rewards, normalization=1):
        s = 1
        for weight, reward in zip(weights, rewards):
            if reward == 1:
                s *= (weight * p)
            else:
                s *= (1 - weight * p)
        return s / normalization

    normalization = integrate.quad(
        lambda x: pdf(x, weights, rewards), 0, 1)[0]
    first_order = integrate.quad(
        lambda x: x * pdf(x, weights, rewards, normalization), 0, 1)[0]
    second_order = integrate.quad(
        lambda x: x * x * pdf(x, weights, rewards, normalization), 0, 1)[0]
    
    # Show a plot of the "Bay-XC" pdf along with the naive beta?
    # REMOVED PLOT CODE
    return first_order, (second_order - first_order ** 2) ** 0.5

def prob_histogram(observation, play_state):
    # Display a histogram of the estimated probabilities during the run
    # and write snapshot values to a csv file.
    global band_means, band_stds, band_dfmeans, band_dfstds
    global band_rewards
    counts, binvals = np.histogram(band_means,range=(0.0,1.0),bins=20)
    fig, ax = plt.subplots(figsize = (7,3))
    ax.hist(band_means,bins=binvals,histtype='step',
             color='lightblue', linewidth=4)
    ax.hist(band_dfmeans,bins=binvals,histtype='step',
             color='orange', linewidth=2)
    title_str = ("Play_state {} completed".format(play_state) + 
                      " at step {}".format(observation.step) +
                      " with reward {}".format(observation.reward))
    ax.set_title(title_str)
    plt.show()
    # Write the values to a dataframe csv
    frame = pd.DataFrame(band_dfmeans, columns=['dfmean'])
    frame['dfstd'] = band_dfstds
    frame['step'] = observation.step
    frame['mean'] = band_means
    frame['std'] = band_stds
    # count the number of times * I * pulled the bandit
    npulls = np.zeros(len(band_rewards))
    for iband in range(len(band_rewards)):
        npulls[iband] = len(band_rewards[iband])
    frame['npulls'] = npulls
    
    print("Saving data frame: play_state_{}.csv".format(play_state))
    print(frame.head(5),"\n\n\n")
    frame.to_csv("play_state_{}.csv".format(play_state),
                index_label='bandit')

def dfmeans_agent(observation, configuration):
    # A 7-state scheme is used:
    # 1,2) rand pull all 1s, mix w/never pulled
    # 3) dfmean, mix w/(1/3) 1 zeros pull
    # 4) dfmean+0.5dfstd, mix w/(1/4) 2 zeros or 30-50,7
    # 5) dfmean+0.5*dfstd, mix w/(1/6) dfstd > 0.06
    # 6) dfmean, require mean>0.20
    # 7) dfmean, require mean>0.25
    # The dfmean/dfstd are calculated using the BayXC function of Xin Cui.
    
    global play_state, last_reward, my_last, other_last, other_lastlast
    # Keep track of the bandit's actions, etc.
    global band_pulls, band_rewards, band_rewdfs
    global band_means, band_stds, band_dfmeans, band_dfstds
    # "figure of merits" - this is used for selection in different ways
    global band_foms

    df = 0.97   # the discount factor
    nband = configuration.banditCount
    
    # - - - - -  Setup - - - - -
    # Set things up if this is the start:
    if observation.step == 0:
        play_state = 1
        last_reward = observation.reward
        got_reward = 0
        # current and previous bandit I pulled
        my_action = -10
        my_last = -20
        # current and last two bandits pulled by the other player
        other_action = -1
        other_last = -2
        other_lastlast = -3
        # Setup values related to the bandits:
        band_pulls = np.zeros(nband)
        # Initially have no rewards info on the bandits,
        # setup the list of lists with flag values.
        band_rewards = [[-1.0]] * nband
        band_rewdfs = [[1.0]] * nband
        # Initial values of the means and stds
        band_means = np.zeros(nband)
        band_stds = np.zeros(nband)
        band_dfmeans = np.zeros(nband)
        band_dfstds = np.zeros(nband)
        band_foms = np.zeros(nband)
    else:
        # - - - - - Digest new information from the pull - - - - -
        # Here we'll get the results from doing the last step's pull
        # and do basic updating and processing
        my_index = observation.agentIndex
        my_action = observation.lastActions[my_index]
        got_reward = int(observation.reward - last_reward)
        last_reward = observation.reward  # save for next time
        # the other player
        other_action = observation.lastActions[1 - my_index]
        # Update the bandit reward info
        if band_rewards[my_action][0] > -0.5:
            # just add on this reward
            band_rewards[my_action] = band_rewards[my_action] + [got_reward]
            band_rewdfs[my_action] = (band_rewdfs[my_action] +
                                  [df**band_pulls[my_action]])
        else:
            # This is the first reward I've gotten for this bandit -
            # set the first rewards and rewdfs appropriately:
            band_rewards[my_action] = [got_reward]
            band_rewdfs[my_action] = [df**band_pulls[my_action]]
                
        # Update the pulls
        band_pulls[my_action] += 1
        band_pulls[other_action] += 1
        
        # Update the df-corrected mean and std for my selected bandit
        mean, std = get_expected_mean_std(band_rewdfs[my_action],
                            band_rewards[my_action], plotit=False)
        # We're interested in the high-side std: since the dist can be skewed,
        # clip the std to be at most 0.5*(1.0 - mean)
        std = min([std, 0.5*(1.0 - mean)])
        # save the estimated intrinsic(initial) mean and std:
        band_means[my_action] = mean
        band_stds[my_action] = std
        # set the new effective dfmeans, dfstds value for this bandit
        band_dfmeans[my_action] = mean*df**band_pulls[my_action]
        band_dfstds[my_action] = std*df**band_pulls[my_action]
        # and adjust the other's selected bandit (if not same as mine)
        if other_action != my_action:
            band_dfmeans[other_action] *= df
            band_dfstds[other_action] *= df
            
        # Assume/include an opponent reward if
        # play_state 1,2,3 and opponent pulled same one again and pulls <= 10
        if( (play_state <= 3) and (other_action == other_last) and
                  (other_action != my_action) and (band_pulls[other_action] <= 10)):
            # Same code as above but for other_action; could be the first reward too.
            if band_rewards[other_action][0] > -0.5:
                band_rewards[other_action] = band_rewards[other_action] + [1]
                band_rewdfs[other_action] = (band_rewdfs[other_action] +
                                  [df**(band_pulls[other_action]-2)])
            else:
                band_rewards[other_action] = [1]
                band_rewdfs[other_action] = [df**(band_pulls[other_action]-2)]
            mean, std = get_expected_mean_std(band_rewdfs[other_action],
                            band_rewards[other_action], plotit=False)
            std = min([std, 0.5*(1.0 - mean)])
            band_means[other_action] = mean
            band_stds[other_action] = std
            band_dfmeans[other_action] = mean*df**band_pulls[other_action]
            band_dfstds[other_action] = std*df**band_pulls[other_action]
            # Could report this... but it's not a pull decision so keep it in background.
            ##print("Step",observation.step,": assume/add opponent's reward for", 
            ##      other_action, ", dfmean =",band_dfmeans[other_action])
        
    # - - - - - Decide on the next pull - - - - -
    # OK, now with the new information digested,
    # decide on a value for the next pull_bandit
    
    # Look for possible poaching opportunities
    poach_it = False
    if ((other_action == other_last) and
                (other_last == other_lastlast)):
        # play_state 1 poaching
        if play_state == 1:
            # Poach it if I have not gotten a 0 reward from this bandit
            if (min(band_rewards[other_action]) > 0.5):
                poach_it = True
        # later play_state poaching
        if play_state == 2:
            # Only poach in play state 2 if the mean is high
            pass  # not using state 2.
        if play_state == 3:
            # play state 3: require a minimal dfmean value
            if (band_dfmeans[other_action] > 0.35):
                poach_it = True
        if play_state >= 4:
            # play state 4 or later: require a minimal dfmean value
            if (band_dfmeans[other_action] > 0.25):
                poach_it = True
                
    # Do the apropriate state's action
    if poach_it:
        pull_bandit = int(other_action)
        print("Step", observation.step, ": poaching",other_action,
                 "w/ dfmean =",band_dfmeans[other_action])
        poach_it = False
    #
    elif play_state == 1:  # 1,2) rand pull all 1s, mix w/never pulled
        # --- DO THIS STATE'S EVAL AND PULLS:
        # Recalculate what's available each time, so don't need to use global band_foms
        # Find all bandits not yet pulled: they have reward as [-1].
        not_pulled = np.zeros(nband) + 0.01*np.random.rand(nband)
        for iband in range(nband):
            if (band_rewards[iband][0] < -0.5):
                not_pulled[iband] = 1.0 + 0.01*np.random.rand()
        # Find all bandits that currently have all 1s
        have_all1s = np.zeros(nband) + 0.01*np.random.rand(nband)
        for iband in range(nband):
            if min(band_rewards[iband]) > 0.5:
                have_all1s[iband] = 1.0 + 0.01*np.random.rand()
        #
        # There are 4 cases for the two types of selected bandits
        # depending on which are available.
        # When exploiting, if the opponent's last pull was one I want
        # to exploit, then do it now.
        #   Explore is only option
        if (max(not_pulled) > 0.5) and (max(have_all1s) < 0.5):
            pull_bandit = int(np.argmax(not_pulled))  # explore
            print("Step",observation.step, ": Explore",pull_bandit)
        #   Exploit is only option
        elif (max(not_pulled) < 0.5) and (max(have_all1s) > 0.5):
            if have_all1s[other_action] > 0.5:
                pull_bandit = int(other_action)  # exploit opponent
                print("Step", observation.step, ": +exploit other's choice",
                      other_action)
            else:
                pull_bandit = int(np.argmax(have_all1s))  # exploit
                print("Step",observation.step, ": +exploit",pull_bandit)
        #   Both available: Exploit, but Explore every nth time
        elif (max(not_pulled) > 0.5) and (max(have_all1s) > 0.5):
            if (observation.step % 3) == 0:
                pull_bandit = int(np.argmax(not_pulled))  # explore
                print("Step",observation.step, ": Explore",pull_bandit)
            else:   
                if have_all1s[other_action] > 0.5:
                    pull_bandit = int(other_action)  # exploit opponent
                    print("Step", observation.step, ": +exploit other's choice",
                          other_action)
                else:
                    pull_bandit = int(np.argmax(have_all1s))  # exploit
                    print("Step",observation.step, ": +exploit",pull_bandit)
        #   Neither available...
        else:
            # --- GO TO NEXT STATE?:
            # Here when there are none to explore and none to exploit,
            # time to go to the next state. Make a choice based on dfmeans:
            pull_bandit = int(np.argmax(band_dfmeans))
            prob_histogram(observation, play_state)
            play_state = 3            
    #
    elif play_state == 3:  # 3) dfmean, mix w/(1/3) 1 zeros pull
        # --- DO THIS STATE'S EVAL AND PULLS:
        # find ones with 1 or less 0 rewards
        band_foms = np.zeros(nband) + 0.01*np.random.rand(nband)
        for iband in range(nband):
            # pull if one or less 0 rewards
            if ( (len(band_rewards[iband]) - sum(band_rewards[iband])) < 2):
                    band_foms[iband] = 1.0 + 0.01*np.random.rand()
        # alternate with pulling the highest dfmean
        if (observation.step % 3 == 0) and (max(band_foms) > 0.5):
            pull_bandit = int(np.argmax(band_foms))
        else:
            pull_bandit = int(np.argmax(band_dfmeans))
        # --- GO TO NEXT STATE?:
        if max(band_foms) < 0.5:
            prob_histogram(observation, play_state)
            play_state = 4
    #
    elif play_state == 4:  # 4) dfmean+0.5dfstd, mix w/(1/4) 2 zeros or 30-50,7
        # --- DO THIS STATE'S EVAL AND PULLS:
        # find ones with 2 or less 0 rewards
        #  OR
        # mean between 0.30, 0.50 with low number of pulls
        band_foms = np.zeros(nband) + 0.01*np.random.rand(nband)
        for iband in range(nband):
            if ( 
                # pull if two or less 0 rewards
                ( ((len(band_rewards[iband]) - sum(band_rewards[iband])) < 3) and
                       (band_means[iband] < 0.60) )
                or
                # mean between 0.30, 0.50 with low number of pulls
                ( (band_means[iband] > 0.30) and (band_means[iband] < 0.50)
                        and (len(band_rewards[iband]) <= 7) ) 
                ):
                band_foms[iband] = 1.0 + 0.01*np.random.rand()
        # alternate with pulling the highest dfmean
        if observation.step % 4 == 0:
            pull_bandit = int(np.argmax(band_foms))
        else:
            pull_bandit = int(np.argmax(band_dfmeans + 0.5*band_dfstds))
        # --- GO TO NEXT STATE?:
        if max(band_foms) < 0.5:
            prob_histogram(observation, play_state)
            play_state = 5
    #
    elif play_state == 5:  # 5) dfmean+0.5*dfstd, mix w/(1/6) dfstd > 0.06
        # --- DO THIS STATE'S EVAL AND PULLS:
        if (observation.step % 6 == 0) and (max(band_dfstds) > 0.06):
            # do a bandit that has dfstd > 0.06
            pull_bandit = int(np.argmax(band_dfmeans * 
                                       (band_dfstds > 0.06)))
        else:
            pull_bandit = int(np.argmax(band_dfmeans + 0.5*band_dfstds +
                                   0.02*np.random.rand(nband)))
        # --- GO TO NEXT STATE?:
        if observation.step > 1500:
            prob_histogram(observation, play_state)
            play_state = 6
    #
    elif play_state == 6:  # 6) dfmean requiring mean>0.20
        # --- DO THIS STATE'S EVAL AND PULLS:
        pull_bandit = int(np.argmax(band_dfmeans * (band_means > 0.20)))
        # --- GO TO NEXT STATE?:
        # Add a state at 1800 to get the histogram
        if observation.step > 1750:
            prob_histogram(observation, play_state)
            play_state = 7
    #
    elif play_state == 7:  # 7) dfmean requiring mean>0.25
        # --- DO THIS STATE'S EVAL AND PULLS:
        pull_bandit = int(np.argmax(band_dfmeans * (band_means > 0.25)))
        # --- GO TO NEXT STATE?:
        if observation.step == 1998:
            prob_histogram(observation, play_state)
            # no next state ;-)
    #
    else:
        print("Shouldn't get to this else.")
    
    # Update past actions
    my_last = 0 + my_action
    other_lastlast = 0 + other_last
    other_last = 0 + other_action
    
    return pull_bandit
```

```python
# How much variation in the score is expected from the random threshold selection?

# Randomly select a set of thresholds and see what the expected optimal-play score is.
# How the threshold values are setup in the environment, from:
# https://github.com/Kaggle/kaggle-environments/blob/b043a9588ef14018816a47b93b841cbcbd9ba906/kaggle_environments/envs/mab/mab.py#L97

# Using the normal random module
def sample():
    """Obtain a value between 0 and sampleResolution to check against a bandit threshold."""
    return random.randint(0, 100)

ngames = 100
game_scores = []
for igame in range(ngames):
    start_thresholds = [sample() for _ in range(100)]
    initial_probs = (np.array(start_thresholds))/100.0
    current_probs = 1.0 * initial_probs
    pulled_probs = []
    for ip in range(4000):
        # choose the highest prob one
        ihigh = np.argmax(current_probs)
        pulled_probs = pulled_probs + [current_probs[ihigh]]
        current_probs[ihigh] *= 0.97
    this_score = sum(pulled_probs)/2.0
    game_scores = game_scores + [this_score]
    ##print("Expected rewards per player = {:.1f}".format(this_score)
```
