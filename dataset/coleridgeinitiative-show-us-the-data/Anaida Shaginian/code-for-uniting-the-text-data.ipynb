## Uniting the texts

```python
def concat(column):
    res = ' '
    for st in column:
        if type(st) == str:
            res += st
    return res
```

```python
train_gr = train.groupby('pub_id')['text'].apply(concat).reset_index()
train_gr.loc[train['pub_id'].isin(train_meta['Id']), 'cleaned_label'] = train_meta.loc[train_meta['Id'].isin(train_gr['pub_id']),'cleaned_label']
```
