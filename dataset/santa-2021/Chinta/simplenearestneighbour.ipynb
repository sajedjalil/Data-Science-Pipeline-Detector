```python
def is_perm(s):
    y = True
    for k in range(1,8):
        y = y&(str(k) in s)
        if not y: break
    return y

permutations = []
for k in range(len(best)-6):
    s = best[k:k+7]
    if (is_perm(s))&(s not in permutations)&(s[:2]!='54'):
        permutations.append(s)
        
print('There are',len(permutations),'that can be split among 3 strings')
```

```python
def hamming_distance(str1, str2):
    return sum( (c1!=c2) for c1, c2 in zip(str1, str2))

def offset(s1, s2):
    assert(len(s1)==len(s2))
    ln = len(s1)
    j = ln
    for k in range(0,ln):
        if hamming_distance(s1[k:],s2[:7-k])==0:
            j=k
            break
    return j
```

```python
# network is the current network from all visited nodes
# nodes has the list of nodes yet to be added to network
# returns a new network, after adding a nearest node
def find_nn_nk(network, nodes, init_node=6):
    distance_M = np.zeros((1, len(nodes)), dtype='int8')
    if (network == ""):
        nodes.remove(nodes[init_node])
        return(nodes[init_node], nodes, 0, 0)
    else:    
        for j in range(len(nodes)):
            if (network[-7:] != nodes[j]):               
                distance_M[0,j] = offset(network[-7:], nodes[j])
            else:
                distance_M[0,j] = 8
        #print(network, np.argmin(distance_M), nodes[np.argmin(distance_M)])
        new_network = network + str(nodes[np.argmin(distance_M)], )[-np.min(distance_M):]
        removed_node = nodes[np.argmin(distance_M)]
        nodes.remove(nodes[np.argmin(distance_M)])
        return(new_network, nodes, np.min(distance_M), removed_node)
```
