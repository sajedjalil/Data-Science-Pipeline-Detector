```python
LETTERS = {
    '1': 'üéÖ',  # father christmas
    '2': 'ü§∂',  # mother christmas
    '3': 'ü¶å',  # reindeer
    '4': 'üßù',  # elf
    '5': 'üéÑ',  # christmas tree
    '6': 'üéÅ',  # gift
    '7': 'üéÄ',  # ribbon
}
```

```python
string_list = list(string_list)
```

```python
movie_schedule = []
for char in string_list:
    movie_schedule.append(LETTERS[char])
```

```python
joined_movie_schedule = ''.join(movie_schedule)
```

```python
submission = [joined_movie_schedule, joined_movie_schedule, joined_movie_schedule]
```

```python
schedule_combination = pd.Series(submission, name='schedule')
schedule_combination.to_csv('submission.csv', index=False)
```
