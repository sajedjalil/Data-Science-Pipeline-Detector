# Assign Permutations To 3 Submission Strings
There are 5040 permutations for `n=7`. The 120 that start with ðŸŽ…ðŸ¤¶ must be contained in every string. There are 4920 remaining. We will assign rougly one third of the remaining 4920 to each string. And we will assign the 120 to every string.

```python
def is_perm(s):
    y = True
    for k in range(1,8):
        y = y&(str(k) in s)
        if not y: break
    return y

permutations = []
for k in range(len(best)-6):
    s = best[k:k+7]
    if (is_perm(s))&(s not in permutations)&(s[:2]!='54'):
        permutations.append(s)
        
print('There are',len(permutations),'that can be split among 3 strings')
```

```python
mandatory = ['54'+''.join(x) for x in itertools.permutations(['1','2','3','6','7'], 5)]
print('There are',len(mandatory),'that must be incluced in all 3 strings')
```

```python
adjust1 = 18
adjust2 = 12

group1 = permutations[:1640-adjust1] + mandatory
group2 = permutations[1640-adjust1:3280+adjust2] + mandatory
group3 = permutations[3280+adjust2:] + mandatory

len(group1), len(group2), len(group3)
```

# Helper Functions

```python
def hamming_distance(str1, str2):
    return sum( (c1!=c2) for c1, c2 in zip(str1, str2))

def offset(s1, s2):
    assert(len(s1)==len(s2))
    ln = len(s1)
    j = ln
    for k in range(0,ln):
        if hamming_distance(s1[k:],s2[:7-k])==0:
            j=k
            break
    return j
```

```python
def reformulate_ATSP_as_TSP(x, inf=100, neg_inf=0):
    np.fill_diagonal(x, neg_inf)
    a = np.concatenate((np.full(x.shape, inf), np.transpose(x)), axis=1)
    b = np.concatenate((x, np.full(x.shape, inf)), axis=1)
    return np.concatenate((a, b), axis=0)

def get_tsp_solution(group):
    
    # CREATE DISTANCE MATRIX
    SIZE_ATSP = len(group)
    M = np.zeros((SIZE_ATSP, SIZE_ATSP), dtype='int8')
    for j in range(SIZE_ATSP):
        #if j%25==0: print(j,', ',end='')
        for k in range(SIZE_ATSP):
            M[j,k] = offset(group[j],group[k])
    
    # Concorde can solve only symmetric TSP
    # So we convert ATSP to TSP
    M = reformulate_ATSP_as_TSP(M)
    SIZE_TSP = SIZE_ATSP*2

    # WRITE PARAMETER FILE
    f = open(f'distances.tsp','w')
    f.write("NAME: distances\n")
    f.write("TYPE: TSP\n")
    f.write("COMMENT: Symmetric TSP\n")
    f.write(f"DIMENSION: {SIZE_TSP}\n")
    f.write("EDGE_WEIGHT_TYPE: EXPLICIT\n")
    f.write("EDGE_WEIGHT_FORMAT: FULL_MATRIX\n")
    f.write("EDGE_WEIGHT_SECTION\n")
    for j in range(SIZE_TSP):
        #if j%25==0: print(j,', ',end='')
        for k in range(SIZE_TSP):
            f.write(f"{M[j,k]:2d} ") 
        f.write("\n")
    f.close()
    
    # EXECUTE TSP SOLVER
    !cd concorde/TSP; ./concorde -x -V ../../distances.tsp
    # READ RESULTING ORDER
    with open('./concorde/TSP/distances.sol') as f:
        lines = f.readlines()
    perms = []
    for ln in lines[1:]:
        lns = ln.strip().split(' ')
        perms.extend([int(x) for x in lns])
    
    # TSP -> ATSP
    perms = [x for x in perms if x < SIZE_ATSP]
    perms = list(set(perms))
    
    # CREATE STRING
    result = group[ perms[0] ]
    for k in range(1,len(perms)):
        s1 = group[ perms[k-1] ]
        s2 = group[ perms[k] ]
        d = offset(s1,s2)
        assert(d!=0)
        result += s2[-d:]
    
    return result
```

# Solve with Concorde TSP Solver

```python
def print_header(x):
    print(); print()
    print('#'*25)
    print('### Computing String',x)
    print('#'*25); print()

print_header(1)
string1 = get_tsp_solution(group1)
print_header(2)
string2 = get_tsp_solution(group2)
print_header(3)
string3 = get_tsp_solution(group3)
```
