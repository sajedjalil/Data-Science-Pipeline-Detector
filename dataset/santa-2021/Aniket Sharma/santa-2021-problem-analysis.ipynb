## distance_matrix.csv

```python
distance_matrix = pd.read_csv('/kaggle/input/santa-2021/distance_matrix.csv', index_col='Permutation')
distance_matrix.head()
```

Distance of 0 means that the two strings are exactly the same, so the diagonal elements of the matrix are all 0.

Distance of 7 means that the two strings do not overlap, therefore 7 is the maximum value in the distance matrix.
