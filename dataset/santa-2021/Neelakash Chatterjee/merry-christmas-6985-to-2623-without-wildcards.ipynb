# The Problem Statement and it's Explanation

Find a set of three strings containing every permutation of the seven symbols ðŸŽ…, ðŸ¤¶, ðŸ¦Œ, ðŸ§, ðŸŽ„, ðŸŽ, and ðŸŽ€ as substrings, subject to the following conditions:

1) Every permutation must be in at least one string.

2) Each permutation beginning with ðŸŽ…ðŸ¤¶ must be in all three strings.

3) Each string may have up to two wildcards ðŸŒŸ, which will match any symbol in a permutation. No string of length    seven containing more than one wildcard will count as a permutation.

4) Your score is the length of the longest of the three strings, with lower scores better.


#### Now what does the above statement mean ? Let's understand the problem with a simple example (consider only 3 symbols):

Let, ðŸŽ…, ðŸ¤¶, ðŸ¦Œ be represented as 1, 2, 3 respectively . Now there can be 6 unique combinations of these symbols (numbers in our example) such as 123, 321, 213 and so on . But how 6 combinations? The answer is simply 3! (!-factorial sign, don't mistake it to be an exclaimation sign), factorial is a mathematical concept which is defined as $n! = n\dot (n-1)\dot (n-2)...1 \hspace{0.25cm} Eg: 7! = 7.6.5.4.3.2.1 = 5040$ . 

Now we have been asked to create 3 strings (here let's say 3 number) made of these permutations , but there are certain conditions:

1) There must be atleast 1 number out of these 3 numbers which has all the permutations in it , all 6 of them . Like 123132213231312321 . 

2) All combinations of first 2 numbers must be there in all the three numbers , which means all such unique combinations which starts with 12.., must be there in all the numbers . In our case it is only 123, but when we will start working with the whole dataset there are going to be lot more and for your string to be valid all such combinations must be there in all 3 strings . Like in our case 123 must be in all our numbers .

3) Now there is a wildcard , represented by ðŸŒŸ which gives us the independence of using any symbol (number here in example) in that place . 

***But the restriction is when string lenght is 7 in the original problem , we can't use more than 1 wildcard but if the string lenght is more than 7 we can use atmost 2 wildcards .***

4) Our score is the length of the largest string (or number here) amongst the 3 strings that we'll be creating .


Say considering only those 3 symbols (or numbers) the strings possible are: 
123132213231312321, 123, 12 . (This set satisfies all conditions, 1st string has all possible combinations of the numbers and also has all combinations of 12..., 2nd string has all combinations of 12.., 3rd string as well has all combinations of 12.... )

**So each and every condition is satisfied** 

Here the score is going to be = 18 , i.e. len(123132213231312321) .

But we can write 123132213231312321 in more and more optimised ways, like say 123121321 - see this also is a valid string and satisfies all conditions , but the length is only 9 . As a result our score becomes 9 which much better than our previous score (lower the better) . **So this is the objective of this problem**

# Super-permutation Problem

**Now, the sample file provided has a very basic solution, we need to optimise the same to get a better score .**

Now whatever I explained in the above cells, like how to reduce the string size yet contain all permutations , that has a special name called **super-permutation**, but suprisingly see the string "super-permutation" can be written as su**PER**mutation . 

- The supermutation of 1 is 1 . length = 1
- The supermutation of 2 is 121 Length = 3, (12 and 21, the first two digits in 121 has one permutation and the last two digit has another one.)
- The supermutation of 3 has Length = 9 . (See the example above in the first cell)
- The supermutation of 4 has Length = 33 . 
- The supermutation of 5 has Length = 153 .

But that's it , supermutation of 5 is the last known to us. But do you see any pattern here ?
Yeah!

- 1! = 1
- 1!+2! = 3
- 1!+2!+3! = 9
- 1!+2!+3!+4! = 33
- 1!+2!+3!+4!+5! = 153

But, then super-permutation of 6 should have been 1!+2!+3!+4!+5!+6! = 873 ? But unfortunately, this didn't hold, recently someone has found the super-permutation of 6 to be **872** . 

```python
# The super-permutation of 7 with above formula turns out to be ?

def factorial(n):
    # Factorial of 0 is 1 (base case)
    if n==0:
        return 1
    else:
        return n*factorial(n-1)

def n_method_supermutation(n):
    sp = 0
    for i in range(1, n+1):
        sp += factorial(i)
    return sp

len_of_sp = n_method_supermutation(7)
print(len_of_sp)
```

The length turns out to be 5913, but still date the shortest has been found to be 5906 . So definitely the pattern does not hold . But yes, it is better than the sample file which has a length of 6985.

```python
# All possible permutations

all_permu = [''.join(i) for i in itertools.permutations(symbols, 7)]
```

```python
# Permutations that are mandatory .

man_per = ['ðŸŽ…ðŸ¤¶'+''.join(j) for j in itertools.permutations(symbols[2:], 5)]
```

The best solution is the best till date , superpermutation of with 7 numbers or digits or symbols less than 5906 has not been found till date , so as we need **3 strings which in total would have all the permutations** , we just need to divide the best solution into 3 parts and try to add to it as needed to satisfy the given conditions . 

```python
# Converting the number string to symbols .

best_sy_sol = ""

for x in best_solution:
    best_sy_sol += symbols[int(x) % 7 - 1]
```

```python
len(best_sy_sol)
```

```python
indices = [len(best_sy_sol)//3, (len(best_sy_sol)//3)*2]
```

```python
indices
```

Let's divide the string into 3 parts using these indices but we'll slice it in such a way so that no combination misses out , but at the same time we also try to keep the strings as short as possible

```python
# Here we are slicing it in such a way, 24 symbols are missing out , but we'll take care of those later .

combos = [best_sy_sol[:indices[0]-12], best_sy_sol[indices[0]+12:indices[1]-12], best_sy_sol[indices[1]+12:]]
```

```python
print(f"Length t1 = {len(combos[0])}, t2 = {len(combos[1])}, t3 = {len(combos[2])}")
```

**Here we are just checking each of the 3 strings if it contains all the mandatory combinations of 'ðŸŽ…ðŸ¤¶', if not we just concatenate those permutations to those strings** 

```python
# Now let's optimize these strings according to our rules. 

for i in range(3):
    for per in man_per:
        if per not in combos[i]:
            combos[i] += per
```

Let's check if we are missing out any permutation when all the 3 strings are taken together.

```python
missed = []

for j in all_permu:
    if j not in combos[0] and j not in combos[1] and j not in combos[2]:
        missed.append(j)
```

```python
# if any permutation is not found in any of these three strings we add that particular string to the missed list.
# which here turns out to be 44 .

len(missed)
```

```python
lengths_ = [len(combos[0]), len(combos[1]), len(combos[2])]
```

```python
# With the missing permutations, the string lengths are :

for i in lengths_:
    print(i)
```

Here after doing all the neccessary additions to all the strings, we are adding the **missed out permutations to the shortest string .** Since our final score is based on our largest string .

```python
if len(missed) > 0:
    index = np.where(min(lengths_))[0][0]
    for i in missed:
        combos[index] += i
```

```python
lengths_ = [len(combos[0]), len(combos[1]), len(combos[2])]
```

```python
# This is our score . 

max(lengths_)
```

```python
# Checking again to see whether we've missed out on some permutations or not . Seems like we didn't . 

for j in all_permu:
    if j not in combos[0] and j not in combos[1] and j not in combos[2]:
        print(j)
```
