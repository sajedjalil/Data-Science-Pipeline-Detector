```python
import plotly as py
import plotly.graph_objs as go
import ipywidgets as widgets
import numpy as np
from scipy import special

py.offline.init_notebook_mode(connected=True)

```

```python
#Layout
x = np.linspace(0, np.pi, 1000)
layout = go.Layout(
    title='My Example',
    yaxis=dict(title='volts'),
    xaxis=dict(title='ns')
)

def update_plot(signals, freq):
    data = []
    for s in signals:
        trace1 = go.Scatter(
            x=x,
            y=special.jv(s, freq * x),
            mode='lines',
            name='bessel {}'.format(s),
            line=dict(shape='spline')
        )
        data.append(trace1)
    #Create Figure
    fig = go.Figure(data=data, layout=layout)
    py.offline.iplot(fig)

signals = widgets.SelectMultiple(options=list(range(6)), value=(0, ), description='Bessel Order')
freq = widgets.FloatSlider(min=20,max=50,value=1,description='Freq')
widgets.interactive(update_plot, signals=signals, freq=freq)
```

```python
#Layout
x = np.linspace(0, np.pi, 1000)
layout = go.Layout(
    title='My Example',
    yaxis=dict(title='volts'),
    xaxis=dict(title='ns')
)

def update_plot(signals, freq):
    data = []
    for s in signals:
        trace1 = go.Scatter(
            x=x,
            y=special.jv(s, freq * x),
            mode='lines',
            name='bessel {}'.format(s),
            line=dict(shape='spline')
        )
        data.append(trace1)
    #Create Figure
    fig = go.Figure(data=data, layout=layout)
    py.offline.iplot(fig)

signals = widgets.SelectMultiple(options=list(range(6)), value=(0,1,2,3,4,5), description='Bessel Order')
freq = widgets.FloatSlider(min=20,max=100,value=20,description='Freq')
widgets.interactive(update_plot, signals=signals, freq=freq)
```
