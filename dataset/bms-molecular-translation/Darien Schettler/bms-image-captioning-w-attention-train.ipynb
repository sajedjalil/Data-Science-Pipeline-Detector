<b style="text-decoration: underline; font-family: Verdana; text-transform: uppercase;">APPROACH OVERVIEW</b>

In this example, we will train a model on a relatively small amount of data to allow for us to see things end-to-end. We do this because the competition data is relatively large. We will also offer the ability to only infer on a smaller part of the InChI string to demonstrate capability.

<h3 style="text-align: font-family: Verdana; font-size: 20px; font-style: normal; font-weight: normal; text-decoration: none; text-transform: none; letter-spacing: 2px; color: navy; background-color: #ffffff;">1.1  THE MODELS</h3>

The overall model architecture we will use is similar to [**Show, Attend and Tell: Neural Image Caption Generation with Visual Attention**](https://arxiv.org/abs/1502.03044). Our pipeline is made up of 3 seperate components.

<br>

<b style="text-decoration: underline; font-family: Verdana; text-transform: uppercase;">COMPONENT 1 - THE FIXED ENCODER</b>
* This is the model that will take the image and encode it as a feature vector.
* We will be using an off-the-shelf model from Tensorflow and we will NOT be training it (Using a model that is more fine-tuned for a task related to what we are doing would be a very good way to improve).
* We will take the raw output of a headless backbone model and combine the smaller two axes together to give us a 2D vector output *(3D including batch)* **`(batch_size, attention_size, final_model_layer_channels_size)`**

<b style="text-decoration: underline; font-family: Verdana; text-transform: uppercase;">COMPONENT 2 - THE TRAINABLE ENCODER</b>
* This is the model that will take the encoded feature vector from the **fixed encoder** and will pass it through a FC network to learn the associated task specific patterns.
* We will be using a single FC layer and we WILL BE training it.
* The output from this network will be the **IMAGE EMBEDDING** and it will have the following 3D Shape (including batch) **`(batch_size, attention_size, embedding_size)`**

<b style="text-decoration: underline; font-family: Verdana; text-transform: uppercase;">COMPONENT 3 - THE TRAINABLE DECODER</b>
* This is the model that will take in multiple inputs and will pass them through a RNN (GRU/LSTM ... Transformer would be best probably) to learn how the image and trailing sequence informs on the desired prediction string.<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;– **INPUT 1**: The feature vector encoded from the image of the molecule<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;– **INPUT 2**: The [**hidden state**](https://ai.stackexchange.com/questions/16133/what-exactly-is-a-hidden-state-in-an-lstm-and-rnn) (i.e. what we've learned in the sequence so far)<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;– **INPUT 3**: The previous word in the sequence (starts with the start token)<br><br>
* We will be using a GRU layer for the Recurrent part of the network.
* The actual network is very similar to [**Show, Attend and Tell Architecture**](https://arxiv.org/pdf/1502.03044.pdf)
* The output from this network will be a vector the length of the vocabulary for every step (until a stop token is hit).

```python
def flatten_l_o_l(nested_list):
    """ Flatten a list of lists """
    return [item for sublist in nested_list for item in sublist]


def tf_load_image(path, img_size=(224,224), tile_to_3_channel=True, invert=False):
    """ Load an image with the correct size and shape """
    img = decode_img(tf.io.read_file(path), img_size, n_channels=1, invert=invert)        
    if tile_to_3_channel:
        return tf.tile(img, tf.constant((1, 1, 3), dtype=tf.int32))
    else:
        return img

    
def pad_to_square(a, constant=255):
    """ Pad a tensor array `a` evenly until it is a square """
    h_src = tf.shape(a)[0]
    w_src = tf.shape(a)[1] 
    if w_src>h_src: # pad height
        n_to_add = w_src-h_src
        top_pad = n_to_add//2
        bottom_pad = n_to_add-top_pad
        a = tf.pad(a, [(top_pad, bottom_pad), (0, 0), (0, 0)], mode='constant', constant_values=constant)
    elif h_src>w_src: # pad width
        n_to_add = h_src-w_src
        left_pad = n_to_add//2
        right_pad = n_to_add-left_pad
        a = tf.pad(a, [(0, 0), (left_pad, right_pad), (0, 0)], mode='constant', constant_values=constant)
    else:
        pass
    return a
    
    
def decode_img(img, img_size=(224,224), n_channels=1, invert=False):
    """ Decode the image by utilizing TF ... pad to square ... and resize """
    
    # convert the compressed string to a 3D uint8 tensor
    img = tf.image.decode_png(img, channels=n_channels)
    
    if invert:
        img = tf.ones_like(img, dtype=tf.uint8)*255-img
        constant_pad=0
    else:
        constant_pad=255
    
    # resize the image to the desired size
    img = pad_to_square(img, constant=constant_pad)
    img = tf.image.resize(img, img_size)
    return tf.cast(img, tf.uint8)


def tokens_to_str(caption_tokens, discard_padding=True):
    """ Should convert a string of token ids to an InChI string """
    if discard_padding:
        return "".join([int_2_char_lex[x] for x in caption_tokens if x!=len(int_2_char_lex)])
    else:
        return "".join([int_2_char_lex[x] for x in caption_tokens])
    
    
def evaluate(image, from_np=False):
    """ TBD """
    attention_plot = np.zeros((MAX_LEN, fixed_encoder.output_shape[1]))
    hidden = tf.zeros((1, RNN_UNITS), tf.float32)

    if not from_np:
        temp_input = tf.expand_dims(tf_load_image(image, img_size=INPUT_SHAPE[:-1]), 0)
        img_tensor_val = fixed_encoder(temp_input)
    else:
        img_tensor_val=image
    
    features = trainable_encoder(img_tensor_val)
    dec = tf.ones((1, 1), tf.uint8)
    result = [int_2_char_lex[1],]

    for i in range(MAX_LEN-1):
        predictions, hidden, attention_weights = dec_model([dec, hidden, features])
        attention_plot[i] = tf.reshape(attention_weights, (-1, )).numpy()
        predicted_id = tf.random.categorical(predictions, 1)[0][0].numpy()
        result.append(int_2_char_lex[predicted_id])
        if int_2_char_lex[predicted_id] == '<END>':
            return result, attention_plot

        dec = tf.expand_dims([predicted_id], 0)

    attention_plot = attention_plot[:len(result), :]
    return result, attention_plot


def plot_attention(image, result, attention_plot):
    """ TBD """    
    temp_image = np.array(Image.open(image))

    fig = plt.figure(figsize=(18, 14))

    len_result = len(result)
    for l in range(len_result):
        temp_att = np.resize(attention_plot[l], (8, 8))
        ax = fig.add_subplot(len_result//2, len_result//2, l+1)
        ax.set_title(result[l])
        img = ax.imshow(temp_image)
        ax.imshow(temp_att, cmap='gray', alpha=0.4, extent=img.get_extent())

    plt.tight_layout()
    plt.show()
    
def test_random_image(style="full"):
    """ TBD """    
    rid = np.random.randint(0, len(val_subset_df))
    path = val_subset_df["img_path"][rid]
    
    if style=="full":
        real_caption = val_subset_df["InChI"][rid][:-1]
    else:
        real_caption = val_subset_df["InChI_chem"][rid][:-1]

    result, attention_plot = evaluate(path)
    result = ''.join(result[:-1])
    print (f"\n\tReal Caption       : {real_caption}")
    print (f"\tPrediction Caption   : {result}")
    print(f"\tLevenshtein Distance  : {Levenshtein.distance(real_caption, result)}\n")
    plot_attention(path, result, attention_plot)
```

We define some parameters including the embedding dimensions and number of RNN units and input shape here.

We also create functions to return the two encoder models we previously discussed: **the fixed encoder** and **the trainable encoder**. We are currently using an **Inception ResNetV2** model, but any backbone would work.

```python
# Basic Training Parameters
BATCH_SIZE = 128
SHUFF_BUFF = 512
EMBED_DIM = 256
RNN_UNITS = 512
AUTOTUNE = tf.data.AUTOTUNE

# Reszize Input Images to...
INPUT_SHAPE = (192,192,3)

# Directory and Checkpoint Parameters
CKPT_PATH = "/kaggle/working/model_training"
os.makedirs(CKPT_PATH, exist_ok=True)
if LOAD_MODEL_FROM_FILE:
    print("\n... LOADING MODEL CHECKPOINT FROM PREVIOUSLY SESSION ...")
    !cp {CKPT_PATH_LOAD}/* {CKPT_PATH}
print(f"\n... Checkpoints will be saved to the following directory:\n\t-->{CKPT_PATH}\n")

# Numpy File Loading Parameters
NPY_DIR = "/kaggle/working/numpy_files"
os.makedirs(NPY_DIR, exist_ok=True)
if LOAD_MODEL_FROM_FILE and TRAIN_AS_WE_GO:
    NPY_DIR = NPY_DIR_LOAD
    print("\n... LOADING NPY FILES FROM PREVIOUSLY SESSION ...\n")
    
def get_encoder_fixed(bb, prep_fn, input_shape, freeze_up_to=""):
    """ Return the fixed encoder - modular with tf.keras.applications """   
    if freeze_up_to!="":
        for l in bb.layers:
            l.trainable=False
            if l.name==freeze_up_to: 
                break
    
    inputs=tf.keras.layers.Input(shape=input_shape)
    prep_inputs = prep_fn(tf.cast(inputs, tf.float32))
    outputs = bb(prep_inputs)
    outputs = tf.keras.layers.Reshape((-1, outputs.shape[-1]))(outputs)
    _model = tf.keras.Model(inputs=[inputs], outputs=[outputs], name="fixed_encoder")
    
    return _model

def get_encoder_trainable(input_shape, embedding_dim):
    """ Return the trainable encoder - Single FC Layer """    
    inputs=tf.keras.layers.Input(shape=input_shape)
    outputs = tf.keras.layers.Dense(embedding_dim, activation="relu")(inputs)
    return tf.keras.Model(inputs=[inputs], outputs=[outputs], name="trainable_encoder")

# Cleanup just in case
gc.collect(); tf.keras.backend.clear_session(); gc.collect(); gc.collect();

# Variable backbone information
model_factory, preprocessing_fn = Classifiers.get('resnet18')
backbone = model_factory(INPUT_SHAPE, weights='imagenet', include_top=False)

# Two parts of the encoder model – fixed and trainable
fixed_encoder = get_encoder_fixed(backbone, preprocessing_fn, INPUT_SHAPE, freeze_up_to="stage4_unit2_relu1")
trainable_encoder = get_encoder_trainable(fixed_encoder.output_shape[1:], EMBED_DIM)
```

<h3 style="text-align: font-family: Verdana; font-size: 20px; font-style: normal; font-weight: normal; text-decoration: none; text-transform: none; letter-spacing: 2px; color: navy; background-color: #ffffff;">5.4 DEFINE DECODER MODEL AND PARAMETERS</h3>

The model architecture is inspired by the [**Show, Attend and Tell** Paper](https://arxiv.org/pdf/1502.03044.pdf).

---

**The big takeaways are:**

* Use attention ([**Bahdanau**](https://arxiv.org/abs/1409.0473) in this case)
* Mask zeros in the embedding layer of the decoder
* Hidden state is both an input and an output (let's us learn from the sequence as we traverse it)
* We return the attention weights so we can visualize later

```python
# Maybe like this
BahdanauAttentionTFA = tfa.seq2seq.BahdanauAttention

# Maybe like this though...
class BahdanauAttention(tf.keras.Model):
    """ TBD """
    def __init__(self, units):
        super(BahdanauAttention, self).__init__()
        self.W1 = tf.keras.layers.Dense(units)
        self.W2 = tf.keras.layers.Dense(units)
        self.V = tf.keras.layers.Dense(1)

    def call(self, features, hidden):
        # features(CNN_encoder output) shape == (batch_size, 64, embedding_dim)

        # hidden shape == (batch_size, hidden_size)
        # hidden_with_time_axis shape == (batch_size, 1, hidden_size)
        hidden_with_time_axis = tf.expand_dims(hidden, 1)

        # attention_hidden_layer shape == (batch_size, 64, units)
        attention_hidden_layer = (tf.nn.tanh(self.W1(features) +
                                             self.W2(hidden_with_time_axis)))

        # score shape == (batch_size, 64, 1)
        # This gives you an unnormalized score for each image feature.
        score = self.V(attention_hidden_layer)

        # attention_weights shape == (batch_size, 64, 1)
        attention_weights = tf.nn.softmax(score, axis=1)

        # context_vector shape after sum == (batch_size, hidden_size)
        context_vector = attention_weights * features
        context_vector = tf.reduce_sum(context_vector, axis=1)

        return context_vector, attention_weights
    
def BahdanauAttn(units, style="custom"):
    """ Get Attention Layer/Model """
    if style is "custom":
        return BahdanauAttention(units)
    elif style is "tfa":
        return BahdanauAttentionTFA(units)
    else:
        raise ValueError("`style` argument must be one of ['custom', 'tfa']")
        
def get_decoder(batch_size, n_units, enc_output_shape, vocab_size, embedding_dim):
    """ TBD """
    input_x = tf.keras.layers.Input(shape=(1,))
    input_hidden = tf.keras.layers.Input(shape=(n_units,))
    input_features = tf.keras.layers.Input(shape=enc_output_shape)
    
    context_vector, attention_weights = BahdanauAttn(n_units)(input_features, input_hidden)
    x = tf.keras.layers.Embedding(vocab_size, embedding_dim, mask_zero=True)(input_x)
    x = tf.keras.layers.Concatenate(axis=-1)([tf.expand_dims(context_vector, 1), x])
    
    rnn_output, state = tf.keras.layers.GRU(n_units, return_sequences=True, return_state=True)(x)
    
    # shape == (batch_size, max_length, hidden_size)
    x = tf.keras.layers.Dense(n_units)(rnn_output)
    
    x = tf.keras.layers.Reshape((-1, x.shape[2]))(x)
    x = tf.keras.layers.Dense(vocab_size)(x)
    
    outputs = tf.keras.layers.Reshape((vocab_size,))(x)
    
    return tf.keras.Model(inputs=[input_x, input_hidden, input_features], outputs=[outputs, state, attention_weights])
    
dec_model = get_decoder(BATCH_SIZE, RNN_UNITS, trainable_encoder.output_shape[1:], VOCAB_SIZE, EMBED_DIM)
print("\n... DECODER INPUTS...")
for i, x in enumerate(dec_model.inputs): print(f"\t--> INPUT {i+1} SHAPE  =  {x.shape}")

print("\n\n\n... DECODER MODEL SUMMARY...\n")
dec_model.summary()
```

<h3 style="text-align: font-family: Verdana; font-size: 20px; font-style: normal; font-weight: normal; text-decoration: none; text-transform: none; letter-spacing: 2px; color: navy; background-color: #ffffff;">5.5 DEFINE OPTIMIZER & LOSS</h3>

Since we are defining a custom training loop later in the tutorial, we will need a loss function that can take into consideration the padding token and mask out it's contribution (otherwise all we would do is predict <PAD>).

```python
# Training configuration
EPOCHS = 3
WARMUP_STEPS = 10000
VERBOSE_FREQ = 100
TRAIN_STEPS = int(np.ceil(N_TRAIN/BATCH_SIZE))
TOTAL_STEPS = EPOCHS * TRAIN_STEPS

# Sparse categorical cross entropy for sparse tokens
loss_object = tf.keras.losses.SparseCategoricalCrossentropy(
    from_logits=True,  
    reduction='none',
)

def loss_function(real, pred):
    """ TBD """
    # Find the padding tokens
    mask = tf.math.logical_not(tf.math.equal(real, 0))
    
    # Calculate loss across everything using Sparse-CC
    loss_ = loss_object(real, pred)
    
    # Remove loss contribution from padding tokens
    mask = tf.cast(mask, dtype=loss_.dtype)
    loss_ *= mask
    
    # Take the average across the batches
    return tf.reduce_mean(loss_)

def lrfn(step, warmup_lr_start, lr_start, lr_final, decays, warmup_steps, total_steps):
    # exponential warmup
    if step < warmup_steps:
        warmup_factor = (step / warmup_steps) ** 2
        lr = warmup_lr_start + (lr_start - warmup_lr_start) * warmup_factor
    # staircase decay
    else:
        power = (step - warmup_steps) // ((total_steps - warmup_steps) / (decays + 1))
        decay_factor =  ((lr_start / lr_final) ** (1 / decays)) ** power
        lr = lr_start / decay_factor

    return round(lr, 8)
# plot the learning rate schedule
def plot_lr_schedule(lr_schedule, name):
    plt.figure(figsize=(15,8))
    plt.plot(lr_schedule)
    schedule_info = f'start: {lr_schedule[0]}, max: {max(lr_schedule)}, final: {lr_schedule[-1]}'
    plt.title(f'Step Learning Rate Schedule {name}, {schedule_info}', size=16)
    plt.grid()
    plt.show()

# Learning rate for encoder
LR_SCHEDULE = [
    lrfn(
        step=step, 
        warmup_lr_start=1e-6, 
        lr_start=15e-4, 
        lr_final=35e-5, 
        decays=EPOCHS, 
        warmup_steps=WARMUP_STEPS, 
        total_steps=TOTAL_STEPS) \
    for step in range(TOTAL_STEPS)
]

# custom learning rate scheduler
class LRReduce():
    def __init__(self, optimizer, lr_schedule):
        self.opt = optimizer
        self.lr_schedule = lr_schedule
        
        # assign initial learning rate
        self.lr = lr_schedule[0]
        
        self.opt.learning_rate.assign(self.lr)
        
    def step(self, step):
        self.lr = self.lr_schedule[step]
        
        # assign learning rate to optimizer
        self.opt.learning_rate.assign(self.lr)
        
    def get_counter(self):
        return self.c
    
    def get_lr(self):
        return self.lr

OPTIMIZER = tf.keras.optimizers.Adam()
LR_reduce = LRReduce(OPTIMIZER, LR_SCHEDULE)
plot_lr_schedule(LR_SCHEDULE, 'Encoder')
```

<h3 style="text-align: font-family: Verdana; font-size: 20px; font-style: normal; font-weight: normal; text-decoration: none; text-transform: none; letter-spacing: 2px; color: navy; background-color: #ffffff;">7.2 A SINGLE TRAIN STEP</h3>

```python
@tf.function
def train_step_wo_encoding(img_tensor, target):
    """ This is ONE step in training """
    # Step 1:
    # --> Initialize the loss for the batch to 0
    loss = 0.0

    # Step 2:
    # --> Initialize the hidden state (INPUT 1)
    # --> Initialize the decoder input (previous word) by setting it to
    #     one. Therefore we get a vector of ones with the shape (batch_size, 1) (INPUT 2)
    _hidden = tf.zeros((target.shape[0], RNN_UNITS), tf.float32)
    _dec = tf.ones((target.shape[0], 1), tf.uint8)

    # Step 3-6 - Use gradient to record operations for automatic differentiation.
    with tf.GradientTape() as tape:
        # Step 3:
        # --> Calculate the feature encoding for the batch of images (INPUT 3)
        #       - NOTE: We only do this once per batch
        _features = trainable_encoder(img_tensor, training=True)
        
        # Step 4-6 - Iterate over the tokens in the ground truth InChI caption 
        #              - Tokenized caption is padded and contains end token
        #              - We start from i==1 as we initialized to i==0 (start token)
        for i in tf.range(1, target.shape[1]):
            # Step 4:
            # --> Pass the inputs to the decoder model
            # --> Capture the predicted next token and the updated hidden state
            #       - NOTE: We will do this N times per batch (N is the max length of a tokenized caption)
            predictions, _hidden, _ = dec_model([_dec, _hidden, _features], training=True)
            
            # Step 5:
            # --> Calculate and add the loss for this token's prediction to the running loss total
            #       - NOTE: We will do this N times per batch (N is the max length of a tokenized caption)
            loss += loss_function(target[:, i], predictions)

            # Step 6:
            # --> Teacher Forcing (use the ground truth from a prior time step as input)
            # --> i.e. We update the word to be passed to the decoder next...
            #       - NOTE: We will do this N times per batch (N is the max length of a tokenized caption)
            _dec = tf.expand_dims(target[:, i], 1)
        
    # Step 7:
    # --> Calculate the average loss across the tokenized caption. This is the batchwise loss.
    total_loss = (loss/tf.cast(target.shape[1], tf.float32))

    # Step 8:
    # --> Capture which variables within the models we want to update on (conduct auto-diff on)
    trainable_variables = trainable_encoder.trainable_variables + dec_model.trainable_variables

    # Step 9:
    # --> Conduct automatic differentiation w.r.t. each variable in training_variables
    gradients = tape.gradient(loss, trainable_variables)

    # Step 10:
    # --> Update the optimizer by applying the gradients
    OPTIMIZER.apply_gradients(zip(gradients, trainable_variables))

    # Step 11:
    # --> Return loss values
    return loss, total_loss
```

```python
@tf.function
def train_step_w_encoding(img_tensor, target):
    """ This is ONE step in training """
    # Step 1:
    # --> Initialize the loss for the batch to 0
    loss = 0.0

    # Step 2:
    # --> Initialize the hidden state (INPUT 1)
    # --> Initialize the decoder input (previous word) by setting it to
    #     one. Therefore we get a vector of ones with the shape (batch_size, 1) (INPUT 2)
    _hidden = tf.zeros((target.shape[0], RNN_UNITS), tf.float32)
    _dec = tf.ones((target.shape[0], 1), tf.uint8)

    # Step 3-6 - Use gradient to record operations for automatic differentiation.
    with tf.GradientTape() as tape:
        # Step 3:
        # --> Calculate the feature encoding for the batch of images (INPUT 3)
        #       - NOTE: We only do this once per batch
        img_embedding = fixed_encoder(img_tensor)
        _features = trainable_encoder(img_embedding, training=True)
        
        # Step 4-6 - Iterate over the tokens in the ground truth InChI caption 
        #              - Tokenized caption is padded and contains end token
        #              - We start from i==1 as we initialized to i==0 (start token)
        for i in tf.range(1, target.shape[1]):
            # Step 4:
            # --> Pass the inputs to the decoder model
            # --> Capture the predicted next token and the updated hidden state
            #       - NOTE: We will do this N times per batch (N is the max length of a tokenized caption)
            predictions, _hidden, _ = dec_model([_dec, _hidden, _features], training=True)
            
            # Step 5:
            # --> Calculate and add the loss for this token's prediction to the running loss total
            #       - NOTE: We will do this N times per batch (N is the max length of a tokenized caption)
            loss += loss_function(target[:, i], predictions)

            # Step 6:
            # --> Teacher Forcing (use the ground truth from a prior time step as input)
            # --> i.e. We update the word to be passed to the decoder next...
            #       - NOTE: We will do this N times per batch (N is the max length of a tokenized caption)
            _dec = tf.expand_dims(target[:, i], 1)
        
    # Step 7:
    # --> Calculate the average loss across the tokenized caption. This is the batchwise loss.
    total_loss = (loss/tf.cast(target.shape[1], tf.float32))

    # Step 8:
    # --> Capture which variables within the models we want to update on (conduct auto-diff on)
    trainable_variables = fixed_encoder.trainable_variables + trainable_encoder.trainable_variables + dec_model.trainable_variables

    # Step 9:
    # --> Conduct automatic differentiation w.r.t. each variable in training_variables
    gradients = tape.gradient(loss, trainable_variables)

    # Step 10:
    # --> Update the optimizer by applying the gradients
    OPTIMIZER.apply_gradients(zip(gradients, trainable_variables))

    # Step 11:
    # --> Return loss values
    return loss, total_loss
```

```python
def evaluate(image, from_np=False):
    """ TBD """
    attention_plot = np.zeros((MAX_LEN, fixed_encoder.output_shape[1]))
    hidden = tf.zeros((1, RNN_UNITS), tf.float32)

    if not from_np:
        temp_input = tf.expand_dims(tf_load_image(image, img_size=INPUT_SHAPE[:-1]), 0)
        img_tensor_val = fixed_encoder(temp_input)
    else:
        img_tensor_val=image
    
    features = trainable_encoder(img_tensor_val)
    dec = tf.ones((1, 1), tf.uint8)
    result = [int_2_char_lex[1],]

    for i in range(MAX_LEN-1):
        predictions, hidden, attention_weights = dec_model([dec, hidden, features])
        attention_plot[i] = tf.reshape(attention_weights, (-1, )).numpy()
        predicted_id = tf.random.categorical(predictions, 1)[0][0].numpy()
        result.append(int_2_char_lex[predicted_id])
        if int_2_char_lex[predicted_id] == '<END>':
            return result, attention_plot

        dec = tf.expand_dims([predicted_id], 0)

    attention_plot = attention_plot[:len(result), :]
    return result, attention_plot


def plot_attention(image, result, attention_plot, limit_n=9):
    """ TBD """    
    temp_image = tf_load_image(image)

    fig = plt.figure(figsize=(20, 20))

    len_result = len(result)
    
    if len_result>limit_n:
        print(f"\n...CAPPING ATTENTION MAP VISUALIZATION TO {limit_n} IMAGES AS DISPLAYING {len_result} IMAGES IS UNWIELDY...\n")
        len_result=limit_n
        attention_plot = attention_plot[:limit_n, :]
    sp_w = min(len_result//2, 3)
    sp_h = int(np.ceil(len_result/sp_w))
    for l in range(len_result):
        temp_att = np.resize(attention_plot[l], (8, 8))
        ax = fig.add_subplot(sp_h, sp_w, l+1)
        ax.set_title(result[l])
        img = ax.imshow(temp_image, cmap="gray")
        ax.imshow(temp_att, cmap='gray', alpha=0.4, extent=img.get_extent())

    plt.tight_layout()
    plt.show()
    
def test_random_image(style="full"):
    """ TBD """    
    rid = np.random.randint(0, len(val_subset_df))
    path = val_subset_df["img_path"][rid]
    
    if style=="full":
        real_caption = val_subset_df["InChI"][rid].replace("<END>", "")
    else:
        real_caption = val_subset_df["InChI_chem"][rid].replace("<END>", "")

    result, attention_plot = evaluate(path)
    result_str = ''.join(result[:-1])
    print (f"\n\tReal Caption       : {real_caption}")
    print (f"\tPrediction Caption   : {result_str}")
    print(f"\tLevenshtein Distance  : {Levenshtein.distance(real_caption, result_str)}\n")
    plot_attention(path, result, attention_plot)
    
def lev_on_validation(ds, from_np=False, from_distribution=False):
    lev_distances = []
    for inp, captions in tqdm(ds):
        # Decoder Inputs
        dec = tf.ones((BATCH_SIZE, 1), tf.uint8)
        hidden = tf.zeros((BATCH_SIZE, RNN_UNITS), tf.float32)
        
        if not from_np:
            inp = fixed_encoder(inp)
        features = trainable_encoder(inp)

        # Initialize prediction array
        pred_captions = [[int_2_char_lex[1],] for b_sub in range(BATCH_SIZE)]

        # Capture ground truth captions
        gt_captions = ["".join([int_2_char_lex[c] for c in cap if c not in [0, 2]]) for cap in captions.numpy()]

        for i in range(MAX_LEN-1):
            predictions, hidden, _ = dec_model([dec, hidden, features])
            if from_distribution:
                predicted_ids = [tf.random.categorical(tf.expand_dims(preds,axis=0), 1)[0][0].numpy() for preds in predictions]
            else:
                predicted_ids = [tf.argmax(preds).numpy() for preds in predictions]
            for b_sub in range(BATCH_SIZE):
                pred_captions[b_sub].append(int_2_char_lex[predicted_ids[b_sub]])
            dec = tf.expand_dims(predicted_ids, -1)

        pred_captions = ["".join([c for c in caption_list if c != "<END>"]) for caption_list in pred_captions]
        lev_distances+=[Levenshtein.distance(gt_c, pred_c) for gt_c, pred_c in zip(gt_captions, pred_captions)]
        
    #     return lev_distances, np.array(lev_distances).mean()
    return np.array(lev_distances).mean()
```
