```python
import numpy as np
import pandas as pd

data=pd.read_json('../input/train.json')
```

```python
#making a copy of the main train dataset
train=data.copy()

#creating a dataframe from band_1 and band_2 columns in train set.
band_1_df=pd.DataFrame(i for i in train['band_1'])
band_2_df=pd.DataFrame(i for i in train['band_2'])

#converting `inc_angle` to numeric values.
train['inc_angle'] = pd.to_numeric(train['inc_angle'], errors='coerce')

#concatenate last 3 dataframes
train_df=pd.concat([band_1_df,band_2_df,train['inc_angle']],axis=1)
```

Figureing out if there is any missing values in the dataset.

```python
null_num=train_df.isnull().sum()
train_df.columns[null_num>0]
```

Only `inc_angle` has missing values. Let's use fancyimpute llibrary of python to impute those missing values. 

[fancyimpute](https://github.com/iskandr/fancyimpute) provides a variety of matrix completion and imputation algorithms that are implemented in Python.

I had issues installing this package. I overcome the installation problems by downloading the source, creating a virtual invironment with conda, and navigating to setup.py file and installing the package.

I will be using KNN, which is:
> Nearest neighbor imputations which weights samples using the mean squared difference on features for which two rows both have observed data.

```python
from fancyimpute import KNN

#fancy impute removes column names.
train_df_cols=list(train_df)

train_df = pd.DataFrame(KNN(k=3).complete(train_df))

train_df.columns=train_df_cols
```

```python
#checking if there is no null value anymore.
null_num=train_df.isnull().sum()
train_df.columns[null_num>0]
```
