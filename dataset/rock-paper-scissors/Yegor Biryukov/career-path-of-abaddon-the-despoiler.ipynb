```python
%%writefile loyalist_random_space_marine.py
# obviously
from random import randint as pathetic_action

def loyalist_random_space_marine(obs, conf):
    """ Adeptus Astartes """
    return pathetic_action(0, 2)
```

# Call forth [Abaddon the Despoiler](https://warhammer40k.fandom.com/wiki/Abaddon), Warmaster of Chaos

```python
%%writefile abaddon_the_despoiler.py
from random import randint as pathetic_action

# as pathetic as they can be
pathetic_actions_of_the_loyalist_random_space_marine = [0, 0, 0]

def abaddon_the_despoiler(obs, conf):
    """ Death to the False Emperor! """
    # count pathetic actions of the loyalist random space marine
    if obs["step"] > 0:
        pathetic_actions_of_the_loyalist_random_space_marine[obs["lastOpponentAction"]] += 1
    else:
        pathetic_actions_of_the_loyalist_random_space_marine[pathetic_action(0, 2)] += 1
    
    # using power of the Chaos Gods and python's max function,
    # figure out current most frequent pathetic action of the loyalist random space marine
    most_frequent_pathetic_action = pathetic_actions_of_the_loyalist_random_space_marine.index(max(pathetic_actions_of_the_loyalist_random_space_marine))
    
    # the cunning plan
    ferocious_action_of_the_abaddon_the_despoiler = most_frequent_pathetic_action + 1
    if ferocious_action_of_the_abaddon_the_despoiler > 2:
        ferocious_action_of_the_abaddon_the_despoiler = 0

    # Chaos Gods will be pleased... otherwise poor old Abaddon is in a big trouble
    return ferocious_action_of_the_abaddon_the_despoiler
    
```

# Evaluate Abaddon the Despoiler

```python
from kaggle_environments import evaluate, make, utils
from time import gmtime, strftime


# total amount of games to play
GAMES_AMOUNT = 1000
# amount of games to play before printing current time
PRINT_TIME_INTERVAL = 50
results = []
games_to_print_time = PRINT_TIME_INTERVAL
# total amount of games played
games_played = 0
# how many games most frequent opponent's action in the entire game
# was as well most frequent opponent's action throughout every step of the game
most_frequent_action_same_in_game_and_steps = 0

print("Started at:", strftime("%H:%M:%S", gmtime()))
for i in range(GAMES_AMOUNT):
    # counter of occurrence of each opponent's action
    actions_of_opponent = [0, 0, 0]
    # how many steps each opponent's action was the most frequent of opponent's actions
    most_frequent_opp_actions = [0, 0, 0]
    
    env = make("rps") #, debug=True)
    env.run(["abaddon_the_despoiler.py", "loyalist_random_space_marine.py"])
    my_reward = 0
    opp_reward = 0

    for j in range(len(env.steps)):
        my_action = env.steps[j][0]["action"]
        opp_action = env.steps[j][1]["action"]
        
        if my_action == (opp_action + 1) or (my_action == 0 and opp_action == 2):
            my_reward += 1
            opp_reward -= 1
        elif my_action == (opp_action - 1) or (my_action == 2 and opp_action == 0):
            my_reward -= 1
            opp_reward += 1
            
        actions_of_opponent[opp_action] += 1
        max_index = actions_of_opponent.index(max(actions_of_opponent))
        most_frequent_opp_actions[max_index] += 1
    
    index_max_actions_of_opponent = actions_of_opponent.index(max(actions_of_opponent))
    index_max_most_frequent_opp_actions = most_frequent_opp_actions.index(max(most_frequent_opp_actions))
    if index_max_actions_of_opponent == index_max_most_frequent_opp_actions:
        most_frequent_action_same_in_game_and_steps += 1

    results.append([my_reward, opp_reward])
    
    games_to_print_time -= 1
    if games_to_print_time <= 0:
        games_to_print_time = PRINT_TIME_INTERVAL
        games_played += PRINT_TIME_INTERVAL
        print(f'Games played: {games_played}; current time: {strftime("%H:%M:%S", gmtime())}')
    
print(f'\nOf {GAMES_AMOUNT} games:\n')

print(f'index_max_actions_of_opponent == index_max_most_frequent_opp_actions: \
{most_frequent_action_same_in_game_and_steps}\n')

won = 0
lost = 0
tie = 0
for result in results:
    reward = result[0]
    if reward >= 20:
        won += 1
    elif reward <= -20:
        lost += 1
    else:
        tie += 1

print(f'won: {won}, lost: {lost}, tie: {tie}\n')

# fate of Abaddon the Despoiler is decided here
if won > lost:
    print('Abaddon: "I am the Arch-fiend, the Despoiler of Worlds, and by my hands shall the false Emperor fall."\n')
else:
    print("Abaddon caught COVID-19 at work and became a Chaos Spawn...\n")

# print("results:")
# for result in results:
#     print(result)
```
