From my basic knowlege of space group and analysis of the data, I would like to edit here about my analysis on one specific feature in this competition: lattice angles. 

**From the analysis, lattice angles will be almost redundant. But with some understanding of materials and Density functional calculation, the strain(percentage of change in bond length) introduced by the slight change of angles might add some modifications to the final results - band gaps and formation energy.   **
