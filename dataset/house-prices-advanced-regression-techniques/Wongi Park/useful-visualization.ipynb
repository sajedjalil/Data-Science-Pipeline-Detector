# Learning_curve
- Visualize how using more training cases affects training performance.
    - 학습사례가 많이질수록 학습 성능에 어떤 영향을 주는지 시각화 해줍니다.
- Overfitting can be identified through visualization.
    - Overfitting 여부를 시각화를 통해서 쉽게 알아볼 수 있습니다.

- Reference : https://www.geeksforgeeks.org/using-learning-curves-ml/

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.neighbors import KNeighborsClassifier
from sklearn.datasets import load_digits
from sklearn.model_selection import learning_curve

dataset = load_digits()
X, y = dataset.data, dataset.target

sizes, training_scores, testing_scores = learning_curve(KNeighborsClassifier(), X, y, cv=10, scoring='accuracy', train_sizes=np.linspace(0.01, 1.0, 50))

testing_scores.shape, training_scores.shape, sizes.shape
```

```python
train_score_mean = np.mean(training_scores, axis = 1)
train_score_std = np.std(training_scores, axis = 1)
test_score_mean = np.mean(testing_scores, axis = 1)
test_score_std = np.std(testing_scores, axis = 1)

plt.figure(figsize = (14,8))
plt.fill_between(sizes, train_score_mean -  train_score_std, train_score_mean + train_score_std , alpha = 0.1, color = 'r')
plt.fill_between(sizes, test_score_mean -  test_score_std, test_score_mean + test_score_std , alpha = 0.1, color = 'g')

plt.plot(sizes, train_score_mean, 'o-', color = 'r', label = 'Training')
plt.plot(sizes, test_score_mean, 'o-', color = 'g', label = 'Validation')
plt.legend(loc = 'best')
plt.xlim(0, 1600, 400)
```

# FacetGrid
- It's useful for visualizing data with different categories.
    - 다양한 범주를 가진 데이터를 시각화 하는데 유용합니다.
- Multi-plot grid for plotting conditiional relationships.
    - 조건부 관계로 다중 plot을 형성합니다.

- Reference : https://seaborn.pydata.org/generated/seaborn.FacetGrid.html

```python
import pandas as pd
import seaborn as sns
data = pd.read_csv('../input/titanic/train.csv')

g = sns.FacetGrid(data, col = 'Pclass', hue = 'Sex')
g = g.map(plt.hist, 'Age', alpha = 0.7)
g.add_legend()
```

```python
g = sns.FacetGrid(data, row = 'Embarked', col = 'Sex', hue = 'Survived')
g = g.map(plt.hist, 'Age', alpha = 0.7)
g.add_legend()
```

# HeatMap
- Plot rectangular data as a color-encoded matrix.
    - 직사각형 행렬에 색상을 입힌 것입니다.
- It use with pandas.corr() in order to figure out the correlation between data.
    - 데이터들간의 상관관계를 파악하기 위한 pandas.corr() 과 주로 사용합니다.

- Reference : https://seaborn.pydata.org/generated/seaborn.heatmap.html?highlight=heatmap#seaborn.heatmap

```python
plt.figure(figsize = (14,10))
g = sns.heatmap(data.corr(), cmap = 'BrBG', annot = True)
```

# PairPlot
- Plot pairwise relationships in a dataset.
    - 데이티간의 쌍관계를 차트로 나타내기 위함이다.
- In Heatmap, the correlation appears numerical, But PairPlot appears Visual
    - HeatMap은 상관관계를 수치로 나타내지만, PairPlot은 시각적으로 나타냅니다.

Reference : https://seaborn.pydata.org/generated/seaborn.pairplot.html?highlight=pairplot#seaborn.pairplot

```python
data_1 = pd.read_csv('../input/house-prices-advanced-regression-techniques/train.csv')
cols = ['SalePrice', 'OverallQual', 'GrLivArea', 'GarageCars']
sns.pairplot(data_1[cols])
```

```python
penguins = sns.load_dataset("penguins")
sns.pairplot(penguins , hue = 'species')
```
