# <span style="color:red">4- Feature Selection (Feature Engineering)</span>


```python
corr = data.corr().sort_values(by='SalePrice', ascending=False)


# We selected over here the 50 most influential features of the data
corr1 = corr['SalePrice'].head(50).index
corr1 = corr1[1:]
```

# <span style="color:red">5- Data Rescaling</span>

```python
# extract [X_train, X_test] from the data after data-preprocessing

X_train = data[corr1].iloc[:955,1:-1]
y_train = data.iloc[:955,-2]

X_test  = data[corr1].iloc[955:,1:-1]
y_test = data.iloc[955:,-2]

# print(X_train.shape) 
# print(y_train.shape)

# print(X_test.shape)
# print(y_test.shape)
```

```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler
ss=StandardScaler()
X_train=ss.fit_transform(X_train)
X_test =ss.fit_transform(X_test)


'''
Another solution
'''
# import numpy as np
# for i in range(1,X_train.shape[1]+1):
#     X_train[X_train.columns[i]]=(X_train[X_train.columns[i]]-X_train[X_train.columns[i]].mean())/np.std(X_train[X_train.columns[i]])

# X_train.tail(15)
```


# <span style="color:red">6- model selection</span>

### <span style="color:green">Linear Model</span>

```python
from sklearn.linear_model import Ridge, Lasso, LinearRegression
lr = LinearRegression()
lr.fit(X_train, y_train)
```

### <span style="color:green">Score</span>

```python
print(lr.score(X_train, y_train))
lr.score(X_test, y_test)
```

###  <span style="color:chocolate">Random Forest</span>

```python
from sklearn.ensemble import RandomForestRegressor

rfr = RandomForestRegressor(n_estimators = 200,
                           max_depth = 12)

rfr.fit(X_train,y_train)
```

### <span style="color:chocolate">Score</span>

```python
print(rfr.score(X_train,y_train))
rfr.score(X_test,y_test)
```

### <span style="color:chocolate">Error</span>

```python
y_pred = rfr.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error

error = mean_squared_error(y_test,y_pred)
error
```

### <span style="color:magenta">XGBoost</span>

```python
from xgboost import XGBRegressor
XGBR = XGBRegressor(n_estimators=90,
                    learning_rate=0.06,
                    random_state=44)

XGBR.fit(X_train, y_train, verbose=False)
```

### <span style="color:magenta">Score</span>

```python
print(XGBR.score(X_train,y_train))
XGBR.score(X_test,y_test)
```

### <span style="color:magenta">Error</span>

```python
y_pred = XGBR.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error

error = mean_squared_error(y_test,y_pred)
error
```
