```python
train = pd.read_csv('/kaggle/input/house-prices-advanced-regression-techniques/train.csv')
test = pd.read_csv('/kaggle/input/house-prices-advanced-regression-techniques/test.csv')
```

```python
train.head()
```

```python
train = train[['MSSubClass', 'LotFrontage', 'LotArea', 'OverallQual',
              'OverallCond', 'MasVnrArea', 'BsmtFinSF1', 'BsmtUnfSF', 
               'TotalBsmtSF', '1stFlrSF', '2ndFlrSF', 'GrLivArea',
              'GarageArea', 'OpenPorchSF', 'SalePrice']]

test = test[['MSSubClass', 'LotFrontage', 'LotArea', 'OverallQual',
              'OverallCond', 'MasVnrArea', 'BsmtFinSF1', 'BsmtUnfSF', 
               'TotalBsmtSF', '1stFlrSF', '2ndFlrSF', 'GrLivArea',
              'GarageArea', 'OpenPorchSF',]]
```

```python
train.fillna(0, inplace = True)
test.fillna(0, inplace = True)
```

```python
X = train.drop('SalePrice', axis = 1)
y = train['SalePrice']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
from sklearn.ensemble import RandomForestRegressor
regr = RandomForestRegressor(max_depth=2, random_state=0)
regr.fit(X, y)
```

```python
y_pred = regr.predict(test)
```
