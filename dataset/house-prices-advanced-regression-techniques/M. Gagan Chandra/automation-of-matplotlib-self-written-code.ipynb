```python
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#Graph Styling
plt.style.use('seaborn-darkgrid')
```

```python
def line_plot(df):
    n = int(input('Press 1 for Single line chart\n' 
                  'Press 2 for Multi Line chart:'))
    if n == 1:
        print('select any column from', df.columns)
        x = input("Enter X-axis column name:")
        y = input("Enter Y- axis column name:")
        c = input('Enter a color which is accepted by mathplotlib:')
        plt.plot(df[x],df[y],color = c)
        plt.xlabel(x) 
        plt.ylabel(y)
        plt.title('LINE PLOT')
        plt.show()
    elif n==2:
        print('select any column from', df.columns)
        x = input("Enter column name in x axis:")
        t = int(input("Enter number of columns in Y axis:"))
        y = []
        j = []
        colors=[]
        for i in range(t):
            d = input("Enter column name in y axis:")
            y.append(d)
            c = input('Enter a color which is accepted by mathplotlib:')
            colors.append(c)
        plt.figure(figsize=(12,5)) # Setting the figure size
        ax = plt.axes()
        ax.set_facecolor("darkgrey") # Setting the background color by using Hex code
        for i in range(len(y)):
            plt.plot(df[x],df[y[i]],colors[i],label = y[i])
        plt.xlabel(x) 
        plt.ylabel(y)
        plt.legend(loc='upper right', fontsize='large',bbox_to_anchor=(1.2, 1.0))
        plt.title('LINE PLOT')
        plt.show()
    else:
        print("Incorrect option")
        
        
def bar_plot(df):
    n = int(input('Press 1 for Basic barplot:\n'
                  'Press 2 for horizontal barplot:\n'
                  'Press 3 for vertical stacked barplot:\n'
                  'Press 4 for horizontal stacked barplot:\n'
                  'Press 5 for grouped barplot:\n'))
    if n == 1:
        print('select any column from', df.columns)
        x = input("Enter X-axis column name:")
        y = input("Enter Y-axis column name:")
        plt.figure(figsize=(8,5)) # Setting the figure size
        ax = plt.axes()
        ax.set_facecolor("#ECF0F1")
        c = input('Enter a color which is accepted by mathplotlib(CSS colors are also accepted):')
        plt.bar(df[x],df[y],color = c)
        plt.xlabel(x)
        plt.ylabel(y)
        for x,y in zip(df[x],df[y]):
            plt.text(x, y+0.05, '%f' % y, ha='center' , va= 'bottom')
            plt.plot()
        plt.title('BAR PLOT')
        plt.show()
    elif n==2:
        print('select any column from', df.columns)
        x = input("Enter X-axis column name:")
        y = input("Enter Y-axis column name:")
        plt.figure(figsize=(8,5)) # Setting the figure size
        ax = plt.axes()
        ax.set_facecolor("#ECF0F1")
        c = input('Enter a color which is accepted by mathplotlib(CSS colors are also accepted):')
        plt.barh(df[x],df[y],color =c)
        plt.xlabel(x)
        plt.ylabel(y)
        for index, value in enumerate(df[y]):
            plt.text(value, index, str(value))
            plt.plot()
        plt.show()
    elif n==3:
        print('select any column from', df.columns)
        x = input("Enter column name in x axis:")
        t = int(input("Enter number of columns in Y axis:"))
        h = []
        for i in df[x]:
            h.append(i)
        y = []
        b=[]
        l=[]
        colors = []
        for i in range(t):
            d = input("Enter column name in y axis:")
            c = input('Enter a color which is accepted by mathplotlib(CSS colors are also accepted):')
            y.append(d)
            colors.append(c)
            p = []
            for i in df[d]:
                p.append(i)
            l.append(p)
        k = l[0]
        b.append(k)
        for i in range(1,t-1):
            b.append(k)
            k = k + l[i]
        plt.figure(figsize=(5,7)) # Setting the figure size
        ax = plt.axes()
        ax.set_facecolor("#ECF0F1")
        plt.bar(h,df[y[0]],label = y[0],color = colors[0])
        for i in range(1,len(y)):
            plt.bar(h,df[y[i]],label = y[i],bottom = b[i-1],color = colors[i])
        plt.xlabel(x)
        plt.ylabel(y)
        plt.title ('$Bar $ $ Chart$')
        if len(y) == 2:
            for x,y in zip(h,l[0]):
                plt.text(x, y, '%f' % y, ha='center' , va= 'bottom')
            for i in range(1,t):
                for x,y,z in zip(h,l[i],l[i-1]):
                    plt.text(x, y+z, '%f' % y, ha='center' , va= 'bottom')
        plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))
        plt.show()
    elif n==4:
        print('select any column from', df.columns)
        y = input("Enter column name in y axis:")
        h = []
        l=[]
        for i in df[y]:
            h.append(i)
        t = int(input("Enter number of columns in X axis:"))
        x = []
        b=[]
        colors = []
        for i in range(t):
            d = input("Enter column name in X axis:")
            c = input('Enter a color which is accepted by mathplotlib(CSS colors are also accepted):')
            x.append(d)
            colors.append(c)
            p = []
            for i in df[d]:
                p.append(i)
            l.append(p)
        k = x[0]
        b.append(k)
        for i in range(1,t-1):
            b.append(k)
            k = k + x[i+1]
        plt.figure(figsize=(8,5)) # Setting the figure size
        ax = plt.axes()
        ax.set_facecolor("#ECF0F1")
        plt.barh(df[y],df[x[0]],label = x[0],color = colors[0])
        for i in range(1,len(x)):
            plt.barh(df[y],df[x[i]],label = x[i],left = df[b[i-1]],color = colors[i])
        plt.xlabel(x)
        plt.ylabel(y)
        if len(x) == 2:
            for x,y in zip(l[0],h):
                plt.text(x-10, y, '%f' % x, ha='center' , va= 'bottom')    
            for i in range(1,t):
                for x,y,z in zip(l[i],h,l[i-1]):
                    plt.text(x+z-10, y, '%f' % x, ha='center' , va= 'bottom')
        plt.title ('$Bar $ $ Chart$')
        plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))
        plt.show()
    elif n==5:
        print('select any column from', df.columns)
        # Displaying values at the top of the Grouped Bar Chart using plt.text()
        plt.figure(figsize=(7,9))
        barWidth = 0.25 # set width of bar
        # set height of bar
        x = input("Enter column name in x axis:")
        h = []
        l = []
        for i in df[x]:
            h.append(i)
        t = int(input("Enter number of columns in Y axis:"))
        y = []
        colors = []
        for i in range(t):
            d = input("Enter column name in y axis:")
            c = input('Enter a color which is accepted by mathplotlib(CSS colors are also accepted):')
            y.append(d)
            colors.append(c)
            p = []
            for i in df[d]:
                p.append(i)
            l.append(p)
        pos = []
        pos.append(np.arange(len(df[y[0]])))
        for i in range(1,t):
            m = [x + barWidth for x in pos[i-1]]
            pos.append(m)
        for i in range(t):
            plt.bar(pos[i], df[y[i]], color= colors[i], width=barWidth, label=y[i])

        # Add xticks on the middle of the group bars
        plt.xlabel('X-axis', fontweight='bold')
        plt.ylabel('Y-axis', fontweight='bold')
        plt.xticks([i + barWidth for i in range(len(df[y[0]]))], h)
        
        for i in range(t):
            for x,y in zip(pos[i],l[i]):
                plt.text(x, y, '%f' % y, ha='center' , va= 'bottom')

        plt.title ('$Grouped $  $ Bar $  $ Chart$')

        # Create legend & Show graphic
        plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))
        plt.show()

def scatter_plot(df):
    n = int(input('Press 1 for one scatter plotting:\n'
                  'Press 2 for two scatter plottings:\n'
                  'Press 3 for morethan two scatter plottings'))
    if n == 1:
        print('select any column from', df.columns)
        x = input("Enter X- axis column name:")
        y = input("Enter Y-axis column name:")
        c = input('Enter a color which is accepted by mathplotlib(CSS colors are also accepted):')
        plt.scatter(df[x],df[y] ,color = c)
        plt.xlabel(x , fontsize = 12)
        plt.ylabel(y , fontsize = 12)
        plt.title ('Scatter Graph')
        plt.show()
        
    if n == 2:
        print('select any column from', df.columns)
        x = input("Enter X- axis column name:")
        y = input("Enter Y-axis column name:")
        y1 = input("Enter another Y- axis column name:")
        c1 = input('Enter a color which is accepted by mathplotlib(CSS colors are also accepted):')
        c2 = input('Enter a color which is accepted by mathplotlib(CSS colors are also accepted):')
        plt.scatter(df[x],df[y] ,label = y,marker='o' , color = c1)
        plt.xlabel(x , fontsize = 12)
        plt.ylabel(y , fontsize = 12)
        plt.title ('Scatter Graph')
        plt.show()
        plt.scatter(df[x],df[y1] ,label = y1,marker='v',color =c2)
        plt.xlabel(x , fontsize = 12)
        plt.ylabel(y , fontsize = 12)
        plt.show()
        plt.scatter(df[x],df[y] ,label = y,marker='o' , color = c1)
        plt.scatter(df[x],df[y1] ,label = y1,marker='v',color =c2)
        plt.xlabel(x , fontsize = 12)
        z = y + y1
        plt.ylabel(z, fontsize = 12)
        plt.title ('Scatter Graph')
        plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))
        plt.show()
    if n == 3:
        print('select any column from', df.columns)
        p = int(input("Enter number of plottings u want:"))
        y = []
        colors = []
        markers = ['o','v','s','p','*']
        x = input("Enter X- axis column name:")
        for i in range(p):
            b = input("Enter Y-axis column name:")
            y.append(b)
        colors = []
        for i in range(p):
            c = input('Enter a color which is accepted by mathplotlib(CSS colors are also accepted):')
            colors.append(c)
        for i in range(p):
            plt.scatter(df[x],df[y[i]],color =colors[i])
            plt.xlabel(x , fontsize = 12)
            plt.ylabel(y[i] , fontsize = 12)
            plt.title ('Scatter Graph')
            plt.show()
        for i in range(p):
            plt.scatter(df[x],df[y[i]] ,marker= markers[i],label = y[i],color =colors[i])
            plt.title ('Scatter Graph')
        plt.xlabel(x , fontsize = 12)
        plt.ylabel(y , fontsize = 12)
        plt.title ('Scatter Graph')
        plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))
        plt.show()

        
def histogram_plot(df):
    n = int(input('Press 1 for Basic Histogram Plot\n'
                  'Press 2 to plot more than 1 Histogram Plot\n'
                  'Press 3 to plot Multiple histogram plots in same grid\n'
                  'Press 4 to plot a cumulative distribution histogram plot'))
    if n==1:
        print('select any column from', df.columns)
        x = input('Enter column name:')
        b = int(input('Enter no.of bins:'))
        c = input('Enter a color which is accepted by mathplotlib(CSS colors are also accepted):')
        e = input('Enter Edge color')
        plt.hist(df[x], bins=b, color=c,edgecolor=e)
        plt.gca().set(title='Histogram', ylabel='Frequency')
        plt.show()
    if n == 2:
        print('select any column from', df.columns)
        num = int(input('Enter number of plots u need'))
        for i in range(num):
            x = input('Enter column name:')
            b = int(input('Enter no.of bins:'))
            c = input('Enter a color which is accepted by mathplotlib(CSS colors are also accepted):')
            e = input('Enter Edge color')
            plt.hist(df[x], bins=b, color=c,edgecolor=e)
            plt.gca().set(title='Histogram', ylabel='Frequency')
            plt.show()
    if n==3:
        print('select any column from', df.columns)
        plt.figure(figsize=(8,11))
        num = int(input('Enter number of plots u need'))
        x = []
        b = []
        c = []
        e = []
        for i in range(num):
            r = input('Enter column name:')
            x.append(r)
            bi = int(input('Enter no.of bins:'))
            b.append(bi)
            co = input('Enter a color which is accepted by mathplotlib(CSS colors are also accepted):')
            c.append(co)
            ed = input('Enter Edge color')
            e.append(ed)
        for i in range(num):
            plt.hist(df[x[i]], bins=b[i], color=c[i] , edgecolor =e[i], alpha=0.6)
        plt.show()
    if n==4:
        print('select any column from', df.columns)
        plt.figure(figsize=(10,8))
        x = input('Enter column name:')
        b = int(input('Enter no.of bins:'))
        c = input('Enter a color which is accepted by mathplotlib(CSS colors are also accepted):')
        plt.hist(df[x], bins=b ,color=c , edgecolor="#639a67",cumulative=True)
        plt.gca().set(title='Histogram', ylabel='Frequency')
        plt.show()
        
        

def pie_charts(df):
    n = int(input('Press 1 for Basic pie plot\n'
                  'Press 2 for Multiple pie plot\n'
                  'Press 3 for Exploding a slice of from a pie plot\n'
                  'Press 4 for Exploding more than one slice from a pie plot'))
    if n==1:
        print('select any column from', df.columns)
        plt.figure(figsize=(8,8))
        x = input('Enter column name:')
        l = input('Enter its label')
        colors = ['#8BC34A','#D4E157','#FFB300','#FF7043','blue','green','magenta','red']
        plt.pie (df[x] , labels= df[l] , colors= colors , startangle=45 , shadow='true', autopct='%1.1f%%', explode=[0,0 , 0 , 0])
        plt.show()
    if n==2:
        print('select any column from', df.columns)
        l = []
        t = int(input('Enter how many pieplots u need:'))
        r = int(input('Enter how many rows u need in subplot'))
        c = int(input('Enter how many columns u need in subplot'))
        if r*c != t:
            print("Invalid rows and columns")
        else:
            x = input('Enter column name:')
            colors = ['#8BC34A','#D4E157','#FFB300','#FF7043','blue','green','magenta','red']
            for i in range(t):
                fig = plt.figure(figsize=(20,13))
                la = input('Enter its Label')
                l.append(la)
            for i in range(t):
                plt.subplot(r,c,i+1)
                plt.pie (df[x] , labels= df[l[i]] , colors= colors[:df[x].count()] , startangle=45 ,shadow =True, autopct='%1.1f%%')
            plt.show()
    if n==3:
        print('select any column from', df.columns)
        plt.figure(figsize=(8,8))
        x = input('Enter column name:')
        l = input('Enter its label')
        colors = ['#8BC34A','#D4E157','#FFB300','#FF7043','blue','green','magenta','red']
        e = []
        for i in range(len(df[l])):
            e.append(0)
        f = int(input('Enter which slice u want to explode:'))
        for i in range(1,len(e)+1):
            if i == f:
                e[i-1] = 0.1
        plt.pie (df[x] , labels= df[l] , colors= colors , startangle=45 , shadow='true', autopct='%1.1f%%', explode=e)
        plt.show()
    if n==4:
        print('select any column from', df.columns)
        plt.figure(figsize=(8,8))
        x = input('Enter column name:')
        l = input('Enter its label')
        colors = ['#8BC34A','#D4E157','#FFB300','#FF7043','blue','green','magenta','red']
        e = []
        for i in range(len(df[l])):
            e.append(0)
        f = int(input('Enter how many slices u want to explode:'))
        m = []
        for i in range(f):
            v = int(input('Enter which slice u want to explode:'))
            m.append(v)
        for i in range(1,len(m)+1):
            for j in range(1,len(e)+1):
                if m[i-1] == j:
                    e[j-1] = 0.1
        print(e)
        plt.pie (df[x] , labels= df[l] , colors= colors , startangle=45 , shadow='true', autopct='%1.1f%%', explode=e)
        plt.show()

        
def area_plot(df):
    n = int(input('Press 1 for basic area plot\n'
                  'Press 2 for stacked area plot'))
    if n == 1:
        print('select any column from', df.columns)
        x = input('Enter X-axis column name:')
        y = input('Enter Y-axis column name:')
        plt.figure(figsize=(16,6))
        print('Enter a color you want in between:')
        c = input('Enter a color which is accepted by mathplotlib(CSS colors are also accepted):')
        plt.fill_between( df[x], df[y], color=c, alpha = 0.3) # Changing transperancy using Alpha parameter
        print('Which color u want on edges:')
        c1 = input('Enter a color which is accepted by mathplotlib(CSS colors are also accepted):')
        plt.plot(df[x], df[y], color=c1)
        plt.title("$ Area $  $ chart $" , fontsize = 16)
        plt.xlabel(x , fontsize = 16)
        plt.ylabel(y , fontsize = 16)
        plt.show()
    if n==2:
        print('select any column from', df.columns)
        x = input('Enter X-axis column name:')
        t = int(input('Enter number of Y-axis columns u need:'))
        l= []
        y = []
        color = []
        for i in range(t):
            col = input('Enter Y-axis column name:')
            l.append(col)
            y.append(df[col])
        plt.figure(figsize=(8,6))
        for i in range(t):
            c = input('Enter a color which is accepted by mathplotlib(CSS colors are also accepted):')
            color.append(c)
        a = float(input('Select your transparency value:'))
        plt.stackplot(df[x],y,labels=l,colors = color,alpha = a )
        plt.legend(loc='upper left')
        plt.xlabel(x , fontsize = 16)
        plt.show()

def donut_plots(df):
    n = int(input('Press 1 for Basic donut plot\n'
                  'Press 2 for Multiple donut plot\n'
                  'Press 3 for Exploding a slice of from a donut plot\n'
                  'Press 4 for Exploding more than one slice from a donut plot'))
    if n==1:
        print('select any column from', df.columns)
        plt.figure(figsize=(8,8))
        x = input('Enter column name:')
        l = input('Enter its label')
        colors = ['#8BC34A','#D4E157','#FFB300','#FF7043','blue','green','magenta','red']
        plt.pie (df[x] , labels= df[l] , colors= colors , startangle=45 , shadow='true', autopct='%1.1f%%', explode=[0,0 , 0 , 0])
        my_circle=plt.Circle( (0,0), 0.7, color='white') # Adding circle at the centre
        p=plt.gcf()
        p.gca().add_artist(my_circle)
        plt.show()
    if n==2:
        print('select any column from', df.columns)
        l = []
        t = int(input('Enter how many donut plots u need:'))
        r = int(input('Enter how many rows u need in subplot'))
        c = int(input('Enter how many columns u need i n subplot'))
        if r*c != t:
            print("Invalid rows and columns")
        else:
            x = input('Enter column name:')
            colors = ['#8BC34A','#D4E157','#FFB300','#FF7043','blue','green','magenta','red']
            for i in range(t):
                fig = plt.figure(figsize=(20,13))
                la = input('Enter its Label')
                l.append(la)
            for i in range(t):
                plt.subplot(r,c,i+1)
                plt.pie (df[x] , labels= df[l[i]] , colors= colors[:df[x].count()] , startangle=45 ,shadow =True, autopct='%1.1f%%')
                my_circle=plt.Circle( (0,0), 0.7, color='white') # Adding circle at the centre
                p=plt.gcf()
                p.gca().add_artist(my_circle)
            plt.show()
    if n==3:
        print('select any column from', df.columns)
        plt.figure(figsize=(8,8))
        x = input('Enter column name:')
        l = input('Enter its label')
        colors = ['#8BC34A','#D4E157','#FFB300','#FF7043','blue','green','magenta','red']
        e = []
        for i in range(len(df[l])):
            e.append(0)
        f = int(input('Enter which slice u want to explode:'))
        for i in range(1,len(e)+1):
            if i == f:
                e[i-1] = 0.1
        print(e)
        plt.pie (df[x] , labels= df[l] , colors= colors , startangle=45 , shadow='true', autopct='%1.1f%%', explode=e)
        my_circle=plt.Circle( (0,0), 0.7, color='white') # Adding circle at the centre
        p=plt.gcf()
        p.gca().add_artist(my_circle)
        plt.show()
    if n==4:
        print('select any column from', df.columns)
        plt.figure(figsize=(8,8))
        x = input('Enter column name:')
        l = input('Enter its label')
        colors = ['#8BC34A','#D4E157','#FFB300','#FF7043','blue','green','magenta','red']
        e = []
        for i in range(len(df[l])):
            e.append(0)
        f = int(input('Enter how many slices u want to explode:'))
        m = []
        for i in range(f):
            v = int(input('Enter which slice u want to explode:'))
            m.append(v)
        for i in range(1,len(m)+1):
            for j in range(1,len(e)+1):
                if m[i-1] == j:
                    e[j-1] = 0.1
        print(e)
        plt.pie (df[x] , labels= df[l] , colors= colors , startangle=45 , shadow='true', autopct='%1.1f%%', explode=e)
        my_circle=plt.Circle( (0,0), 0.7, color='white') # Adding circle at the centre
        p=plt.gcf()
        p.gca().add_artist(my_circle)
        plt.show()

def stem_plot(df):
    print('select any column from', df.columns)
    x = input('Enter X-axis column name:')
    y = input('Enter Y-axis column name:')
    plt.figure(figsize=(8,8))
    c = input('Enter a color which is accepted by mathplotlib(CSS colors are also accepted):')
    plt.xlabel(x)
    plt.ylabel(y)
    plt.stem(df[x],df[y],linefmt=c)
    plt.show()

def box_plot(df):
    plt.figure(figsize =(10, 7))
    n = int(input('Press 1 for single column box plot\n'
                  'Press 2 for morethan one  column box plot'))
    if n==1:
        print('select any column from', df.columns)
        x = input('Enter column name:')
        plt.boxplot(df[x])
        plt.ylabel(x)
        plt.show()
    if n==2:
        print('select any column from', df.columns)
        d = []
        x = []
        e = int(input('Enter how many columns u want:'))
        for i in range(e):
            b = input('Enter column name:')
            x.append(b)
            d.append(df[b])
        plt.boxplot(d)
        plt.ylabel(x)
        plt.show()
    
def violin_plot(df):
    plt.figure(figsize =(10, 7))
    n = int(input('Press 1 for single column violin plot\n'
                  'Press 2 for morethan one violon box plot'))
    if n==1:
        print('select any column from', df.columns)
        x = input('Enter column name:')
        plt.violinplot(df[x])
        plt.ylabel(x)
        plt.show()
    if n==2:
        print('select any column from', df.columns)
        d = []
        x = []
        e = int(input('Enter how many columns u want:'))
        for i in range(e):
            b = input('Enter  column name:')
            x.append(b)
            d.append(df[b])
        plt.violinplot(d)
        plt.ylabel(x)
        plt.show()
```
