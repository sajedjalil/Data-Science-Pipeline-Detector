```python
import pandas as pd 
import numpy as np 
df = pd.read_csv("../input/commonlitreadabilityprize/train.csv")
df2 = pd.read_csv("../input/commonlitreadabilityprize/test.csv")
```

```python
texts = df['excerpt'].to_list() + df2['excerpt'].to_list()
y = df['target'].values


from sklearn.feature_extraction.text import TfidfVectorizer

vectorizer = TfidfVectorizer(stop_words='english')
X = vectorizer.fit_transform(texts)

Xtrain = X[:len(df)]
Xtest = X[len(df):]

from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
X_train, X_test, y_train, y_test = train_test_split(Xtrain, y, test_size=0.05, random_state=42)

from sklearn.svm import SVR
from sklearn.ensemble import GradientBoostingRegressor

regr =  GradientBoostingRegressor()
#regr =  SVR()
regr.fit(X_train,y_train)


mean_squared_error(y_train,regr.predict(X_train))

mean_squared_error(y_test,regr.predict(X_test))

import matplotlib.pyplot as plt

plt.scatter(y_test, regr.predict(X_test))


df2['target'] = regr.predict(Xtest)

df2[['id','target']].to_csv ('submission.csv', index = False, header=True)
```
