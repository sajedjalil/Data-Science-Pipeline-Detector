```python
remove_non_alphabets =lambda x: re.sub(r'[^a-zA-Z]',' ',x)
```

```python
tokenize = lambda x: word_tokenize(x)
```

```python
ps = PorterStemmer()
stem = lambda w: [ ps.stem(x) for x in w ]
```

```python
lemmatizer = WordNetLemmatizer()
leammtizer = lambda x: [ lemmatizer.lemmatize(word) for word in x ]
```

#You know that wind and water can shape Earthâ€™s land. But did you know that glaciers can too?

![](https://www.whatihavelearnedteaching.com/wp-content/uploads/2016/02/Processes-that-Shape-the-Earth-2A-Wind-and-Water-Change-Land.jpeg)whatilearnedteaching.com

```python
print('Processing : [=', end='')
df['excerpt'] = df['excerpt'].apply(remove_non_alphabets)
print('=', end='')
df['excerpt'] = df['excerpt'].apply(tokenize) # [ word_tokenize(row) for row in data['MESSAGE']]
print('=', end='')
df['excerpt'] = df['excerpt'].apply(stem)
print('=', end='')
df['excerpt'] = df['excerpt'].apply(leammtizer)
print('=', end='')
df['excerpt'] = df['excerpt'].apply(lambda x: ' '.join(x))
print('] : Completed', end='')
df.head()
```

```python
max_words = 10000
cv = CountVectorizer(max_features=max_words, stop_words='english')
sparse_matrix = cv.fit_transform(df['excerpt']).toarray()
```

```python
x_train, x_test, y_train, y_test = train_test_split(sparse_matrix, np.array(df['standard_error']))
```

#I wanted to use target however it resulted in error when I ran the model and epochs below. 
Therefore I changed to standard error, which is probably another error.

```python
class LogisticRegression(nn.Module):
    def __init__(self):
        super(LogisticRegression, self).__init__()
        self.linear1 = nn.Linear(10000, 100)
        self.linear2 = nn.Linear(100, 10)
        self.linear3 = nn.Linear(10, 2)
        
    def forward(self, x):
        x = F.relu(self.linear1(x))
        x = F.relu(self.linear2(x))
        x = self.linear3(x)
        return x
```

```python
model = LogisticRegression()
```

```python
criterion = nn.CrossEntropyLoss()
optimizer = torch.optim.Adam(params=model.parameters() , lr=0.01)
```

```python
x_train = Variable(torch.from_numpy(x_train)).float()
y_train = Variable(torch.from_numpy(y_train)).long()
```

```python
epochs = 20
model.train()
loss_values = []
for epoch in range(epochs):
    optimizer.zero_grad()
    y_pred = model(x_train)
    loss = criterion(y_pred, y_train)
    loss_values.append(loss.item())
    pred = torch.max(y_pred, 1)[1].eq(y_train).sum()
    acc = pred * 100.0 / len(x_train)
    print('Epoch: {}, Loss: {}, Accuracy: {}%'.format(epoch+1, loss.item(), acc.numpy()))
    loss.backward()
    optimizer.step()
```
