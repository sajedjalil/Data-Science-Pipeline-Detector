```python
#Lobna Mohamed Abd ElAziz
import pandas as pd
import numpy as np
data=pd.read_csv("../input/kaggle-survey-2021/kaggle_survey_2021_responses.csv")
data.head()
```

```python
data.drop(index=data.index[0], axis=0, inplace=True)
data
```

```python
#1
#value_count:function returns a Series containing counts of unique values,, A new index label named level_0 has been generated.
D1=data['Q1'].value_counts().reset_index()
D1=D1.sort_values(by='index')
```

```python
import seaborn as sns
from matplotlib import pyplot as plt
Y=(D1['Q1']/D1['Q1'].sum())*100
myexplode=(0.0,0.1,0.2,0.2,0.1,0.2,0.1,0.1,0.2,0.1,0.2)
plt.style.use("fivethirtyeight")
mylabel=D1['index']
colors={'#9b2226','#ae2012','#bb3e03','#ca6702','#ee9b00','#e9d8a6','#94d2bd','#0a9396','#005f73','#001219','#99d98c'}
plt.pie(Y,labels=mylabel,autopct="%1.1f%%",startangle=15,explode=myexplode,shadow=True,colors=colors)
plt.axis("equal")
plt.title("Age distribution")
plt.legend(mylabel)
#TO maximize graph size
plt.gcf().set_size_inches(15,8)
plt.show()
```

```python
#2
D2=data['Q2'].value_counts().reset_index()
D2=D2.sort_values(by='Q2')
print(D2)
```

```python
x=D2['index']
y=(D2['Q2']/D2['Q2'].sum())*100
plt.style.use("fivethirtyeight")
colors={'Man':'#606c38','Woman':'#283618','Prefer not to say':'#ef233c','Nonbinary':'#dda15e','Prefer to self-describe':'#bc6c25'}
plots=sns.barplot(data=data,y=y,x=x,palette=colors,ci=None)
plt.xlabel("Gender Distribution")
plt.ylabel("% of respondents")
plt.title("Gender distribution")
labels = list(colors.keys())
handles = [plt.Rectangle((0,0),1,1, color=colors[label]) for label in labels]
plt.legend(handles, labels)
plt.gcf().set_size_inches(14,6)
for bar in plots.patches:
    plots.annotate(format(bar.get_height(),'.2f'),
                  (bar.get_x()+bar.get_width()/2,
                  bar.get_height()),ha='center',va='center',
                  size=15,xytext=(0,5),
                  textcoords='offset points')
plt.show()
```

```python
#3
D3=data['Q3'].value_counts().reset_index()
D3=D3.sort_values(by='Q3')
print(D3)
```

```python
#4
import plotly.express as px

#to work more faster we group the columns we will work on it 
df_groupby = data[['Q2','Q3']]
#When we reset the index, the old index is added as a column, and a new sequential index is used
visual_3_grp = df_groupby.groupby(['Q3','Q2']).count().reset_index()
visual_3 = px.choropleth(visual_3_grp , locations="Q3",color='Q2',
                         locationmode='country names',animation_frame='Q2',
                         title = 'Participation based on the Countries & Gender')
#Remove Whitespace Margins From Plotly graph
visual_3.update_layout(margin={"r":0,"t":0,"l":0,"b":0})
visual_3.show()
#del visual_3_grp
 
```
