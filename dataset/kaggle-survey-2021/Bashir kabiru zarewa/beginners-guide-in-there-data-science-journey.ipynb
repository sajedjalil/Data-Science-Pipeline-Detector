### What is Data Science?
<a id = 'What_ds'></a>
We all hear about the word **Data science**, but what does it really means?

**Data science** is a broad field that refers to the collective processes, concepts, theories, tools and technologies that enables the analysis, review and extraction of valuable knowledge and information from raw data. it geared towards helping individuals and organizations make better decisions from consumed, stored and managed data. 
Data science blend various tools. it makes use of algorithms and machine learning principles with the goal to discover hidden patterns from raw data.


### Who is Data Scientist?
<a id = 'whos_DS'></a>
Here i will use the kagglers responses on the question that was asked about **the activities that make up an important part of your role at work** to describe who a data scientist is. Now let's analyse the responses and draw our description.

```python
#annotation fuction
def gen_annotations(annot):
    """
    Generates annotations to insert in the chart
    """
    if annot is None:
        return []
    
    annotations = []
    # Adding labels
    for d in annot:
        annotations.append(dict(xref='paper', x=d['x'], y=d['y'],
                           xanchor='left', yanchor='bottom',
                           text= d['text'],
                           font=dict(size=13,
                           color=d['color']),
                           showarrow=False))
    return annotations

#creating a dictionary to store the multiple columns for all that applys
role_important={}

role_important['Analyze and understand data']=data_2021.Q24_Part_1.value_counts().sum()
role_important['Build infrastructure for analyzing data']=data_2021.Q24_Part_2.value_counts().sum()
role_important['Build prototypes ML']=data_2021.Q24_Part_3.value_counts().sum()
role_important['Build machine learning service']=data_2021.Q24_Part_4.value_counts().sum()
role_important['improve existing ML models']=data_2021.Q24_Part_5.value_counts().sum()
role_important['advances the state of the art of ML']=data_2021.Q24_Part_6.value_counts().sum()
#role_important['None of these activities']=data_scientist.Q24_Part_7.value_counts().sum()

roll1= round(100*pd.DataFrame.from_dict(role_important, orient='index')/len(data_2021),2)
roll1.reset_index(inplace=True)
roll1.rename(columns={"index": "Importance", 0 : "percentage"}, inplace=True)
roll1.sort_values(by='percentage', ascending=False, inplace=True)

# Add title and annotationsit
annot = [{'x': 0.60, 'y': 30, 'text': "<b>49%</b> of the data scientist work is <b>understanding</b><br> and  <b>analysing</b> data.",'color': 'crimson'},
        {'x': 0.60, 'y': 22, 'text': "While <b>51%</b> of the work is building and managing<br> machine learning models</b>",'color': 'lightslategray'}]

#highlight colors
color = ['lightslategray',]*6
color[0] = 'crimson'
color[2] = 'crimson'
fig = go.Figure()
fig.add_trace(go.Bar(x = roll1['Importance'], y = roll1['percentage'], marker_color = color))
fig.update_layout(plot_bgcolor = "white",
                  title = dict(text = "<b>Important part of Data Scientist Work</b>"),
                  xaxis = dict(title = "Important activities"),
                  yaxis = dict(title = "Percentage of Responses"),
                 width=700,
                 height = 500,
                 annotations = gen_annotations(annot))
fig.show()
```

As the analysis shows, the important work of data scientist which is **analysis and building machine learning models** are almost 50/50. So we can describe a **Data scientist** as a person who use raw data to understand, analyse and build machine learning models for business decision.
