#### In DataScience, There are not solved problem that What language is better at DataScience, Python? or R? 
#### A lot of beginners say that, "So, What's the difference with two language?", "So, What should i learn to start DataScience?"
#### This time I will try analysis, who use Python? and who use R? 

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
import matplotlib.font_manager as fm

data = pd.read_csv("../input/kaggle-survey-2021/kaggle_survey_2021_responses.csv")[1:]
font = fm.FontProperties(fname='../input/acmeregular/Acme-Regular.ttf')

only_python = data[(data['Q7_Part_1']=='Python')& (data['Q7_Part_2']!='R')]
only_R = data[(data['Q7_Part_2']=='R')& (data['Q7_Part_1']!='Python')]
both = data[(data['Q7_Part_1']=='Python') & (data['Q7_Part_2']=='R')]

lan_data = {}
for column in data.columns[7:20]:
    if column == 'Q7_Part_1' or column == 'Q7_Part_2':
        lan_data[data[column].unique()[0]] = len(data[~data[column].isna()])
    else:
        lan_data[data[column].unique()[1]] = len(data[~data[column].isna()])
        
del(lan_data['None'])
del(lan_data['Other'])
```

```python
fig, ax = plt.subplots(figsize=(25,10),facecolor="white")
ax.patch.set_facecolor('white')
ax.axis("off")

spec = gridspec.GridSpec(ncols=10, nrows=1, figure=fig)
ax1 = fig.add_subplot(spec[0, :6],facecolor="white")
ax2 = fig.add_subplot(spec[0, 6:],facecolor="white")

ax1.bar(x=list(lan_data.keys()), height= list(lan_data.values()),width=0.6,color='#93BFFF')
ax1.spines['right'].set_visible(False)
ax1.spines['top'].set_visible(False)
ax1.spines['left'].set_visible(False)
ax1.axvspan(xmin = -0.5,xmax= 1.5, fc='#F4F4D6', alpha=0.3)

explode = [0.05, 0.05, 0.05]
ax2.pie([len(only_python),len(both),len(only_R)],explode=explode, shadow=True, startangle=90,colors=['#93BFFF','#F7FAC8','#F0B3AA'])
ax2.axis("off")
ax1.set_xticks("")

for i,name in enumerate(list(lan_data.keys())):
    ax1.text(s=name, x=i, y=-1000, ha='center', font =font, fontsize= 20)
    if i ==0 or i == 1:
        ax1.text(s=f"{round(list(lan_data.values())[i]/ sum(list(lan_data.values()))*100,2)}%", x=i, y=list(lan_data.values())[i]+500, ha='center', font =font, fontsize= 25)
ax1.text(s="<--- I want to focus this two language.",x=2, y=20000,font=font, fontsize=20)
ax1.text(s = "What programming languages do you use on a regular basis?",x=-0.5,y=25000, font=font, fontsize=30)
ax2.text(s="Only using Python",x=0.1,y=-0.2,font=font,color='#F5E9F5', va="center",ha="right",fontsize=30)
ax2.text(s=f"{round(len(only_python)/sum([len(only_python), len(both), len(only_R)])*100,1)}%",x=-0.5,y=-0.4,font=font,color='#F5E9F5', va="center",ha="center",fontsize=45)

ax2.text(s="Using both",x=0.8,y=0.6,font=font,color='#93BFFF', va="center",ha="right",fontsize=30)
ax2.text(s=f"{round(len(both)/sum([len(only_python), len(both), len(only_R)])*100,1)}%",x=0.55,y=0.4,font=font,color='#93BFFF', va="center",ha="center",fontsize=45)

ax2.text(s="Only using R",x=0.5,y=1.2,font=font,color='#5C2340', va="center",ha="right",fontsize=30)
ax2.text(s=f"{round(len(only_R)/sum([len(only_python), len(both), len(only_R)])*100,1)}%",x=0.8,y=1.2,font=font,color='#5C2340', va="center",ha="center",fontsize=45)
ax2.text(s = "Python, R user's rate",x=0,y=1.5, font=font, fontsize=30,va='center',ha='center')

plt.show()
```

#### And Using Python and R user's rate are more than 20%, R user's rate is only 3.3%

```python
a = only_python['Q5'].value_counts().sort_index()
a = pd.DataFrame(a)
python_job = a.reset_index()
python_job['percentage'] = python_job['Q5']/sum(a.values)

b = only_R['Q5'].value_counts().sort_index()
b = pd.DataFrame(b)
R_job = b.reset_index()
R_job['percentage'] = R_job['Q5']/sum(b.values)

job_compare = pd.merge(left=R_job, right=python_job, on='index',how='outer')
job_compare = job_compare.fillna(0)
job_compare['percentage_y'] = job_compare['percentage_y']*-1

job_compare['percentage_x'] = job_compare['percentage_x']*100
job_compare['percentage_y'] = job_compare['percentage_y']*100
job_compare['percentage_x'] = job_compare['percentage_x']+8
job_compare['percentage_y'] = job_compare['percentage_y']-8

plt.figure(figsize=(20,20))
plt.barh(y=job_compare['index'], width=job_compare['percentage_x'],color='#F0B3AA')
plt.barh(y=job_compare['index'], width=job_compare['percentage_y'],color='#93BFFF')
plt.barh(y=job_compare['index'], width=[8]*len(job_compare),color = 'white')
plt.barh(y=job_compare['index'], width=[-8]*len(job_compare),color = 'white')
plt.axis('off')
for i,text in enumerate(list(job_compare['index'])):
    plt.text(s=text, x=0, y=i, font=font, fontsize=20,va='center',ha='center',alpha=0.8)
    
plt.text(s="Python", x = -15, y =15, font = font, fontsize=40,color = '#93BFFF',ha='center', va='center')
plt.text(s="R", x = 15, y =15, font = font, fontsize=40,color = '#F0B3AA',ha='center', va='center')

for i,value in enumerate(list(job_compare['percentage_x'])):
    plt.text(s=f"{round(value-8,1)}%", x=value+1, y=i, font=font, fontsize=20,va='center',ha='left')
    
for i,value in enumerate(list(job_compare['percentage_y'])):
    plt.text(s=f"{round(value+8,1)*-1}%", x=value-1, y=i, font=font, fontsize=20,va='center',ha='right')

plt.text(s="What is your job?",x=0, y=16, font=font, fontsize=35,va='center',ha='center')
plt.show()
```

```python
def split_data1(st):
    tmp = str(st).split(", ")[0]
    return tmp[3:-2]
def split_data2(st):
    tmp = str(st).split(", ")[1]
    if tmp[1:-1] =='':
        return 0
    else:
        return tmp[1:-1]
a = []
for i in ['Q14_Part_1','Q14_Part_2','Q14_Part_3','Q14_Part_4','Q14_Part_5','Q14_Part_6','Q14_Part_7','Q14_Part_8','Q14_Part_9','Q14_Part_10','Q14_Part_11','Q14_OTHER']:
    a.append([*only_python[i].value_counts().items()])
    
a = pd.DataFrame(a)
a.columns = ['s']
a['lib'] = a['s'].apply(split_data1)
a['count'] = a['s'].apply(split_data2)
a = a.drop(['s'],axis=1)

b = []
for i in ['Q14_Part_1','Q14_Part_2','Q14_Part_3','Q14_Part_4','Q14_Part_5','Q14_Part_6','Q14_Part_7','Q14_Part_8','Q14_Part_9','Q14_Part_10','Q14_Part_11','Q14_OTHER']:
    b.append([*only_R[i].value_counts().items()])
    
b = pd.DataFrame(b)
b.columns = ['s']
b['lib'] = b['s'].apply(split_data1)
b['count'] = b['s'].apply(split_data2)
b = b.drop(['s'],axis=1)

vis = pd.merge(right=a, left=b,on='lib', how='outer')
vis['count_x'] = pd.to_numeric(vis['count_x'])
vis['count_y'] = pd.to_numeric(vis['count_y'])

vis['count_x'] = vis['count_x']/sum(vis['count_x'])* 100
vis['count_y'] = vis['count_y']/sum(vis['count_y'])* 100

R_others = vis[(vis['lib']=='D3 js')|(vis['lib']=='Altair')|(vis['lib']=='Bokeh')|(vis['lib']=='Geoplotlib')|(vis['lib']=='Leaflet / Folium')].sum()['count_x'] + vis.loc[11,'count_x']
R_none = vis.loc[10,'count_x']
Python_others = vis[(vis['lib']=='D3 js')|(vis['lib']=='Altair')|(vis['lib']=='Bokeh')|(vis['lib']=='Geoplotlib')|(vis['lib']=='Leaflet / Folium')].sum()['count_y'] + vis.loc[11,'count_y']
Python_none = vis.loc[10,'count_y']

others = pd.DataFrame(['Others', R_others, Python_others]).T
others.columns = vis.columns
none = pd.DataFrame(['None', R_none, Python_none]).T
none.columns = vis.columns

vis = pd.concat([vis.iloc[:5],others,none])
vis = vis.reset_index(drop=True)

vis['count_x'] = vis['count_x'] +10
vis['count_y'] = vis['count_y']*-1 -10

plt.figure(figsize=(20,13))
plt.barh(y=vis['lib'], width=vis['count_x'],color='#F0B3AA')
plt.barh(y=vis['lib'], width=vis['count_y'],color='#93BFFF')
plt.barh(y=vis['lib'], width=[10]*len(vis),color = 'white')
plt.barh(y=vis['lib'], width=[-10]*len(vis),color = 'white')
plt.axis('off')
for i,text in enumerate(list(vis['lib'])):
    plt.text(s=text, x=0, y=i, font=font, fontsize=20,va='center',ha='center',alpha=0.8)
    
plt.text(s="Python", x = -12, y =7, font = font, fontsize=40,color = '#93BFFF',ha='right', va='center')
plt.text(s="R", x = 12, y =7, font = font, fontsize=40,color = '#F0B3AA',ha='left', va='center')

for i,value in enumerate(list(vis['count_x'])):
    plt.text(s=f"{round(value-10,1)}%", x=value+1, y=i, font=font, fontsize=20,va='center',ha='left')
    
for i,value in enumerate(list(vis['count_y'])):
    plt.text(s=f"{round(value+10,1)*-1}%", x=value-1, y=i, font=font, fontsize=20,va='center',ha='right')

plt.text(s="What data visualization libraries or tools do you use on a regular basis?",x=0, y=8, font=font, fontsize=35,va='center',ha='center')
plt.show()
```
