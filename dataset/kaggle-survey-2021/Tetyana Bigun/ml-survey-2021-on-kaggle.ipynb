```python
#import all needed libraries 

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
#open and reading the data

df=pd.read_csv("/kaggle/input/kaggle-survey-2021/kaggle_survey_2021_responses.csv")
df.head(10)
```

```python
list_questions = []                 #list of questions
for i in df.iloc[0]:
    list_questions.append(i)
print(list_questions[:10])
```

```python
df.columns
#df = df.rename(columns=df.iloc[0]) # the questions are now the names of columns
df = df.drop(index = 0,axis = 0)    # delete the double column 
df.head()
```

```python
#nan checking
df.isna().mean()
```

```python
df.describe()
```

```python
#What is your age (# years)?

age_value = df["Q1"].value_counts().reset_index()
age_value.columns = ["age","count"]
#print(age_value)


#Visulisation

plt.style.use("dark_background")  #style

myexplode=(0.15,0.15,0.15,0.15,0.15,0.15,0.15,0.15,0.15,0.15, 0.15)
mypalette = ["#d7e1ee", "#cbd6e4", "#bfcbdb", "#b3bfd1", "#a4a2a8", 
             "#df8879", "#c86558", "#b04238", "#991f17",
             "#146356", "#6b506b"]

fig_age, ax_age = plt.subplots(figsize=(12,8)) #creating figure

age_value.plot(kind="pie", y="count", legend = True, title ="Age of participants", ax=ax_age, 
                   labeldistance=1.1, pctdistance= 0.7, shadow = True, labels=age_value["age"],
                    fontsize = 15, explode=myexplode, colors=mypalette,
                    wedgeprops = {'linewidth':1, 'edgecolor':'black'},
                   autopct='%1.1f%%', startangle = 192,
              ylabel="");

plt.legend(loc = "lower right", fontsize = 12) #location of legend


fig_age.tight_layout()
```

```python
#What is your gender? - Selected Choice

gender_value = df["Q2"].value_counts(normalize=True).reset_index()
gender_value.columns = ["gender","count"]
#print(gender_value)


#Visualisation 

plt.style.use("dark_background")  #style
x_labels = gender_value["gender"] # names of ticklabels 

fig_gender, ax_gender = plt.subplots(figsize=(10,4))
gender_value.plot(kind="barh", legend = False, title ="GENDER", ax=ax_gender, 
                  grid=True, color="#991f17",
                 edgecolor = "black", width = 1.0,linewidth = 0.8, hatch='-|');

ax_gender.set_yticklabels(x_labels,fontsize = 13); 

fig_gender.tight_layout()
```

```python
#In which country do you currently reside?

list_questions[3] #original question Q3

country = df["Q3"].value_counts(normalize=True).reset_index()

country.index=country["index"]                               #country name as index
country = country.sort_values(by = "Q3", ascending = False) #sort countries

top_20_countries = country.head(20)             #top 20 countries



#Visualisation of top 20 countries

plt.style.use("fivethirtyeight")
fig_country, ax_country = plt.subplots(figsize=(10,6))

top_20_countries.plot(kind="barh",legend = False, #title =,
                    ax=ax_country, grid=True,color="#df8879",
                 edgecolor = "black", width = 0.8 ,linewidth = 0.8);

ax_country.tick_params(colors='#146356', which='both')   # ticks color for both axes

plt.title("Top 20 countries of residence",               #title style
          loc = "center", fontstyle='italic', 
          fontweight=10, color = '#146356')

ax_country.set_ylabel(None)
ax_country.set_xlabel("count in %", color = '#146356' )

plt.setp(ax_country.get_yticklabels(), fontsize=14);  # yticklabels bigger 
```

```python
#EDUCATION

list_questions[4] #original question Q4

education_value = df["Q4"].value_counts(normalize=True).reset_index()

education_value['index'].replace({"Some college/university study without earning a bachelor’s degree":"Some university/college",
                     "No formal education past high school":'No formal education'},inplace=True)       #replace some indexes
education_value


#Visualisation 
plt.style.use("grayscale")

fig_education, ax_education = plt.subplots(figsize=(10,6))

education_value.plot(kind="barh",legend = False, title ="Education of participants", ax=ax_education, grid=False,
                 edgecolor = "black", width = 0.7,linewidth = 0.8,color="#CA965C");

ax_education.set_xlabel("count in %", fontsize = 15)
ax_education.set_ylabel(None)

y_labels = education_value["index"] # names of ticklabels 
ax_education.set_yticklabels(y_labels,fontsize = 13); # labels y

plt.setp(ax_education.get_yticklabels(), fontsize=14);  # yticklabels bigger 

```

```python
#Select the title most similar to your current role (or most recent title if retired): - Selected Choic

list_questions[5]   #original question Q5

#Current role

job_value = df["Q5"].value_counts(normalize=True).reset_index()

job_value['index'].replace({"Currently not employed":"Not employed","Machine Learning Engineer":'ML Engineer',"Program/Project Manager":'Project Manager',
                    "DBA/Database Engineer":"Database Engineer","Developer Relations/Advocacy":'Advocacy'},inplace=True)

job_value

#Visualisation 
plt.style.use("grayscale")
fig_jobs, ax_jobs = plt.subplots(figsize=(10,6), dpi= 80)
job_value.plot(kind="barh",legend = False, ax=ax_jobs, grid = False,
                 edgecolor = "black", width = 0.7,linewidth = 0.8,color="#bfcbdb", hatch='--');

ax_jobs.set_xlabel("count", fontsize = 15)
ax_jobs.set_ylabel(None)

plt.title("Current role of participants",               #title style
          loc = "center", fontstyle='oblique', 
          fontweight="bold", color = '#146356')


ax_jobs.tick_params(colors='#146356', which='both')

y_labels = job_value["index"]                # names of ticklabels 
ax_jobs.set_yticklabels(y_labels,fontsize = 13);      # labels y
ax_jobs.set_xlabel("count in %", color = '#146356', fontstyle = "oblique")

plt.setp(ax_jobs.get_yticklabels(), fontsize=14);  # yticklabels bigger


```
