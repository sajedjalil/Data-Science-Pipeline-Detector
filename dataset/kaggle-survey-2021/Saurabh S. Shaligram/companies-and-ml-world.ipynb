```python
headings = df_survey.columns.values.tolist()
headings_description = df_survey.values[0]
unique_fields = [value for value in headings if "_" not in value]
df_survey_without_desc = df_survey.drop(df_survey.index[[0]])

for dict in dict_input:
    r1 = re.compile(dict["column"])

    question_wise_headings = list(filter(r1.match, headings))  # Columns for each question
    # print(question_wise_headings)

    df_topic = df_survey_without_desc[question_wise_headings]
    topic_cnt = df_topic.count()

    topic_cnt = topic_cnt.rename(dict["company_mapping"], axis=1)
    df_topic_cnt = topic_cnt.to_frame()
    # print(df_topic_cnt)
    df_topic_cnt['index'] = df_topic_cnt.index
    df_topic_cnt = df_topic_cnt.rename({0: "Percentage Share"}, axis=1)
    df_topic_cnt_sum = df_topic_cnt.groupby(level=0).sum()  # .rename({0: "count_1"}, axis=1)
    d_percentage = round((df_topic_cnt_sum / df_topic_cnt_sum.sum()) * 100, 2)
    d_percentage_display = d_percentage.copy()
    d_percentage_display['Company Name'] = d_percentage.index
    d_percentage_display = d_percentage_display[["Company Name", "Percentage Share"]]
    
    display(HTML(dict["title"]))
    display(HTML(dict["description"]))

    d_percentage.plot(kind='pie', subplots=True, textprops={'fontsize': 5}, ylabel='')
    plt.legend(loc='upper left', bbox_to_anchor=(1.1, 1.0))
    plt.show()
    
    display(HTML("<br/><h5><u>Data</u><h5>"))
    display(HTML(d_percentage_display.sort_values(by='Percentage Share', ascending=False).to_html(index=False)))

    

    if (dict["column"] == "Q27_A"):
        df_27_A = df_survey_without_desc[["Q5", "Q27_A_Part_11"]].copy()        
        topic_cnt_sum_27_A = df_27_A.groupby(["Q5", "Q27_A_Part_11"]).size()
        
        df_topic_cnt_sum_27_A = topic_cnt_sum_27_A.to_frame()
        df_topic_cnt_sum_27_A['index'] = [v[0] for v in df_topic_cnt_sum_27_A.index if 'None' in v] #df_topic_cnt_sum_27_A.index.levels[0].values
        df_topic_cnt_sum_27_A = df_topic_cnt_sum_27_A.rename({0: "Count"}, axis=1)
        
        lst_percentage_27_A = (df_topic_cnt_sum_27_A["Count"].values / df_topic_cnt_sum_27_A.sum().values[0] * 100)
        df_percentage_display_27_A = pd.DataFrame()
        df_percentage_display_27_A["Profession"] = df_topic_cnt_sum_27_A['index'].values
        df_percentage_display_27_A["Count"] = df_topic_cnt_sum_27_A["Count"].values
        df_percentage_display_27_A["Percentage Share"] = lst_percentage_27_A

        display(HTML("<h3>Availability of infrastructure to Students<h3>"))
        display(HTML("<h5>Following analysis shows that Students have complete access to infrastucture. Zero students in 'None' category." +
                     " Good investment in infrastructure in education sector is really admirable.</h5>")) 
        
        display(HTML(df_percentage_display_27_A.sort_values(by='Percentage Share', ascending=False).to_html(index=False)))
        
        
        df_27_A = df_survey_without_desc[["Q1", "Q27_A_Part_11"]].copy()        
        topic_cnt_sum_27_A = df_27_A.groupby(["Q1", "Q27_A_Part_11"]).size()
        
        df_topic_cnt_sum_27_A = topic_cnt_sum_27_A.to_frame()
        df_topic_cnt_sum_27_A['index'] = [v[0] for v in df_topic_cnt_sum_27_A.index if 'None' in v] #df_topic_cnt_sum_27_A.index.levels[0].values
        df_topic_cnt_sum_27_A = df_topic_cnt_sum_27_A.rename({0: "Count"}, axis=1)
        
        lst_percentage_27_A = (df_topic_cnt_sum_27_A["Count"].values / df_topic_cnt_sum_27_A.sum().values[0] * 100)
        df_percentage_display_27_A = pd.DataFrame()
        df_percentage_display_27_A["Age Group"] = df_topic_cnt_sum_27_A['index'].values
        df_percentage_display_27_A["Count"] = df_topic_cnt_sum_27_A["Count"].values
        df_percentage_display_27_A["Percentage Share"] = lst_percentage_27_A

        display(HTML("<h3>Availability of infrastructure to Youth<h3>"))
        display(HTML("<h5>Following analysis shows that majority youth of age 22 to 35 (approx 49% in None category) donot have access to Cloud Computing Platforms. </h5>"))
        display(HTML(df_percentage_display_27_A.sort_values(by='Percentage Share', ascending=False).to_html(index=False)))
        
    if (dict["column"] == "Q32_A"):
        df_32_A = df_survey_without_desc[["Q5", "Q32_A_Part_20"]].copy()
        topic_cnt_sum_32_A = df_32_A.groupby(["Q5", "Q32_A_Part_20"]).size()
        df_topic_cnt_sum_32_A = topic_cnt_sum_32_A.to_frame()
        df_topic_cnt_sum_32_A['index'] = [v[0] for v in df_topic_cnt_sum_32_A.index if 'None' in v] #df_topic_cnt_sum_32_A.index.levels[0].values
        df_topic_cnt_sum_32_A = df_topic_cnt_sum_32_A.rename({0: "Count"}, axis=1)
        
        lst_percentage_32_A = (df_topic_cnt_sum_32_A["Count"].values / df_topic_cnt_sum_32_A.sum().values[0] * 100)
        df_percentage_display_32_A = pd.DataFrame()
        df_percentage_display_32_A["Profession"] = df_topic_cnt_sum_32_A['index'].values
        df_percentage_display_32_A["Count"] = df_topic_cnt_sum_32_A["Count"].values
        df_percentage_display_32_A["Percentage Share"] = lst_percentage_32_A

        display(HTML("<h3>Availability of Big Data Products to Students<h3>"))
        display(HTML("<h5>Following analysis shows that Students have complete access to Big Data Products. Zero students in 'None' category." +
                     " Good investment in Big Data infrastructure in education sector is really admirable</h5>"))        
        display(HTML(df_percentage_display_32_A.sort_values(by='Percentage Share', ascending=False).to_html(index=False)))
        
        df_32_A = df_survey_without_desc[["Q1", "Q32_A_Part_20"]].copy()
        topic_cnt_sum_32_A = df_32_A.groupby(["Q1", "Q32_A_Part_20"]).size()
        df_topic_cnt_sum_32_A = topic_cnt_sum_32_A.to_frame()
        df_topic_cnt_sum_32_A['index'] = [v[0] for v in df_topic_cnt_sum_32_A.index if 'None' in v] #df_topic_cnt_sum_32_A.index.levels[0].values
        df_topic_cnt_sum_32_A = df_topic_cnt_sum_32_A.rename({0: "Count"}, axis=1)
        
        lst_percentage_32_A = (df_topic_cnt_sum_32_A["Count"].values / df_topic_cnt_sum_32_A.sum().values[0] * 100)
        df_percentage_display_32_A = pd.DataFrame()
        df_percentage_display_32_A["Age Group"] = df_topic_cnt_sum_32_A['index'].values
        df_percentage_display_32_A["Count"] = df_topic_cnt_sum_32_A["Count"].values
        df_percentage_display_32_A["Percentage Share"] = lst_percentage_32_A

        display(HTML("<h3>Availability of Big Data Products to Youth<h3>"))
        display(HTML("<h5>Following analysis shows that majority youth of age 22 to 35 (approx. 49% in None category) do not have access to Big Data Products. </h5>"))
        display(HTML(df_percentage_display_32_A.sort_values(by='Percentage Share', ascending=False).to_html(index=False)))
        
    display(HTML('<hr/>'))
```
