```python
#Creating class to make dataset
class Making_data:
    def __init__(self,col):
        self.col=col
    def making_data(self,df):
        list_mach=[self.col]+([i for i in df.columns])
        data=pd.melt(India_reponses[list_mach],
                      id_vars=[self.col],
                      value_vars=India_reponses[[i for i in df.columns]].columns.tolist())

        return data

    def making_data1(self,list1,list2,list3):
        data=pd.melt(India_reponses[list(set(list1.columns.tolist())-set([ elem for elem in list2.columns.tolist()[1:]]))],
                     id_vars=[self.col],value_vars=list3.columns.tolist())
        return data
    
    def making_data2(self,list4):
        df_1=single_ques[self.col].isin(list4)
        return df_1
    
def making_data3(col,list1,list2,list3):
        data=Making_data(col).making_data1(list1,list2,list3).value_counts().to_frame()
        return data
```
