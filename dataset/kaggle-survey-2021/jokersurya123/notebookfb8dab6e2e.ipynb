```python
# relationship between programming languages and current role.

# creating a lang vs current role data frame. This is done by taking each role and counting no of people using a particular language for all languages
prog_lang_cols = ['Q7_Part_1','Q7_Part_2','Q7_Part_3','Q7_Part_4','Q7_Part_5','Q7_Part_6','Q7_Part_7','Q7_Part_8','Q7_Part_9','Q7_Part_10','Q7_Part_11','Q7_Part_12','Q7_OTHER']
current_roles = list(df['Q5'].value_counts().index)
current_roles = current_roles[:-1]
lang_2d = []
for lang in prog_lang_cols:
    temp = []
    for current_role in current_roles:
#         print(lang)
#         print(current_role)
        if len(df[df['Q5'] == current_role][lang].value_counts()) == 0:
            temp.append(0)
            continue
        temp.append(df[df['Q5'] == current_role][lang].value_counts()[0])
    lang_2d.append(temp)
    
Python = lang_2d[0]
R = lang_2d[1]
SQL = lang_2d[2]
C = lang_2d[3]
Cplusplus = lang_2d[4]
Java = lang_2d[5]
Javascript = lang_2d[6]
Julia = lang_2d[7]
Swift = lang_2d[8]
Bash = lang_2d[9]
MATLAB = lang_2d[10]
No_lang = lang_2d[11]
Other = lang_2d[12]

lang_vs_role_df = pd.DataFrame(list(zip(Python, R, SQL, C, Cplusplus, Java, Javascript, Julia, Swift, Bash, MATLAB, No_lang, Other)),
               columns =['Python', 'R', 'SQL', 'C', 'Cplusplus', 'Java', 'Javascript', 'Julia', 'Swift', 'Bash', 'MATLAB', 'No_lang', 'Other'],  index =current_roles)
lang_vs_role_df
lang_vs_role_df.plot(kind='bar', stacked=True,
        title='Stacked Bar Graph for lang vs role')
plt.legend(bbox_to_anchor = (1.05, 0.6))
plt.show()
```

### Imp Observations
- Seems like over all for all roles python seems to be mostly followed by the community which might indicate for any aspiring individual, it is advised to learn python as the above results are showing that python is a skill which all follow
- Also for Data Scientists, Data Analysts , it seems to be that both SQL and R are also followed widely.
- For Software Engineers, R language is not widely used but it seems java script is used extensively by software engineers.
- Also based on the above graph, it is clear that data scientists who use/know C, Cplus plus are very less in number as most ML algorithms and data analysis cannot be done in C and Cplus plus.
- Machine Learning Engineers seem to be using python and SQL a lot but rarely they use R, which also is interesting as machine learning people mostly spend time on building models but less on statistics hence usage of R is less.
- Also Business Analysts seem to be using programming languages , but not that much which is also evident from the role as business analysts spend much time on gathering requirements , and basically understand customer requirement and co-ordinate work between customers and data scientists/ analysts.
- Also usage of programming languages for statisticians seem to be very very less, they seem to be using mostly python and R but that too very few
