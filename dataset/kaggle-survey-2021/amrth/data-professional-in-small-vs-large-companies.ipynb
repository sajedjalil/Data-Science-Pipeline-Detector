```python
# import library
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
# import dataset
df=pd.read_csv("../input/kaggle-survey-2021/kaggle_survey_2021_responses.csv",low_memory=False)
df=df[1:] #Remove first row as i contains longer text
df.head()
```

```python
# create dummy column to label our company sizes based on Q21 answer
conditions = [
    (df['Q21'] == '0-49 employees'),
    (df['Q21'] == '50-249 employees'),
    (df['Q21'] == '250-999 employees'),
    (df['Q21'] == '1000-9,999 employees'),
    (df['Q21'] == '10,000 or more employees')
    ]

# create a list of the values we want to assign for each condition
company_size = ['XS','S','M','L','XL']

# create a new column and use np.select to assign values to it using our lists as arguments
df['company size'] = np.select(conditions, company_size)
```

```python
# Filter only Data Professional data
# Get indexes of Q24 where the column has value of Data Professional jobdesc
indexNames = df[~(
    (df['Q24_Part_1'] == 'Analyze and understand data to influence product or business decisions') |
    (df['Q24_Part_2'] == 'Build and/or run the data infrastructure that my business uses for storing, analyzing, and operationalizing data')|
    (df['Q24_Part_3'] == 'Build prototypes to explore applying machine learning to new areas') |
    (df['Q24_Part_4'] == 'Build and/or run a machine learning service that operationally improves my product or workflows') |
    (df['Q24_Part_5'] == 'Experimentation and iteration to improve existing ML models') |
    (df['Q24_Part_6'] == 'Do research that advances the state of the art of machine learning')
    )].index
# Delete these row indexes from dataFrame
df.drop(indexNames , inplace=True)
```

Using this rules, we got 12267 respondents data!

# **2. Company Size Distribution**
At first, how's the company size distribution?
From about 12k respondent, around 3.7k of them are working in the Extra Small Company (0-49 employees). Wow! I didn't expect this actually



```python
#to print Company Size Distribution plot
x = df['company size'].value_counts().values #set x values
y = df['company size'].value_counts().index #set y index

sns.set(font_scale = 1) #set font scale
sns.set(rc={'axes.facecolor':'#FFFFFF', 'figure.facecolor':'#FFFFFF'}) #set background of the plot to white 

plt.figure(figsize=(6,6)) #set the plot size
ax = sns.barplot(x=x, y=y, linewidth=2, zorder=5, color='#FFCDD2', 
                 edgecolor='black',order = ['XS','S','M','L','XL']) #set the countplot value

#give label with count value of each bars
for i in ax.patches:
    ax.annotate(format(i.get_width(), '.0f'), 
                   (i.get_width(), i.get_y() + i.get_height() / 2 ), 
                   ha = 'center', va = 'center', 
                   xytext = (20, 0), 
                   textcoords = 'offset points')

ax.set(xlabel='Count', ylabel='Company Size') #set x and y axis label
plt.title('Company Size Distribution',fontsize=15, weight='bold') #set the plot title
plt.show() #show the plot
```

## 3.1 Role Distribution of each Company Sizes
Then, what about the distribution of our role if we look it grouped by company's size? Is Data Scientist and Data Analyst still in the top on all size?

```python
#function to show percentage annotate above the countplot, percentage is grouped by each company's size
def percentage_annotate(ax, font_size, font_position,font_rotation):
    sum_heights = [[p.get_height() for p in bars] for bars in ax.containers]
    for bars in ax.containers:
        for i, p in enumerate(bars):
            total = sum(xgroup[i] for xgroup in sum_heights)
            percentage = f'{(100 * p.get_height() / total) :.1f}%'
            ax.annotate(percentage, (p.get_x() + p.get_width() / 2, p.get_height()), 
                   ha = 'center', va = 'center', 
                   xytext = font_position, 
                   rotation = font_rotation,
                   size = font_size,
                   textcoords = 'offset points')

#set colour palette order for plot with <=7 bars of each groups
colour_palette_short = ["#F994D1", "#F8F994","#F39A6E", "#94F9AE","#94CBF9","#B4450E","#3B60F7"]
#set colour palette order for plot with >7 bars of each groups
colour_palette_long = ["#F994D1", "#F13214", "#F8F994",
                  "#94F9AE","#F39A6E", "#94CBF9",
                  "#B4450E","#ECDDFD","#3B60F7",
                  "#CACACA","#000000","#FFFFFF",
                  "#A60EB5","#178C40","#838584","#FC8005","#4FC4B8"]
```

```python
#to print role distribution in each companies plot
df.rename(columns={'Q5': 'Role'}, inplace=True) #to rename column Q5
plt.figure(figsize=(22,9)) #set plot size

ax = sns.countplot(x="company size", hue="Role", data=df, order = ['XS','S','M','L','XL'],
                   palette=colour_palette_long,edgecolor='black',linewidth=2,zorder=5) #set the countplot value
    
plt.title('Role Distribution',fontsize=15, weight='bold') #set the plot title
ax.set(xlabel='Company Size', ylabel='Count') #set the x and y label of the axis
percentage_annotate(ax,8,(0,20),90) #call function to print %value, grouped by each company's size 
plt.show() #print the plot
```

Our hypothesis is true! Data Scientist and Data Analyst are the top 2 in all company size with similar percentage compared to all roles in each size.
