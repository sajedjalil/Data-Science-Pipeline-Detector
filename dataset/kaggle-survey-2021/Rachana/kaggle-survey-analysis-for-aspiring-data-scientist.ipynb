```python
#lets start by renaming the first six columns:
response_df.columns = response_df.columns.to_series().replace({'Q1': 'age', 'Q2':'gender', 'Q3' : 'nationality', 'Q4':'education', 'Q5':'job_des', 'Q6':'coding_exp', 'Q20': 'industry_employed'})
response_df.head(3)
```

```python
#distribution of AGE in the kaggle survey respondents in %:
age_users = response_df.age.value_counts()
age_users = age_users/response_df.shape[0]*100
age_users.plot.bar()
plt.title('Age Distribution')
plt.ylabel('Users Percentage')
plt.xlabel('Age(yrs)')
```

```python
#gender wise distribution of kaggle respondents:
gender_users = response_df.gender.value_counts()
gender_users = gender_users/response_df.shape[0]*100
```

```python
#majority infact more than 75% respondents are male and only 20% are females:
gender_users.plot.bar()
plt.title('Gender Distribution')
plt.ylabel('% of Users')
plt.xlabel('Gender')
```

```python
# we will drop all gender categories except male and female:
response_new = response_df.copy()
# we will drop 400 rows based on gender questions:
response_new = response_new.drop(response_new[(response_new["gender"] == 'Prefer not to say')].index)
response_new = response_new.drop(response_new[(response_new["gender"] == 'Nonbinary')].index)
response_new = response_new.drop(response_new[(response_new["gender"] == 'Prefer to self-describe')].index)
response_new = response_new.drop(response_new[(response_new["gender"] == 'What is your gender? - Selected Choice')].index)
response_new.shape
```

```python
# top ten countries based on users
nationality_users = response_new.nationality.value_counts().head(10)
nationality_users = nationality_users/response_new.shape[0]*100
nationality_users.plot.bar()
plt.title('Top Ten Countries')
plt.ylabel('% of Users')
plt.xlabel('Country Name')
```

The demographic analysis of the survey response by the kaggle users shows:
