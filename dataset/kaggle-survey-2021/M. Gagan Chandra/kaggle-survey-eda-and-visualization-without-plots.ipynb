# Basic Data Cleaning

#### Here I am focusing on the questions where most of the people answered so droping columns with null values greather than 50% 

```python
df.isnull().sum() > 12987.0
# checking if columns has null values number greather than 12987 and found out that there are some columns so i will delete those columns now
```

```python
n = []
for i in df:
    if df[i].isnull().sum() > 12987.0:
        n.append(i)
```

```python
from IPython.core.display import HTML

def multi_table(table_list):
    ''' Acceps a list of IpyTable objects and returns a table which contains each IpyTable in a cell
    '''
    return HTML(
        '<table><tr style="background-color:white;">' + 
        ''.join(['<td>' + table._repr_html_() + '</td>' for table in table_list]) +
        '</tr></table>')
```
