# Preference of programming languages

Which languages are most popular among kagglers? 

```python
language_preference_columns=original_data.iloc[:,7:20][1:]
language_preference_columns
```

```python
total_participants=original_data.shape[0]-1
language_preference_columns=original_data.iloc[:,7:20][1:]
python_counts=language_preference_columns.iloc[:,0].count()
r_counts=language_preference_columns.iloc[:,1].count()
sql_counts=language_preference_columns.iloc[:,2].count()
c_counts=language_preference_columns.iloc[:,3].count()
c_plus_plus_counts=language_preference_columns.iloc[:,4].count()
java_counts=language_preference_columns.iloc[:,5].count()
Javascript_counts=language_preference_columns.iloc[:,6].count()
Julia_counts=language_preference_columns.iloc[:,7].count()
Swift_counts=language_preference_columns.iloc[:,8].count()
Bash_counts=language_preference_columns.iloc[:,9].count()
matlab_counts=language_preference_columns.iloc[:,10].count()
none_counts=language_preference_columns.iloc[:,11].count()
other_counts=language_preference_columns.iloc[:,12].count()
```

```python
data=[python_counts,r_counts,sql_counts,c_counts,c_plus_plus_counts,java_counts,Javascript_counts,Julia_counts,
                                 Swift_counts,Bash_counts,matlab_counts ,none_counts,other_counts]
data_percentage=[x*100/total_participants for x in data]
columns=['Python','R','SQL','C','C++','Java','Javascript','Julia','Swift','Bash','MATLAB','None','Other']
data_dict={x:y for x,y in zip(columns,data_percentage)}
language_counts=pd.DataFrame(data_dict,index={0})
```

```python
language_counts
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
fig=plt.figure(figsize=(10,6))
ax = sns.barplot(data=language_counts)
ax.set_title("Language preference",fontsize=15)
ax.set_xlabel("Language categories",fontsize=10)
ax.set_ylabel("Percentage/%",fontsize=10)
```

It is clear that Python is the most popular programming language. Over **80 percent** kagglers use Python on a regular basis, which is not a    surprise. SQL is the second, over **40 percent**. This is also one of the most commonly used programming languages for data scientists. Other mainstream languages, such as R, C, C++, Java and Javascript, take up about **20%**. 

```python
language_preference_columns.to_csv('language_preference_table.csv')
```

# Chinese kagglers' preference

As a Chinese, it is natural to compare Chinese kagglers' preference with the whole community's preference. The comparison results are as follows.

## programming language

```python
chinese_frame=original_data.loc[original_data['Q3']=='China']
```

```python
chinese_frame
```

```python
language_preference_columns_c=chinese_frame.iloc[:,7:20]
language_preference_columns_c
```

```python
count_c=[]
for i in range(language_preference_columns_c.shape[-1]):
    count_col=language_preference_columns_c.iloc[:,i].count()
    count_c.append(count_col)
```

```python
data_percentage_c=[x*100/814 for x in count_c]
columns=['Python','R','SQL','C','C++','Java','Javascript','Julia','Swift','Bash','MATLAB','None','Other']
data_dict_c={x:y for x,y in zip(columns,data_percentage_c)}
language_counts_c=pd.DataFrame(data_dict_c,index={'China'})
```

```python
tick_labels=['Python','R','SQL','C','C++','Java','Javascript','Julia','Swift','Bash','MATLAB','None','Other']
bar_width=0.35
x=np.arange(13)
fig=plt.figure(figsize=(10,6))
ax1=fig.add_subplot(111)
ax1.bar(x,data_percentage,bar_width,color='c',align="center",label="World",alpha=0.5)
ax1.bar(x+bar_width,data_percentage_c,bar_width,color='b',align="center",label="China",alpha=0.5)
plt.xticks(x+bar_width/2,tick_labels,family="Times New Roman",weight="normal",size=10)
from matplotlib import ticker
loc=ticker.MultipleLocator(10)
ax1.yaxis.set_major_locator(loc)
for ticklabel in ax1.yaxis.get_ticklabels():
    ticklabel.set_fontsize(10)
ax1.legend(loc='best',fontsize=15)
ax1.set_ylabel("Percentage/%",fontsize=10)
ax1.set_xlabel("Language categories",fontsize=10)
ax1.set_title("Language preference: World vs China",fontsize=15)
```

Compared with the whole community,Chinese kagglers have a greater preference for Python, accounting for **over 90% compared with 84%**.    
Besides, Chinese kagglers **prefer C, C++ and MATLAB** languages while show **less interest in SQL and R languages**. It is also interesting to see Chinese kagglers **pefer Java to Javascript**,which is **contrary** to the taste of the whole kaggle community.
