```python
df = pd.read_csv("/kaggle/input/kaggle-survey-2021/kaggle_survey_2021_responses.csv")
df.head()
```

# Age wise Distribution

```python
grouped_age =  df.groupby(['Q1'],as_index=False).Q2.count()
from matplotlib.pyplot import figure
figure(figsize=(10, 8), dpi=80)
import matplotlib.pyplot as plt
import seaborn as sns
plt.bar(grouped_age['Q1'], grouped_age['Q2'])
```

## Nice to know that, most of us are university students 



# Country wise Distribution

```python
grouped_country =  df.groupby(['Q3'],as_index=False).Q1.count()
from matplotlib.pyplot import figure
figure(figsize=(20, 8), dpi=80)
import matplotlib.pyplot as plt
import seaborn as sns
plt.bar(grouped_country['Q3'], grouped_country['Q1'])
plt.xticks(rotation=90)
plt.show()
```
