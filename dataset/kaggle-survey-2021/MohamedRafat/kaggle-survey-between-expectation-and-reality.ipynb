```python
import numpy as np
import pandas as pd
import seaborn as sn
import matplotlib.pyplot as plt
sn.set_theme(style="darkgrid", color_codes=True)
plt.figure(figsize=(20,20))
sn.set_palette(sn.color_palette("Paired", 15))
```

```python
kaggle_survey = pd.read_csv('../input/kaggle-survey-2021/kaggle_survey_2021_responses.csv')
```

### Remove questions row from survey answers

```python
kaggle_survey.rename({'Time from Start to Finish (seconds)': 'survey_time'}, axis=1, inplace=True)
questions_list = list(kaggle_survey.iloc[0])
kaggle_survey.drop([0], axis=0, inplace=True)
print("Questions are")
for q in questions_list: print(q)
```

### Data Convertion & Preparation

```python
kaggle_survey['survey_time'] = pd.to_numeric(kaggle_survey['survey_time']) / (60 * 60)
kaggle_survey.head()
```

# Gender & Education Level

## Which gender is taking time to answer the survey questions ?

```python
plt.figure(figsize=(12,6))
sn.barplot(data=kaggle_survey, y='Q2', x='survey_time', ci=None)
plt.ylabel("Gender")
plt.xlabel("Survey Time")
plt.title("Spent time by genders.")
sn.despine()
plt.show()
```

#### it is obviously that the people who prefre to self-describe their gender spent less time to answer the survey. (are they answer all questions ?)

#### But, the people who didn't say their gender spent a lot of time.

# What is their education level ?

```python
plt.figure(figsize=(20,7))
sn.barplot(data=kaggle_survey, x='Q2', y='survey_time', hue='Q4', ci=None)
plt.xlabel("Gender")
plt.ylabel("Survey Time")
plt.title("Spent time by genders and their education level")
sn.despine()
plt.show()
```

#### It is logically for the men with bachelor, master and doctoral degree almost took the same time to finish the survey.(the higher education is, the fast answering shoud be.)

#### But, the strange thing is the men with professional doctorate took quite long time the the survey. (are they disteracted during the answering ?)

#### people who prefered not to say their gender specially who took some college/university study without a bachelor's degree spent a lot of time in answering the survey. (are those people take a lot of time to interpret the survey questions & answers more than others ?)


# Participation

# Which gender contributed majorly for answering this survey ?

```python
plt.figure(figsize=(10,5))
sn.countplot(data=kaggle_survey, y='Q2', order=kaggle_survey['Q2'].value_counts().index)
plt.ylabel("Gender")
plt.xlabel("Count")
plt.title("Major gender contribution")
sn.despine()
plt.show()
```

#### We can say that, most of the contributions in this survey are men.(More than 20K)

#### There is a huge difference between the men and women contribution (Less than 5K). (are the women don't the surveys ? or women community in kaggle is too small ?)

# Men Jobs

# Which job most of men do ?

```python
men_survey = kaggle_survey[kaggle_survey['Q2'] == 'Man']

plt.figure(figsize=(10,10))
sn.countplot(data=men_survey, y='Q5', order=men_survey['Q5'].value_counts().index)
plt.ylabel("Men job title")
plt.xlabel("Count")
plt.title("Job titles men like them.")
sn.despine()
plt.show()
```

#### The majority of the men are students. It seems that the students very active with kaggle. (what is the degree that they are studying ?)

#### The statisticians men are a small community in kaggle.

# What is the degree that these men students are studying ?

```python
student_men_survey = kaggle_survey[(kaggle_survey.Q2 == 'Man') & (kaggle_survey.Q5 == 'Student')]

plt.figure(figsize=(10,7))
sn.countplot(data=student_men_survey, y='Q4', order=student_men_survey['Q4'].value_counts().index)
plt.ylabel("Studying Degree")
plt.xlabel("Count")
plt.title("Men studying degrees")
sn.despine()
plt.show()
```

#### Most of the men students are studying bachelor and master degrees although the majority of them are still in the bachelor

# Let's take a look at programming languages.

# Which programming language the men students like to use ?

```python
lang_counts = [student_men_survey["Q7_Part_1"].value_counts()[0], student_men_survey["Q7_Part_2"].value_counts()[0], student_men_survey["Q7_Part_3"].value_counts()[0], student_men_survey["Q7_Part_4"].value_counts()[0], student_men_survey["Q7_Part_5"].value_counts()[0], student_men_survey["Q7_Part_6"].value_counts()[0], student_men_survey["Q7_Part_7"].value_counts()[0], student_men_survey["Q7_Part_10"].value_counts()[0], student_men_survey["Q7_Part_11"].value_counts()[0]]
features = ["Python", "R", "SQL", "C", "C++", "Java", "Javascript","Bash", "MATLAP"]
fig, ax = plt.subplots(1, 1, figsize=(19, 10))
ax.pie(lang_counts, labels=features, autopct='%1.1f%%', shadow=False, explode=[0.01]*9, pctdistance=0.5)
ax.axis('equal')
plt.legend()
plt.show()
```

### It is obviously the men students prefer to use python as a regular programming langauge.

### SQL, C and C++ have a quite good intersts from men students side.

### It is intiuitive to see python got the first place in data analytics but, R has less interest from the students although it can provide good capability for data analytics. (is R a hard language ?)
