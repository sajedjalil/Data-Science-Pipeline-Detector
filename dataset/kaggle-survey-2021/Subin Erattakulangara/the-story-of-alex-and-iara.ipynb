```python
# extract the alex and iara charectors form the dataframe
alex_df = df21[df21['Q2']=='Man']
Iara_df = df21[df21['Q2']=='Woman']

#fucntion for plotting donut graph
def plot_donut(df,colname=None, title='X distribution of Kaggle Users ',colors=''):
    plt.figure(figsize=(15,10))
    explode_values = (0, 0, 0.2, 0, 0, 0, 0.2, 0, 0, 0, 0, 0.2,0.2)
    
    circle = plt.Circle((0,0), 0.7, color='white')
    plt.rcParams['text.color'] = 'black'
    
    plt.pie(df[colname].value_counts().values,
        labels=df[colname].value_counts().index,
        colors=colors,shadow=True,autopct='%.1f%%' )
    
    p = plt.gcf()
    p.gca().add_artist(circle)
    
    plt.title(title, size=20)
    plt.show()

#Plotting the donut graph
plot_donut(df=alex_df, colname='Q4', title='Formal level education distribution of Alex in 2021'
            ,colors=Bamako_12.hex_colors)
plot_donut(df=Iara_df, colname='Q4', title='Formal level education distribution of Iara in 2021'
            ,colors=Bamako_12.hex_colors)
```

```python
#function for the vertical bar graph
def VerticaBarGraph(df,colname=None, title='X distribution of Kaggle Users ',y_label='',shift1=100,shift2=100):
    sns.set_theme(style="darkgrid")

    f, ax = plt.subplots(figsize=(10, 16))


    sns.set_color_codes("pastel")
    sns.barplot(x=df[colname].value_counts().values,
                y=df[colname].value_counts().index,
                label="Percentage of user", color="g",edgecolor='black', linewidth=0.5, saturation = 0.5)

    ax.legend(ncol=2, loc="lower right", frameon=True)
    ax.set( ylabel="",xlabel="User count")
    sns.despine(left=True, bottom=True)
    
    plt.title(title, weight='bold',fontsize=15)
    plt.xlabel("# Users", weight='semibold',fontsize=15)
    plt.ylabel(y_label, weight='semibold',fontsize=15)

    z=df[colname].value_counts().values
    for n, i in enumerate(df[colname].value_counts().index):    
        ax.text(df[colname].value_counts().values[n]+shift1,n,s=f'{round(z[n]/df.shape[0]*100,1)}%,',color='black') 
        ax.text(df[colname].value_counts().values[n]+shift2,n,s=round(z[n],1),color='black')
        
#Plot the bar graph
VerticaBarGraph(df=alex_df, colname='Q5', title='Title distribution of kaggle users',
            shift1 =100, shift2=600, y_label = 'Designation')
VerticaBarGraph(df=Iara_df, colname='Q5', title='Title distribution of kaggle users',
            shift1 =100, shift2=600, y_label = 'Designation')
```

```python
# Read all the column names and extract question 7
arsenel = []
for i in df21.columns:
    if "Q7" in i:
        arsenel.append(i)
arsenel_list = df21[arsenel]

# Create list of arsenel and the parameters
arsenel_list_desc = arsenel_list.describe(include='all').T.sort_values('freq', ascending=False)
arsenel_list_desc['Proportion'] = arsenel_list_desc['freq'] / arsenel_list.shape[0]

# Plotting a bar graph based of the list of arsenel and proportion
plt.figure(figsize=(12,6))
sns.barplot(x='top',y='Proportion',data=arsenel_list_desc, edgecolor='black')
plt.title('List of Arsenel used and proportion among Alex and Iara')
plt.show()
```
