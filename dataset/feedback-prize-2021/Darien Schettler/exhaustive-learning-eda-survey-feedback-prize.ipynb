In order to use **`TPU`**, we use **`TPUClusterResolver`** for the initialization which is necessary to connect to the remote cluster and initialize cloud TPUs. Let's go over two important points


````python
print(f"\n... ACCELERATOR SETUP STARTING ...\n")

# Detect hardware, return appropriate distribution strategy
try:
    # TPU detection. No parameters necessary if TPU_NAME environment variable is set. On Kaggle this is always the case.
    TPU = tf.distribute.cluster_resolver.TPUClusterResolver()  
except ValueError:
    TPU = None

if TPU:
    print(f"\n... RUNNING ON TPU - {TPU.master()}...")
    tf.config.experimental_connect_to_cluster(TPU)
    tf.tpu.experimental.initialize_tpu_system(TPU)
    strategy = tf.distribute.experimental.TPUStrategy(TPU)
else:
    print(f"\n... RUNNING ON CPU/GPU ...")
    # Yield the default distribution strategy in Tensorflow
    #   --> Works on CPU and single GPU.
    strategy = tf.distribute.get_strategy() 

# What Is a Replica?
#    --> A single Cloud TPU device consists of FOUR chips, each of which has TWO TPU cores. 
#    --> Therefore, for efficient utilization of Cloud TPU, a program should make use of each of the EIGHT (4x2) cores. 
#    --> Each replica is essentially a copy of the training graph that is run on each core and 
#        trains a mini-batch containing 1/8th of the overall batch size
N_REPLICAS = strategy.num_replicas_in_sync
    
print(f"... # OF REPLICAS: {N_REPLICAS} ...\n")

print(f"\n... ACCELERATOR SETUP COMPLTED ...\n")
````

---


**XLA** (Accelerated Linear Algebra) is a domain-specific compiler for linear algebra that can accelerate TensorFlow models with potentially no source code changes. **The results are improvements in speed and memory usage**.

<br>

When a TensorFlow program is run, all of the operations are executed individually by the TensorFlow executor. Each TensorFlow operation has a precompiled GPU/TPU kernel implementation that the executor dispatches to.

XLA provides us with an alternative mode of running models: it compiles the TensorFlow graph into a sequence of computation kernels generated specifically for the given model. Because these kernels are unique to the model, they can exploit model-specific information for optimization.<br><br>

<div class="alert alert-block alert-danger" style="margin: 2em; line-height: 1.7em; font-family: Verdana;">
    <b style="font-size: 16px;">üõë &nbsp; WARNING:</b><br><br>- XLA can not currently compile functions where dimensions are not inferrable: that is, if it's not possible to infer the dimensions of all tensors without running the entire computation<br>
</div>

<div class="alert alert-block alert-info" style="margin: 2em; line-height: 1.7em; font-family: Verdana;">
    <b style="font-size: 16px;">üìå &nbsp; NOTE:</b><br><br>- XLA compilation is only applied to code that is compiled into a graph (in <b>TF2</b> that's only a code inside <b><code>tf.function</code></b>).<br>- The <b><code>jit_compile</code></b> API has must-compile semantics, i.e. either the entire function is compiled with XLA, or an <b><code>errors.InvalidArgumentError</code></b> exception is thrown)
</div>

We will collect the essay details at this step. We will need these details for later EDAs
* Information such as unique words, length of essay (character and word count), etc.

````python
def flatten_l_o_l(nested_list):
    """ Flatten a list of lists """
    return [item for sublist in nested_list for item in sublist]

def get_essay_details(df, essay_paths):
    """
    
    This function will take a list of paths to essays and return the
    lengths of the respective essays as well as all unique words contained within
    that essay
    
    this code chunk is heavily inspired from Rob Mulla's code
    https://www.kaggle.com/robikscube/student-writing-competition-twitch
    
    
    Args:
        df (pd.DataFrame): The dataframe to update
        txt_paths (list of strings): The paths to all of the text files to probe
    
    Returns:
        Updated pd.DataFrame
    """
    # Copy to prevent unwanted changes
    _df = df.copy()
    
    char_len_map, word_len_map, unique_word_map = dict(), dict(), dict()
    
    for essay_path in tqdm(essay_paths, total=len(essay_paths)):
        with open(essay_path, "r") as essay:
            
            # Get ID to use for mapping
            _id = essay_path[:-4].rsplit("/", 1)[-1]
            
            # Get words and character information
            data = essay.read()
            words = [x.strip() for x in data.strip().split()]
            c_len = len(data.strip())
            w_len = len(words)
            
            # Update respective id maps
            char_len_map[_id] = c_len
            word_len_map[_id] = w_len
            unique_word_map[_id] = set(words)
            
    _df["essay_c_len"] = _df["id"].map(char_len_map)
    _df["essay_w_len"] = _df["id"].map(word_len_map)
    _df["essay_unique_words"] = _df["id"].map(unique_word_map)
    
    return _df

train_df = get_essay_details(train_df, train_df.txt_path.unique())
train_df
````

````python
# EDA SPECIFIC FUNCTIONS - SOME REWRITING MOSTLY @THEDRCAT AND @ERIKBRUIN
def visualize_2(df, _id=None):
    """ Function to visualize a single discourse with displacy and include gaps
    
    Args:
        df (pd.DataFrame): The dataframe to pull examples from.
        _id (str, optional): The string index used to identify the relevant body text
            to pull the discourse from. If no id is provided than a random one from the
            provided df will be selected.
    
    Returns:
        None; An example will be displayed.
    """
    _df = df.copy()
    
    # Pick a random id if none is passed to the function
    if _id is None: _id = _df.sample(1).id.values[0]
    
    # Create the entity list which is a collection of discourse segments
    # where each discourse segment is described by it's start and end character
    # position as well as the discourse type label for that respective segment.
    entities = []
    for i, row in _df[_df['id']==_id].iterrows():
        
        if (row.discourse_absolute_position==0 and row.discourse_start>0):
            entities.append({
                        'start': 0, 
                         'end': row['discourse_start'],
                         'label': "Gap Text"
                    })
            
        entities.append({
                        'start': row['discourse_start'], 
                         'end': row['discourse_end'], 
                         'label': row['discourse_type']
                    })
        
        if row.chunk_gap_length!=0:
            entities.append({
                        'start': row['discourse_end'], 
                         'end': row['discourse_end']+row.chunk_gap_length, 
                         'label': "Gap Text"
                    })
        elif (row.discourse_relative_position>=1.0 and row.essay_c_len>row.discourse_end):
            entities.append({
                        'start': row['discourse_end'], 
                         'end': row['essay_c_len'],
                         'label': "Gap Text"
                    })
        
    # Open the correct text file
    with open(row["txt_path"], 'r') as file: raw_txt_file = file.read()
        
    # Provide the required document details as a single dictionary
    #    - "text": The Raw Text File
    #    - "ents": A list of dictionaries describing the relevant discourse chunks
    #    - "title": The title to ascribe to the visualization
    doc_details = {
        "text": raw_txt_file,
        "ents": entities,
        "title": f"ID: {_id}"
    }
    
    # A mapping of options that will allow us to customize the rendering that displacy will
    # generate. We leverage our previously defined colour map and class list.
    options = {"ents": DISCOURSE_CLASSES+["Gap Text",], "colors": d2c_map}
    
    # Displacy's render function that we control with specific arguments
    displacy.render(docs=doc_details, style="ent", options=options, manual=True, jupyter=True)

# Create a dataframe with information about the gaps in the essays and display
gap_df = train_df.groupby("id").agg({'essay_c_len': 'first', 'chunk_gap_length': 'sum', 'discourse_end':'last'})
gap_df["gap_at_end"] = gap_df.apply(lambda row: 0 if row.discourse_end>=row.essay_c_len else row.essay_c_len-row.discourse_end, axis=1)
gap_df["total_unclassified"] = gap_df["gap_at_end"]+gap_df["chunk_gap_length"]
gap_df["percent_unclassified"] = (100*gap_df["total_unclassified"]/gap_df[["essay_c_len", "discourse_end"]].max(axis=1)).round(3)
gap_df = gap_df.sort_values(by="percent_unclassified", ascending=False)

print("\n... INFORMATION ABOUT THE GAPS (NON-CLASSIFIED TEXT) IN THE STUDENT ESSAYS ...\n\n")
display(gap_df.head(10)); print("\n\n")

visualize_2(train_df, "C278EDC82048"); print("\n\n")
visualize_2(train_df, "D753B3C6BDDA"); print("\n\n")
visualize_2(train_df, "129497C3E0FC"); print("\n\n")
visualize_2(train_df, "BBE0D2A103B3"); print("\n\n")
visualize_2(train_df); print("\n\n")
````

[**Erik Bruin (erikbruin)**](https://www.kaggle.com/erikbruin)

> <i>"Making n_grams for each discourse type. [...] I wanted to make a function to compose Top-10 n_grams per discount type by using **`CountVectorizer()`**. This function should also work for the single words (just run it with **`n_grams=1`**)."</i>



````python
def get_n_grams(df, n_items, top_n=10):
    """
    
    Retrieve the most common n-grams for each respective discourse type
    
    Original Function:
        Created By     : Erik Bruin (https://www.kaggle.com/erikbruin)
        Created Within : NLP On Student Writing EDA (https://www.kaggle.com/erikbruin/nlp-on-student-writing-eda)
    
    I will simply be providing some extra annotation to help my understanding.
    
    
    """
    df_words = pd.DataFrame()
    for dt in tqdm(DISCOURSE_CLASSES):
        # Get the relevant texts and create our count vectorizer
        texts = df[df.discourse_type==dt]["discourse_text"].tolist()
        count_vec = CountVectorizer(
            lowercase=True, stop_words='english', ngram_range=(n_items, n_items)
        ).fit(texts)
        
        # Use the count vectorizer to transform the text
        bag_of_words = count_vec.transform(texts)
        
        # Get statistics and relationships from bag of words
        sum_of_words = bag_of_words.sum(axis=0)
        words_freq = [(word, sum_of_words[0, idx]) for word, idx in count_vec.vocabulary_.items()]
        
        cvec_df = pd.DataFrame.from_records(words_freq, columns= ['words', 'counts']).sort_values(by="counts", ascending=False)
        cvec_df.insert(0, "discourse_type", dt)
        cvec_df = cvec_df.iloc[:top_n,:]
        
        df_words = df_words.append(cvec_df)
    return df_words

N_GRAMS_TO_GO_THROUGH=5
TOP_N = 3
for i in range(N_GRAMS_TO_GO_THROUGH):
    train_ngram_df = get_n_grams(train_df, n_items=i, top_n=TOP_N)
    display(train_ngram_df.head(N_GRAMS_TO_GO_THROUGH*TOP_N))
````

> <i>"The metric is a version of micro f1 where matches are made between the predicted text and the ground truth when the overlap is >= 0.5. The metric page states that this happens in the process:</i>

> * <i>We will purposefully make the label not overlap and show how it will be a false positive.</i>
> 
> <br><i>[...] Note `@cdeotte` noted in the comments. The scoring metric for this competition is **macro_f1** and **`thescore_feedback_comp`** function now computes the **macro f1 score to follow the evaluation page**."</i>

**I will go through the functions and double check my understanding and potentially modify or add details/things as I see fit... all credit goes to Rob. I will also share a snippet from a <a href="https://datascience.stackexchange.com/questions/15989/micro-average-vs-macro-average-performance-in-a-multiclass-classification-settin" style="color: darkred; fontweight: bold;">StackExchange Question/Answer</a> that explains a bit more about Macro v. Micro scores...**

> <i style="color: darkred;">"A macro-average will compute the metric independently for each class and then take the average (hence treating all classes equally), whereas a micro-average will aggregate the contributions of all classes to compute the average metric. In a multi-class classification setup, micro-average is preferable if you suspect there might be class imbalance (i.e you may have many more examples of one class than of other classes)."</i>

````python
def calc_overlap(row):
    """
    
    Calculates the overlap between prediction and
    ground truth and overlap percentages used for determining
    true positives.
    
    Args:
        row (pd.Series): A single row from a dataframe containing
            both the prediction and ground truth string
        
    Returns:
        A list containing the two overlap lengths
    
    """
    
    # Get unique words in the prediction string and ground truth string
    set_pred, set_gt = set(row.predictionstring_pred.split(" ")), set(row.predictionstring_gt.split(" "))
    
    # Length of each set as well as the intersection length
    len_gt, len_pred = len(set_gt), len(set_pred)
    inter = len(set_gt.intersection(set_pred))
    overlap_1, overlap_2 = inter/len_gt, inter/len_pred
    
    return [overlap_1, overlap_2]


def score_feedback_comp_micro(pred_df, gt_df):
    """
    A function that scores for the kaggle Student Writing Competition

    Uses the steps in the evaluation page here:
        https://www.kaggle.com/c/feedback-prize-2021/overview/evaluation
        
    Args:
        pred_df (pd.DataFrame): Dataframe containing prediction values
        gt_df (pd.DataFrame): Dataframe containing ground truth values
        
    Returns:
        The value of the micro F1 score comparing the two dataframes
    """
    
    # Reduce pandas dataframe to the relevant columns, reset the indices and prevent shallow copying
    gt_df = (gt_df[["id", "discourse_type", "predictionstring"]].reset_index(drop=True).copy())
    pred_df = pred_df[["id", "class", "predictionstring"]].reset_index(drop=True).copy()
    
    # Create a column to hold the original index
    gt_df["gt_id"], pred_df["pred_id"] = gt_df.index, pred_df.index

    ######################################################################
    ######################################################################
    ############       MAIN STEPS TO CALCULATE MICRO F1       ############
    ######################################################################
    ######################################################################
    
    ######################################################################
    # STEP 1. Compare Ground Truth w/ Predicted Value For A Given Class  #
    ######################################################################
    
    # Simple outer merge to bring together the two dataframes
    joined = pred_df.merge(
        gt_df,
        left_on=["id", "class"],
        right_on=["id", "discourse_type"],
        how="outer",
        suffixes=("_pred", "_gt"),
    )
    
    # Fill the nan values with a blank to prevent errors
    joined["predictionstring_gt"] = joined["predictionstring_gt"].fillna(" ")
    joined["predictionstring_pred"] = joined["predictionstring_pred"].fillna(" ")

    # Utilize our calc overlap function to determine the respective overlap
    # values for each discourse chunk (row) in the pred/gt dataframes
    joined["overlaps"] = joined.apply(calc_overlap, axis=1)

    ######################################################################
    #   STEP 2. Conditional Handling For Overlap and Multiple Matches    #
    ######################################################################
    #      - If the overlap between the ground truth and prediction is   # 
    #        >= 0.5, and the overlap between the prediction and the      #
    #        ground truth >= 0.5, the prediction is a match and          #
    #        considered a true positive.                                 #
    #                                                                    #
    #      - If multiple matches exist, the match with the highest       #
    #        pair of overlaps is taken.                                  #
    ######################################################################
    
    # Get the percentage overlap and save it to two new columns (respectively)
    joined["overlap1"] = joined["overlaps"].apply(lambda x: eval(str(x))[0])
    joined["overlap2"] = joined["overlaps"].apply(lambda x: eval(str(x))[1])
    
    # Look for potential true positives by identifying rows where
    # the overlap value in both overlap columns is greater (or equal) to 50%
    # We also capture the maximum overlap between the two respective overlap columns 
    # and save that to a new column
    joined["potential_TP"] = (joined["overlap1"] >= 0.5) & (joined["overlap2"] >= 0.5)
    joined["max_overlap"] = joined[["overlap1", "overlap2"]].max(axis=1)
    
    # Now we capture the ids for which we have succesfully predicted with a
    # high enough overlap. If more than one prediction we take the `.first()` one.
    tp_pred_ids = (
        joined.query("potential_TP")
        .sort_values("max_overlap", ascending=False)
        .groupby(["id", "predictionstring_gt"])
        .first()["pred_id"]
        .values
    )

    ######################################################################
    #         STEP 3. Handle False Positive and False Negatives          #
    ######################################################################
    #      - Any unmatched ground truths are false negatives             # 
    #      - Any unmatched predictions are false positives               #
    ######################################################################

    # Assume all submitted predictions which are not TPs are FP
    # i.e. predicted incorrectly
    fp_pred_ids = [p for p in joined["pred_id"].unique() if p not in tp_pred_ids]

    # All remaining unmatched ground truth segments are now classified as FN
    # i.e. failed to predict
    matched_gt_ids = joined.query("potential_TP")["gt_id"].unique()
    unmatched_gt_ids = [c for c in joined["gt_id"].unique() if c not in matched_gt_ids]

    # Get the numbers found for each type
    TP = len(tp_pred_ids)
    FP = len(fp_pred_ids)
    FN = len(unmatched_gt_ids)
    
    
    ######################################################################
    #                     STEP 4. Calculate MicroF1                      #
    ######################################################################
    
    # Calculate the F1 score (micro)
    my_f1_score = TP / (TP + 0.5 * (FP + FN))
    
    ######################################################################
    ######################################################################
    ####################### MAIN STEPS FINISHED ##########################
    ######################################################################
    ######################################################################
    
    return my_f1_score


def score_feedback_comp(pred_df, gt_df, return_class_scores=False):
    """ Function to calculate the `score` used in this Feedback Competition 
    
    Args:
        pred_df (pd.DataFrame): Dataframe containing prediction values
        gt_df (pd.DataFrame): Dataframe containing ground truth values
        return_class_scores (bool, optional): Whether or not to 
            return the individual classwise scores
    
    Returns:
        Either returns only the Macro F1 Score or returns both
        the individual Classwise Macro F1 Scores as well as the Overall
        Macro F1 Score
        
    """
    
    # Initialize a Classwise Scoring Dictionary
    class_scores = {}
    
    # Grab only the part of the dataframe that is relevant to calculating the score
    pred_df = pred_df[["id", "class", "predictionstring"]].reset_index(drop=True).copy()
    
    # Iterate over a given discourse type and the respective ground truth df subset 
    #    - tqdm for visual progressbar
    for discourse_type, gt_subset in tqdm(gt_df.groupby("discourse_type"), total=len(gt_df.groupby("discourse_type"))):
        # Get the pred subset to compare to the ground truth 
        # subset for the respective discourse type
        pred_subset = (
            pred_df.loc[pred_df["class"] == discourse_type]
            .reset_index(drop=True)
            .copy()
        )
        
        # Calculate the classwise score (Micro F1) and store in dictionary
        class_score = score_feedback_comp_micro(pred_subset, gt_subset)
        class_scores[discourse_type] = class_score
        
    # To calculate Macro we simply average the classwise Micro scores.
    f1 = np.mean([v for v in class_scores.values()])
    
    # Only return the classwise information if argument is passed
    if return_class_scores:
        return f1, class_scores
    else:
        return f1
    
gt_df = train_df.copy()
pred_df = gt_df.copy()
pred_df = pred_df.rename(columns={"discourse_type": "class"})

microf1_score = score_feedback_comp_micro(pred_df, gt_df)
macrof1_score, class_scores = score_feedback_comp(
    pred_df, gt_df, return_class_scores=True
)

print(
    f"Using the ground truth to predict the micro f1 score is {microf1_score:0.4f} and macro f1 score {macrof1_score:0.4f}"
)
print("The individual class scores are:")
print(class_scores)
````

> <i>"<b>The Secret of Convert is Use Char Array</b><br>In this competition, **Kaggle provides <span style="color: red;">LABELS</span> as <span style="color: red;">CHARACTERS</span> (`discourse_start`, `discourse_end`) AND <span style="color: red;">WORDS</span> (`predictionstring`).** Let's imagine that the target (**`discourse_text`**) is **`"my name is Chris Deotte."`**. Therefore we need to convert these labels to token labels. **The easiest way to convert to tokens or from tokens is to use an intermediate character array.**"</i>

![cdeotte token image 2](https://raw.githubusercontent.com/cdeotte/Kaggle_Images/main/Dec-2021/token2.png)


````python
MAX_LEN = 10
demo_d_text = "my name is Chris Deotte."
demo_d_type = "Claim"
demo_d_start = 7
demo_d_end = 31
demo_pred_str = "1 2 3 4 5"

print(f"\n... DEMO STRING ...\n\t--> '{demo_d_text}'\n")

'''
From @cdeotte - reformatted for code comment to improve inline readability

    - Let's say that we know our target's discourse_start and discourse_end 
      and we want to create tokens with targets. 
    - First create an array of what we want, i.e token array. 
         - In this example, 0 will designate no target
         - In this example 1 will be a positive target
    - We can then replace 1's with our target class numbers
    - Next we create a character array and place Kaggle's target into it
    - Next, iterate over the token array and grab the respective 
      target from character array!
    - Now token_array has a 1 where there exists a target and 0 otherwise. 
    - We can train our model with this MAX_LEN sized token array.
'''

# STEP 1 - CREATE DESTINATION TOKEN ARRAY
demo_token_array = np.zeros(MAX_LEN)

# STEP 2 - PUT TARGET INTO ORIGIN CHARACTER ARRAY
demo_char_array = np.zeros((len(demo_d_text)))
demo_char_array[demo_d_start:demo_d_end] = 1
print(f"\n... DEMO CHARACTER ARRAY ...\n\t--> {demo_char_array}\n")

# STEP 3 - FILL TOKEN ARRAY FROM CHARACTER ARRAY
demo_encoding = tokenizer(demo_d_text, max_length=MAX_LEN, padding='max_length', 
                     truncation=True, return_offsets_mapping=True)
print(f"\n... DEMO ENCODING ...\n\t--> {demo_encoding}\n")

demo_map_token_to_char = demo_encoding['offset_mapping']
print(f"\n... DEMO TOKEN TO CHAR MAP ...\n\t--> {demo_map_token_to_char}\n")

for k in range(MAX_LEN):
    demo_token_array[k] = demo_char_array[demo_map_token_to_char[k][0]]
print(f"\n... DEMO TOKEN ARRAY ...\n\t--> {demo_token_array}\n")
````

<br>

[**Chris Deotte (cdeotte)**](https://www.kaggle.com/cdeotte)

> <i>"<b>Inference - Convert Tokens to Words</b><br> After inference, our predictions are in a token array. We want to convert them to a word array to submit to LB. First create an array of what you want, i.e. word array. We now have a word_array which has 1's where the target is and 0's otherwise. We then generate the `predictionstring`. That's it! Using an intermediate character array makes converting from tokens to words and words to tokens easy! The only missing code is making the map_word_to_char. We [show that below too]"</i>



````python
# For testing purposes...
class DumbModel:
    def __init__(self,):
        pass
    def predict(self, text):
        return list(np.round(np.random.random(len(text))))
    
def create_word2char_map(text, blank_chars=(' ','\n','\xa0','\x85',)):
    """ Function to create word to char mapping
    
    From @cdeotte - reformatted as a function to improve readability/funcitonality
    
    Args:
        text (str): String to create the mapping from
        blank_chars (tuple of strings): The chars to consider as blank space
        
    Returns:
        The prediction formatted correctly as a word array
    
    """
    
    # Initialize
    word_start = True
    word2char_map = []
    
    # Loop to create mapping
    for i, char in enumerate(text):
        if (char not in blank_chars)&(word_start==True):
            word2char_map.append(i)
            word_start=False
        elif char in blank_chars:
            word_start=True
    return word2char_map

def text_inference(model, text, tokenizer, max_len=MAX_LEN, return_pred_str=True):
    """ Function to perform inference and format appropriately
    
    From @cdeotte - reformatted as a function to improve readability/funcitonality
    
    Args:
        model (): The model to use for inference
        text (str): String to perform inference on
        tokenizer (PreTrainedTokenizerFast): The tokenizer that was
            fit on the respective text corpus for this task
        max_len (int, optional): The maximum allowed number of tokens
        return_pred_str (bool, optional): Whether to return the mask
            array or the respective prediction string.
        
    Returns:
        The prediction formatted correctly as a word array
    
    """

    # STEP 1 - CREATE DESTINATION WORD ARRAY
    word_array = np.zeros((len(text.split())), dtype=np.int32)

    # STEP 2 - CREATE INTERMEDIATE CHARACTER ARRAY
    char_array = np.zeros((len(text)), dtype=np.int32)

    # STEP 3 - FILL CHAR ARRAY FROM PREDICTION TOKEN ARRAY
    preds = model.predict(text) 
    
    _encoding = tokenizer(text, max_length=max_len, padding='max_length', truncation=True, return_offsets_mapping=True)
    _map_token_to_char = _encoding['offset_mapping']
    _word2char_map = create_word2char_map(text)
    
    for k in range(max_len): 
        char_array[_map_token_to_char[k][0]:_map_token_to_char[k][1]] = preds[k]

    # STEP 4 - FILL WORD ARRAY FROM CHARACTER ARRAY
    word_array = np.array([char_array[_word2char_map[k]] for k in range(len(word_array))], dtype=np.int32)
    
    if not return_pred_str:
        return word_array
    else:
        return ' '.join([str(n) for n in np.where(word_array==1)[0]])
    

_input_string = "my name is Chris Deotte."
print(f"\n... INPUT STRING : {_input_string}")

pred_raw = text_inference(DumbModel(), _input_string, tokenizer, return_pred_str=False)
print(f"... RAW BINARY PREDICTION : {pred_raw}")

pred_str = ' '.join([str(n) for n in np.where(pred_raw==1)[0]])
print(f"... PREDICTION STRING     : '{pred_str}'")
print()
````

> <i>"<b><br>Kaggle's Preprocess Algorithm</b><br><br>After brainstorming in the forums with others (especially helpful was [@hkaggleqrdl](https://www.kaggle.com/hkaggleqrdl) ), we have discovered Kaggle's preprocess function for creating the **`predictionstring`** from discourse text using Python's **`split()`**.</i>
>
><i>The algorithm below explains the **mystery of why we are seeing 2% of `predictionstring` that do not match discourse text in the train data**. The **main cause** of the discrepancy is that **the first "word" (which may be a single punctuation mark like a comma) of discourse text is sometimes a portion of a longer "word" in the full text**. (Here "word" means connected sequence of characters without blank space).<br><br><br><b>Mystery Algorithm</b><br><br>The following algorithm produces 100% of train's predictionstring's perfectly when using discourse_start and discourse_end. So the mystery is solved! üéâ</i>

````python
def mystery_algorithm(full_text, discourse_start, discourse_end):
    """ Algorithm reverse engineered by @cdeotte and team """
    
    char_start = discourse_start
    char_end = discourse_end
    word_start = len(full_text[:char_start].split())
    word_end = word_start + len(full_text[char_start:char_end].split())
    word_end = min( word_end, len(full_text.split()) )
    predictionstring = " ".join( [str(x) for x in range(word_start,word_end)] )
    
    return predictionstring

def calc_word_indices(full_text, discourse_start, discourse_end):
    """ Algorithm shared by host """
    start_index = len(full_text[:discourse_start].split())
    token_len = len(full_text[discourse_start:discourse_end].split())
    output = list(range(start_index, start_index + token_len))
    if output[-1] >= len(full_text.split()):
        output = list(range(start_index, start_index + token_len-1))
        
    return output

example_full_text_1 = "If you want to learn data science, then you should join Kaggle. There are opportunities to discuss, code, and compete."
discourse_text = ", then you should join Kaggle."

print(discourse_text)
print(mystery_algorithm(example_full_text_1, example_full_text_1.index(discourse_text), example_full_text_1.index(discourse_text)+len(discourse_text)))
print(calc_word_indices(example_full_text_1, example_full_text_1.index(discourse_text), example_full_text_1.index(discourse_text)+len(discourse_text)))
print(np.array(example_full_text_1.split())[np.array(calc_word_indices(example_full_text_1, example_full_text_1.index(discourse_text), example_full_text_1.index(discourse_text)+len(discourse_text)))])
````

KMeans clustering is an algorithm that aims to partition n observations into k clusters in which each observation belongs to the cluster with the nearest center. Cluster centers (or cluster centroid) are simply the cluster‚Äôs mean. KMeans require us to provide the number of cluster k beforehand. Ideally, we want to choose the number of clusters carefully. One technique to find a good number of clusters is the Elbow Method. Essentially, the method consists of plotting the explained variation as a function of the number of clusters, and picking the elbow of the curve as the number of clusters. However, for simplicity, I‚Äôll arbitrarily set k = 10 as it seems to be decent for this dataset.

‚ÄòElbow‚Äô Method
With the Elbow method, we plot the within-cluster sum of squares (RSS) for different K values and try to see what looks like an elbow. Let‚Äôs test K being 1 to 8.

````python
def preprocess_text(input_strs , filters=None , stopwords=None):
    """
    
    Function to preprocess a column of text from a CuDF dataframe
    
    Code from NVIDIA blog post 
        --> https://developer.nvidia.com/blog/nlp-and-text-precessing-with-rapids-now-simpler-and-faster/
      
    --------------------------------------------
      STEPS:
    --------------------------------------------
        * filter punctuation
        * to_lower
        * remove stop words (from nltk corpus)
        * remove multiple spaces with one
        * remove leading spaces    
    ----------------------------------------
        
    Args:
        input_strs (CuDF Series): Strings to be preprocessed
        filters (list of strs, optional): Which symbols to remove
            - If no list is provided a default list will be used.
        stopwords (list of strs, optional): Which words to remove
            - If no list is provided a default from nltk will be used.
        
    Returns:
        CuDF series of strings (i.e. a single column) after preprocessing
        conducted on the GPU
    
    """
    
    # Defaults
    _STOPWORDS = nltk.corpus.stopwords.words('english')    
    _FILTERS   = [  '!', '"', '#', '$', '%', '&', '(', ')', '*', 
                    '+', '-', '.', '/',  '\\', ':', ';', '<', '=', 
                    '>', '?', '@', '[', ']', '^', '_', '`', '{', 
                    '|', '}', '\~', '\t','\\n',"'",",",'~' , '‚Äî'  ]
    
    # Assign default if no specific arg is passed passed
    if stopwords is None:
        stopwords=_STOPWORDS
    if filters is None:
        filters=_FILTERS
    
    # filter punctuation and case conversion
    input_strs = input_strs.str.lower().replace_multi(FILTERS, ' ', regex=False)
        
    # place stopwords on device to accelerate stopword removal
    stopwords_gpu = nvstrings.to_device(stopwords)
    input_strs = nvtext.replace_tokens(input_strs.data, stopwords_gpu, ' ')
    input_strs = cudf.Series(input_strs)
        
    # replace multiple spaces with single one and strip leading/trailing spaces
    input_strs = input_strs.str.replace(r"\s+", ' ', regex=True)
    input_strs = input_strs.str.strip(' ')
    
    return input_strs

def preprocess_text_df(df, text_cols=('discourse_text',), **kwargs):
    """
    
    Function to preprocess a CuDF dataframe and the respective text columns
    
    Code from NVIDIA blog post 
        --> https://developer.nvidia.com/blog/nlp-and-text-precessing-with-rapids-now-simpler-and-faster/
    
    Args:
        df (CuDF Dataframe): The dataframe to preprocess
        text_cols (tuple, optional): The columns containing text that
            we want to be preprocessed.
        
    Returns:
        The preprocessed CuDF dataframe
        
    """
    _df = df.copy()
    
    for col in text_cols:
        _df[col] = preprocess_text(_df[col], **kwargs)
    return  _df

# preprocess_text_df(train_df)
````

````python
# Create TFIDF vectorizer from the RAPIDS CuML library
#    - N_FEATURES refers to the number of dimensions we will use
#      to express the datacontained within our corpus of text
N_FEATURES = 5_000
tfidf_vec = TfidfVectorizer(stop_words='english', max_features=N_FEATURES)
cudf_corpus = cudf.Series(train_df["discourse_text"])
tfidf_matrix = tfidf_vec.fit_transform(cudf_corpus)

# This is to convert our sparse matrix to a dense one and subset if necessary
#   - Change SUBSAMPLE_LEN to a smaller constant for a smaller dense subset
SUBSAMPLE_LEN = tfidf_matrix.shape[0] # 100_000
tfidf_dense_sample = tfidf_matrix[:SUBSAMPLE_LEN].todense()

print(f"\n... TFIDF MATRIX SHAPE: {tfidf_matrix.shape} ...\n")

# KMEANS Variables
num_clusters = 25 # arbitrary
kmeans_model = KMeans(n_clusters=num_clusters, n_init=1, max_iter=1000)

# Calculate clusters and representative distances
kmeans = kmeans_model.fit(tfidf_dense_sample)
kmeans_clusters = kmeans.predict(tfidf_dense_sample)
kmeans_distances = kmeans.transform(tfidf_dense_sample)

# Using the previously computed TF-IDF, we can look at the most important words 
# for each of our clusters (which are the words with the highest tf-idf value 
# in each of the cluster centers). We filter terms that are also present in 
# other clusters as they do not help us distinguish different topics. 
sorted_centroids = kmeans.cluster_centers_.argsort()[:, ::-1]
terms = tfidf_vec.get_feature_names()

TOP_N = 100
clusters_terms = sorted_centroids[:, :TOP_N].get()

for i, c1 in enumerate(clusters_terms):
    cluster = set(c1)
    for j, c2 in enumerate(clusters_terms):
        if i == j:
            continue
        cluster -= set(c2)
    cluster = c1[np.isin(c1, list(cluster))][:10]
    
    k_cluster = train_df[(kmeans_clusters==i).get()]
    print(f"\n\n--------------- CLUSTER {i+1} ---------------\n" \
          f"\n\tCommon Terms: " \
          f"{', '.join(terms[cluster].to_arrow().to_pylist())}")
    
    print(f"\n\tLabel Distribution: {k_cluster.discourse_type.value_counts().to_dict()}")
    
    for j in range(3):
        d_ex = k_cluster.sample(1)
        print(f"\n\tCluster Example #{j+1}" \
              f"\n\t\tDiscourse TYPE: --> {d_ex['discourse_type'].values[0]}" \
              f"\n\t\tDiscourse TEXT: --> {d_ex['discourse_text'].values[0]}")
        
# ######################################################### #
# ---###################=== t-SNE ===###################--- #
# ######################################################### #
# KMeans helps us find clusters for our data, 
# but the resulting matrix is highly-dimensional with 
# 30 components which is hard to represent on a 2D plane. 
# In order to visualize those clusters, we will use the t-SNE 
# algorithm to reduce the dimensionality of our data. 
# t-SNE models each high-dimensional object by a n-dimensional 
# point (for instance n=2) in such a way that similar objects 
# are modeled by nearby points and dissimilar objects are modeled 
# by distant points with high probability. 
# In our case, t-SNE will help us find a low dimensional data representation 
# that keeps close points together in the 2D space. 
# With this 2D representation, we will see more clearly if our KMeans worked 
# as clusters should be visible on the plot.
# ######################################################### #
tsne_model = TSNE(n_components=2, verbose=1, n_iter=1000)
tsne_kmeans = tsne_model.fit_transform(kmeans_distances)
````
