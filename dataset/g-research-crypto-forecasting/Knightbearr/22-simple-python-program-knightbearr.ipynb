> # **1. BMI Converter**

```python
weight = int(input('Weight: ')) # Kg
height = float(input('Height: ')) # Cm

bmi = (weight / height) ** 2

if bmi < 18.5:
    print("Underweight")
elif bmi >= 18.5 and bmi <= 25:
    print("Normal")
elif bmi >= 25 and bmi <= 30:
    print("Overweight")
else:
    print("Obesity")
```

> # **4. Stack**

```python
# Stack Wrapper Class

class Stack(): # LIFO = last in first out
    def __init__(self): 
        self.stack = list()

    def push(self, item): # Push an item onto the stack
        self.stack.append(item)

    def pop(self): # Pop an item off of the stack
        if len(self.stack) > 0:
            return self.stack.pop()
        else:
            return None

    def peek(self): # get an item on top of stack, without removing it
        if len(self.stack) > 0:
            return self.stack[len(self.stack)-1]
        else:
            return None

    def __str__(self): # magic funct that for changing the bool to str
        return str(self.stack)

mystack = Stack()
mystack.push(4)
mystack.push(5)
mystack.push(6)
mystack.pop()
print(mystack)
```

> # **5. Perimeter and Square Area**

```python
def perimeter_square_area(width, length):
    around = 2 * (width + length)
    large = (width * length)
    return around, large

circumference_result, broad_result = perimeter_square_area(10, 20)
print(f'Circumference result: {circumference_result}, Broad Result: {broad_result}')
```

> # **13. Counting a Number**

```python
total = count = average = 0

while True:
    num_input = input('Enter a number: ')
    if num_input == 'done':
        print(f'Total: {total}, Count: {count}, Average: {average}')
        break
    try:
        count += int(num_input)
        total += 1
        average = total/count
    except ValueError:
        print('Invalid input')
```

> # **14. Looping Through Strings**

```python
fruit, index = 'banana', 0

while index < len(fruit):
    letter = fruit[index]
    print(f'index: {index}, letter: {letter}\n')
    index += 1

# Prettier way
for letter in fruit:
    print(f'prettier: {letter}\n')

# Loop and counting
word, count = 'banana', 0

for letter in word:
    if letter == 'a':
        count += 1
print(f'loops: {count}\n')
```

> # **19. Binary Search**

```python
def binary_search(list, item):
    low = 0
    high = len(list)-1
    
    while low <= high:
        mid = (low + high) // 2
        guess = list[mid]
        if guess == item:
            return mid
        if guess > item:
            high = mid - 1
        else:
            low = mid + 1
    return None

mylist = [1, 3, 5, 7, 9]
print(f'the number is in index: {binary_search(mylist, 7)}')
```

> # **20. Longest Word Search Program**

```python
n = input('')

list_word = n.split(' ')

longest_word = list_word[0]

for word in list_word[0:]:
    if len(word) > len(longest_word):
        longest_word = word

print(f'longest word: {longest_word}')
```
