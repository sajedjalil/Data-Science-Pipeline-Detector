```python
##------------------
#import dependencies
##------------------
import torchvision.transforms as transforms

import augly.image as imaugs
import augly.utils as utils

from PIL import Image
import numpy as np
import matplotlib.pyplot as plt

import os
import requests
```

```python
##--------------------
#Download sample image
##--------------------

URL = "https://upload.wikimedia.org/wikipedia/en/thumb/7/7d/Lenna_%28test_image%29.png/330px-Lenna_%28test_image%29.png"
IMPATH = "./lena.jpg"
response = requests.get(URL)

file = open("lena.jpg","wb")
file.write(response.content)
file.close()
```

<h1 style="font-weight:'bold'; color:blue; font-family:monospace; text-align :center;"> List of Augmentations</h1>

<img src="https://i.postimg.cc/Ghd2J7gp/carbon.png" height="600" width="600" style="display: block;margin-left: auto;margin-right: auto;">

```python
orig = Image.open(IMPATH)
aug1 = imaugs.blur(orig)
aug2 = imaugs.color_jitter(orig)
aug3 = imaugs.contrast(orig)
aug4 = imaugs.hflip(orig)
aug5 = imaugs.vflip(orig)
aug6 = imaugs.rotate(orig)
aug7 = imaugs.sharpen(orig)
aug8 = imaugs.shuffle_pixels(orig)

img_ls = [orig,aug1,aug2,aug3,aug4,aug5,aug6,aug7,aug8]
titles = ["original","blur","color jitter","contrast","horizontal flip","vertical flip","rotation","sharpen","shuffle pixels"]
```

```python
##---------------------------------------------
#let's view results with standard Lenna image :)
##---------------------------------------------

plt.figure(figsize=(20,20))

for i in range(9):
    plt.subplot(3,3,i+1)
    plt.imshow(img_ls[i])
    plt.axis("off")
    plt.title(titles[i])
plt.show()
```
