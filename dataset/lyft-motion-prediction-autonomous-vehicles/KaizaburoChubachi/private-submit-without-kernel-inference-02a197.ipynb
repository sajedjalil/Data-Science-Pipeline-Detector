# Import already calculated submission.csv as Dataset

We have 2 choices to prepare dataset:

 - 1. Upload calculated submission.csv file (maybe from local machine) as Dataset and use it
 - 2. Use other kernel's dataset (it can be either your own kernel, or any other public kernel)

For me the most easiest way to find the kernel is to upvote the kernel you want to use, and filter it as "Your Favorites" at search botton. Upvote his kernel! :)
