```python
train_dat =  pd.read_csv("../input/train.tsv",delimiter="\t")
test_dat = pd.read_csv("../input/test.tsv",delimiter="\t")
```

```python
test_dat = test_dat[["test_id","shipping"]]
predicted = train_dat.groupby("shipping").mean().reset_index()
submission = test_dat.merge(predicted)[["test_id","price"]]
```

```python
submission.to_csv("submission.csv",index=False)
```
