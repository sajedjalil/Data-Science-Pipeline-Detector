```python
# importing all csv files from engagement folder as one dataframe

starting_directory = os.getcwd()    # to get back to default directory

path = "../input/learnplatform-covid19-impact-on-digital-learning/engagement_data"
os.chdir(path)

lis = list()
for file in os.listdir():
    df = pd.read_csv(file)
    file_id = file.split(".")[0]
    df['district_id'] = file_id
    lis.append(df)

engage_df = pd.concat(lis)

os.chdir(starting_directory)


# reading dataframe
engage_df.reset_index(inplace=True)
engage_df.drop("index", axis=1, inplace=True)

engage_df.info() 
```

<p>Although, before changing dtypes, checking if there are any null or duplicate values in the dataframe. If any, decisions regarding removal or modification of the rows will be made depending upon the impact it'll make on the dataframe.</p>

```python
def null_count(df, col_list):
    for col in col_list:
        null_v = df[col].isnull().sum()
        print(f"There are {null_v} null values in column '{col}'")
        
def check_null_dup(df):
    null_col_list = list()
    print("Missing Value Percentage :")
    for col in df.columns:
        value = np.mean(df[col].isnull())
        if value == 0: continue            # ignoring columns that do have any missing values    
        null_col_list.append(col)
        print(f"{col}\t-\t{value}%")
    if len(df[df.duplicated()]) == 0:
        print("\nDataframe contains no duplicate values!\n")
    else:
        dup = df[df.duplicated()]
        print('Found following duplicated rows\n')
        return dup
    null_count(df, null_col_list)

check_null_dup(engage_df)
```


<p><b>To handle the missing values we can either use univariate or multivariate imputation techniques or choose to remove the values. These will differ from column to column.</b></p>

---
<div class="h4">lp_id</div>
<p class="para">Though this column has around 2.5% missing values we cannot impute values using statistical method such as "mode" since it will create bias over the highly occured unique product identifier. Hence, removing rows.</p>

<div class="h4">pct_access & engagement_index</div>
<p class="para">Both of these columns have less than 1% null values and ideally we can choose to remove these null values from our df without affecting our dataset. But, we can also fill in missing values wherever possible:</p>
<ul><li><b>pct_access</b> -- Has a low number of missing values, thus, we can fill it using mean values without affecting the variance</li>
<li><b>engagement_index</b> -- Missing values on the other hand are on a higher side. Along with this it also includes lp_id that are not available in product_info.csv, thus removing those values.<br>We will fill the missing values of unique product ids that are present in product_info.csv with the average engagement index of that product id.</li></ul>

```python
path = "../input/learnplatform-covid19-impact-on-digital-learning/products_info.csv"
product_df = pd.read_csv(path)


# removing ip_id null value rows and changing dtype to int
null_index = engage_df[engage_df["lp_id"].isnull()].index
engage_df.drop(null_index, inplace=True)

# replacing pct_access null values with mean of that column
engage_df['pct_access'].fillna(np.mean(engage_df['pct_access']), inplace=True)

# handling engagement_index missing values
miss_val = engage_df[engage_df["engagement_index"].isnull()]["lp_id"].unique()
uniq_id = product_df["LP ID"].unique()

id_avg = dict()

for val in miss_val:
    if val in uniq_id:
        avg = engage_df[engage_df["lp_id"] == val]["engagement_index"].mean()
        id_avg[val] = avg
        
# this will map average engagement_index for the product id's that are present in 
# product_info.csv respectively
engage_df.engagement_index = engage_df["engagement_index"].fillna(engage_df["lp_id"].map(id_avg))

# removing missing values whose product id's are not present in product_info.csv
null_index = engage_df[engage_df["engagement_index"].isnull()].index
engage_df.drop(null_index, inplace=True)


# changing dtypes
engage_df.time = pd.to_datetime(engage_df["time"])
engage_df.district_id = engage_df["district_id"].astype("int64")
engage_df.lp_id = engage_df["lp_id"].astype("int64")
```

<div class="h2">Student with disabilities</div>
<p class="para">Initially on my level I assumed that since most of the products with either of the primary essential functionalities are not designed keeping in mind students with disabilities. They will have a huge loss of learning during the pandemic.<br> Accodring to <b>United States Department of Education</b>, children with disabilities and their families were dispropotionately affected by the pandemic compared to their peers without disabilities. Even with the support of special education and related services required by their <b>IEPs (indivisualised eduction programs)</b>, some children were unable to progress towards acheiving the functional and academic goals included in their IEPs.<br></p>
<p class="statement">Therefore, reopening schools safely is of utmost importance for the progress of students with disabilities</p>
<ul><div class="h2">Individuals with Disabilitites Education Act</div>
    <p class="para">Though enacted in 1975, the Individuals with Disabilitites Education Act <b>IDEA</b> formerly known as <b>Education for all handicapped children Act</b>, mandates the provision of free and appropriate public school for eligible students aged 3-21. This eligibility is determined by a disability that adversly affects academic performance.<br>In 2019-2020, number of students ages 3-21 who received special education services under the IDEA was 7.3 million or 14 percent of all public school students. Among the students receiving special education services, the most common category of disability (33 percent) was specific learning disabilities.<br><b>The following image has been taken from <a href="https://nces.ed.gov/programs/coe/indicator/cgg" target="_blank">NCES COE - Students with Disabilities</a></b>, which gives the statistics of these students how they were attend most of their classes, that is the enviornment.</p><img src="https://nces.ed.gov/programs/coe/figures/CGG/Figure_1_CGG-flat.svg"></img></ul>
    <ul><ul><div class="h3">Fundings</div>
    <li><b>Part B:</b> Part B of IDEA provides funds to the states to assist them in providing <b>FAPE</b>(free appropriate public education) to children ages 3-12 with disabilities who are in need of special education and related services.</li>
    <li><b>Part C:</b> Part C of IDEA provides funds to the states to assit them in developing and implementing state wide, comprehensive, co-ordinated, multidisciplinary, intragency systems to make early intervention  services available to all children from birth through age two with disabilities and their families.</li></ul><br>
    <div class="h4">Difficulties</div>
    <p class="para">The assumption underlying the Individuals with
Disabilities Education Act (IDEA) and its
predecessor legislation is that, on average, the cost
of educating children with disabilities is twice the
average cost (measured as the national average per
pupil expenditure or APPE) of educating other
children. Congress determined that the federal
government would pay up to 40 percent of this
"excess" cost, which is referred to as full funding.
Since 1981, the first year for which full funding
was 40 percent of APPE, the federal share has
remained less than half of the federal commitment
based on regular appropriations. Each year the
federal government fails to fully fund IDEA, it
shifts the costs for educating students with special
needs to states and school districts. <a href="https://www.nea.org/sites/default/files/2021-01/IDEA%20Funding%20Gap%20by%20State%20FY%202020.pdf" target="_blank">Source</a><br></p>
    <p class="statement">Hence to implement an infrastructure as desired for the students with disabilities, the federal government fundings if payed as promised will make a huge difference.</p>
