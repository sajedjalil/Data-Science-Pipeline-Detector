```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)

paths = []
import os
for dirname, _, filenames in os.walk('/kaggle/input'):
    for filename in filenames:
        path = os.path.join(dirname, filename)
        paths.append(path)

import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import cufflinks as cf
%matplotlib inline
from plotly.offline import download_plotlyjs, init_notebook_mode, plot, iplot
import plotly.graph_objects as go
init_notebook_mode(connected = True)
cf.go_offline()

products = pd.read_csv("/kaggle/input/learnplatform-covid19-impact-on-digital-learning/products_info.csv", index_col = 0)

new = products['Primary Essential Function'].str.split(' - ', expand = True)
products['Category'] = new[0]

lpid = products.index.to_list()

districts = pd.read_csv("/kaggle/input/learnplatform-covid19-impact-on-digital-learning/districts_info.csv", index_col = 0)

districts['ccr'] = districts['county_connections_ratio'].replace(dict((('[0.18, 1[', 'low'), ('[1, 2[', 'high'))))

districts = districts.drop('county_connections_ratio', axis =1)

cols = ['pct_black/hispanic', 'pct_free/reduced', 'pp_total_raw']
districts[cols] = districts[cols].replace({'\[$': ']'},regex=True)

pct_conv = {'[0, 0.2]': "Below 20%", '[0.2, 0.4]': "20-40%", '[0.4, 0.6]': "40-60%", '[0.8, 1]': "Above 80%",
       '[0.6, 0.8]': "60-80%" }
mapcol = ['pct_black/hispanic', 'pct_free/reduced']
districts[mapcol] = districts[mapcol].applymap(lambda x: pct_conv.get(x)).fillna('missing', axis = 0)

districts_cleaned = districts[districts['state'].notna()]
districts_cleaned_id = districts_cleaned.index.to_list()

code = dict((('AZ', 'Arizona'), ('CT', 'Connecticut'), ('IL','Illinois'), ('UT', 'Utah'), ('NC','North Carolina'),
      ('MO','Missouri'), ('WA','Washington'), ('MA','Massachusetts'), ('NY','New York'), ('IN','Indiana'), ('VA','Virginia'),
      ('OH','Ohio'), ('NJ', 'New Jersey'), ('CA','California'), ('CO', 'District Of Columbia'), ('TX', 'Texas'),
      ('TN','Tennessee'), ('FL','Florida'), ('WI','Wisconsin'), ('NH','New Hampshire'), ('MI','Michigan')
     ))

code = dict(zip(code.values(), code.keys()))

districts_cleaned['code'] = districts_cleaned['state'].replace(code)

def makedf():
    df = pd.DataFrame()
    df1 = pd.DataFrame()
    discarded = 0
    for path in paths[3:]:
        districtId = path.split('/')[-1][:-4]
        x = pd.read_csv(path)
        x['month'] = pd.to_datetime(x['time'], yearfirst=True).dt.month
        if x['month'].nunique() != 12:
            discarded += 1
            continue
        month = x.groupby(by = ['lp_id', 'month'], dropna=True, as_index = False)[['pct_access', 'engagement_index']].median()
        month = month.loc[month['lp_id'].isin(lpid)]
        x = x.loc[x['lp_id'].isin(lpid)]
        month['district_id'] = int(districtId)
        df = pd.concat([df, month], axis = 0)
        df1 = pd.concat([df1, x], axis = 0)
    return df, discarded, df1

df_new, discarded, products_whole = makedf()

products_daily = products_whole.copy()

products_daily = products_daily[products_daily['lp_id'].notna()]
products_daily['lp_id'] = products_daily['lp_id'].astype('int32').astype("category")

products_daily = products_daily.loc[products_daily['lp_id'].isin(lpid)]

products_daily = products_daily.groupby(by = ['time', 'lp_id'], as_index=False).mean()
products_daily['prod'] = products_daily['lp_id'].apply(lambda x: products.loc[x, 'Product Name'])
products_daily['category'] = products_daily['lp_id'].apply(lambda x: products.loc[x, 'Category'])

df_new = df_new.loc[df_new['district_id'].isin(districts_cleaned_id)]
df_new['lp_id'] = df_new['lp_id'].astype('int64')
df_new['state'] = df_new['district_id'].apply(lambda x: districts_cleaned.loc[x, 'code'])
df_new['prod'] = df_new['lp_id'].apply(lambda x: products.loc[x, 'Product Name'])
```

```python
prod_monthly = df_new.groupby(by = ['prod', 'month'], as_index = False)[['pct_access', 'engagement_index']].median()

temp = prod_monthly[['prod', 'engagement_index']].groupby(by = 'prod', as_index = False).mean().sort_values(by = "engagement_index", ascending=False).iloc[:20]
Top20Tools = temp['prod']
fig = px.bar(data_frame=temp, y="prod", x = "engagement_index", orientation='h')
fig.update_layout(
    bargap = 0.5,
    title={
        'text': "Top Digital Learning Tools in 2020",
        'y': 0.97,
        'x':0.5,
        'xanchor': 'center',
        'yanchor': 'top', 
    },
    xaxis_title="Engagement index",
    yaxis_title="",
    yaxis=dict(
        showgrid=False,
        zeroline=False,
        showline=False,
        showticklabels=True,
        categoryorder= 'total ascending'
    ),
    xaxis=dict(
        showline=True,
        showgrid=False,
        showticklabels=True,
        linecolor='rgb(204, 204, 204)',
        linewidth=2,
        ticks='outside',
        tickfont=dict(
            family='Arial',
            size=12,
            color='rgb(82, 82, 82)',
        ),
    ),
    autosize=True,
    font=dict(
        size=14,
    ),
    height = 700,
    margin=dict(
        autoexpand=False,
        l=280,
        t=110,
        r = 280,
        pad = 30
    ),
    showlegend=False,
    plot_bgcolor='white'
)
fig.show()
```
