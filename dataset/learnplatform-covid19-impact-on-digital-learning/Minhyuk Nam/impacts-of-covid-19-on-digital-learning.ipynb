### <font color="#6C8255">3.1.1. Districts Information</font>
We started from collecting districts information. The data provides `state`, `locale`, `pct_black/hispanic`, `pct_free/reduced`, `county_connections_ratio`, `pp_total_raw`.

* **Dropped some districts information** which does not provide state information. Its frequency is 57, which is nearly 25%.
* Added state **abbreviation column** for further convenience.
* There is a **very small variation in `country_connections_ratio`**, where there is only 1 district with `[1,2[`, while others are all `[0.18, 1[`, which we would not use on our analysis.
* **Converted `pp_total_raw` columns** into 7 groups as it is separated in detailed level. (See [Section 4.1.1.](#--Share-of-High-Speed-Internet-Connection) for detailed distribution)

```python
districts_info_raw = pd.read_csv("../input/learnplatform-covid19-impact-on-digital-learning/districts_info.csv")

# Drop observation which does not contain state information
districts_info = districts_info_raw[districts_info_raw.state.notna()].reset_index(drop=True)

# Rename columns
districts_info.columns = ['leaid', 'state', 'locale', 'blkhis_ratio', 'freereduced_ratio', 'connection_ratio', 'pp_totexp']

# Add a column for state name abbreviation
districts_info['st'] = districts_info['state'].apply(lambda x: us_state_abbrev[str(x)])

# Add a column for region of each states based on Census Regions
# NE, MW, S, W order
region = [['NH', 'MA', 'CT', 'NY', 'NJ', 'PA'], ['OH', 'MI', 'WI', 'IL', 'MN', 'ND', 'MO', 'IN'], ['DC', 'VA', 'NC', 'TN', 'FL', 'TX'], ['AZ', 'UT', 'CA', 'WA' ]]
districts_info['region'] = districts_info['st'].apply(lambda x: 'NE' if x in region[0] else ('MW' if x in region[1] else ('S' if x in region[2] else 'W')))

# Convert categorical column into CategoricalDtype
districts_info['region'] = districts_info['region'].astype(pd.api.types.CategoricalDtype(list(districts_info['region'].unique())))
districts_info['locale'] = districts_info['locale'].astype(pd.api.types.CategoricalDtype(list(districts_info['locale'].unique())))

# Convert interval data into single value
districts_info['blkhis_ratio'] = districts_info['blkhis_ratio'].apply(lambda x: float(str(x).split(',')[0][1:]) if str(x) != 'nan' else np.nan)
districts_info['freereduced_ratio'] = districts_info['freereduced_ratio'].apply(lambda x: float(str(x).split(',')[0][1:]) if str(x) != 'nan' else np.nan)
districts_info['connection_ratio'] = districts_info['connection_ratio'].apply(lambda x: float(str(x).split(',')[0][1:]) if str(x) != 'nan' else np.nan)
districts_info['pp_totexp'] = districts_info['pp_totexp'].apply(lambda x: int(str(x).split(',')[0][1:]) if str(x) != 'nan' else np.nan)

# Convert pp_totexp into 6 categories (delta: 4000)
districts_info['pp_totexp_raw'] = districts_info['pp_totexp']
districts_info['pp_totexp'] = districts_info['pp_totexp'].apply(lambda x: 4000 if x<=6000 else (x if ((8000 <= x) & (x < 18000)) else (18000 if x>=18000 else np.nan)))

print(f'There are {len(districts_info_raw)} of districts, while \
{len(districts_info_raw[districts_info_raw.state.isna()])} have dropped due to missing values.\n')
print(f'Distributions of connection_ratio: ')
print(districts_info_raw['county_connections_ratio'].value_counts(),'\n')
districts_info.head()
```

### <font color="#6C8255">3.1.2. Engagement Data</font>
The data consists student's daily engagment data on each product (`pct_access`, `engagement_index`). Since the data does not consist of `district_id` we added from the title of each data (`leaid`).

* Aggregated the data into `district x product x month` level, which will allow us to analyze the impact of Covid-19 more in a more robust manner (i.e., reduce noise due to high-frequency data)

```python
PATH = '../input/learnplatform-covid19-impact-on-digital-learning/engagement_data' 
engagement_raw = pd.DataFrame()

for district in districts_info.leaid.unique():
    df = pd.read_csv(f'{PATH}/{district}.csv', index_col=None, header=0)
    df['leaid'] = district
    
    # Dropped data whose engagement index is np.nan
    df = df[df.engagement_index.notna()].reset_index(drop=True)
    
    engagement_raw = pd.concat([engagement_raw, df])

engagement_day = engagement_raw #may be drop na??

# Generate `year`, `month`, `day` variables
engagement_day['year'] = engagement_day['time'].apply(lambda x: int(str(x)[:4]) if str(x) != 'nan' else np.nan)
engagement_day['month'] = engagement_day['time'].apply(lambda x: int(str(x)[5:7]) if str(x) != 'nan' else np.nan)
engagement_day['day'] = engagement_day['time'].apply(lambda x: int(str(x)[8:]) if str(x) != 'nan' else np.nan)
engagement_day['time'] = pd.to_datetime(engagement_day['time'], format='%Y-%m-%d', errors='coerce')
engagement_day['week'] = engagement_day['time'].dt.week

# Aggregate into `district x product x month` level
engagement_month = engagement_day.groupby(['year', 'month', 'leaid', 'lp_id'])['pct_access', 'engagement_index'].mean().reset_index()

engagement_month.head()
```

* Split the `primary_function` into 2 levels as it is separated in a very detailed level.
* Synchronized `primary_function` value which is not consistent even they are alike (e.g., 'Sites, Resources & References' -> 'Sites, Resources & Reference')

```python
products_info_raw = pd.read_csv("../input/learnplatform-covid19-impact-on-digital-learning/products_info.csv")
products_info = products_info_raw[['LP ID', 'Product Name', 'Provider/Company Name', 'Sector(s)',
       'Primary Essential Function']]

# Split 'Primary Essential Function'
products_info = products_info[products_info['Primary Essential Function'].notna()].reset_index(drop=True) #drop products which has no info
products_info['mainfunction'] = products_info['Primary Essential Function'].apply(lambda x: str(x).split(' - ')[0])
products_info['mainfunction'] = products_info['mainfunction'].astype(pd.api.types.CategoricalDtype(list(products_info['mainfunction'].unique())))
products_info['subfunction'] = products_info['Primary Essential Function'].apply(lambda x: str(x).split(' - ')[0]+":"+str(x).split(' - ')[1])
products_info['subfunction'] = products_info['subfunction'].astype(pd.api.types.CategoricalDtype(list(products_info['subfunction'].unique())))

# Generate dummies for each sector
sectors_dummies = products_info['Sector(s)'].str.get_dummies(sep="; ")
sectors_dummies.columns = [f"{re.sub(' ', '', x)}" for x in sectors_dummies.columns]
products_info = products_info.join(sectors_dummies)
products_info.drop("Sector(s)", axis=1, inplace=True)

# Synchronize values in 'Primary Essential Function LEV2'
products_info['subfunction'] = products_info['subfunction'].replace({'LC:Sites, Resources & References' : 'LC:Sites, Resources & Reference'})
products_info.drop("Primary Essential Function", axis=1, inplace=True)

# Rename columns
products_info.columns = ['lp_id', 'product_name', 'provider_name', 'mainfunction', 'subfunction', 'corporate', 'higher', 'prek12']

products_info.head()
```

### <font color="#6C8255">3.1.4. COVID-19 Infection: Economic Tracker (Opportunity Insights)</font>
The data consists of the number of Covid-19 infections (`case_count`, `death_count`) in each state on daily basis.
* Replace `.` as `np.nan`
* Aggregated the data into `state x month` level, which will allow us to analyze the impact of Covid-19 in `monthly` level

```python
infection_raw = pd.read_csv("../input/opportunity-insights-real-time-economic-tracker-us/data/COVID - State - Daily.csv")
infection_raw = infection_raw[infection_raw['year']==2020]
infection_day = infection_raw[['year', 'month', 'day', 'statefips', 'case_count', 'death_count']]
infection_day = infection_day.rename(columns = {'case_count': 'cum_cases', 'death_count': 'cum_deaths'})
infection_day['date'] = infection_day[['year', 'month', 'day']].apply(lambda x: '-'.join(x.values.astype(str)), axis="columns")
infection_day['date'] = pd.to_datetime(infection_day['date'], format='%Y-%m-%d', errors='coerce')
infection_day['week'] = infection_day['date'].dt.week

# Convert `statefips` into `st`
geoid = pd.read_csv("../input/opportunity-insights-real-time-economic-tracker-us/data/GeoIDs - State.csv")
geoid = geoid[['statefips', 'stateabbrev', 'state_pop2019']]
geoid = geoid.rename(columns = {'stateabbrev': 'st', 'state_pop2019': 'st_pop'})
# geoid = geoid[['statefips', 'stateabbrev']]
# geoid = geoid.rename(columns = {'stateabbrev': 'st'})
infection_day = pd.merge(infection_day, geoid, on = 'statefips')
infection_day = infection_day[infection_day['st'].isin(sample_us_state_abbrev)].reset_index(drop=True)

# Replace '.' as np.nan
infection_day['cum_cases'] = infection_day['cum_cases'].apply(lambda x: np.nan if str(x) == "." else int(x))
infection_day['cum_deaths'] = infection_day['cum_deaths'].apply(lambda x: np.nan if str(x) == "." else int(x))

# To construct the monthly data, keep observations which are recorded at the last day of each month
infection_day = infection_day.join(infection_day.groupby(['statefips', 'month'])['day'].max(), on=['statefips', 'month'], rsuffix='_max')
infection_month = infection_day[(infection_day['day'] == infection_day['day_max']) & (infection_day['year'] == 2020)]

# Compute new cases and death in state x month level
infection_month = infection_month.sort_values(by = ['statefips', 'month']).reset_index(drop=True)
infection_month['new_cases'] = infection_month.groupby(['statefips'])['cum_cases'].diff()
infection_month['new_death'] = infection_month.groupby(['statefips'])['cum_deaths'].diff()

infection_day = infection_day.sort_values(by = ['statefips', 'month', 'day']).reset_index(drop=True)
infection_day['new_cases'] = infection_day.groupby(['statefips'])['cum_cases'].diff()
infection_day['new_death'] = infection_day.groupby(['statefips'])['cum_deaths'].diff()

# Rename and cleean dataset
infection_day = infection_day.drop(columns = ['day_max'])
infection_month = infection_month.drop(columns = ['day_max', 'day', 'date', 'week'])

infection_month.head()
```

### <font color="#6C8255">3.1.5. COVID-19 Policy Measure: OxCGT (Blavatnik School of Government, U of Oxford)</font>
The dataset tracks 18 government policy responses in three different policy areas. We kept only containment related policies (i.e., 8 policies `school`, `work`, `publicEvent`, `gathering`, `publicTrans`, `stayHome`, `interMove`, `travelControl`) and three indices which measures the degree for each area - containment, economic support and overall index (`stringency`, `govtResponse`, `econSupport`) in daily basis.

* Replace `.` as `np.nan`
* Aggregated the data into `state x month` level, which will allow us to analyze the impact of Covid-19 in `monthly` level. (e.g., If the given `state x month` has the value `0.45` in `school` column, then it implies that school was closed for `45%` of the month in a given state.)

```python
policy_raw = pd.read_csv("../input/uscovidpolicy/data/OxCGRT_US_latest.csv")
policy_day = policy_raw[['RegionCode', 'Date', 'C1_School closing', 'C2_Workplace closing', \
                    'C3_Cancel public events', 'C4_Restrictions on gatherings', \
                    'C5_Close public transport', 'C6_Stay at home requirements', \
                    'C7_Restrictions on internal movement', 'C8_International travel controls', \
                    'StringencyIndex']]

# Drop national level observation
policy_day = policy_day[policy_day['RegionCode'].notna()].reset_index(drop=True)

# Rename columns
policy_day.columns = ['st', 'date', 'school', 'work', 'publicEvent', 'gathering', \
                  'publicTrans', 'stayHome', 'interMove', 'travelControl', \
                  'stringency']

# Generate variables (date)
policy_day['st'] = policy_day['st'].apply(lambda x: str(x)[3:5])
policy_day['year'] = policy_day['date'].apply(lambda x: int(str(x)[:4]))
policy_day['month'] = policy_day['date'].apply(lambda x: int(str(x)[4:6]))
policy_day['day'] = policy_day['date'].apply(lambda x: int(str(x)[6:]))
policy_day['date'] = pd.to_datetime(policy_day['date'], format='%Y%m%d', errors='coerce')
policy_day['week'] = policy_day['date'].dt.week

# Convert flag variable into dummy
for var in policy_day.columns[2:10]:
#     policy[var + '_alter'] =  policy[var].apply(lambda x: 1 if x >= 1 else 0)
    policy_day[var] =  policy_day[var].apply(lambda x: 1 if x >= 2 else 0)

policy_day = policy_day.sort_values(by = ['st', 'year', 'month', 'day']).reset_index(drop=True)
policy_month = policy_day.groupby(['year', 'month', 'st'])\
            ['school', 'work', 'publicEvent', 'gathering', 'publicTrans', 'stayHome', 'interMove', \
             'travelControl', 'stringency'].mean().reset_index()

policy_month.head()
```

## <font color="#6C8255">4.2.1. Covid-19 and Related Policy</font>

### <font color="#6C8255">- New Covid-19 Confirmed Cases</font>
The following figures depict the relationship between new monthly Covid-19 confirmed cases with the monthly average `percent access` and `engagement index` respectively. The above two graphs show the `mean` and `median` of each variable of each decile group defined by `number of monthly average new Covid-19 cases`. Latter two scatter plots depict the relationship between our variable of interest and Covid-19. 

<br>

As shown in the figure, there is **a decreasing relationship between Covid-19 and both `percent access` and `engagement index`**. This is different from our general conjecture that Covid-19 might increase online learning products engagement.

This may due to the fact that there might be other factors which is correlated with the Covid-19 spread, and affects this product, we should further control other factors to rule out such a case, and to confirm this trend. To do so, we are going to implement a multivariate linear regression model in [section 5.1](#5.1.-Impact-of-COVID-19)

```python
plots = [['pct_access', 'Access Ratio (Unit: %)'], ['engagement_index', 'Engagement Index (Unit: page-loads per 1K students)']]

for p in plots:
    ### bin plot
    data = main_day.groupby(['lp_id'])['new_cases', p[0]].mean().reset_index()
    breaks = np.quantile(data['new_cases'], q=np.linspace(0, 1, 11))
    groups = data.groupby([pd.cut(data['new_cases'], bins=breaks, duplicates='drop')])[p[0]].agg(['mean','median','size'])
    fig, ax = plt.subplots(figsize=(22, 8))
    fig.suptitle(f'Distribution of {p[1]} by New Covid-19 Cases', fontsize=20)
    groups[["mean", "median"]].plot(kind="line", ax=ax)
    groups["size"].plot(kind="bar", ax=ax, rot=45, secondary_y=True,
                        color="grey", alpha=0.3, grid=True)
    ax.set(ylabel=p[1])
    ax.right_ax.set_ylabel("Observations in each bin")
    plt.show()

for p in plots:
    data = main_day.groupby(['lp_id'])['new_cases', p[0]].mean().reset_index()
    ### scatter plot
    sns.jointplot(x='new_cases', y=p[0], data=data, dropna=True, kind='reg',
                 height=6)
    plt.show()
```

### <font color="#6C8255">- School Closure Policy</font>
Here we plotted the relationship between the `school closure` policy and `student's engagement` toward the online learning products. Each observation is the annual average of each variable. Again, the upper two figures show the `mean` and `median` of each variable of each decile group defined by `percentage of days that the school was closed`. Latter two scatter plots depict the relationship between our variable of interest and the policy.

<br>

As shown in the figure, there is **an increasing relationship between the policy and both `percent access` and `engagement index`**, which is similar to our general conjecture that making students attend less school has increased students' engagement on online learning product. # <font color="#6C8255">4.3. Evolution of Online Learning Engagement: How trend changed over time?</font>
<font size=3> In this sub-section, we have investigated <b>how 'acess ratio' and 'engagment index' have been changed over time relative to 'Covid-19 spread' and 'school closure' policy</b> which seems to be the most direct factor that affects the online learning engagement behavior. </font>

<br>

**To exclude the weekday fixed effect** (i.e., Since student tends to engage less during weekends, there will be oscillations in graph making difficult to read the overall trends), we **aggregated up the data into weekly level**. Hence each row contains `weekly average engagement information`, and `weekly new Covid-19 confirmed cases`.

```python
# For x-axis labels
month={1: (4, ''), 2: (6,'Feb'), 3: (10,'Mar'),
      4: (15,'Apr'), 5: (19,'May'),6: (23, 'Jun'),
      7: (28,'Jul'), 8: (32,'Aug'), 9: (37,'Sep'),
      10: (41,'Oct'), 11: (45,'Nov'), 12: (50,'Dec')}
xticks = [i+4 for i in range(50)]
xtick_label = ['' for i in range(50)]
for i in range(12):
    xtick_label[int(month[i+1][0])-4] = month[i+1][1]

# Covid-19 trends
data2 = infection_day.groupby(['week']).agg({'new_cases': 'sum'}).reset_index()

#for policy dependent background
data3 = policy_day.groupby(['week']).agg({'school': 'mean'}).reset_index()
data2['background'] = data3['school'].apply(lambda x: 0 if x<0.50 else (1 if x<0.75 else 2)) #50%, 75%
data2['tmp'] = data2['background'].diff().apply(lambda x: 1 if (str(x) == 'nan') or (x != 0) else 0)
tmp = data2[data2['tmp'] == 1]
tmp['background_new'] = np.nan
for i in range(len(tmp)):
    tmp['background_new'].iloc[i] = i
data2 = pd.concat([data2, tmp['background_new']], axis=1).fillna(method='ffill')
ranges = data2.groupby('background_new').agg({'week': ['min', 'max'], 'background': 'mean'})
ranges['label'] = ranges.index
ranges['label'] = ranges['label'].apply(lambda x: 0 if x in [0, 1, 3] else 1)
labels_background = {0: '<50% school closed', 
                     1: '50-75% school closed',
                     2: '>=75% closure policy adopted'}
```

* Both indexes **varies along with semester**; which indicates that the index increased with a new semester, and decreased as vacation starts.

* Overall online engagement has been **increased compared to the first semester (i.e., Jan-Jun) with pandemic**; This implies that students become more familiar as time passes.
    
*  **Difficult to detect the direct relationship of spread of Covid-19 and school closure policy toward engagement** without controlling other factors.

```python
#national
sns.set(rc = {'figure.figsize':(15,8)}, style = 'white')
data = main_day.groupby(['week']).agg({'engagement_index': 'mean', 'pct_access': 'mean', 'new_cases': 'sum'}).reset_index()

#draw lineplot
fig, ax = plt.subplots(figsize = (15, 8))
ax = sns.lineplot(data=data, x='week', y='engagement_index', color="b", linewidth=2, label='Engagement Index')
ax2=ax.twinx()
sns.lineplot(data=data2, x='week', y='new_cases', color="r", linewidth=5, ax=ax2, label='Weekly New Covid-19 Confirmed')
for i, row in ranges.iterrows():
    ax.axvspan(xmin=row[('week','min')], xmax=row[('week','max')]+1, 
               facecolor='C'+row[('background', 'mean')].astype(int).astype(str), 
               alpha=0.3, label="_"*int(row[('label')]) + labels_background[int(row[('background', 'mean')])])
plt.grid()
ax.legend(loc='lower left')
ax.set_xticks(xticks)
ax.set_xticklabels(xtick_label)
ax.set_xlabel("Months")
ax.set_ylabel(r"Engagement Index (Unit: page-loads per 1K students)")
ax2.set_ylabel(r"New Covid-19 Confirmed Case (Unit: 1M)")
plt.title('Evolution of National Weekly Average Engagement Index', size=20)
plt.show()

#draw lineplot
fig, ax = plt.subplots(figsize = (15, 8))
ax = sns.lineplot(data=data, x='week', y='pct_access', color="b", linewidth=2, label='% Access')
ax2=ax.twinx()
sns.lineplot(data=data2, x='week', y='new_cases', color="r", linewidth=5, ax=ax2, label='Weekly New Covid-19 Confirmed')
for i, row in ranges.iterrows():
    ax.axvspan(xmin=row[('week','min')], xmax=row[('week','max')]+1, 
               facecolor='C'+row[('background', 'mean')].astype(int).astype(str), 
               alpha=0.3, label="_"*int(row[('label')]) + labels_background[int(row[('background', 'mean')])])
plt.grid()
ax.legend(loc='lower left')
ax.set_xticks(xticks)
ax.set_xticklabels(xtick_label)
ax.set_xlabel("Months")
ax.set_ylabel(r"Access Ratio")
ax2.set_ylabel(r"New Covid-19 Confirmed Case (Unit: 1M)")
plt.title('Evolution of National Weekly Average Acess Ratio', size=20)
plt.show()
```

# <font color="#6C8255">5.1. Impact of COVID-19</font>
<font size = 3><b>"What is the effect of the COVID-19 pandemic on online and distance learning?"</b></font>

## <font color="#6C8255">- Model</font>
In this subsection, we investigated whether the degree of Covid-19 have affected the online learning engagement of studnets, if so what extent. To do so, we implemetned following model as follows:

$$ln(Y_{i,s,p,m})=\alpha + \beta \cdot ln(Covid\ 19_{i,s,p,m}) + \mathbf{\theta\ X}_{i,s,p,m}+\iota_m+\kappa_s +\epsilon_{i,s,p,m}$$

where $Y_{i,s,p,m}$ indicates `access raito` and `engagement index` in distict `i`, state `s`, product `p`, and month `m`. 

* Our variable of interest is `Coivd-19` in this sub-section which estimates the impact of new monthly Covid-19 confirmed cases. As we have log-transformed both independent and dependent variables, **this can be interpreted as a 1% increase in new monthly confirmed cases increase a dependent variable by $\beta$ %** 

* We further included **socioeconomic factors** (e.g.,`share of black or hispanic studnets`, `share of free or reduced priced lunch program eligible students`, `totel expenditure on education per pupil`, and `locale`), **product characteristics** (e.g., `function of products`, `sector` as a vector $\mathbf{X}_{i,s,p,m}$. 

* Finally, $\iota_m$ and $\kappa_s$ capture month-, state-fixed effects to rule out unobserved characteristics within month and state.

* Standard errors are **clustered in state-level**, which considers the heteroskedasticity within a state cluster. Furthermore, as each observation is considered equally even each row represents districts with different size, we, therefore, **weighted each observation by the population of the state** where the district is belonged to.

## <font color="#6C8255">- Findings</font>

* As shown in the table, the new monthly Covid-19 infection case is **positively correlated with student's engagement behavior** unlike our previous graphical analysis in [section 4.2.1](#--New-Covid-19-Confirmed-Cases). 

* However, it is worth noting that it is **not statistically significant on conventional level**. This implies that Covid-19 itself has **high probability that there is no impact on online learning engagement on average**.

```python
result = list()

for y in ['ln_pct_access', 'ln_engagement_index']:
    result.append(sm.OLS.from_formula(f'{y} ~ ln_new_cases + C(locale) + C(blkhis_ratio) + C(freereduced_ratio) + C(pp_totexp) + C(mainfunction) + C(subfunction) + work + stayHome + stringency + corporate + higher + prek12 + C(month) + C(st)',
                  data=data, missing='drop').fit(cov_type='cluster', cov_kwds={'groups': data['st']}))
```

```python
regressors = ['ln_new_cases', 'N', 'R2']

table = summary_col([result[0], result[1]], float_format='%.3f', stars=True, 
                  model_names=['% Access\n(1)','Engagement\n(2)'],
                  info_dict={'N':lambda x: "{0:d}".format(int(x.nobs)),
                             'R2':lambda x: "{:.2f}".format(x.rsquared)},
                  regressor_order=regressors)
table.tables[0]['regressor'] = [i for sublist in [[j]*2 for j in table.tables[0].index[0::2]] for i in sublist]
table.tables[0] = table.tables[0][table.tables[0]['regressor'].isin(regressors)]
del table.tables[0]['regressor']
print(table)
```

# <font color="#6C8255">5.2. Impact of COVID-19 Policy</font>
<font size=3><b>"Do certain state interventions, practices or policies correlate with the increase or decrease online engagement?"</b></font>

## <font color="#6C8255">- Model</font>
In this subsection, we further analyzed whether the degree of Covid-19 and **policy related to pandemic** have affected the online learning engagement of students, if so what extent. We added several control variables -- `School`, `work`, `stayHome`, and `Stringency` on the equation described above.

<br>

$$ln(Y_{i,s,p,m})=\alpha + \beta \cdot ln(Covid\ 19_{i,s,p,m}) + \gamma\cdot School\ Closure_{i,s,p,m} +\mathbf{\theta\ X}_{i,s,p,m}+\iota_m+\kappa_s+\epsilon_{i,s,p,m}$$

$$ln(Y_{i,s,p,m})=\alpha + \beta \cdot ln(Covid\ 19_{i,s,p,m}) + \gamma\cdot School\ Closure_{i,s,p,m} + \eta\cdot School\ Closure_{i,s,p,m} \times ln(Covid\ 19_{i,s,p,m})+\mathbf{\theta\ X}_{i,s,p,m}+\iota_m+\kappa_s+\epsilon_{i,s,p,m}$$

<br>

* Our variable of interest in this model is `School Closure` which measures the share of closed dates of school among a given month. **This can be interpreted as closing school completely for an entire month would increase a dependent variable by $\gamma \times 100$ %**, while $\eta$ estimates the heterogeneous impact of policy by Covid-19 spread.

* We further included **log of `Coivd-19`**, **socioeconomic factors** (e.g.,`share of black or Hispanic students`, `share of free or reduced-priced lunch program eligible students`, `total expenditure on education per pupil`, and `locale`), and **product characteristics** (e.g., `function of products`, `sector` as a vector $\mathbf{X}_{i,s,p,m}$ as the previous model. 

* Finally, $\iota_m$ and $\kappa_s$ capture month-, state-fixed effects to rule out unobserved characteristics within month and state.

* Standard errors are **clustered in state-level**, and **weighted each observation by the population of the state** where the district is belonged to, as previous analysis.

## <font color="#6C8255">- Findings</font>

Odd columns show the results without interaction term while even columns include interaction of Covid-19 and school closure policy. 

* In columns (1) and (3), where there is no interaction term, both **`Covid-19` and `School Closure` has positively correlated with online engagement**, although it is still not statistically significant at the conventional level, except for `School Closure` in column (3). This implies that **closing school for a whole month would increase 21.9% more engagement on average**.

* Remaining columns (2), (4), we introduced **the interaction term between `Covid-19` and `school closure` to capture whether the impact varies along other impact** (i.e., check whether impact of school closure varies if the spread of Covid-19 differs). By doing so, the coefficient of `Covid-19` and `school closure` become negative and statistically significant on conventional level. However this does not necesaily implies that Covid-19 decreased student's engagment, rather we should also take into account the additional term - `Covid-19 x School Closure` which pulls out the interaction impact when both variable appears together by adding all in following manner:
    $$Impact = \beta \cdot ln(Covid\ cases_{i,s,p,m}) + \gamma\cdot School\ Closure_{i,s,p,m} + \eta\cdot School\ Closure_{i,s,p,m} \times ln(Covid\ cases_{i,s,p,m})$$

* The coefficient for the interaction term indicates that **additional 1% increase in new monthly `Covid-19` cases would increase the impact of `school closure` by 0.11%** and this is **statistically significant at 1% level**, which clearly shows that **the impact of `school closure` would be increased as more `Covid-19` case appeases.**

```python
result = list()

for y in ['ln_pct_access', 'ln_engagement_index']:
    result.append(sm.OLS.from_formula(y + ' ~ ln_new_cases + school + C(locale) + C(blkhis_ratio) + C(freereduced_ratio) + C(pp_totexp)+  C(subfunction) + work + stayHome + stringency + corporate + higher + prek12 + C(month) + C(st)',
                  data=data, missing='drop').fit(cov_type='cluster', cov_kwds={'groups': data['st']}))

    result.append(sm.OLS.from_formula(y + ' ~ ln_new_cases + school + ln_new_cases * school + C(locale) + C(blkhis_ratio) + C(freereduced_ratio) + C(pp_totexp)+ C(subfunction) + work + stayHome + stringency + corporate + higher + prek12 + C(month) + C(st)',
                  data=data, missing='drop').fit(cov_type='cluster', cov_kwds={'groups': data['st']}))
```

```python
regressors = ['ln_new_cases', 'school', 'ln_new_cases:school', 'N', 'R2']

table = summary_col([result[0], result[1], result[2], result[3]], float_format='%.3f', stars=True, 
                  model_names=['% Access\n(1)','% Access\n(2)','Engagement\n(3)', 'Engagement\n(4)'],
                  info_dict={'N':lambda x: "{0:d}".format(int(x.nobs)),
                             'R2':lambda x: "{:.2f}".format(x.rsquared)},
                  regressor_order=regressors)
table.tables[0]['regressor'] = [i for sublist in [[j]*2 for j in table.tables[0].index[0::2]] for i in sublist]
table.tables[0] = table.tables[0][table.tables[0]['regressor'].isin(regressors)]
del table.tables[0]['regressor']
print(table)
```

# <font color="#6C8255">5.4. Heterogeneity of socio-economic factor</font>
<font size=3><b>"How does student engagement with online learning platforms relate to different geography?"</b></font>

## <font color="#6C8255">- Model</font>
We investigated **whether the degree of Covid-19 and policy related to pandemic affects differently by each socio-economic group** by interacting dummy variables for a socioeconomic group with $Covid-19 \times School\ Closure$. This coefficient would capture the difference between the given group and the baseline group (i.e., omitted group).

<br>

$$ln(Y_{i,s,p,m})=\alpha + \beta \cdot ln(Covid\ 19_{i,s,p,m}) + \gamma\cdot School\ Closure_{i,s,p,m} + \Sigma_{j\in Group}\gamma_j\cdot School\ Closure_{i,s,p,m}\times 1(i \in Group_j) +\mathbf{\theta\ X}_{i,s,p,m}+\iota_m+\kappa_s+\epsilon_{i,s,p,m}$$

$$ln(Y_{i,s,p,m})=\alpha + \beta \cdot ln(Covid\ 19_{i,s,p,m}) + \gamma\cdot School\ Closure_{i,s,p,m} + \Sigma_{j\in Group}\gamma_j\cdot School\ Closure_{i,s,p,m}\times 1(i \in Group_j) + \theta\cdot School\ Closure_{i,s,p,m} \times ln(Covid\ 19_{i,s,p,m}) + \Sigma_{j\in Group}\theta_j\cdot School\ Closure_{i,s,p,m} \times ln(Covid\ 19_{i,s,p,m}) \times 1(i \in Group_j) +\mathbf{\theta\ X}_{i,s,p,m}+\iota_m+\kappa_s + \epsilon_{i,s,p,m}$$

<br>

* Our variable of interest is $ School\ Closure\ \times\ 1(i\in Group_j)$ which captures the different impact of `school closure` policy relative to the base group. **This can be interpreted as closing school completely for an entire month increase dependent variable by $\gamma_j \times 100$ % more relative to the base group**

* While, $ School\ Closure\times ln(Covid-19) \times 1(i \in Group_j) $ estimates **the heterogeneous impact of the policy by Covid-19, realative to the ommitted group.**

* We further included **socioeconomic factors** (e.g.,`share of black or Hispanic students`, `share of free or reduced-priced lunch program eligible students`, `total expenditure on education per pupil`, and `locale`), **product characteristics** (e.g., `function of products`, `sector` as a vector `$\mathbf{X}$'. 

* Finally, $\iota_m$ and $\kappa_s$ capture month-, state-fixed effects to rule out unobserved characteristics within month and state.

* Standard errors are **clustered in state-level**, and **weighted each observation by the population of the state** where the district is belonged to, as previous analysis.

## <font color="#6C8255">- Results</font>
Similar to our previous results, **`Covid-19` itself is negatively correlated with online engagement**, and **`school closure` is positively correlated with overall online learning engagement behavior**. This implies that online engagement increases when students cannot go to school.

In columns (2) and (4), we introduced the heterogeneous interaction term between `Covid-19` and `school closure`. However, the variable of interest (i.e., `Covid-19` and `School Closure`) does not change its direction, which **guarantees the robustness of our model**. Followings are the overall results by each variable:

* **Ethnicity**: **Composition of ethnicity does not affect the student's online learning participation** unless the district is consists of the majority (over 80%) of black and Hispanic students. However, the impact is substantial in the district with over 80% are black and Hispanic students.

* **Economic Status**: Districts with **more students who are economically disadvantaged (proxied by the eligibility of free lunch program) tend to engage less than those who are not**. However, the trend is reversed for districts with the majority (over 80%) are eligible for the free lunch program.

* **Education Expenditure**: Unlike our previous figures, there is **no impact of getting more total expenditure on education per pupil on online learning behavior**. # <font color="#6C8255">5.5. Heterogenious Time trend along Products</font>
<font size=3><b>"How does student engagement with different types of education technology change over the course of the pandemic?"</b></font>

## <font color="#6C8255">- Model</font>
We investigated **whether the degree of Covid-19 and policy related to pandemic affects differently by each product type** by interacting dummy variables for product type with $School\ Closure$. 

Furthermore, we included product type-specific linear time trends to see **whether there are different trends along with each product type** would capture the slope difference between the given product type and the baseline group (i.e., omitted group).

<br>

$$ln(Y_{i,s,p,m})=\alpha + \beta \cdot ln(Covid\ 19_{i,s,p,m}) + \gamma\cdot School\ Closure_{i,s,p,m} + \Sigma_{j\in Product_j}\gamma_j\cdot School\ Closure_{i,s,p,m}\times 1(i \in Product_j) +\mathbf{\theta\ X}_{i,s,p,m}+\iota_m+\kappa_s+\epsilon_{i,s,p,m}$$

$$ln(Y_{i,s,p,m})=\alpha + \beta \cdot ln(Covid\ 19_{i,s,p,m}) + \gamma\cdot School\ Closure_{i,s,p,m} + \Sigma_{j\in Product_j}\gamma_j\cdot School\ Closure_{i,s,p,m}\times 1(i \in Product_j) + \theta_j \cdot 1(i \in Product_j) \times m +\mathbf{\theta\ X}_{i,s,p,m}+\iota_m+\kappa_s + \epsilon_{i,s,p,m}$$

<br>

* Our variable of interest is $ School\ Closure_{i,s,p,m}\times 1(i \in Product_j) $ which captures the different impact of `school closure` policy relative to the base group. **This can be interpreted as closing school completely for an entire month increase dependent variable by $\gamma_j \times 100$ % more relative to the base group**

* While, $ \theta_j $ estimates **the heterogeneous linear time trend of each product type relative to the omitted group.**

* We further included **socioeconomic factors** (e.g.,`share of black or Hispanic students`, `share of free or reduced-priced lunch program eligible students`, `total expenditure on education per pupil`, and `locale`), **product characteristics** (e.g., `function of products`, `sector` as a vector `$\mathbf{X}$'. 

* Finally, $\iota_m$ and $\kappa_s$ capture month-, state-fixed effects to rule out unobserved characteristics within month and state.

* Standard errors are **clustered in state-level**, and **weighted each observation by the population of the state** where the district is belonged to, as previous analysis.

## <font color="#6C8255">- Findings</font>

Odd columns show the results without linear trend for each product type while even columns include linear trend for each product type. **Baseline group is the Learning and Curriculum product type.**

* In column (1) and (3), where there is no product type-specific linear trend term, both **`Covid-19` and `School Closure` has positively correlated with online engagement**, even it is still not statistically significant at the conventional level. 'School and Districts Operation' products such as Google Classroom tend to be used more on average, but this gets decreased as a school has been closed by 10.8% in access ratio. On the other hand, 'Classroom Management Tools' such as 'Zoom' and 'Google Hangout' are likely to have lower usage on average but it increases when school is closed by 12.6% in access ratio and 65.7&.

* Remaining columns (2), (4), we introduced **product-specific time trends to capture whether each product has different time trends along as pandemic progress**. Even we include such variables, the result holds constant which implies the model is robust. One thing to note is that 'School and Districts Operation' products tend to decrease most fast compared to the baseline group (i.e., Learning and Curriculum product)

```python
result = list()

for y in ['ln_pct_access', 'ln_engagement_index']:
    result.append(sm.OLS.from_formula(f'{y} ~ ln_new_cases + school + C(mainfunction) * school + C(locale) + C(blkhis_ratio) + C(freereduced_ratio) + C(pp_totexp)+ C(mainfunction) + work + stayHome + stringency + corporate + higher + prek12 + C(month) + C(st)',
                  data=data,  missing='drop').fit(cov_type='cluster', cov_kwds={'groups': data['st']}))

    result.append(sm.OLS.from_formula(f'{y} ~ ln_new_cases + school + C(mainfunction) * school + C(locale) + C(blkhis_ratio) + C(freereduced_ratio) + C(pp_totexp)+ C(mainfunction)*month + work + stayHome + stringency + corporate + higher + prek12 + C(month) + C(st)',
                  data=data,  missing='drop').fit(cov_type='cluster', cov_kwds={'groups': data['st']}))
```

```python
regressors = ['ln_new_cases', 'school', 'C(mainfunction)[T.CM]', 'C(mainfunction)[T.SDO]', 
              'C(mainfunction)[T.LC/CM/SDO]', 'C(mainfunction)[T.CM]:school', 'C(mainfunction)[T.SDO]:school', 
              'C(mainfunction)[T.LC/CM/SDO]:school', 'C(mainfunction)[T.CM]:month', 'C(mainfunction)[T.SDO]:month', 
              'C(mainfunction)[T.LC/CM/SDO]:month', 'N', 'R2']

table = summary_col([result[0], result[1], result[2], result[3]], float_format='%.3f', stars=True, 
                    model_names=['% Access\n(1)','% Access\nLinear Trend\n(2)', 'Engagement\n(3)', 'Engagement\nLinear Trend\n(4)'],
                  info_dict={'N':lambda x: "{0:d}".format(int(x.nobs)),
                             'R2':lambda x: "{:.2f}".format(x.rsquared)},
                  regressor_order=regressors)
table.tables[0]['regressor'] = [i for sublist in [[j]*2 for j in table.tables[0].index[0::2]] for i in sublist]
table.tables[0] = table.tables[0][table.tables[0]['regressor'].isin(regressors)]
del table.tables[0]['regressor']
print(table)
```
