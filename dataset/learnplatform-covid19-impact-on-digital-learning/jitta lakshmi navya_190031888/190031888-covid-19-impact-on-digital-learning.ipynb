```python
import pandas as pd
# load DISTRICT data
district_data= pd.read_csv("/kaggle/input/learnplatform-covid19-impact-on-digital-learning/districts_info.csv")

# load PRODUCT data
product_data= pd.read_csv("/kaggle/input/learnplatform-covid19-impact-on-digital-learning/products_info.csv")


# DISTRICT data table
fig= go.Figure(data=[go.Table(header= dict(values= list(district_data.columns),
               fill_color= 'paleturquoise',
               align= 'left'),
               cells= dict(values= [district_data['district_id'],district_data['state'],
               district_data['locale'],district_data['pct_black/hispanic'],
               district_data['pct_free/reduced'],district_data['county_connections_ratio'],
               district_data['pp_total_raw']],
               fill_color= [['white', 'lavender']*len(district_data)],
#              fill_color= 'lavender',
               align= 'left'))])
fig.show()


# PRODUCT data table
fig= go.Figure(data=[go.Table(header= dict(values= list(product_data.columns),
               fill_color= 'paleturquoise',
               align= 'left'),
               cells= dict(values= [product_data['LP ID'],product_data['URL'],
               product_data['Product Name'],product_data['Provider/Company Name'],
               product_data['Sector(s)'],product_data['Primary Essential Function']],
               fill_color= [['white', 'lavender']*len(product_data)],
               align= 'left'))])
fig.show()
```

```python
# ENGAGEMENT data
all_csv= []
for district in district_data["district_id"]:
    df= pd.read_csv(f'/kaggle/input/learnplatform-covid19-impact-on-digital-learning/engagement_data/{district}.csv')
    df.insert(0, 'district_id', district)
    all_csv.append(df)
engagement_data= pd.concat(all_csv)
engagement_data.head()

# # data table
# fig= go.Figure(data=[go.Table(header= dict(values= list(engagement_data.columns),
#                fill_color= 'paleturquoise',
#                align= 'left'),
#                cells= dict(values= [engagement_data['district_id'],engagement_data['time'],
#                engagement_data['lp_id'],engagement_data['pct_access'],engagement_data['engagement_index']],
#                fill_color= [['white', 'lavender']*len(engagement_data)],
# #              fill_color= 'lavender',
#                align= 'left'))])
# fig.show()
# print(f'engagement_data has {engagement_data.shape[0]} rows and {engagement_data.shape[1]} columns')
```

```python
# checking missing values
print('-'*45)
print("percentage of missing values in DISTRICT DATA")
print('-'*45)
print(district_data.isnull().sum()/len(district_data)*100)
print('-'*45)
print("percentage of missing values in PRODUCT DATA")
print('-'*45)
print(product_data.isnull().sum()/len(product_data)*100)
print('-'*45)
print("percentage of missing values in ENGAGEMENT DATA")
print('-'*45)
print(engagement_data.isnull().sum()/len(engagement_data)*100)

# dropping rows with missing values
district_data.dropna(subset= ['state'],axis=0,inplace= True)

# also dropping "pp_total_raw" column as it has ~50% null values
district_data.drop(columns=["pp_total_raw"], axis=1, inplace= True)
```

```python
# states with number of school districts
a= district_data["state"].value_counts()
fig= px.scatter(a, x=a.index, y=a.values, size=a.values, color=a.index, hover_name= a.index, size_max=60, 
                title="States with number of school districts")
fig.update_layout()
fig.show()
```

```python
# locale with number of school districts
locale= district_data["locale"].value_counts()
fig= px.scatter(locale, x=locale.index, y=locale.values, size=locale.values, color=locale.index, hover_name= locale.index, size_max=60, 
                title="Locale with number of school districts")
fig.update_layout()
fig.show()
```

```python
state_locale=district_data.groupby("state")["locale"].value_counts().to_frame().rename(columns={"locale": "number of school districts"}).reset_index()
fig = px.treemap(state_locale, path=['state', 'locale'], values='number of school districts',color='number of school districts', 
                 color_continuous_scale='piyg',
                title=('State-Wise School District Distribution in Locale'))
fig.update_layout(title_x=0.5)
                
fig.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
engagement_data.rename(columns={"lp_id": "LP ID"}, inplace=True)
merged=pd.merge(engagement_data, product_data, on= "LP ID")
m=merged.groupby("Product Name")["pct_access"].mean().sort_values(ascending=False).head(10)

engagement_data.rename(columns={"lp_id": "LP ID"}, inplace=True)
merged=pd.merge(engagement_data, product_data, on= "LP ID")
n=merged.groupby("Product Name")["engagement_index"].sum().sort_values(ascending=False).head(10)

# plot
plt.figure(figsize=(15,4))

plt.subplot(121)
plt.bar(m.index, m.values, color=["#6930c3","#5e60ce","#0096c7","#48cae4","#ade8f4","#ff7f51","#ff9b54","#ffbf69"])
plt.xlabel('Product Name')
plt.xticks(rotation=90)
plt.ylabel('Mean percentage of students')
plt.title("With atleast one-page load event")

plt.subplot(122)
plt.bar(n.index, n.values, color=["#4f000b","#720026","#ce4257","#ff7f51","#ff9b54"])
plt.xlabel('Product Name')
plt.xticks(rotation=90)
plt.ylabel('Page-load per 1000 students')
plt.title("With number of page-load per 1000 students")
```

```python
class_doc=merged[(merged["Product Name"]=="Google Classroom")|(merged["Product Name"]=="Google Docs")]
pct=class_doc.groupby(["time", "Product Name"])["pct_access"].mean().to_frame().reset_index()
eng=class_doc.groupby(["time", "Product Name"])["engagement_index"].sum().to_frame().reset_index()
# plot
fig = px.line(pct, x="time", y="pct_access", color='Product Name',title='Percentage of students with atleast one-page load event on a given day',
              template="ggplot2", width=800, height=400)
fig.show()

fig = px.line(eng, x="time", y="engagement_index",title='Sum of number of page-load per 1000 students on a given day', color='Product Name',
              template="seaborn", width=800, height=400)
fig.show()
```

```python
product_data["Basic function"]= product_data["Primary Essential Function"].str.extract('(LC|CM|SDO)')
function_dict ={'LC' : 'Learning & Curriculum', 'CM' : 'Classroom Management', 'SDO' : 'School & District Operations'}
product_data['Basic_function_full_form'] = product_data["Basic function"].map(function_dict)
product_data["Sub Basic function"]= product_data["Primary Essential Function"].str.lstrip('(LCCMSDO')
product_data = product_data.replace('-', '', regex=True)

colors = ['#AFEEEE', 'hotpink', '800080']
fig = go.Figure(data=[go.Pie(labels=product_data["Basic_function_full_form"].value_counts().index, values=product_data["Basic_function_full_form"].value_counts().values)])
fig.update_traces(hoverinfo='label+value', textinfo='label+percent', textfont_size=12,
                  marker=dict(colors=colors, line=dict(color='#808080', width=1)))
fig.show()
```

```python
LC=product_data[product_data["Basic function"]=="LC"]

colors = ["#aa4465","#dd2d4a","#f26a8d","#f49cbb","#ffcbf2","#e2afff","#ff86c8","#ffa3a5",
          "#20B2AA","#00FFFF","#40E0D0", "#7FFFD4","#AFEEEE","#008080", "#B0E0E6", "#3CB371", "#66CDAA"]
fig = go.Figure(data=[go.Pie(labels=LC["Sub Basic function"].value_counts().index, values=LC["Sub Basic function"].value_counts().values, 
                             hole=.75)])
fig.update_traces(hoverinfo='label+value', textinfo='percent', textfont_size=12,
                  marker=dict(colors=colors, line=dict(color='#808080', width=1)))
fig.update_layout(
    title_text="Sub Basic function Distribution",
    annotations=[dict(text='Learning & Curriculum', x=0.50, y=0.5, font_size=12, showarrow=False)])
fig.show()
```
