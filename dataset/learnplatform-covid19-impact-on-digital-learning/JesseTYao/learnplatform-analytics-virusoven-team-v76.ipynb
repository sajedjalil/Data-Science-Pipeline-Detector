<h1> 1. Executive Summary </h1>
<body>
<p>We are the team of college student, highschooler, and middle schooler under the guidance of Prof. Dong. Our first hand experience in the pandemic motivated us to participate in this competition. One of our members, whose school has a 38% in free/reduced lunch programs, experienced an an “absolute tragedy” at the start of the pandemic. There were no zoom meetings and zero lectures. It was essentially just a list of assignments from google classroom. When zoom meetings finally came about, his school had an average attendance rate of 43%. </p>
 
<p>To gain a better understanding of the effect of COVID-19 on online learning, we split the data into four parts: Before COVID (January - March), Beginning of COVID (March -June), summer (July-August), New School Year (September-December).</p>
<p>Considering the specialty of the summer time period (July-August), in our analysis, we focus on the other three time periods. </p>
<p>Our initial data exploration discovers a striking decline of online learning access during the initial period of the pandemic (March – June), which was later recovered after the new school year starts. Although it is consistent with one of our team members’ personal experiences during that time period, it attracts our attention to this special time period when the break of the pandemic initially strikes. </p>
<p>Throughout this study, we focus on two variables measuring online activities at the district level, including</p>
<ul>
     <li>Number of students accessing the online learning platform, which is measured as the percentage of students who accessed at least one page per 1000 students in the school district. </li>
     <li>Total number of pages accessed on the learning platform, which is measured as total engagement index. </li>
</ul>
<ul>
    <li>Heterogeneous impacts of the pandemic through data visualization</li>
    <li>Difference in difference analysis</li>
    <li>Causal analysis evaluating the internet access policy implemented by Connecticut</li>
    </ul>
<h2>Our findings can be summarized as the following: </h2>
    <h3>First, Initial Response to the Pandemic hit</h3>
    <p>When pandemic first hit, across all school districts, we see increases in total access to the learning platform, and further increases when the new school year starts. However, in most of the school districts, the number of students accessing the online platform was actually decreased at the onset of the pandemic, although it climbed back during the new school year. The decrease is especially dramatic among the districts with lower economic resources, as measured by the following three metrics: </p>
    <ol type='a'>
        <li>percentage of students with free or reduced lunch</li>
        <li>total expenditure per pupil</li>
        <li>percentage of students that are Hispanic or black</li>
    </ol>
    <p>Further investigations shows that the number of products accessed at the beginning of COVID also decreased except for the wealthier school districts. While this may be the result of two possible action changes, including：
        <ul>
            <li>Due to a financial hit, some school district may have reduced the number of products accessible to their students. Considering the short response time period, this is unlikely.</li>
            <li>Additional readings on <a href = 'https://www.theguardian.com/world/2020/apr/13/coronavirus-covid-19-exposes-cracks-us-digital-divide'>news articles</a> reveal that while internet access can be a luxury to some families, they were leveraging the free offering from some public places, like café . The shutdown of those public places due to the break of the pandemic dramatically hindered those families in their access to internet.</li>
    </ul>
    </p>
    <h3>Second, State Level Case Study</h3>
    <p>Aggregating the data into State level, we found that among the top four states covered in the data, Connecticut stands out as the one being hit the least. <a href = 'https://portal.ct.gov/Coronavirus/Information-For/Internet-Access'> Connecticut's COIVD-19 Response page</a> indicates that this may be due to an emergency internet access assistance program implemented at the state level. Consistent with our conjecture above, the internet accessibility can be a key driver to the student’s access to the learning platform. </p>
    <h3>Third, Weekend vs. Weekdays</h3>
    <p>An interesting finding from our time series analysis reveals that although the difference in access to the platform during the weekdays vs. weekends was much higher before the pandemic hits, that difference is significantly decreased after. This may be the result of the additional flexibility brought by the pandemic, and the reduced outdoor opportunities during the weekend due to the shutdown orders and concerns of big crowds. </p>
    <h3>Forth, Difference in Difference (DID) Analysis</h3>
    <p>Leveraging the DID approach, we measure the differential impact on school district with different economic situations, before and after the pandemic. We find that:</p>
 <ol>
    <li>School districts with higher free lunch rate (poorer schools) have lower product access rate during in pandemic period than school districts with lower free lunch rate (more resourceful schools). This may have many different causes:
        <ul>
        <li>Schools from poorer regions are less prepared to move online, due to the school's lack of IT resources and the teacher's lack of IT pedagogical skills.
        </li>
        <li>Students from poorer regions are less prepared to move online, due to the lack of resources such as<a href："https://blogs.microsoft.com/on-the-issues/2019/07/31/signs-of-progress-on-broadband-mapping-but-more-work-still-to-do/">broadband</a>,and <a href:"https://www.frontiersin.org/articles/10.3389/fpsyg.2020.592670/full">self motivation</a>.
        </li>
        </ul>
    </li>
    <li>School districts with the highest free lunch rate and highest black hispanic ratio have the lowest product access rate. This means that in terms of regions with the highest free lunch rate, those whom have a higher black hispanic ratio perform worse.</li>
    
</ol>
    <h3>Fifth, Causal Analysis</h3>
    <p>Based on the previous findings, the unique help to internet access offered by the state of Connecticut seems to play an important role in reducing the negative impact of the pandemic hit. To understand the causal influence of such a policy, we employed the approach of synthetic control approach to understand the causal influence of such special state level policy. Leveraging the free access of the <a href = 'https://pypi.org/project/pycausalimpact/'>Causal Impact package</a>, we found that the positive effect of pct_access_sum by Connecticut observed during the intervention period (the beginning of COVID) is statistically significant and unlikely to be due to random fluctuations.</p>

<h2>Suggestions for Education Policy</h2>
<p>We suggest that following policy change could significanly close the gap of the online learning engagement.</p>
<ol>
   <li>Equity investment to different district. Focus funding to school districts with less expenditure.</li>
    <li>Internet accessibility is a "Human Right". Increase support to provide internet access to families, especially in poor communities with high free/reduce lunch ratio.</li>
    <li>In coordination with increase support of internet access, schools in high free/reduce lunch ratio districts should also provide training in basic IT skills and training in digital citizenship. Communities in poor districts should acknowledge their digital rights.
</ol>

```python
class DayType(Enum):
  WEEKEND = "weekend"
  WEEKDAY = "weekday"
  ALL = "all"

def get_days(start, end, day_type = None):
  date_format = "%Y-%m-%d"
  a = datetime.strptime(start, date_format)
  b = datetime.strptime(end, date_format)
  total = None
  if (day_type == DayType.WEEKEND):
    total = np.busday_count( a.date(), b.date(), weekmask = [0,0,0,0,0,1,1] )
  elif (day_type == DayType.WEEKDAY):
    total = np.busday_count( a.date(), b.date(), weekmask = [1,1,1,1,1,0,0] )
  else:
    delta = b - a
    total = delta.days

  return total

def get_days_date(start, end, day_type = None):
  a = start
  b = end
  total = None
  if (day_type == DayType.WEEKEND):
    total = np.busday_count( a.date(), b.date(), weekmask = [0,0,0,0,0,1,1] )
  elif (day_type == DayType.WEEKDAY):
    total = np.busday_count( a.date(), b.date(), weekmask = [1,1,1,1,1,0,0] )
  else:
    delta = b - a
    total = delta.days

  return total
```

```python
def label_pandemic_date(state, date):
  # print(f"{state}. {type(state)}")
  if (state == 'nan' or state is np.nan or not state or not date or pd.isna(state)):
    return None;

  dt = date.to_pydatetime()
  
  if (dt >= SECOND_HALF_START ):
    return AFTER_P;
  elif (dt < SECOND_HALF_START and dt >= FIRST_HALF_END):
    return SUMMER;
  else:
    if state == 'District Of Columbia':
      state = 'District of Columbia'
    state_policy = policy_json[state]

    if (dt >= state_policy['CLSCHOOL']):
      return IN_P;
    else:
      return PRE_P

WEEKDAY = "Weekday"
WEEKEND = "Weekend"

def label_weekday(date):
  weekno = date.weekday()

  if weekno < 5:
      return WEEKDAY;
  else:  # 5 Sat, 6 Sun
      return WEEKEND
```

<h1> 3. Heterogeneous Impact of COVID-19 </h1>

<p>Although the break of the pandemic has unavoidable impact on students' online learning activities across the nation, the severity of its impacts vary dramatically across school districts. To understand the differential impacts, and its associations with the social-economic factors, we deploy a series of plots to visualize our findings.</p>
<p> In the following plots, we leverage three socio-economic variables from the school district data,  including Percentage of students who received reduced or free lunch, Percentage of black/hispanic students, Total expenditure per pupil.</p>
<p>The studies focused on the following three activity variables summarized from the engagement data, including pct_access_nunique (copy the current paragraph in the notebook)
pct_access_sum_avg(copy the current paragraph in the notebook)
engagement_index_sum_avg(copy the current paragraph in the notebook)</p>
<p>We use difference in difference for this analysis. The following process was conducted beforehand:</p>
<ol>
    <li>Labeled the engagement data by period of the pandemic. We divided the year 2020 into 4 periods:: 
        <ul>
            <li><b>Before pandemic (BP)</b>: From the begining of the year to the time when the state closed the school.</li>
            <li><b>In pandemic (IP)</b>: From the time when school closed to the Jun. 15th which is the begining of the summer.</li>
            <li><b>Summer</b>: From the Jun. 15th to Sept. 1st. This period has no significant activities, so we excluded this period.</li>
            <li><b>New School Year (NSY)</b>: From Sept. 1st to end of the year.</li>
        </ul> 
    </li>
    <li>We also want see the <b>difference between the weekday and weekend</b>, so we label the engagement data by weekday as well. </li>
    <li>Aggregate the engagement data by district_id, weekday, pandemic period, state. Calculate count, sum, unique, mean of pct_access and count, sum, mean of engagement_index.</li>
    <li>Because each period has a different amount of days, we calculated the daily average of pct_access_count, pct_access_sum, engagement_index_count and engagement_index_sum.</li>
    <li>Some categories in the <b>district data have close to no data</b>. For example, "pct_free/reduced" has very few districts in "\[0.8, 1\[" category. So, we have <b>combined some of these categories with others.</b></li>    
</ol>
<p>​
We researched the following engagement data variables: 'pct_access_nunique', 'pct_access_count','pct_access_count_avg', 'pct_access_sum_avg', and 'engagement_index_sum_avg'. For each of the data variables above, we sliced the data using "pct of free/reduced lunch", "pct black/hispanic", "per-pupil total expenditure", and "state". <b>We examined and compared the difference between each pandemic period based on these variables.</b></p>
<ul>
    <li><b>pct_access_nunique</b> represents how many unique products are in a particular district in a particular period.
```python
def divided_by_period_days(df, col, state_col, pan_col, wd_col = None):
  state = df[state_col]
  pandemic = df[pan_col]
  day_type = DayType.ALL
  if wd_col: 
    day_type = DayType.WEEKDAY if df[wd_col] == 'Weekday' else DayType.WEEKEND

  if ( (not isinstance(state, str)) or state == 'nan' or state is np.nan or not state or not pandemic or not df[col]):
    print(f"{df['district_id']}, {state_col} {pandemic} {df[col]}")
    return None;

  if state == 'District Of Columbia':
      state = 'District of Columbia'
  # print(f"{state}. {type(state)}")
  state_policy = policy_json[state]

  num_day = state_policy[pandemic][day_type]
  if (not num_day):
    print(f"{df['district_id']}, {state_col} {pandemic} {df[col]}  No num_day")
    return None;
  
  # print(f"{df['district_id']}, {state_col} {pandemic} {df[col]} {float(df[col]) / float(num_day)}")
  return float(df[col]) / float(num_day)
```

```python
def rename_category_value(df, col):
  prefix = ""
  if (col == 'pp_total_raw'):
    prefix = 'ptr'
  elif col == 'pct_black/hispanic':
    prefix = 'pbh'
  elif col == 'pct_free/reduced':
    prefix = 'pfr'
  elif col == 'county_connections_ratio':
    prefix = 'ccr'
  valstr = df[col]
  if ( (not isinstance(valstr, str)) or valstr == 'nan' or valstr is np.nan or not valstr ):
    return None;
  var = prefix + "_" + valstr.strip("[]0").replace("0","").replace(".","").replace(",","_").strip("_")
  return var
```

```python
# https://towardsdatascience.com/how-to-make-bar-and-hbar-charts-with-labels-using-matplotlib-b701ce70ba9c
def add_bat_label(bars, ax, labels):
  for idx, bar in enumerate(bars):
    height = bar.get_height()
    label_x_pos = bar.get_x() + bar.get_width() / 2
    ax.text(label_x_pos, height, s=f"{int(round(labels[idx]))}%", ha='center', va='bottom')
```

```python
def autolabel(ax, rects, data_labels):
    """Attach a text label above each bar in *rects*, displaying its height."""
    first = len(data_labels) == 0
    currentLbl = []
    # for rect in rects:
    for idx, rect in enumerate(rects):
        height = rect.get_height()
        currentLbl.append(height)
        if(not first):
          pre = data_labels[idx]
          cur = height
          ax.annotate(f"{(cur-pre)*100/pre:.0f}%",
                      xy=(rect.get_x() + rect.get_width() / 2, height),
                      xytext=(0, 3),  # 3 points vertical offset
                      textcoords="offset points",
                      ha='center', va='bottom')
    return currentLbl
```

<h1> 4. Difference in Difference (DID) Analysis </h1>

<p> At the end of the kernal our team decided to run a least squares regression analysis. The aim of running the regression does not attempt to make predictions of student engagement on learning platforms, but rather to better understand the relationship between the different parameters and student engagement on learning platforms. We hope to reensure that the variables that we identify as important influencers of the disparity between the different pct_access_sum across different states can be further supported. The variables incorporated in the model are as follows:
</p>
<ol>
    <li><b>pct_free/reduced:</b> Percentage of students in the districts eligible for free or reduced-price lunch.</li>
    <li><b>pandemic_stage:</b> The stage of pandemic as divided with our difference in difference method</li>
    <li><b>pct_black_hispanic:</b> Percentage of students in the districts identified as Black or Hispanic.</li>
    <li><b>week_day: </b>Type of day, since there is a lot of difference between the engagement during week days and the engagement during weekends.</li>
</ol>
<p>All of these input variables are set as dummy variables of different degrees. While for the output variable we can have pct_access_count and engagement_index_mean; which can be denoted with $Y_i$ for $i = 0,1$. We can suppose the relationship between the input variables and the output:
    $$Y_i = \alpha_0 + \beta_l L_i + \beta_t T_i + \beta_r R_i + \beta_d D_i + \epsilon$$
    </p>
<p>The notations are uppercased, because they are matrices of dummy variables with multiple classes. To make the notations comprehensible, let $L_i$ indicate pct_free/reduced lunch which has four dummy variables:</p>

$$
\begin{array}{cols} 
E[\beta_lL_i|l_{1i}=0,l_{2}=0,l_{3i}=0] & = & \text{pct_free/reduced [0.0-0.2]},\\ 
E[\beta_lL_i|l_{1i}=1,l_{2}=0,l_{3i}=0] & = & \text{pct_free/reduced [0.2-0.4]},\\
E[\beta_lL_i|l_{1i}=0,l_{2}=1,l_{3i}=0] & = & \text{pct_free/reduced [0.4-0.6]},\\
E[\beta_lL_i|l_{1i}=0,l_{2}=0,l_{3i}=1] & = & \text{pct_free/reduced [0.6-1.0]}.
\end{array}
$$
<p> Let $T_i$ be a time variable that indicates pandemic_stage:</p>
$$
\begin{array}{cols} 
E[\beta_tT_i|t_{1i}=0,t_{2}=0] & \mbox{=} & \text{before_COVID},\\ 
E[\beta_tT_i|t_{1i}=1,t_{2}=0] & \mbox{=} & \text{beginning_of_COVID},\\
E[\beta_tT_i|t_{1i}=0,t_{2}=1] & \mbox{=} & \text{new_school_year}.
\end{array}
$$
<p> Let $R_i$ indicate the race variable, pct_black_hispanic:</p>
$$
\begin{array}{cols} 
E[\beta_rR_i|r_{1i}=0,r_{2}=0] & \mbox{=} & \text{pct_black_hispanic [0.0-0.2]},\\ 
E[\beta_rR_i|r_{1i}=1,r_{2}=0] & \mbox{=} & \text{pct_black_hispanic [0.2-0.4]},\\
E[\beta_rR_i|r_{1i}=0,r_{2}=1] & \mbox{=} & \text{pct_black_hispanic [0.4-1.0]}.
\end{array}
$$
<p> Let $D_i$ indicate the day type variable differentiating weekday and weekend:</p>
$$
\begin{array}{cols} 
E[\beta_dD_i|d_{1i}=0] & \mbox{=} & \text{Week_Day},\\
E[\beta_dD_i|d_{1i}=1] & \mbox{=} & \text{Week_End}.
\end{array}
$$
<p> As a result, the baseline formula for the model when all dummies are zero is:
$$ 
\alpha_0  =  \text{pct_free/reduced [0.0-0.2]} + \text{pct_black_hispanic [0.0-0.2]} + \text{before_COVID} + \text{Week_Day}
$$

```python
# merging the three data sets districts_df, products_df and eng_all_df.

li = []
df = pd.merge(eng_all_df, districts_df[['district_id','state','pct_free/reduced','pct_black/hispanic']], how='inner', on = 'district_id')
df = pd.merge(df, products_df[['LP ID','primary_function_main','primary_function_sub']], how = 'right', right_on = 'LP ID', left_on = 'lp_id')
li.append(df)
reg_df = pd.concat(li)
del li

# dropping all the mssing values in pct_free/reduced
reg_df = reg_df.replace('Missing', np.nan)
reg_df =reg_df.dropna(subset=["pct_free/reduced"])

# defining functions to sort dataset
def get_one_period_dist_id(eng_df, query):
   period_df = eng_df.loc[query]
   agg_df = period_df.groupby(['district_id']).agg(
     {
      'pct_access': ['count', 'mean', 'sum'],
      'engagement_index':['count', 'mean','sum']
     }
   ).reset_index()
   agg_df.columns = ["_".join(a) for a in agg_df.columns.to_flat_index()]
   # print(agg_df.info())
   return agg_df

def combine_freelunch (row):
   if  row['pct_free/reduced'] in ['[0.6, 0.8[', '[0.8, 1[']:
      return '[0.6, 1['
   return row['pct_free/reduced']

def Diff_in_Diff_Categories(eng_df,day_type):
    jan_query = (eng_df['time'] >= Pre_Start) & (eng_df['time'] <= Pre_End)
    jan_df = get_one_period_dist_id(eng_df, jan_query)
    ls = []
    for i in range(jan_df.shape[0]):
        ls.append(0)
    ls = pd.DataFrame(ls).rename(columns = {0 : 'Pandemic_Stage'}, inplace = False)
    jan_df = pd.merge(jan_df, ls, left_index = True, right_index = True)
    del ls
    jan_df['pct_access_sum'] = jan_df["pct_access_sum"]/Pre_Days(day_type)
    jan_df['engagement_index_sum'] = jan_df["engagement_index_sum"]/Pre_Days(day_type)
    jan_df['pct_access_mean'] = jan_df["pct_access_mean"]
    jan_df['engagement_index_mean'] = jan_df["engagement_index_mean"]
    jan_df['pct_access_count'] = jan_df["pct_access_count"]/Pre_Days(day_type)
    jan_df['engagement_index_count'] = jan_df["engagement_index_count"]/Pre_Days(day_type)
    
    mar_query = (eng_df['time'] >= Pan_Start) & (eng_df['time'] <= Pan_End)
    mar_df = get_one_period_dist_id(eng_df, mar_query)
    ls = []
    for i in range(mar_df.shape[0]):
        ls.append(1)
    ls = pd.DataFrame(ls).rename(columns = {0 : 'Pandemic_Stage'}, inplace = False)
    mar_df = pd.merge(mar_df, ls, left_index = True, right_index = True)
    del ls
    mar_df['pct_access_sum'] = mar_df["pct_access_sum"]/Pan_Days(day_type)
    mar_df['engagement_index_sum'] = mar_df["engagement_index_sum"]/Pan_Days(day_type)
    mar_df['pct_access_mean'] = mar_df["pct_access_mean"]
    mar_df['engagement_index_mean'] = mar_df["engagement_index_mean"]
    mar_df['pct_access_count'] = mar_df["pct_access_count"]/Pre_Days(day_type)
    mar_df['engagement_index_count'] = mar_df["engagement_index_count"]/Pre_Days(day_type)
    
    sept_query = (eng_df['time'] >= Post_Start) & (eng_df['time'] <= Post_End)
    sept_df = get_one_period_dist_id(eng_df, sept_query)
    ls = []
    for i in range(sept_df.shape[0]):
        ls.append(2)
    ls = pd.DataFrame(ls).rename(columns = {0 : 'Pandemic_Stage'}, inplace = False)
    sept_df = pd.merge(sept_df, ls, left_index = True, right_index = True)
    del ls
    sept_df['pct_access_sum'] = sept_df["pct_access_sum"]/Post_Days(day_type)
    sept_df['engagement_index_sum'] = sept_df["engagement_index_sum"]/Post_Days(day_type)
    sept_df['pct_access_mean'] = sept_df["pct_access_mean"]
    sept_df['engagement_index_mean'] = sept_df["engagement_index_mean"]
    sept_df['pct_access_count'] = sept_df["pct_access_count"]/Pre_Days(day_type)
    sept_df['engagement_index_count'] = sept_df["engagement_index_count"]/Pre_Days(day_type)
    
    merged_df = pd.merge(jan_df, mar_df, how = 'outer')
    merged_df = pd.merge(merged_df, sept_df, how = 'outer')
    merged_df.rename({'district_id_': 'district_id'}, axis=1, inplace=True)
    merged_df = pd.merge(merged_df, districts_df, how = 'inner', on = 'district_id')
    
    return merged_df

def prim_func_merge(eng_df):
    weekend_query = eng_df['time'].dt.dayofweek.isin([5,6])
    weekday_query = eng_df['time'].dt.dayofweek.isin([0,1,2,3,4])
    weekday_eng_df = eng_df.loc[weekday_query]
    weekend_eng_df = eng_df.loc[weekend_query]
    weekday_merged_df = Diff_in_Diff_Categories(weekday_eng_df,'weekday')
    weekend_merged_df = Diff_in_Diff_Categories(weekend_eng_df,'weekend')
    
    ls = []
    for i in range(weekday_merged_df.shape[0]):
        ls.append(0)
    ls = pd.DataFrame(ls).rename(columns = {0 : 'Week_End'}, inplace = False)
    weekday_merged_df = pd.merge(weekday_merged_df, ls, left_index = True, right_index = True)
    del ls
    
    ls = []
    for i in range(weekend_merged_df.shape[0]):
        ls.append(1)
    ls = pd.DataFrame(ls).rename(columns = {0 : 'Week_End'}, inplace = False)
    weekend_merged_df = pd.merge(weekend_merged_df, ls, left_index = True, right_index = True)
    del ls
    
    allweek_merged_df = pd.merge(weekday_merged_df, weekend_merged_df, how = 'outer')

    return allweek_merged_df

#running merge functions

merge_reg_df = prim_func_merge(reg_df)
```

<h1> 5. Causal Analysis for State Help of Internet Access </h1>

<p>As reveled in the analysis in the above section, students in the state of Connecticut seems to be impacted less at the onset of the pandemic. We conjecture that the help offered by the state to ensure more residents can access internet from home play a critical role. To evaluate the potential causal impact of such a state level policy, we conduct the causal analysis using the approach of synthetic control. Leveraging the free Causal Impact courtesy of Google, we present our results here. </p>
<p>On March 14th, very early into the pandemic, the FCC launched the <b>Keep Americans Connected Pledge</b> to ensure that internet service providers would ensure connectivity during the pandemic by not terminating business or residential service for anyone who can’t pay their bill due to hardship caused by COVID-19. According to Connecticut's <a href:"https://portal.ct.gov/Coronavirus/Information-For/Internet-Access">official state website</a>, in addition to the <b>Keep Americans Connected Pledge</b>, several internet providers in CT took a step further by providing <b>free internet for low-income households</b> and providing 60 days free broadband services for students who have not already subscribed to cope with online education. These initiatives could be some of the reasons that explain the unique data pattern of CT during the initial period of the pandemic.</p>
    <p>In addition, Connecticut also houses a respectable broadband (25 megabits per second download and 3 megabits per second upload) availability and usage number in comparison to other states. According to <a href:"https://blogs.microsoft.com/on-the-issues/2019/07/31/signs-of-progress-on-broadband-mapping-but-more-work-still-to-do/">microsoft's research in 2019</a>, broadband is available to 93.5% of the people in US, yet only 52% of the people use the internet at broadband speed. Different counties in different states have disparate broadband usage percentage. For example, while the average broadband usage in California is 61.6%, the lowest of its counties is 23.6%. When these disparaties are common across most states, Connecticut stands out with an average bandwith of 52.6% and a lowest bandwith of 45.5%. The gap is at a low 7.1% in contrast to California's 38.0%.Under such circumstances broadband reachout programs in response to COVID-19 pandemic may benefit online education positively. Disadvantaged family's inability to pay expensive mobile data does seem to have a higher correlation with the digital divide.</p>
<p>The treatment group is consisted of a time series of pct_access_sum of Connecticut state. The control group consists  of time series pct_access_sum of California, Illinois, Washington, Wisconsin and Utah. Using these states as control variables does not necessary mean that these other states did not respond to the internet inequality as quickly as Connecticutt.

```python
Connecticut_df = reg_df[reg_df['state']=='Connecticut']
California_df = reg_df[reg_df['state']=='California']
Illinois_df = reg_df[reg_df['state']=='Illinois']
Utah_df = reg_df[reg_df['state']=='Utah']
Wisconsin_df = reg_df[reg_df['state']=='Wisconsin']
Washington_df = reg_df[reg_df['state']=='Washington']
def agg_state_data(state_df):
    state_agg_df = state_df.groupby(['time']).agg(
        {
        'pct_access':['count','mean','sum'],
        'engagement_index':['count','mean','sum']
        }
    )
    state_agg_df.columns = ["_".join(a) for a in state_agg_df.columns.to_flat_index()]
    return state_agg_df
Connecticut_agg_df = agg_state_data(Connecticut_df)
California_agg_df = agg_state_data(California_df)
Illinois_agg_df = agg_state_data(Illinois_df)
Utah_agg_df = agg_state_data(Utah_df)
Wisconsin_agg_df = agg_state_data(Wisconsin_df)
Washington_agg_df = agg_state_data(Washington_df)

data = Connecticut_agg_df['pct_access_sum']
data = pd.merge(data,California_agg_df['pct_access_sum'],left_index=True,right_index=True)
data = data.rename(columns = {'pct_access_sum_x':'Connecticut','pct_access_sum_y':'California'})
data = pd.merge(data,Illinois_agg_df['pct_access_sum'],left_index=True,right_index=True)
data = pd.merge(data,Utah_agg_df['pct_access_sum'],left_index=True,right_index=True)
data = data.rename(columns = {'pct_access_sum_x':'Illinois','pct_access_sum_y':'Utah'})
data = pd.merge(data,Wisconsin_agg_df['pct_access_sum'],left_index=True,right_index=True)
data = pd.merge(data,Washington_agg_df['pct_access_sum'],left_index=True,right_index=True)
data = data.rename(columns = {'pct_access_sum_x':'Wisconsin','pct_access_sum_y':'Washington'})
pre_period = ['2020-01-01','2020-03-15']
post_period = ['2020-03-16','2020-06-01']
end_period = ['2020-09-01','2020-12-31']
ci_Conn = causalimpact.CausalImpact(data, pre_period, post_period, nseasons=[{'period': 7, 'harmonics': 2}])
```
