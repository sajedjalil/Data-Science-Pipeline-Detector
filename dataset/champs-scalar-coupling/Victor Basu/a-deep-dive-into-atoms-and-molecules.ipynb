**This kernel is based on the kaggle competition [Predicting Molecular Properties](https://www.kaggle.com/c/champs-scalar-coupling/overview). This Kernel consists of Visualzation of all the datasets in the Data segment of the competition for better understanding of properties of Atoms and Molecules.**
* **Atom** - The atom is the smallest object that retains the properties of an element. Atoms are composed of electrons and a nucleus. The nucleus contains protons and neutrons. 
* **Molecules** - A molecule is a particle made up of two or more atoms that are chemically bonded together; the number of atomic nuclei making up a molecule is a determinate number. 
* **Dipole Moment** - A dipole moment is a measurement of the separation of two opposite electrical charges. Dipole moments are a vector quantity. The magnitude is equal to the charge multiplied by the distance between the charges and the direction is from negative charge to positive charge: Œº = q ¬∑ r where Œº is the dipole moment, q is the magnitude of the separated charge, and r is the distance between the charges.
* **Potential energy** - Potential energy is that energy which an object has because of its position. It is called potential energy because it has the potential to be converted into other forms of energy, such as kinetic energy.

Source of inspiration for the code - [Tod Newman](https://www.kaggle.com/todnewman/keras-neural-net-for-champs), [Andrew Lukyanenko](https://www.kaggle.com/artgor/molecular-properties-eda-and-models)

```python
def map_atom_info(df, atom_idx):
    df = pd.merge(df, structures, how = 'left',
                  left_on  = ['molecule_name', f'atom_index_{atom_idx}'],
                  right_on = ['molecule_name',  'atom_index'])
    
    df = df.drop('atom_index', axis=1)
    df = df.rename(columns={'atom': f'atom_{atom_idx}',
                            'x': f'x_{atom_idx}',
                            'y': f'y_{atom_idx}',
                            'z': f'z_{atom_idx}'})
    return df
```

```python
def make_features(df):
    df['dx']=df['x_1']-df['x_0']
    df['dy']=df['y_1']-df['y_0']
    df['dz']=df['z_1']-df['z_0']
    df['distance']=(df['dx']**2+df['dy']**2+df['dz']**2)**(1/2)
    return df
```

```python
train = map_atom_info(train, 0)
train = map_atom_info(train, 1)

test = map_atom_info(test, 0)
test = map_atom_info(test, 1)
```

```python
train_=make_features(train)
test_=make_features(test) 
```

```python
train_.drop("molecule_name", axis=1, inplace=True)
test_.drop("molecule_name", axis=1, inplace=True)
```

```python
test_id = test_['id']
train_.drop("id", axis=1, inplace=True)
test_.drop("id", axis=1, inplace=True)
```

```python
train_type = pd.get_dummies(train_['type'])
test_type = pd.get_dummies(test_['type'])
```

```python
train_new = pd.concat([train_, train_type], axis=1)
train_new.drop("type", axis=1, inplace=True)
test_new = pd.concat([test_, test_type], axis=1)
test_new.drop("type", axis=1, inplace=True)
```

```python
train_new.head()
```

```python
train_new['atom_0'] = train_new['atom_0'].astype("category").cat.codes
train_new['atom_1'] = train_new['atom_1'].astype("category").cat.codes
test_new['atom_0'] = test_new['atom_0'].astype("category").cat.codes
test_new['atom_1'] = test_new['atom_1'].astype("category").cat.codes
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import explained_variance_score,mean_absolute_error,mean_squared_error,r2_score
from sklearn.neighbors import KNeighborsRegressor
from catboost import CatBoostRegressor
```

```python
y = train_new["scalar_coupling_constant"]
train_new.drop("scalar_coupling_constant", axis=1, inplace=True)
X = train_new
```

```python
x_train, x_test, y_train, y_test = train_test_split(X, y, random_state=42)
```

## Lets see which regressor works better CatBoost or my all time favourite KNN regressor...

### KNN Model

```python
model_KNN = KNeighborsRegressor(n_neighbors=75, algorithm='auto', n_jobs=-1,p=2)
model_KNN.fit(x_train, y_train)
```

```python
y_pred = model_KNN.predict(x_test)
```

```python
print("Variance_Score(KNN_Regressor)\t:"+str(explained_variance_score(y_test,y_pred)))
print("Mean_Absolute_Error(KNN_Regressor)\t:"+str(mean_absolute_error(y_test,y_pred)))
print("Mean_Squared_Error(KNN_Regressor)\t:"+str(mean_squared_error(y_test,y_pred)))
print("R2-Score(KNN_Regressor)\t:"+str(r2_score(y_test,y_pred)))
```

## CatBoost Modelüê±

```python
model_cat = CatBoostRegressor(iterations=3000,depth= 13,random_seed = 23,
                           task_type = "GPU")
model_cat.fit(x_train, y_train)
```

```python
y_pred_cat = model_cat.predict(x_test)
```

```python
print("Variance_Score(cat_Regressor)\t:"+str(explained_variance_score(y_test,y_pred_cat)))
print("Mean_Absolute_Error(cat_Regressor)\t:"+str(mean_absolute_error(y_test,y_pred_cat)))
print("Mean_Squared_Error(cat_Regressor)\t:"+str(mean_squared_error(y_test,y_pred_cat)))
print("R2-Score(cat_Regressor)\t:"+str(r2_score(y_test,y_pred_cat)))
```
