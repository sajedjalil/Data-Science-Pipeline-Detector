```python
# Standard
import gc
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder, StandardScaler
from scipy.sparse import csr_matrix
from sklearn.decomposition import PCA, TruncatedSVD
# Graph Stuff
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
df = pd.read_csv('../input/train.csv')
df.head()
```

```python
df_test = pd.read_csv('../input/test.csv')
df_test.head()
```

```python
df_structures = pd.read_csv('../input/structures.csv')
df_structures.head()
```

```python
le = LabelEncoder()
df_structures.atom = le.fit_transform(df_structures.atom)
```

```python
def map_atom_info(df, atom_idx):
    df = pd.merge(df, df_structures, how = 'left',
                  left_on  = ['molecule_name', f'atom_index_{atom_idx}'],
                  right_on = ['molecule_name',  'atom_index'])
    
    df = df.drop('atom_index', axis=1)
    df = df.rename(columns={'atom': f'atom_{atom_idx}',
                            'x': f'x_{atom_idx}',
                            'y': f'y_{atom_idx}',
                            'z': f'z_{atom_idx}'})
    return df

df = map_atom_info(df, 0)
df = map_atom_info(df, 1)
df_test = map_atom_info(df_test, 0)
df_test = map_atom_info(df_test, 1)

del df_structures
gc.collect()
```

```python
df['dist'] = np.linalg.norm(df[['x_0', 'y_0', 'z_0']].values - df[['x_1', 'y_1', 'z_1']].values, axis=1)
df_test['dist'] = np.linalg.norm(df_test[['x_0', 'y_0', 'z_0']].values - df_test[['x_1', 'y_1', 'z_1']].values, axis=1)
df['dist'] = np.exp(-(df['dist']**2))
df_test['dist'] = np.exp(-(df_test['dist']**2))
```

```python
geofeats = ['molecule_name','atom_index_0','atom_index_1','atom_0','atom_1','dist']
a = pd.concat([df[geofeats],df_test[geofeats]])
print(a.shape)
del df
del df_test
gc.collect()
```

```python
a = a.set_index(['molecule_name','atom_0','atom_1','atom_index_0','atom_index_1'])
```
