```python
# å¤§å®¶å°è¯•ä¸€ä¸‹å§~

# ä»£ç å†™åœ¨ä¸‹æ–¹ï¼Œç›®æ ‡ç”Ÿæˆä¸€ä¸ªdfï¼Œå‘½åä¸ºmeta

print(len(train_data))

var_len = len(train_data)

var = train_data.columns

#æ•°æ®è§’è‰²
var_role = []

#å˜é‡ç±»å‹ ï¼ˆnomial
var_type = []

#æ•°æ®ç±»å‹ (int64, )
data_type = train_data.dtypes
#keep or not? 
keep_or_not = []
#unique value
num_unique_value = []

#missing value
num_missing_value = []

#

var_role.append("åºåˆ—å·ï¼ˆä¸å±äºç‰¹å¾ï¼‰")
var_role.append("ç›®æ ‡æ ‡ç­¾ï¼ˆæ˜¯å¦å‘èµ·ç´¢èµ”)")
keep_or_not.append(False)
keep_or_not.append(True)
for i in range(2, len(var)):
    var_role.append("ç‰¹å¾")
    keep_or_not.append(True)

for i in range(len(var)):
    if "bin" in var[i] or var[i] == "target":
        var_type.append("binary")
    elif "cat" in var[i] or var[i] =="id":
        var_type.append("nominal")
    elif "float" in str(data_type[i]):
        var_type.append("interval")
    elif "int" in str(data_type[i]):
        var_type.append("ordinal")
    if var[i] == 'id':
        num_unique_value.append("ä¸ç»Ÿè®¡")
    else:
        num_unique_value.append(len(pd.unique(train_data[var[i]]))- 1)

    t = len(train_data[train_data[var[i]]==-1])
    percentage = round(t/var_len*100,2)
    num_missing_value.append((str(t) + "(" + str(percentage) + "%" + ")"))

data = {
    'æ•°æ®è§’è‰²': var_role,
    'å˜é‡ç±»å‹' :var_type,
    'æ•°æ®ç±»å‹' :data_type,
    'æ˜¯å¦ä¿ç•™' : keep_or_not,
    'å”¯ä¸€å€¼æ•°é‡' : num_unique_value,
    'ç¼ºå¤±å€¼æ•°é‡' : num_missing_value
}

meta = pd.DataFrame(data, index = var)
                                
                                
                            
                            

    
    
        

    

```
