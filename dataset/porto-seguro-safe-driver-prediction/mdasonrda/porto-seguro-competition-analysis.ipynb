# Data Exploration
In this section, I'll investigate what data looks like and peform statistic observations. 

```python
#load the train dataset
data = pd.read_csv("../input/train.csv")
#display dataset info
data.info()
```

This dataset has a total of 59 columns and 595212 entries. The Target is identified by the name target. There aren't columns with Null values. Although, values of -1 indicate that the feature was missing from the observation. Additionally, all columns have numeric values. Now, let's look at what the data looks like.

```python
data.head(10)
```

```python
data.describe()
```

As we can se above, the target column present only 0 an 1 values. we already can see that the features ps_ind_02_cat, ps_ind_04_cat, ps_ind_05_cat has -1 values. Then, let's see the other features with -1 values. 

```python
#Display all columns with -1
Missing_Values_col = data.loc[:].columns[(data.loc[:] == -1).any()]
print ("There are {} columns with missing values".format(len(Missing_Values_col)))
print (data[data == -1][Missing_Values_col].count())
```

Above we can see the columns with -1 values and their respective amount of -1 values. Below, we can better visualize the missing values in the dataset. Most of the missing values occur in the columns suffixed with _cat. ps_reg_03, ps_car_03_cat and ps_car_05_cat are the top three fetures with missing values, more than 18% of the samples are missing values. And 6 features have less than 500 missing values, that is, less than 0.08% of data.

```python
Missing_data = data[Missing_Values_col]
Missing_data = Missing_data.replace(-1,np.NaN)
import missingno as msno
msno.matrix(Missing_data)
```
