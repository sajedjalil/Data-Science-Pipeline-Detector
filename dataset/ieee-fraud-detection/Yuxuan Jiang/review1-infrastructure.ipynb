**The First one-Google Cloud**

>Google cloud is really generous! If you have a google account, you can get 300 dollars for free and create a [instance](https://cloud.google.com/compute/docs/instances/). The following is the detailed process of creating a instance in google cloud.

*1. sign up*

google "google cloud" and visit its official website:
![](https://i.imgur.com/xlTllyb.png)
click get started for free and use your google account to sign up
Then it will ask you for your credit card account to symbolically take away one dollar from you. After that, you can enjoy your google cloud!

*2. create an instance*

To see the complete introduction of instance, learn from the [docs](https://cloud.google.com/compute/docs/) and [videos](https://www.youtube.com/googlecloud). For kagglers, we need a virtual machine (VM) with big memory space, so we will use the VM Instance service.
click the button and create a new project
![project](https://i.imgur.com/XyaR91B.png)
choose VM Instance from computer engine
![](https://i.imgur.com/7S9udCl.png)
then we go into the edit interface.First, name your instance. Second, choose your region. Third, choose your machine type.
![](https://i.imgur.com/Wn2lON5.png)
I chose general purpose and 8v cpu, 52GB memory.You can view the detailed information [here](https://cloud.google.com/compute/docs/machine-types).
As we know, GPU is extremely fast for catboost and xgboost, so if you want to have GPU, click the " CPU platform and GPU" link below.
Now it's time to choose our operation system image. If you know how to use Ubuntu, strongly recommend Ubuntu, because you only need "pip install xgboost" to install xgboost. However, I don't know it at all, so I chose windows 2008 R2,which I was quite familiar with.
![](https://i.imgur.com/oTmupt6.png)
Then just choose what I chose.
![](https://i.imgur.com/tCC5nn6.png)
now you have an instance!
![](https://i.imgur.com/sqrD5SR.png)

*3. set static external IP and firewall rules*

the external IP address is dynamic by default, so we have to change the type from Ephemeral to Static
![](https://i.imgur.com/sLTVjjj.png)
now click on the "Firewall Rules" below "External IP addresses" and create firewall rule,you can specify your port number.
![](https://i.imgur.com/8cIxxWg.png)

*4. start the instance and install the coding environment*

Finally! It's time to use your instance!
My connect method is RDP, if your method is SSH, you can see the instructions [here](https://medium.com/towards-data-science/running-jupyter-notebook-in-google-cloud-platform-in-15-min-61e16da34d52).
Reset your password and download the RDP file. 
![](https://i.imgur.com/lQhjYtL.png)
The RDP file is for remote desktop connection, clear instruction is [here](https://filext.com/file-extension/RDP). Lucky that we don't have to set anything because it contains all the information. 
start your instance
![](https://i.imgur.com/2ZSjmcP.png)
Just click the RDP file and connect your instance remotely!
![](https://i.imgur.com/Kf0W9tI.png)
The first connection need password, the later ones don't. When you connect succefully(remember to close the VPN), you can see a interface telling you " preparing for your desktop".
This is my desktop, but at the beginning it was not that pretty. The icons were very small.Besides, it didn't have google chorme, annaconda and jupyter! So download them just in the way you did on your own PC.
[](https://i.imgur.com/ZMqFaus.png)

*5. save your "money"*
Although we didn't pay too much on this, we still have to be enconomy. Stop your instance as soon as you don't need it!
