**App Summary:** Throughout the app, you will be able to **plot RNAs**, **generate predictions** for degradation targets, **generate some ARNIE features** and also **analyze the data** via plots. ARNIE is a Python API developed by DasLab to compute RNA energetics and do structure prediction across multiple secondary structure packages. In the Custom Sample section, you will not need the uploaded dataset. You can just enter any sequence, structure, and predicted loop type info and get both RNA plot and predictions.
