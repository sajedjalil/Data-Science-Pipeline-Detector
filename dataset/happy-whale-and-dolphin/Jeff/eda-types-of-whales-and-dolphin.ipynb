Creating function to display the some images from dataset of a particular type

```python
def Species_9(name):
    species_list = []
    t = train["species"] == name
    species_list = [i for i, x in enumerate(t) if x]
    top_9_species = species_list[:9]
    return top_9_species
```

```python
def display_9_images(top_9_images,heading):    
    
    figure, ax = plt.subplots(nrows=3,ncols=3,figsize=(16,8))
    plt.suptitle(heading)
    for index,image in enumerate(train.image[top_9_images]):
            img_array = np.array(Image.open(f"../input/happy-whale-and-dolphin/train_images/{image}"))
            try:
                ax.ravel()[index].imshow(img_array)
                ax.ravel()[index].set_axis_off()
            except:
                continue;
    plt.tight_layout()
    plt.show()
```
