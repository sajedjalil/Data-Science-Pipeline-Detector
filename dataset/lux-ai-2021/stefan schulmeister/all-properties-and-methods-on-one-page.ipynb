```python
%%writefile agent.py
from lux.game import Game
from lux.game_map import Cell, RESOURCE_TYPES
from lux.constants import Constants
from lux.game_constants import GAME_CONSTANTS
from lux import annotate
import pandas as pd
import math


def print_object_information(name, obj):
    """
    Small function to print methods and properties of a given object.
    """
    attr = vars(obj)
    methods = [method_name for method_name in dir(obj)
          if callable(getattr(obj, method_name))]
    methods = [method_name for method_name in methods if method_name[0:2] !="__"]
    print("")
    print("")
    print("Name:", name)
    print("Type:", type(obj))
    print(20*"-")
    print("Properties:")
    for key, value in attr.items():
        if(len(str(value)) > 700):
            print(key, ": ", str(value)[:700])
        else:
            print(key, ": ", value )
    print(20*"-")
    print("Methods:")
    print(methods)
    print("")
    print("")

"""
We simply use a empty agent and print out some runtime information.
"""
def agent(observation, configuration):
    global game_state

    ### Do not edit ###
    if observation["step"] == 0:
        game_state = Game()
        game_state._initialize(observation["updates"])
        game_state._update(observation["updates"][2:])
        game_state.id = observation.player
    else:
        game_state._update(observation["updates"])
    
    actions = []

    ### AI Code goes down here! ### 
    player = game_state.players[observation.player]
    opponent = game_state.players[(observation.player + 1) % 2]
    width, height = game_state.map.width, game_state.map.height

    """
    In each step there are two observations. One for each player. Since we do not want to print everything twice, 
    we will limit ourselves to the observations of player 1.
    """
    if observation.player == 1:
        """
        There seems to be some kind of time limit per observation and it is not able to print everything in one step. 
        Therefore I have split the output of each object into different steps.
        """
        if observation["step"] == 0:
            print_object_information("observation" , observation)      
        if observation["step"] == 1:
            print_object_information("player" , game_state.players[1])   
        if observation["step"] == 2:
            print_object_information("unit" , game_state.players[1].units[0])
        if observation["step"] == 3:
            print_object_information("cargo" , game_state.players[1].units[0].cargo)
        if observation["step"] == 4:
            print_object_information("cities" , game_state.players[1].cities["c_2"])
        if observation["step"] == 5:
            print_object_information("citytiles" , game_state.players[1].cities["c_2"].citytiles[0])
        if observation["step"] == 6:
            print_object_information("game_state" , game_state) 
        if observation["step"] == 7:
            print_object_information("map" , game_state.map) 
        if observation["step"] == 8:
            print_object_information("cell" , game_state.map.map[0][0])
        if observation["step"] == 9:
            print(50*"*")
            print("OBSERVATION UPDATES:")
            print(50*"*")
    
    """
    It is interesting to observe the observation.update property over several steps.
    """
    if ((observation["step"] in range(9,13)) and (observation.player == 1)):
        print(15*"-")
        print("observation_updates_turn", observation["step"])
        print(observation.updates) 
        print(15*"-")
    return actions
```
