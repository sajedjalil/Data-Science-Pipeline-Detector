**Model**

```python
def auc(y_true, y_pred):
    auc = tf.metrics.auc(y_true, y_pred)[1]
    K.get_session().run(tf.local_variables_initializer())
    return auc
```

```python
def modelir():
    model = Sequential()

    #input 
    model.add(Dense(200, input_dim=200, kernel_initializer = 'uniform', 
                    kernel_regularizer=regularizers.l2(0.005), kernel_constraint = max_norm(5.)))
    model.add(Activation("relu"))
    model.add(Dropout(0.2))

    #1 
    model.add(Dense(200, kernel_initializer = 'uniform', 
                    kernel_regularizer=regularizers.l2(0.005), kernel_constraint=max_norm(5)))
    model.add(Activation("relu"))
    model.add(Dropout(0.1))
    
    #2
    model.add(Dense(100, kernel_initializer = 'uniform', 
                    kernel_regularizer=regularizers.l2(0.005), kernel_constraint=max_norm(5)))
    model.add(Activation('relu'))
    model.add(Dropout(0.1))
    
    #2
    model.add(Dense(50, kernel_initializer = 'uniform', 
                    kernel_regularizer=regularizers.l2(0.005), kernel_constraint=max_norm(5)))
    model.add(Activation('tanh'))
    model.add(Dropout(0.1))
    
    #output
    model.add(Dense(1, activation = 'sigmoid'))

    model.compile(loss="binary_crossentropy", optimizer="adam", metrics=["accuracy", auc])
    return model
#print(model.summary())
```

```python
def simple_blend1(X, y, test):
    model = modelir()
    pred = pd.DataFrame()
    for i in range(1, 10):
        X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = i)
        earlystopper = EarlyStopping(patience=5, verbose=1)
        history = model.fit(X_train, y_train, batch_size=16384, epochs=500, validation_split=0.2, verbose=2, callbacks=[earlystopper], shuffle=True)
        scores = model.evaluate(X_test, y_test, verbose=0)
        print("accuracy for test data: %.2f%%" % (scores[1]*100))
        plt.plot(history.history['acc'], label='accuracy for train data')
        plt.plot(history.history['val_acc'], label='validation data accuracy')
        plt.xlabel('epochs')
        plt.ylabel('accuracy')
        plt.legend()
        plt.show()
        y_pred = model.predict(X_test)
        y_pred = (y_pred > 0.7)
        cm = confusion_matrix(y_test, y_pred)
        print(cm)
        y_pred_t = model.predict(test)
        print(y_pred_t.T[0])
        pred[i] = y_pred_t.T[0]
    return pred
```

```python
pr1 = simple_blend1(X, y, test)
```

```python
pr1['mean'] = pr1.mean(axis=1)
pr1.head()
```

```python
def simple_blend2(X, y, test):
    pred = pd.DataFrame()
    for i in range(1, 10):
        model = modelir()
        X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = i)
        earlystopper = EarlyStopping(patience=5, verbose=1)
        history = model.fit(X_train, y_train, batch_size=512, epochs=500, validation_split=0.2, verbose=2, callbacks=[earlystopper], shuffle=True)
        scores = model.evaluate(X_test, y_test, verbose=0)
        print("accuracy for test data: %.2f%%" % (scores[1]*100))
        plt.plot(history.history['acc'], label='accuracy for train data')
        plt.plot(history.history['val_acc'], label='validation data accuracy')
        plt.xlabel('epochs')
        plt.ylabel('accuracy')
        plt.legend()
        plt.show()
        y_pred = model.predict(X_test)
        y_pred = (y_pred > 0.7)
        cm = confusion_matrix(y_test, y_pred)
        print(cm)
        y_pred_t = model.predict(test)
        print(y_pred_t.T[0])
        pred[i] = y_pred_t.T[0]
    return pred
```

```python
pr2 = simple_blend2(X, y, test)
```

```python
pr2['mean'] = pr2.mean(axis=1)
pr2.head()
```
