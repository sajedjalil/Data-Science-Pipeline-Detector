# Blend Boosting study on dataset of the Santander Customer Transaction Prediction:

Here I share with you a systematic blend boosting study on dataset of the Santander Customer Transaction Prediction (https://www.kaggle.com/c/santander-customer-transaction-prediction). I just collect some submission files (20 distincts, but skip some of them).

Basically, I start to analysis of correlations, then decide to sort them according to their sum of correlation values in between. This lets me divide 16 scores into 3 subgroups. Then I make internal linear calibration in each subgroup by considering their scores on the Kaggle. Finally I make recalling between subgroups to achieve higher scores on the Kaggle by resubmission. Of course, if you spend much more time, you can always achieve betters scores, but I stop it here because it is already highest score on Kaggle ;-).  

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# loading dummy submission file
sub_file = pd.read_csv('../input/santander-customer-transaction-prediction/submission_blend_1.csv')

# loading data including 16 best scores
df_sub = pd.read_csv(r'../input/santander-customer-transaction-prediction/best_blend_2.csv')

# a rough correlation based visualization of 16 best scores
plt.figure(figsize=(10,10))
sns.heatmap(df_sub.corr(), cmap='Spectral')
plt.ylabel('file index numbers')
plt.xlabel('file index numbers')
plt.show()
```

```python
# basic analysis and visualization of subgroups in different color.
plt.figure(figsize=(12, 5))
df_mean_corr = pd.DataFrame({'mean_corr': df_sub.corr().mean()})
df_mean_corr = df_mean_corr.sort_values('mean_corr', ascending=False)
df_mean_corr = df_mean_corr.reset_index()

plt.plot(df_mean_corr.index[:8], df_mean_corr['mean_corr'].values[:8], 'o', ms=10)
plt.plot(df_mean_corr.index[8:13], df_mean_corr['mean_corr'].values[8:13], 'o', ms=10)
plt.plot(df_mean_corr.index[13:14], df_mean_corr['mean_corr'].values[13:14], 'o', ms=10)
plt.plot(df_mean_corr.index[14:15], df_mean_corr['mean_corr'].values[14:15], 'o', ms=10)
plt.plot(df_mean_corr.index[15:16], df_mean_corr['mean_corr'].values[15:16], 'o', ms=10)

plt.xticks([*range(len(df_mean_corr))], df_mean_corr['index'].tolist())
plt.title('determination of sub_groups')
plt.ylabel('a correlation ralated index')
plt.xlabel('file index numbers')
plt.show()
```

```python
# a linear combination to achieve much better scores
df_sub['weighted_avg'] = abs(1 * (
    100 * ( 100 * df_sub['0'] + 30 * df_sub['3'] + 20 * df_sub['4'] + 10 * df_sub['5'] +
           10 * df_sub['7'] + 1 * df_sub['8'] + 1 * df_sub['11'] + 1 * df_sub['13']) / 173 +

    5 * ( 50 * df_sub['1'] + 50 * df_sub['2'] + 5 * df_sub['10'] + 5 * df_sub['12'] + 1 * df_sub['14']) / 111 +
    1 * ( 20 * df_sub['6'] + 5 * df_sub['9'] + 1 * df_sub['15']) / 26 + 200 * df_sub['16'] ) / 306 )


# create the final submission file
submission = pd.DataFrame({'ID_code': sub_file.ID_code, 'target': df_sub['weighted_avg'].tolist()})
submission.to_csv(r'submission_file.csv', index=False)
```

## It gets a 0.92694 AUC as public score, and looks the best score on Kaggle so far ;-)
