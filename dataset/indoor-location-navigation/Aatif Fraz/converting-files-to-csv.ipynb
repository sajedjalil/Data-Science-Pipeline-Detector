Given is the code to convert all given information to CSV files, for further use.

I have only converted the test paths and the associated 24 sites, but you may choose to convert all of them if you so choose, but keep in mind some sites are different/poorly formatted, so the below code will not work on those. But it works for the sites that we are supposed to predict, so here goes: 

```python
import pandas as pd
import numpy as np
import os
from IPython.display import clear_output, display
from dataclasses import dataclass
@dataclass
class ReadData:
    acce: np.ndarray        # TS, X, Y, Z, Accuracy
    acce_uncali: np.ndarray # TS, X, Y, Z, X_uncali,Y_uncali, Z_uncali, Accuracy 
    gyro: np.ndarray        # TS, X, Y, Z, Accuracy
    gyro_uncali: np.ndarray # TS, X, Y, Z, X_uncali, Y_uncali, Z_uncali, Accuracy
    magn: np.ndarray        # TS, X, Y, Z, Accuracy
    magn_uncali: np.ndarray # TS, X, Y, Z, X_uncali, Y_uncali, Z_uncali, Accuracy
    ahrs: np.ndarray        # TS, X, Y, Z, Accuracy
    wifi: np.ndarray        # TS, ssid, bssid, rssi, frequency, lastseen_ts
    ibeacon: np.ndarray     # TS, uuid, majorid, minorid, txpower, rssi, distance, mac, unix_time
    waypoint: np.ndarray    # TS, X, Y


def read_data_file(data_filename):
    acce = [[0, 0, 0, 0, 0]]
    acce_uncali = [[0, 0, 0, 0, 0, 0, 0, 0]]
    gyro = [[0, 0, 0, 0, 0]]
    gyro_uncali = [[0, 0, 0, 0, 0, 0, 0, 0]]
    magn = [[0, 0, 0, 0, 0]]
    magn_uncali = [[0, 0, 0, 0, 0, 0, 0, 0]]
    ahrs = [[0, 0, 0, 0, 0]]
    wifi = [[0, 0, 0, 0, 0, 0]]
    ibeacon = [[0, 0, 0, 0, 0, 0, 0, 0, 0]]
    waypoint = [[0, 0, 0]]

    with open(data_filename, 'r', encoding='utf-8') as file:
        lines = file.readlines()

    for line_data in lines:
        line_data = line_data.strip()
        if not line_data or line_data[0] == '#':
            continue

        line_data = line_data.split('\t')
        try:
            if line_data[1] == 'TYPE_ACCELEROMETER':
                ts = int(line_data[0])
                x, y, z = float(line_data[2]), float(line_data[3]), float(line_data[4])
                accuracy = float(line_data[5])
                acce.append([ts, x, y, z, accuracy])
                continue

            if line_data[1] == 'TYPE_ACCELEROMETER_UNCALIBRATED':
                ts = int(line_data[0])
                x, y, z = float(line_data[2]), float(line_data[3]), float(line_data[4])
                x_uncali, y_uncali, z_uncali = float(line_data[5]), float(line_data[6]), float(line_data[7])
                accuracy = float(line_data[8])
                acce_uncali.append([ts, x, y, z, x_uncali, y_uncali, z_uncali, accuracy])
                continue

            if line_data[1] == 'TYPE_GYROSCOPE':
                ts = int(line_data[0])
                x, y, z = float(line_data[2]), float(line_data[3]), float(line_data[4])
                accuracy = float(line_data[5])
                gyro.append([ts, x, y, z, accuracy])
                continue

            if line_data[1] == 'TYPE_GYROSCOPE_UNCALIBRATED':
                ts = int(line_data[0])
                x, y, z = float(line_data[2]), float(line_data[3]), float(line_data[4])
                x_uncali, y_uncali, z_uncali = float(line_data[5]), float(line_data[6]), float(line_data[7])
                accuracy = float(line_data[8])
                gyro_uncali.append([ts, x, y, z, x_uncali, y_uncali, z_uncali, accuracy])
                continue

            if line_data[1] == 'TYPE_MAGNETIC_FIELD':
                ts = int(line_data[0])
                x, y, z = float(line_data[2]), float(line_data[3]), float(line_data[4])
                accuracy = float(line_data[5])
                magn.append([ts, x, y, z, accuracy])
                continue

            if line_data[1] == 'TYPE_MAGNETIC_FIELD_UNCALIBRATED':
                ts = int(line_data[0])
                x, y, z = float(line_data[2]), float(line_data[3]), float(line_data[4])
                x_uncali, y_uncali, z_uncali = float(line_data[5]), float(line_data[6]), float(line_data[7])
                accuracy = float(line_data[8])
                magn_uncali.append([ts, x, y, z, x_uncali, y_uncali, z_uncali, accuracy])
                continue

            if line_data[1] == 'TYPE_ROTATION_VECTOR':
                ts = int(line_data[0])
                x, y, z = float(line_data[2]), float(line_data[3]), float(line_data[4])
                accuracy = float(line_data[5])
                ahrs.append([ts, x, y, z, accuracy])
                continue

            if line_data[1] == 'TYPE_WIFI':
                ts = int(line_data[0])
                ssid = line_data[2]
                bssid = line_data[3]
                rssi = line_data[4]
                freq = line_data[5]
                lastseen_ts = line_data[6]
                wifi_data = [ts, ssid, bssid, rssi, freq, lastseen_ts]
                wifi.append(wifi_data)
                continue

            if line_data[1] == 'TYPE_BEACON':
                ts = int(line_data[0])
                uuid = line_data[2]
                major = line_data[3]
                minor = line_data[4]
                txpower = int(line_data[5])
                rssi = int(line_data[6])
                distance = float(line_data[7])
                macaddress = line_data[8]
                unixtime = int(line_data[9])
                ibeacon_data = [ts, uuid, major, minor, txpower, rssi, distance, macaddress, unixtime]
                ibeacon.append(ibeacon_data)
                continue

            if line_data[1] == 'TYPE_WAYPOINT':
                waypoint.append([int(line_data[0]), float(line_data[2]), float(line_data[3])])
        except:
            continue
            
    acce = np.array(acce)
    acce_uncali = np.array(acce_uncali)
    gyro = np.array(gyro)
    gyro_uncali = np.array(gyro_uncali)
    magn = np.array(magn)
    magn_uncali = np.array(magn_uncali)
    ahrs = np.array(ahrs)
    wifi = np.array(wifi)
    ibeacon = np.array(ibeacon)
    waypoint = np.array(waypoint)

    return ReadData(acce, acce_uncali, gyro, gyro_uncali, magn, magn_uncali, ahrs, wifi, ibeacon, waypoint)


def get_site_from_test(tfile):
    with open(tfile, 'r', encoding='utf-8') as file:
        lines = file.readlines()
    for line in lines:
        line_data = line.split('\t')
        if (not line_data) or line_data[0] != '#':
            continue
        for word in line_data:
            if word[:6].lower() == 'siteid':
                crap, sid = word.split(':')
                return sid

def get_unique_ts(arr):
    ts = set()
    for item in arr:
        for k in item:
            ts.add(int(k))
    return sorted(list(ts))
```
