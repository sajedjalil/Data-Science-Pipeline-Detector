# 1) Find Max of f_x_y

```python
temp=0
np_max_fxy = np.ones((len(mall_list),3))
np_max_fxy[:,0] = 1
np_max_fxy[:,1] = 3
np_max_fxy[:,2] = 3
```

```python
floor_max_num = []
for mall in mall_list:
    floor_num = []
    for key in list(os.walk('../input/indoor-location-navigation/train/'+mall))[0][1]:
        try:
            floor_num.append(floor_txt_num[key])
        except:
            floor_num.append(1)
    floor_max_num.append(max(floor_num))
np_max_fxy[:,0]=np.array(floor_max_num)
```

```python
mall_max_x_y=[]
temp=0
for mall in mall_list:
    mall_train_path = glob.glob('../input/indoor-location-navigation/train/'+mall+'/*/*')
    max_x_y = [2,2]
    for path in mall_train_path:
        try:
            mwi_data_xy = (read_data_file(path)).waypoint
            if (mwi_data_xy != []) and  (max_x_y[0] < max(mwi_data_xy[:,1])):
                max_x_y[0]= max(mwi_data_xy[:,1])
            if (mwi_data_xy != []) and  (max_x_y[1] < max(mwi_data_xy[:,2])):
                max_x_y[1]= max(mwi_data_xy[:,2])
        except:
            continue
    mall_max_x_y.append(max_x_y)
    print(temp," out of 204")
    temp+=1
```

```python
np_max_fxy[:,1:3]=np.array(mall_max_x_y)
np_max_fxy = np_max_fxy.astype('int')
df_max_fxy = pd.DataFrame(np_max_fxy)
df_max_fxy.columns = ["max_floor","max_x","max_y"]
df_max_fxy["mall"]=mall_list
df_max_fxy.to_csv('max_fxy.csv')
```
