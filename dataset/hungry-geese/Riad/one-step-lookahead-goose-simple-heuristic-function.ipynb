```python
def get_win_percentages(agent1,agent2,agent3,agent4, n_rounds=100):   
    outcomes = evaluate("hungry_geese", [agent1, agent2],  n_rounds//2)
    # Agent 2 goes first (roughly) half the time      
    outcomes += [[b,a,c,d] for [a,b,c,d] in evaluate("hungry_geese",[agent1,agent2,agent3,agent4,],n_rounds)]
    print("Agent 1 Win Percentage:", outcomes)
```
