```python
x = torch.Tensor(np.array([conv_batchnorm_output]*BATCH).reshape(BATCH,FEATURE,7,11))

print("conv_batchnorm_output")
print(x)

print("heads")
heads_ = torch.Tensor(np.array([heads]*BATCH).reshape(BATCH,4,7,11))
print(heads_)

print("conv_batchnorm_output * heads")
x_ = (x.unsqueeze(1) * heads_.unsqueeze(2))
print(x_.shape)
print(x_)

print("h_head")
h_head = (x.unsqueeze(1) * heads_.unsqueeze(2)).sum(4, keepdim=True).sum(3, keepdim=True)
print(h_head.shape)
print(h_head)
```

```python
print("bodies")
bodies_ = torch.Tensor(np.array([bodies]*BATCH).reshape(BATCH,4,7,11))
h_body = (h_head * bodies_.unsqueeze(2)).sum(1)
print(bodies_)

print("h_body")
print(h_body.shape)
print(h_body)
```
