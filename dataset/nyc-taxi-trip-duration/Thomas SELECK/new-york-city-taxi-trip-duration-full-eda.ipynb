The submissions are evaluated using the RMSLE (Root Mean Squared Logarithmic Error) defined as follows:
$$RMSLE = \sqrt(\frac{1}{n}\sum\limits_{i=1}^n (log(p_i + 1) - log(a_i + 1))^2)$$
