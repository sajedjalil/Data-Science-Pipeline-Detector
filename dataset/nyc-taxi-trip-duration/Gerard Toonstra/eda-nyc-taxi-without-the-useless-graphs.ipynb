And here we see the challenge: "Given a specific pick up location and a drop off location and a time of pick up, what will be the expected trip duration?".  The vendor id and the flag may be able to help us to refine the answer to that question later.

 This gives us a couple of options already:
- the vendor doing specific trips many times has a lot of local knowledge, so they may be able to shave off some trip minutes. For example, instead of going over the freeway, they may use back-roads to get to a frequently visited location.
- passenger count isn't necessarily that indicative, but we allow the model to determine that. 
- the flag is just something relating to the way the trip was executed and a mobile connection or not. 

The date/time +lat/lon is going to be crucial, because that essentially describes the traffic movement for particular days and construction work that could have taken place. Traffic on any given day can be really different and there are no really good indicators for that. So what this model "ought"  to do in a real life scenario is figure out ways to express traffic on a given day in such a way that it can be predicted.
