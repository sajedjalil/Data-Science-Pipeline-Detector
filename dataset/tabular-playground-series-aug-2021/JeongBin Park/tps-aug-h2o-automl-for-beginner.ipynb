```python
import pandas as pd
import numpy as np
import h2o
from h2o.automl import H2OAutoML
h2o.init()
```

```python
train = pd.read_csv("../input/tabular-playground-series-aug-2021/train.csv")
test = pd.read_csv("../input/tabular-playground-series-aug-2021/test.csv")
sample_submission = pd.read_csv("../input/tabular-playground-series-aug-2021/sample_submission.csv")
```

```python
train = train.drop('id',axis=1)
test = test.drop('id',axis=1)

train = h2o.H2OFrame(train)
test = h2o.H2OFrame(test)
```

#### You should change data to H2OFrame to use H2O AutoML.

```python
x = train.columns
y = "loss"
x.remove(y)

aml = H2OAutoML(max_models=10, seed=1)
aml.train(x=x, y=y, training_frame=train)
```

#### Train !
#### It takes some time.... 

```python
lb = aml.leaderboard
lb.head(rows=lb.nrows)
```

#### And you can check train result from leader board.

```python
preds = aml.leader.predict(test)
sample_submission['loss'] = h2o.as_list(preds)['predict']
sample_submission.to_csv("submission.csv",index=None)
```
