# Portrait of a successful Kaggler vs usual Kaggler

Kaggle has over 5 million registered users. For simplicity of naming, we will call the first group of successful Kagglers - ***successfull Kaggler***. The second group is a group of ordinary highly motivated Kaggler - ***usual Kaggler***.

# **Split into two groups**

To split it down into two groups, I suggest looking at the other two groups first - working and student users. The fact is that there is some imbalance in the received data from the survey and we need to decide what to do with it.

```python
data_salaries = data.copy()

salaries_order = ['Not defined', '$0-999', '1,000-1,999', '2,000-2,999', '3,000-3,999', '4,000-4,999', '5,000-7,499', '7,500-9,999',
'10,000-14,999','15,000-19,999', '20,000-24,999', '25,000-29,999', '30,000-39,999', '40,000-49,999', '50,000-59,999', '60,000-69,999', '70,000-79,999', '80,000-89,999', '90,000-99,999',
'100,000-124,999', '125,000-149,999',  '150,000-199,999', '200,000-249,999',  '250,000-299,999', '300,000-500,000', '> $500,000']

salaries_order_more_than_1000 = ['1,000-1,999', '2,000-2,999', '3,000-3,999', '4,000-4,999', '5,000-7,499', '7,500-9,999',
'10,000-14,999','15,000-19,999', '20,000-24,999', '25,000-29,999', '30,000-39,999', '40,000-49,999', '50,000-59,999', '60,000-69,999', '70,000-79,999', '80,000-89,999', '90,000-99,999',
'100,000-124,999', '125,000-149,999',  '150,000-199,999', '200,000-249,999',  '250,000-299,999', '300,000-500,000', '> $500,000']

# insert the new column with int values of salaries
data_salaries.insert(0, 'Q24_int', np.zeros((data_salaries['Q1'].count(),1), dtype='int64'))

# transform the string to an int to perform simpler calculations later
data_salaries.Q24_int = pd.Series(data_salaries.Q24, dtype="string").str.split("-").str.get(0).str.replace("$", "").str.replace(">", "").str.replace(',', '').fillna(-1).astype(int)
data_salaries.Q24 = pd.Series(data_salaries.Q24, dtype="string").fillna('Not defined')

data_studying_salaries = data_salaries[(data_salaries.Q5 == 'Student') |
                                       (data_salaries.Q5 == 'Currently not employed') | 
                                       (data_salaries.Q5 == 'Other')]

data_working_salaries = data_salaries[(data_salaries.Q5 != 'Student') & 
                                      (data_salaries.Q5 != 'Currently not employed') & 
                                      (data_salaries.Q5 != 'Other')]

group_studying_salaries = data_studying_salaries['Q24'].value_counts()[salaries_order]
group_working_salaries = data_working_salaries['Q24'].value_counts()[salaries_order]

#---------------------

data_studying_salaries_for_graph = data_studying_salaries["Q24"].dropna()
data_working_salaries_for_graph = data_working_salaries["Q24"].dropna()

percent_group_studying_salaries = data_studying_salaries_for_graph.value_counts(normalize=True)[salaries_order]
percent_group_working_salaries = data_working_salaries_for_graph.value_counts(normalize=True)[salaries_order]
count_group_studying_salaries = data_studying_salaries_for_graph.value_counts(normalize=False).astype('float64')[salaries_order]
count_group_working_salaries = data_working_salaries_for_graph.value_counts(normalize=False).astype('float64')[salaries_order]

data_studying_salaries_for_graph = pd.DataFrame(
    np.hstack([percent_group_studying_salaries.values.reshape(-1, 1), count_group_studying_salaries.values.reshape(-1, 1)]),
    index=percent_group_studying_salaries.index,
    columns=["Percent", "Count"],
).reset_index()

data_working_salaries_for_graph = pd.DataFrame(
    np.hstack([percent_group_working_salaries.values.reshape(-1, 1), count_group_working_salaries.values.reshape(-1, 1)]),
    index=percent_group_working_salaries.index,
    columns=["Percent", "Count"],
).reset_index()


fig = px.bar(data_studying_salaries_for_graph, x="index", y="Count", text="Percent", range_y=[0,7100])

fig.update_traces(
    hovertemplate="Salary: %{x}<br>Count: %{y:,.0f}",
    marker_color='#CD5B45',
    marker_opacity=0.7,
    texttemplate="%{text:.0%}",
    textfont=textfont,
    textposition='inside'
)

fig.update_layout(
    xaxis=dict(
        title_text="", 
        tickfont=tickfont,
        titlefont=axistitlefont
    ),
    yaxis=dict(
        tickfont=tickfont,
        titlefont=axistitlefont,
        tickformat=",", 
        title_text="Counts"),
    hoverlabel=hoverlabelformat,
    margin=dict(pad=5, t=100),
    height=500,
    width=800,
    title=dict(text="Salary of studying users (Total)", font=titlefont),
)
fig.show()

#------------------------

fig = px.bar(data_working_salaries_for_graph, x="index", y="Count", text="Percent", range_y=[0,2400])

fig.update_traces(
    hovertemplate="Salary: %{x}<br>Count: %{y:,.0f}",
    marker_color='#4C9141',
    marker_opacity=0.7,
    texttemplate="%{text:.0%}",
    textfont=textfont,
    textposition='inside'
)

fig.update_layout(
    xaxis=dict(
        title_text="", 
        tickfont=tickfont,
        titlefont=axistitlefont
    ),
    yaxis=dict(
        tickfont=tickfont,
        titlefont=axistitlefont,
        tickformat=",", 
        title_text="Counts"),
    hoverlabel=hoverlabelformat,
    margin=dict(pad=5, t=100),
    height=500,
    width=800,
    title=dict(text="Salary of working users (Total)", font=titlefont),
)
fig.show()
```

As you can see from the first chart, students are not paid. The rest of the values of users who study Data Science most likely consist of salaries from other activities. Some of the users from the first graph are those who decided to change their field of activity to Data Science. Accordingly, they have a salary, but their income has nothing to do with their activities in Data Science.

On the second chart, everything is much more interesting, since the part of users who did not want to report their income is clearly expressed. Unfortunately, we have to remove them from our further research. There are also some respondents whose annual salaries are too low under $ 1000 for the Data Science. On this score, I have two assumptions. The first is that these people work in areas not related to Data Science and want to retrain. Second, the respondents did not see that they were asked about the annual salary and indicated their monthly income. For example, in Russia it is not customary to carry out calculations based on the annual salary, we are always talking about the monthly salary. In any case, since we do not know for sure, and we have no right to guess in the study, we will also have to remove these respondents from the dataset.

Next, we will build an adequate salary chart for working users. 

```python
data_not_defined_salaries = data_salaries[(data_salaries.Q24_int < 1000) &
                                          (data_salaries.Q5 != 'Student') & 
                                          (data_salaries.Q5 != 'Currently not employed') & 
                                          (data_salaries.Q5 != 'Other')]

data_working_salaries = data_salaries[(data_salaries.Q24_int != 0) & 
                                    (data_salaries.Q24_int != -1) & 
                                    (data_salaries.Q5 != 'Student') & 
                                    (data_salaries.Q5 != 'Currently not employed') & 
                                    (data_salaries.Q5 != 'Other')]

group_not_defined_salaries = data_not_defined_salaries.groupby('Q24')[['Q1']].count()
group_not_defined_salaries.rename(columns={"Q1": "count"},inplace=True)

group_working_salaries = data_working_salaries['Q24'].value_counts()[salaries_order_more_than_1000]

#-----------------------
data_working_salaries_for_graph = data_working_salaries["Q24"].dropna()

percent_group_working_salaries = data_working_salaries_for_graph.value_counts(normalize=True)[salaries_order_more_than_1000]
count_group_working_salaries = data_working_salaries_for_graph.value_counts(normalize=False).astype('float64')[salaries_order_more_than_1000]

data_working_salaries_for_graph = pd.DataFrame(
    np.hstack([percent_group_working_salaries.values.reshape(-1, 1), count_group_working_salaries.values.reshape(-1, 1)]),
    index=percent_group_working_salaries.index,
    columns=["Percent", "Count"],
).reset_index()

#------------------------

fig = px.bar(data_working_salaries_for_graph, x="index", y="Count", text="Percent", range_y=[0,600])

fig.update_traces(
    hovertemplate="Salary: %{x}<br>Count: %{y:,.0f}",
    marker_color='#4C9141',
    marker_opacity=0.7,
    texttemplate="%{text:.0%}",
    textfont=textfont,
    textposition='inside'
)

fig.update_layout(
    xaxis=dict(
        title_text="", 
        tickfont=tickfont,
        titlefont=axistitlefont
    ),
    yaxis=dict(
        tickfont=tickfont,
        titlefont=axistitlefont,
        tickformat=",", 
        title_text="Counts"),
    hoverlabel=hoverlabelformat,
    margin=dict(pad=5, t=100),
    height=500,
    width=800,
    title=dict(text="Salary of working users (without not defined users)", font=titlefont),
)
fig.show()
```

Now the graph of distribution of salaries of working users has become more visual, but not more informative. The level of salaries varies greatly depending on the country of residence, position, work experience, and much more. Below, when we select two groups of users and go through all the questions from the survey, the picture becomes clearer.

```python
sum_studying_users = group_studying_salaries.sum()
sum_working_users = group_working_salaries.sum()
sum_working_not_defined_users = group_not_defined_salaries['count'].sum()
sum_counts = [sum_studying_users, sum_working_users, sum_working_not_defined_users]

labels = ['Studying DS Users', 'Working DS Users', 'Not defined DS Users']
fig = go.Figure()
fig.add_trace(go.Pie(values=sum_counts, 
                     labels=labels, 
                     textinfo='label+percent', 
                     showlegend=False, 
                     insidetextorientation='radial', 
                     hole=0.3))

fig.update_layout(
    title="Studying/working DS Kaggle users",
    title_x = 0.5,
    margin=dict(l=0, r=0, t=40, b=0),
    legend_orientation="h")
fig.show()
```

This pie allows us to understand that further we are working with only about 80% of the data from all responses from survey. Now let's calculate the average prize pool for this year's competitions and calculate the share of successfull Kagglers and usual Kagglers.

```python
# calculate the average prize pool for this year's competitions
data_comp = data_competitions.copy()
data_comp = data_comp[(pd.to_datetime(data_comp['DeadlineDate']) > pd.to_datetime('2020-01-01 23:59:59')) & 
                      (pd.to_datetime(data_comp['DeadlineDate']) < pd.to_datetime('2021-01-01 23:59:59'))]
data_comp = data_comp[(data_comp.RewardQuantity.notnull()) & 
                  (data_comp.RewardType == 'USD')]

# remove competition with $1,000,000 prizes
data_comp = data_comp.RewardQuantity.sort_values(ascending=False).iloc[1:]
data_1st_place_comp = data_comp/2

average_1st_prize = round(data_1st_place_comp.sum()/data_1st_place_comp.count())

#-------------------------------------------
# Dataset Successful/Usual DS Kaggle users that participated in survey

data_usual_users_working = data_working_salaries[(data_working_salaries.Q24_int < average_1st_prize*2)]
data_usual_users_studying = data_studying_salaries[(data_studying_salaries.Q24_int < average_1st_prize*2)]
data_successful_users_working = data_working_salaries[(data_working_salaries.Q24_int >= average_1st_prize*2)]
data_successful_users_studying = data_studying_salaries[(data_studying_salaries.Q24_int >= average_1st_prize*2)]

# dataframe with usual users
data_u = data_usual_users_working.append(data_usual_users_studying)

# dataframe with successful users
data_s = data_successful_users_working.append(data_successful_users_studying)

count_data_usual = data_u['Q1'].count()
count_data_successful = data_s['Q1'].count()

sum_users = [count_data_usual, count_data_successful]

#-------------------------------------------

labels = ['Usual DS Users', 'Successful DS Users']
fig = go.Figure()
fig.add_trace(go.Pie(values=sum_users, 
                     labels=labels, 
                     textinfo='label+percent', 
                     showlegend=False, 
                     insidetextorientation='radial', 
                     hole=0.3))

fig.update_layout(
    title="Successful/Usual DS Kaggle users",
    title_x = 0.5,
    margin=dict(l=0, r=0, t=40, b=0),
    legend_orientation="h")
fig.show()
```

The Usual Kaggler group is consist of students and a portion of the working users whose annual salaries are below twice the average prize of all competitions in 2020. The average prize for all competitions in 2020 is approximately 17,000 dollars. 
Accordingly, all working users, whose annual income is less than 34,000 dollars, fall into the category of Usual Kagglers. The group of Successful Kagglers includes almost all working users, whose annual income is above 34,000 dollars.

Thus, we got two groups of interest to us - Successful Kagglers (24.6\%) and Usual Kagglers (75.4\%). Next, we will go through all the issues in terms of these two groups and draw up a portrait of each of them. Let's start!

To compare salaries by country, below I calculated the average salaries of working survey participants, depending on the country.
```python
countries = data_working_salaries[(data_working_salaries.Q3 != 'Other')].groupby('Q3')[['Q1']].count().sort_values(by='Q1', ascending=False)
countries.reset_index(inplace=True)

# TOP-20 countries by amount of participants
countries = countries[(countries['Q1'] > 80)]

avg_country_salaries = pd.DataFrame(columns = ['Average_Salary'], dtype='int64')


for country in countries['Q3']:
    data_country_salaries = data_working_salaries[(data_working_salaries.Q3 == country)]
    data_country_salaries = data_country_salaries.groupby('Q24_int')[['Q1']].count()
    data_country_salaries.rename(columns={"Q1": "count"},inplace=True)
    data_country_salaries.reset_index(inplace=True)
    data_country_salaries = data_country_salaries[data_country_salaries['count'] > data_country_salaries['count'].mean(0)].sort_values('count',ascending=False)
    avg_country_salaries.loc[country] = round((data_country_salaries.prod(1).sum(0))/data_country_salaries['count'].sum(0))
       
avg_country_salaries = avg_country_salaries.sort_values(by='Average_Salary', ascending=False)

fig = px.bar(avg_country_salaries, y=avg_country_salaries.index, x="Average_Salary",text="Average_Salary", orientation="h", range_x=[0,120000])

fig.update_traces(
    hovertemplate="Country: %{y}<br>Summa: %{x:,.f}",
    marker_color='#4682B4',
    marker_opacity=0.7,
    texttemplate="%{text:.}",
    textfont=textfont,
    textposition='inside'
)
fig.update_layout(
    xaxis=dict(
        title_text="Annual average salaries, USD", 
        tickfont=tickfont,
        titlefont=axistitlefont
    ),
    yaxis=dict(
        categoryorder= "total ascending",
        tickfont=tickfont,
        titlefont=axistitlefont,
        tickformat=",", 
        title_text=""),
    hoverlabel=hoverlabelformat,
    margin=dict(pad=5, t=100),
    height=700,
    width=800,
    title=dict(text="Average salaries of working users by countries (TOP-20)", font=titlefont),
)
fig.show()
```
