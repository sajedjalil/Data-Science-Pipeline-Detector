```python
df = pd.read_csv('../input/kaggle-survey-2020/kaggle_survey_2020_responses.csv')
df = df.drop(0, axis=0)
```

```python
# Libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
# Function used to collect data with several columns
def coldat0(serie):
    X0 = df[[i for i in df.columns if serie in i]]
    X = pd.Series(dtype='int')
    for i in X0.columns:
        X[X0[i].value_counts().index[0]] = X0[i].count()
    return X.sort_values()
```

Let's start with the question : "Which of the hosted notebook products do you use on a regular basis?

```python
X = coldat0('Q10')
fig = px.treemap(X, path=[X.index], labels=X.index, values=X, title='Regular basis hosted notebook products')
fig.show()
```

Unsurprisingly, most of the nootbooks are hosted on Colab and Kaggle. Most of them are young Data Analyst as shown in the following Graph.

```python
index = [i for i in df.columns if 'Q10' in i]
X = pd.DataFrame([])
for ind in index:
    Inter = df.groupby(by=[ind,'Q1']).count()['Q3'].reset_index()
    Inter.columns = ['Q10','Q1','Q3']
    X = X.append(Inter)

fig = px.funnel(X, x='Q10', y='Q3', color='Q1', title = 'Age of Data Analyst and hosted notebook products')
fig.show()
```

So, which cloud computing platforms do Data Analyst expect to become more familiar with in the next 2 years?

```python
X = coldat0('Q26_B')
fig = px.pie(X, values=X, names=X.index)
fig.update_traces(textposition='inside')
fig.update_layout(uniformtext_minsize=12, uniformtext_mode='hide', title= 'Expected cloud computing platforms in next 2 years')
fig.show()
```

Yes, the answer for this question is clear. Although there is comparable competition between AWS and GCP plafroms, the MA's share is also significant.

```python
index = [i for i in df.columns if 'Q39' in i]
X = pd.DataFrame([])
for ind in index:
    Inter = df.groupby(by=[ind,'Q2']).count()['Q1'].reset_index()
    Inter.columns = ['Q39','Q2','Q1']
    X = X.append(Inter)
fig = px.histogram(X, y='Q1', x='Q39', color='Q2', marginal=None, title= 'Favorite media sources that report on data science topics')
fig.show()
```

Congratulations Kaggle ! on having the largest 'clients' of all media sources. This result matches the survey where the Data Analysts host their notebooks. But, data analysis or ML are shared the most on GitHub, not Kaggle.
