```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

# Gender

```python
def gender_distribution():
    data_1["Q2"][1:].replace(['Prefer to self-describe', 'Prefer not to say','Nonbinary'], 'Other', inplace=True)
    data = data_1["Q2"][1:].value_counts()
    figure, ax = plt.subplots(1,1, figsize=(10,10))
    explode = (0.1, 0.1, 0.1) 
    wedges = plt.pie(data, explode=explode, labels=data.index, autopct='%1.1f%%',
        shadow=True, startangle=90, colors=['lightsteelblue', 'firebrick', 'dimgray'], textprops={'fontsize': 25})
    plt.show()
gender_distribution()
```

**Notes**
* Majority of Kagglers are men.

# Ages exploration

```python
def ages_distribution():
    sns.set_style('ticks')
    figure, ax = plt.subplots(1,1,figsize=(20,8))
    data = data_1['Q1'][1:].value_counts()
    colors = ['lightsteelblue' if (x < max(data)) else 'firebrick' for x in data]
    graph = sns.barplot(y=data, x=data.index, palette=colors, data=data_1, edgecolor=(0, 0, 0), linewidth=1.2)
    sns.despine(offset=15, trim=True)
    ax.yaxis.labelpad = 25
    ax.xaxis.labelpad = 25
    plt.title('Ages distribuiton', fontdict={'fontsize' : 25})
    plt.ylabel('Quantity', fontdict={'fontsize':20})
    plt.xlabel('Age', fontdict={'fontsize':20})
    plt.show()

ages_distribution()
```

**Notes**
* Most kagglers seem to be around 18-29, biggest representation is in 25-29 group. This indicates growing intrest in data science among young people.


```python
def ages_vs_jobs():
    data_1['count'] = 1
    data = pd.pivot_table(data_1.iloc[1:,:], values='count', index='Q5', columns='Q1', aggfunc=np.sum)
    figure, ax = plt.subplots(1, 1, figsize=(15,15))
    heat = sns.heatmap(data, square=True, linewidth=.5, cmap='RdGy', ax=ax, cbar=False, annot=True, fmt="d")
    heat.tick_params(labelsize=15)
    ax.yaxis.labelpad = 25
    ax.xaxis.labelpad = 25
    plt.title('Ages vs jobs', fontdict={'fontsize': 25}, pad=20)
    plt.ylabel('Jobs', fontdict={'fontsize':15})
    plt.xlabel('Ages', fontdict={'fontsize':15})
    plt.show()
    
ages_vs_jobs()
```

**Notes**

* As expected people between the ages of 18-29 study a lot, especially the youngest group ( 18-21 ).
* Most jobs seem to be occupied with people aged at around 22-29. ( Later I'll return to this )

# Education

```python
def education():
    # Two values are very long, I am going to shorten them for looks
    data_1['Q4'][1:].replace("Some college/university study without earning a bachelorâ€™s degree", "Unfinished bachelor's", inplace=True)
    data_1['Q4'][1:].replace("No formal education past high school", "High school", inplace=True)
    data = pd.concat([data_1['Q4'][1:].value_counts()[:5], data_1['Q4'][1:].value_counts()[6:]])
    sns.set_style('ticks')
    figure, ax = plt.subplots(1,1,figsize=(18,8))
    colors = ['lightsteelblue' if (x < max(data)) else 'firebrick' for x in data]
    graph = sns.barplot(y=data.index, x=data, palette=colors, orient='h', edgecolor=(0, 0, 0), linewidth=1.2)
    graph.tick_params(labelsize=15)
    ax.yaxis.labelpad = 25
    ax.xaxis.labelpad = 25
    sns.despine(offset=15, trim=True)
    plt.title('Education levels distribution', fontdict={'fontsize': 25}, pad=20)
    plt.xlabel('Quantity', fontdict={'fontsize':15})
    plt.show()
education()
```

**Notes**
* Master's seems to be a standard.
* This correlates to fact that most of Kagglers are 25-29. Becouse 25-29 are majority, master's degree is also a majority ( people with master's are usually older than bachelors )

```python
def ages_vs_education():
    data_1['count'] = 1
    fig, ax = plt.subplots(1, 1, figsize=(18,18))
    data = pd.pivot_table(data_1.iloc[1:,:], values='count', index='Q4', columns='Q1', aggfunc=np.sum)
    heatmap = sns.heatmap(data, square=True, linewidth=.5, ax=ax, cbar=False, annot=True, fmt="d", cmap='RdGy')
    heatmap.tick_params(labelsize=15)
    ax.yaxis.labelpad = 25
    ax.xaxis.labelpad = 25
    plt.title('Age vs education', fontdict={'fontsize' : 25}, pad=20)
    plt.xlabel('Ages', fontdict={'fontsize':15})
    plt.ylabel('Eduacation', fontdict={'fontsize':15})
    plt.show()
    
ages_vs_education()
```

**Notes**
* Expectation from previous chart is fulfiled here - 25-29 is the most numerous group with master's degree.
* Group 22-24 has almost the same ammount of people with bachelor's and master's. That information further pushes the idea of growing intrest in data science among younger people, and shows that people really care about education in this field.


# Jobs

```python
def job():
    data = data_1['Q5'][1:].value_counts()
    sns.set_style("ticks")
    figure, ax = plt.subplots(1,1,figsize=(18,12))
    colors = ['lightsteelblue' if (x < max(data)) else 'firebrick' for x in data]
    graph = sns.barplot(y=data.index, x=data, palette=colors, orient='h', edgecolor=(0, 0, 0), linewidth=1.2)
    graph.tick_params(labelsize=15)
    ax.yaxis.labelpad = 25
    ax.xaxis.labelpad = 25
    sns.despine(offset=15, trim=True)
    plt.title('Jobs distribution', fontdict={'fontsize' : 25}, pad=20)
    plt.xlabel('Quantity', fontdict={'fontsize':15})
    plt.show()

job()
```

**Notes**
* Majority of Kagglers are still students
* Top 3 jobs ( excluding students ) are Data Scientists, Software Engineers and Other. By other i understand that these are people who are in collage and don't have a tech job yet, are freelancing, have their own company...
* High place of Software Engineers implies that many people were / are studying Computer Science, rather that Econmics, Finances etc.

```python
def education_vs_job():
    figure, ax = plt.subplots(1,1,figsize=(20,20))
    data_1['count'] = 1
    data = pd.pivot_table(data_1.iloc[1:,:], values='count', index=['Q4'], columns=['Q5'], aggfunc=np.sum)
    data.fillna(0, inplace=True)
    graph = sns.heatmap(data, cmap='RdGy', linewidth=.5, cbar=False, annot=True, fmt=".1f", square=True)
    graph.tick_params(labelsize=15)
    graph.set_xticklabels(rotation=90, labels=data.columns)
    ax.yaxis.labelpad = 25
    ax.xaxis.labelpad = 25
    plt.title('Education vs job', fontdict={'fontsize':25}, pad=20)
    plt.xlabel('Education', fontdict={'fontsize':15})
    plt.ylabel('Job', fontdict={'fontsize':15})
    plt.show()

education_vs_job()
```

**Notes**
* This chart explains why master's is so popular. It seems to be a requirement if you want a Data Science job. Generally looking, people with master's just have more ( 587 unemplyed vs 706 bachelors ) and nicer jobs.
* More bachelors have job as a Software Engineer than masters, it makes me feel like Software Engineer is just a bridge job between collage and Data Science job.
* 32% of people without higher education than high school have a job in tech ( which brings me hope for some kind of job in near future :D )

# Job distribution between the ages of 18 and 24

*Looking for all the undergraduates 

```python
def job_dist():
    data = []
    data_2 = []
    
    for i, age in enumerate(data_1['Q1']):
        if str(age) == '18-21':
            if str(data_1['Q5'][i]) != 'nan':
                data.append(data_1['Q5'][i])
                
    for i, age in enumerate(data_1['Q1']):
        if str(age) == '22-24':
            if str(data_1['Q5'][i]) != 'nan':
                data_2.append(data_1['Q5'][i])
                
    df = pd.DataFrame({'freq' : data})
    df_2 = pd.DataFrame({'freq' : data_2})
    sns.set_style("ticks")
    fig, ax = plt.subplots(1,1, figsize=(15,10))
    graph = sns.barplot(x=df_2['freq'].value_counts(), y=df_2['freq'].value_counts().index, color='firebrick', label='Ages: 22-24', edgecolor=(0, 0, 0), linewidth=1.3)
    sns.barplot(x=df['freq'].value_counts(), y=df['freq'].value_counts().index, color='lightsteelblue', label='Ages: 18-21', edgecolor=(0, 0, 0), linewidth=1.2)
    ax.legend(ncol=2, loc="lower right", frameon=True)
    ax.yaxis.labelpad = 25
    ax.xaxis.labelpad = 25
    graph.tick_params(labelsize=15)
    sns.despine(offset=15, trim=True)
    plt.title('Jobs before graduation', fontdict={'fontsize':25}, pad=20)
    plt.xlabel('Quantity', fontdict={'fontsize':15})
    plt.show()
    
job_dist()
```

**Notes**
* 42% of people between the ages of 18 and 24 have found a job in tech and 12% have found a job in Data Science or related field ( Machine Learning Engineer, Data Analyst ).
* 25% of people between the ages of 18 and 21 also have found a job in tech and 6% have found a job in Data Science or related field.( this one makes me really glad!! )

# Youngest kagglers

```python
def find_youngest(get_df=None):
    data = pd.DataFrame()
    countries = data_1['Q3'][1:].unique()
    nums = []
    for country in countries:
        s = 0
        for i, age in enumerate(data_1['Q1'][1:]):
            if age == '18-21' and data_1['Q3'][i+1] == country:
                s += 1
        nums.append(s)
        
    d = dict(zip(countries, nums))
    l = len(d.values())
    if not get_df:
        while True:
            try:
                index = list(d.values()).index(max(d.values()))
                yield max(d.values()), list(d.keys())[index]
                del d[list(d.keys())[index]]
            except ValueError:
                break
    else:
        df = pd.DataFrame.from_dict(d, orient='index')
        df.columns = ['Amount']
        return df
    
            
def find_your_country(country):       
    for i, x in enumerate(find_youngest()):
        if country in x:
            print(f"{country}'s place: {i+1}")
            break
```

# Most commonly used programming language

```python
def foo(part):
    ans = []
    for i in range(len(data_1[f'Q7_Part_{part}'][0])):
        if data_1[f'Q7_Part_{part}'][0][-i-1] != ' ':
            ans.append(data_1[f'Q7_Part_{part}'][0][-i-1])
        else:
            break
    return ''.join(reversed(ans))

def most_commonly_used():
    data = pd.DataFrame()
    for i in range(1,12):
        data[f'{foo(i)}'] = [len([x for x in data_1[f'Q7_Part_{i}'][1:] if x == foo(i)])]
    return data

mcu = most_commonly_used()

def plot_langs():
    sns.set_style("ticks")
    figure, ax = plt.subplots(1,1, figsize=(20,8))
    colors = ['lightsteelblue' if (x < max(mcu.iloc[0, :])) else 'firebrick' for x in mcu.iloc[0, :]]
    graph = sns.barplot(x=mcu.columns, y=mcu.iloc[0, :], palette=colors, edgecolor=(0, 0, 0), linewidth=1.2)
    for p in graph.patches:
        graph.annotate(format(p.get_height(), '.2f'), 
                   (p.get_x() + p.get_width() / 2., p.get_height()), 
                   ha = 'center', va = 'center', 
                   xytext = (0, 9), 
                   textcoords = 'offset points')
    graph.tick_params(labelsize=15)
    ax.yaxis.labelpad = 25
    ax.xaxis.labelpad = 25
    sns.despine(offset=15, trim=True)
    plt.ylabel('Quantity', fontdict={'fontsize':15})
    plt.title('Most commonly used programming language', fontdict={'fontsize':25})
    plt.show()
plot_langs()
```

**Notes**
* Python > R.
* Second most commonly used language is SQL. Which logicaly makes a valuable tool in Data Scientist's toolkit.
* C, C++, Java and Javascript are also quite popular. I think it's becouse that most people have a degree in Computer Science, becouse this languages aren't that usefull for Data Science.
