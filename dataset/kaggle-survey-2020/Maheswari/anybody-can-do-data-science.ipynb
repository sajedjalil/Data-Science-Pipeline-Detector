```python
def create_category_wise_countplot_df(df,column_pattern,across_cols = True):
    if across_cols == True:
        subset_df = df[df.filter(like = column_pattern).columns].describe()
        count_df = subset_df.T[['top','count']].reset_index(drop=True).rename(columns = {'top':'index'})
    else:
        count_df = df[df.filter(like = column_pattern).columns][column_pattern].value_counts().to_frame().reset_index()
        count_df.rename(columns={ count_df.columns[1]: "count" },inplace = True)
    return(count_df)
```

## What comes out to be the most preferred programming language out of all?
> *Python* wins the race!<br><br>

```python
reg_prog_lang_count_df = create_category_wise_countplot_df(kaggle_survey_datamatrix,'Q7',across_cols=True)
first_prog_lang_count_df = create_category_wise_countplot_df(kaggle_survey_datamatrix,'Q8',across_cols=False)

plt.figure(figsize=(12,7))
sns.set_style('whitegrid')
plt.tight_layout()
plt.suptitle('In their words - What is the go-to programming language? \n\n',fontsize=25, y=1)

plt.subplot(1,2,1);
total = reg_prog_lang_count_df['count'].sum()
values = reg_prog_lang_count_df.sort_values('index')['count'].astype('int')
colors = ['#9ecae1' if (x < max(values)) else '#3182bd' for x in values ]
ax = sns.barplot(x="count", y="index", data=reg_prog_lang_count_df.sort_values('index'),palette=colors)
ax.set(ylabel='Programming Language', xlabel='# Data Science Enthusiasts')
ax.set_title("Regularly used Programming Language", fontsize=15)
ax.set_xticklabels(ax.get_xticklabels(), rotation=90)
for p in ax.patches:
        percentage = '{:.1f}%'.format(100 * p.get_width()/total)
        x = p.get_x() + p.get_width() + 0.02
        y = p.get_y() + p.get_height()/2
        ax.annotate(percentage, (x, y))

plt.subplot(1,2,2);
total = first_prog_lang_count_df['count'].sum()
values = first_prog_lang_count_df.sort_values('index')['count'].astype('int')
colors = ['#9ecae1' if (x < max(values)) else '#3182bd' for x in values ]
ax = sns.barplot(x="count", y="index", data=first_prog_lang_count_df.sort_values('index'),palette=colors)
ax.set(ylabel='Programming Language', xlabel='# Data Science Enthusiasts')
ax.set_title("Recommened Programming Language", fontsize=15)
ax.set_xticklabels(ax.get_xticklabels(), rotation=90)
# ax.yaxis.tick_right()
for p in ax.patches:
        percentage = '{:.1f}%'.format(100 * p.get_width()/total)
        x = p.get_x() + p.get_width() + 0.02
        y = p.get_y() + p.get_height()/2
        ax.annotate(percentage, (x, y))

plt.subplots_adjust(wspace=0.5)
plt.show()
```

## Where do you find the comfort for coding?
> *Jupyter* and *VSCode* seem to be the go-to IDE among the Data Enthusiasts.<br><br>
While, in terms of Hosted Notebook Product, *Collab* and *Kaggle Notebooks* are leading.

```python
plt.figure(figsize=(16,10))
sns.set_style('whitegrid')
plt.tight_layout()
plt.suptitle('C for Code, C for Comfort',fontsize=25, y=1)

ide_count_df = create_category_wise_countplot_df(kaggle_survey_datamatrix,'Q9',across_cols=True)
nb_count_df = create_category_wise_countplot_df(kaggle_survey_datamatrix,'Q10',across_cols=True)

plt.subplot(1,2,1);
ide_dict = dict(zip(ide_count_df['index'],ide_count_df['count']))
squarify.plot(sizes = ide_dict.values(),label=ide_dict.keys(),color=sns.color_palette('Set2')[0:len(ide_dict.values())],alpha = 0.7,
             text_kwargs={'rotation':90})
plt.axis('off')
plt.title("Integrated Development Environments (IDEs) of choice", fontsize=15)

plt.subplot(1,2,2);
total = nb_count_df['count'].sum()
ax = sns.barplot(x="count", y="index", data=nb_count_df.sort_values('index'),palette='Set3')
ax.set(ylabel='Notebook Product', xlabel='# Data Science Enthusiasts')
ax.set_title('Hosted Notebook Product of choice',fontsize=15)
ax.set_xticklabels(ax.get_xticklabels(), rotation=90)
# ax.yaxis.tick_right()
sns.despine(offset=10, trim=True)
for p in ax.patches:
        percentage = '{:.1f}%'.format(100 * p.get_width()/total)
        x = p.get_x() + p.get_width() + 0.02
        y = p.get_y() + p.get_height()/2
        ax.annotate(percentage, (x, y))

plt.subplots_adjust(wspace=0.8)
plt.show()
```

## What are the computing platforms and specialized hardwares adopted?
> We see that most individuals opt for their *personal computer or laptop* as default computing platform, followed by *cloud computing platforms* and *deep learning workstations*.<br>This rightly points us to the inference that not all have access to cloud computing platforms or workstations, and that for simple tasks or day-to-day activities local systems are the most prevalent option. Something that pretty much everyone has access to.<br><br>
Looking into the specialized hardwares, we see a wide acceptance for *GPUs*

```python
plt.figure(figsize=(12,10))
sns.set_style('whitegrid')
plt.tight_layout()
plt.suptitle('Computing Platforms and Specialized Hardwares',fontsize=25, y=1)
plt.title('Not all have access to cloud computing platforms or workstations, and that for simple tasks or day-to-day activities \nlocal systems are the most prevalent option')

compute_platform_count_df = create_category_wise_countplot_df(kaggle_survey_datamatrix,'Q11',across_cols=False)
special_hw_count_df = create_category_wise_countplot_df(kaggle_survey_datamatrix,'Q12',across_cols=True)
tpu_count_df = create_category_wise_countplot_df(kaggle_survey_datamatrix,'Q13',across_cols=False).reindex([0,2,1,3,4])

plt.subplot(1,2,1);
# total = nb_count_df['count'].sum()
ax = sns.barplot(x="count", y="index", data=compute_platform_count_df.sort_values('index'),palette='Set3')
ax.set(ylabel='Computing Platform', xlabel='# Data Science Enthusiasts')
ax.set_title('Computing Platform preferred',fontsize=15)
ax.set_yticklabels(['A cloud computing platform \n (AWS, Azure, GCP, hosted notebooks, etc)',
 'A deep learning workstation \n (NVIDIA GTX, LambdaLabs, etc)',
 'A personal computer or laptop',
 'None',
 'Other'])
ax.set_xticklabels(ax.get_xticklabels(), rotation=90)
sns.despine(offset=10, trim=True)
for p in ax.patches:
        percentage = '{:.1f}%'.format(100 * p.get_width()/total)
        x = p.get_x() + p.get_width() + 0.02
        y = p.get_y() + p.get_height()/2
        ax.annotate(percentage, (x, y))

plt.subplot(2,2,2);
special_hw_dict = dict(zip(special_hw_count_df['index'],special_hw_count_df['count']))
plt.pie(x=special_hw_dict.values(),colors=sns.color_palette('Set2')[0:len(special_hw_dict.values())],
       autopct='%.2f%%', pctdistance=1.1, labeldistance=1.2)
plt.axis('equal')
plt.title('In case, specialized hardwares are used - \n What are they?',fontsize=15)
plt.legend(labels=special_hw_dict.keys(),bbox_to_anchor=(1,0.8), loc='upper left',fontsize=10,title = 'Special Hardware')

plt.subplot(2,2,4);
ax = sns.lineplot(data=tpu_count_df, x="index", y="count",sort = False)
ax.set(xlabel='Frequency', ylabel='Number of Data Science Enthusiasts')
ax.set_title("If TPUs were used - \n How many times they were?", fontsize=15)
ax.set_xticklabels(tpu_count_df['index'], rotation=45)

plt.subplots_adjust(wspace=0.5)
plt.show()
```
