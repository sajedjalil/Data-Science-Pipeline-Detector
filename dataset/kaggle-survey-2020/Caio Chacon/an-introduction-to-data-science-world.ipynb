Now you may be asking yourself, "omg what functions are these? I thought I would be a Data Scientist". So, my dear friend, in a way, they are all "Data Scientists", because they all work with data, but each one does something different. I will explain to you what each one does:

* **Data Scientist**: Data science is an inter-disciplinary field that uses scientific methods, processes, algorithms and systems to extract knowledge and insights from many structural and unstructured data. Data science is related to data mining, machine learning and big data.

* **Software Engineer**: Software engineering is the systematic application of engineering approaches to the development of software. Software engineering is a computing discipline.

* **Data Analyst**: The Data Analyst is the professional whose focus of analysis and problem solving relates to data, types of data, and relationships among data elements within a business system or IT system.

* **Research Scientist**: Research scientists are responsible for designing, undertaking and analysing information from controlled laboratory-based investigations, experiments and trials.

* **Statistician**: A statistician is a person who works with theoretical or applied statistics. The profession exists in both the private and public sectors. statisticians are concerned with the collection, analysis, interpretation and presentation of quantitative information

* **Product/Project Manager**: Products/Project Manager is responsible for the project coordination, planning and research required for the identification and introduction of new products and busines opportunities.

* **Machine Learning Engineer**: Machine learning engineers sit at the intersection of software engineering and data science. They leverage big data tools and programming frameworks to ensure that the raw data gathered from data pipelines are redefined as data science models that are ready to scale as needed.

* **Business Analyst**: Business analyst help guide businesses in improving processes, products, services and software through data analysis. These agile workers straddle the line between IT and the business to help bridge the gap and improve efficiency. 

* **DBA/Database Engineer**: Database engineers create and maintain databases and database programs. Most companies that digitally store large amounts of information hire database engineers to design and maintain their database systems.

* **Data Engineer**: A data engineer is a worker whose primary job responsibilities involve preparing data for analytical or operational uses. The specific tasks handled by data engineers can vary from organization to organization but typically include building data pipelines to pull together information from different source systems; integrating, consolidating and cleansing data.

As we can see, most people are still students, and those who work are well divided between Data Scientist, Software Engineer, Data Analyst or Research Scientist, but notice that the graph is well divided. This means that you can follow the area that you feel most comfortable with!
