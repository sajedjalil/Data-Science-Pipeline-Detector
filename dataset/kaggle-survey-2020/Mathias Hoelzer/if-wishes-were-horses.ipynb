# Input data

First of course the input data need to be taken into the Notebook. In that step the "No / None" answers for the professionals has been replaced by "None" to make the plotting and data exploration easier afterwards.

```python
Surv2020 = pd.read_csv('/kaggle/input/kaggle-survey-2020/kaggle_survey_2020_responses.csv')
Surv2020.replace(to_replace="No / None", value="None", inplace=True)

Question_Text = Surv2020.iloc[0,:]
Surv2020.drop(0,axis=0,inplace=True)
```

```python
Surv2020.head()
```

Now we need to separate both groups to work further with them. According to the provided documentation participants that have answered to be Student or Not Employed or having spend no Money in the cloud are considered as Non Professionals.  

```python
NonProf = Surv2020[(Surv2020['Q5'] == "Student") | (Surv2020['Q5'] == "Currently not employed") | (Surv2020['Q25'] == "$0 ($USD)") | (Surv2020['Q25'] == "NaN")]
Prof = Surv2020[(Surv2020['Q5'] != "Student") & (Surv2020['Q5'] != "Currently not employed") & ((Surv2020['Q25'] != "$0 ($USD)") | (Surv2020['Q25'] == "NaN"))]
```

The next step and first investigate who are Professional and Non-Professionals according to the survey, so the following parameters are to be investigated:

1. Age ranges for each group;
2. Position held.

## Professionals vs. Non-Professionals


So here is the first exploration of the two groups.

```python
Prof_Group = pd.DataFrame(Prof['Q1'].value_counts())
NonProf_Group = pd.DataFrame(NonProf['Q1'].value_counts())

pie_Cl1A = go.Pie(labels=Prof_Group.index,values=Prof_Group.Q1,name="Professionals",hole=0.7,domain={'x': [0,0.4]})
pie_Cl2A = go.Pie(labels=NonProf_Group.index,values=NonProf_Group.Q1,name="Non-Prof",hole=0.7,domain={'x': [0.5,1]})

layout = dict(title = 'Age distribution amongst both groups', font=dict(size=15), legend=dict(orientation="h"),
              annotations = [dict(x=0.12, y=0.5, text='Professionals', showarrow=False, font=dict(size=20)),
                             dict(x=0.87, y=0.5, text='Non-Professionals', showarrow=False, font=dict(size=20)) ])

fig = dict(data=[pie_Cl1A, pie_Cl2A], layout=layout)
py.iplot(fig)
```

As it can be seen above (and not surprising) the Non profesionals are younger than the Professionals. For that reason and for the fact that most of the Non-Professionals are students or currently unemployed we could call them Next Generation Data Scientists.

```python
Prof_Group = pd.DataFrame(Prof['Q5'].value_counts())
NonProf_Group = pd.DataFrame(NonProf['Q5'].value_counts())

pie_Cl1A = go.Pie(labels=Prof_Group.index,values=Prof_Group.Q5,name="Professionals",hole=0.7,domain={'x': [0,0.4]})
pie_Cl2A = go.Pie(labels=NonProf_Group.index,values=NonProf_Group.Q5,name="Non-Prof",hole=0.7,domain={'x': [0.5,1]})

layout = dict(title = 'Job distribution amongst both groups', font=dict(size=15), legend=dict(orientation="h"),
              annotations = [dict(x=0.12, y=0.5, text='Professionals', showarrow=False, font=dict(size=20)),
                             dict(x=0.87, y=0.5, text='Non-Professionals', showarrow=False, font=dict(size=20)) ])

fig = dict(data=[pie_Cl1A, pie_Cl2A], layout=layout)
py.iplot(fig)
```

# The analysis

In this chapter the answers will be analyzed for bith groups individually to find out further trends and insights on future developments.

## Cloud computing products and platform use

```python
Quest = list({'Q26', 'Q27'})
Title = {"Q26": "Cloud computing platforms used (left) or planned to be used (right)",
        "Q27": "Cloud computing products used (left) or planned to be used (right)",
        "Q28": "ML products used (left) or planned to be used (right)",
        "Q29": "Big Data products used (left) or planned to be used (right)",
        "Q31": "Business Intelligence tools used (left) or planned to be used (right)",
        "Q33": "Automated ML tools used (left) or planned to be used (right)",
        "Q35": "Tools to manage ML experiements used (left) or planned to be used (right)",}
for i in Quest:   
    Cloud1 = Prof.filter(regex=i)
    Cloud2 = NonProf.filter(regex=i)
    Cloud1_A = Cloud1.apply(pd.value_counts)
    Cloud2_A = Cloud2.apply(pd.value_counts)
    Cloud1_A["sum"] = Cloud1_A.sum(axis=1)
    Cloud2_A["sum"] = Cloud2_A.sum(axis=1)
    Cloud1_A["sum_Norm"] = Cloud1_A["sum"]/len(Cloud1)*100
    Cloud2_A["sum_Norm"] = Cloud2_A["sum"]/len(Cloud2)*100

    Cloud1_A.sort_values(by = "sum_Norm", axis=0, ascending=False, inplace=True)

    trace1 = go.Bar(
        x=Cloud1_A.index,
        y=Cloud1_A['sum_Norm'],
        name='Professionals'
    )
    trace2 = go.Bar(
        x=Cloud2_A.index,
        y=Cloud2_A['sum_Norm'],
        name='Non-Professionals'
    )

    data = [trace1, trace2]
    printtitle = Title[i]
    py.offline.iplot({
        "data": data,
        "layout": go.Layout(barmode='group', title=printtitle,width=800,height=500,)
    })
```

As we can observe from the above graphs there is a great interest of Non-Professionals in all type of Computing platforms, this is mainly due to the fact that as this group of participants is not yet employed there is no need to focus on a specific product and/or platform.

With regards to Big Data and ML products there are some intersting features that could indicate what will be the new products to be used.

For the Big Data products MySQL is still the first choice for Professionals as well as Non-Professionals to be learnt. On the other hand other products are catching interest of the community such as MongoDB or GoogleCloud SQL for which a high interest of Non-Professionals exist to get to know these tools.

Regarding ML products the most surprising result might be that more than a quarter of Professionals do not use these tools at all in their daily work and that more than 12% of Non-Professionals have no plan to learn either of these products. When it comes to plans for learing these tools Google Cloud AI Platform as well as other Google products are the clear winner in this race.

The last two questions in the survey were about automated ML tools as well as tools to manage these ML experiements. 

What we can see here is quite interesting and might reflect two realities: 

The most obvious result is that a large relative mayority of Professionals is not using either of these tools. Additonally ML Management tools are not planned to be learnt by Non-Professionals either.
