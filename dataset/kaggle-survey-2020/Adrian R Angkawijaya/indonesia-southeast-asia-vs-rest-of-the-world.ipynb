# 2020 Kaggle ML & DS Survey

In this survey dataset answered by Kagglers all over the world, I am very curious to see the characteristics of kagglers from developing countries, especially in Indonesia (since I am an Indonesian myself). I would like to know if they are so different than kagglers from the U.S, Europe, India, etc.

I will compare kagglers from Indonesia to the rest of Southeast Asia and also the rest of the world. Let's see if we can find something interesting!

```python
import numpy as np
import pandas as pd

%matplotlib inline
%config InlineBackend.figure_format = 'retina'
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings("ignore")
```

```python
data = pd.read_csv('../input/kaggle-survey-2020/kaggle_survey_2020_responses.csv')
```

```python
data.head()
```

```python
##Filter out the first row with questions
data = data.iloc[1:]
```

**Distribution of countries where Kagglers who completed the survey are from:**

```python
sns.set(rc={'figure.figsize':(12,8)})
sns.set_style("whitegrid", {'axes.grid' : False})

##Country
sns.countplot(y = 'Q3', data =data, order = pd.value_counts(data['Q3']).iloc[:20].index, palette='tab10')
plt.title("Where are the Kagglers from?", fontweight="bold", fontsize=18)
plt.xlabel("")
plt.ylabel("")
plt.yticks(fontsize = 14)
plt.show()
```

Most of the Kagglers are from <b>India</b>. The number of Kagglers from Indonesia in the dataset is more than I first expected. I will be able to use this sample for exploration. 

```python
##Gender
sns.countplot(y = 'Q2', data =data, order = pd.value_counts(data['Q2']).index, palette='tab10')
plt.title("Gender Distribution of the Kagglers", fontweight="bold", fontsize=18)
plt.xlabel("")
plt.ylabel("")
plt.yticks(fontsize = 14)
plt.show()
```

```python
##Job Title
sns.countplot(y = 'Q5', data =data, order = pd.value_counts(data['Q5']).index, palette='tab10')
plt.title("What are the Kagglers working as?", fontweight="bold", fontsize=18)
plt.xlabel("")
plt.ylabel("")
plt.yticks(fontsize = 14)
plt.show()
```

Most of the Kagglers are <b>students</b>. I will filter out the students later because I want to get to know the difference in characteristics for Kagglers that are employed.

Next, let's divide the datasets to see the comparison of Indonesian kagglers with the rest of Southeast Asia & the world:

- <b>Indonesia</b>: Indonesia only
- <b>Southeast Asia</b>: Singapore, Vietnam, Malaysia, Thailand, Phillipines
- <b>Rest of the World</b>: Other countries outside of the ones mentioned above

```python
##Filter out students 
data = data[data['Q5'] != 'Student']

##Indonesia
indo = data[data['Q3'] == 'Indonesia']

##Asia
southeast_asia_list = ['Singapore', 'Viet Nam', 'Malaysia', 'Thailand', 'Philippines']

se_asia = data[data['Q3'].isin(southeast_asia_list)]

##Rest of the world
not_included = ['Indonesia', 'Singapore', 'Viet Nam', 'Malaysia', 'Thailand', 'Philippines']

world = data[~data['Q3'].isin(not_included)]
```

- **Age & Education level**

```python
fig, axs = plt.subplots(nrows = 3, ncols = 2, figsize=(20,10))
sns.countplot(y = 'Q1', data = indo, order = pd.value_counts(indo['Q1']).index, ax = axs[0][0], palette = 'Set1').set_title('Indonesia (Age)', fontweight="bold", size=18)
sns.countplot(y = 'Q1', data = se_asia, order = pd.value_counts(se_asia['Q1']).index, ax = axs[1][0], palette = 'Dark2').set_title('Southeast Asia (Age)', fontweight="bold", size=18)
sns.countplot(y = 'Q1', data = world, order = pd.value_counts(world['Q1']).index, ax = axs[2][0],palette = 'tab10').set_title('Rest of the World (Age)', fontweight="bold", size=18)
sns.countplot(y = 'Q4', data = indo, order = pd.value_counts(indo['Q4']).index, ax = axs[0][1], palette = 'Set1').set_title('Indonesia (Education)', fontweight="bold", size=18)
sns.countplot(y = 'Q4', data = se_asia, order = pd.value_counts(se_asia['Q4']).index, ax = axs[1][1], palette = 'Dark2').set_title('Southeast Asia (Education)', fontweight="bold", size=18)
sns.countplot(y = 'Q4', data = world, order = pd.value_counts(world['Q4']).index, ax = axs[2][1],palette = 'tab10').set_title('Rest of the World (Education)', fontweight="bold", size=18)

for ax in axs.flat:
    ax.set_ylabel('') 
    ax.set_xlabel('')
    ax.tick_params(labelsize=16)
    
plt.tight_layout()
```

Even though most of the Kagglers from Indonesia seem younger than the rest of the world & Southeast Asia, the age difference is not that significant.

The difference can be seen in the education level where most of Kagglers from Indonesia only have a bachelor's degree, but Master's degree is the most popular education level for the other parts of the world.

There are also only few Indonesia kagglers that have Doctoral degree if compared to the rest of the world & Southeast Asia.
