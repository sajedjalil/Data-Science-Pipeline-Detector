```python
# Create function to splite dataframe to subdataframes
def split_dataframe(data, parts=None, chunks=None):
    """parts arg is the number of parts to be partitioned data into,
       chunks is the size of each piece of data, that is, the number of lines in each segment"""
    data_list = []
    if parts == None:
        if chunks == None:
            raise('ValueError : \'parts\' and \'chunks\' are None')
        else:
            if type(chunks) != int:
                raise('ValueError : type of \'chunks\' must be integer, but it \'%s\''%type(chunks))
            else:
                rows   = data.shape[0]
                n_part = chunks
                start  = 0
                stop   = chunks
                for i in range(0, parts):
                    subdata = data.iloc[start:stop, :]
                    data_list.append(subdata)
                    start += n_part
                    stop  += n_part
                return data_list
    else:
        if type(parts) != int:
            raise('ValueError : type of \'parts\' must be integer, but it \'%s\''%type(parts))
        else:
            rows   = data.shape[0]
            n_part = int(np.ceil(rows/parts))
            start  = 0
            stop   = n_part
            for i in range(0, parts):
                subdata = data.iloc[start:stop, :]
                data_list.append(subdata)
                start += n_part
                stop  += n_part
            return data_list
```

```python
gender_country = gender_country.sort_values('Man', ascending=False)
df_list        = split_dataframe(gender_country, parts=8)
ndx            = [x.index for x in df_list]
```

```python
fig, ax = mpp.subplots(8, 1, figsize=(10, 18), constrained_layout=True)
# ax return array contents subplots try printing it.
for i in range(len(df_list)):
    ax[i].bar(ndx[i], df_list[i]['Man'], color='blue', width=0.2, align='center')
    ax[i].bar(ndx[i], df_list[i]['Woman'], color='red', width=0.2, align='center')
    ax[i].bar(ndx[i], df_list[i]['Other'], color='yellow', width=0.2, align='center')

mpp.show()
```

<h2 style="color:#fff;background-color:blue;padding:5px;text-align;left;">Languages Users</h2>

```python
cols  = survey.loc[:, 'Q7_Part_1':'Q9_OTHER'].columns
total = survey.shape[0]
Ndx   = []
Cnt   = []
Pcn   = []
for col in cols:
    s = survey[col].value_counts()
    Ndx.append(s.index[0])
    Cnt.append(s[0])
    Pcn.append(round((s[0]/total)*100, 2))
    #print(f'{s.index[0]} \n{s[0]} \n{round((s[0]/total)*100, 2)}% \n')
lang_stc = pd.DataFrame({'Counts':Cnt[:13], 'Percentage':Pcn[:13]}, index=Ndx[:13])
lang_stc
```

```python
fig, ax   = mpp.subplots(1, 1, figsize=(12, 8))
bars_num  = lang_stc.shape[0]
color_map = sb.color_palette('hls', bars_num)
lang_stc  = lang_stc.drop('None', axis=0)
ax.bar(lang_stc.index, lang_stc['Counts'], color=color_map)

for i in lang_stc.index:
    ax.annotate(f"{lang_stc.at[i, 'Counts']}", xy=(i, lang_stc.at[i, 'Counts']+200), va='center', ha='center', 
                color='#222')

for b in ['top', 'left', 'right']:
    ax.spines[b].set_visible(False)

ax.set_yticks([])
fig.text(0.1, 0.9, 'Languages Users', fontsize=28, fontweight='bold', fontfamily='serif')
mpp.show()
```
