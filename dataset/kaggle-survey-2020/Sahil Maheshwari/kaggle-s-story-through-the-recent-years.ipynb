```python
from math import pi
from bokeh.io import output_file, show
from bokeh.palettes import RdYlBu,Viridis
from bokeh.plotting import figure
from bokeh.transform import cumsum


x = {
      "India":5851,
      "United States of America":2237,
      "Other":1388,
      "Brazil":694,
      "Japan":638,
      "Russia":582,
      "United Kingdom and Ireland": 489,
      "Nigeria":476,
      "China":474,
      "Germany":404
}

data = pd.Series(x).reset_index(name='value').rename(columns={'index':'country'})
data['angle'] = data['value']/data['value'].sum() * 2*pi
data['color'] = RdYlBu[len(x)]

gp = figure(plot_height=400, title="Distribution of Kagglers in Top Ten Countries-2020", toolbar_location=None,
           tools="hover", tooltips="@country: @value", x_range=(-0.5, 1.0),background_fill_color="beige", plot_width=600)

gp.wedge(x=0, y=1, radius=0.4,
        start_angle=cumsum('angle', include_zero=True), end_angle=cumsum('angle'),
        line_color="white", fill_color='color', legend_field='country', source=data)

gp.axis.axis_label=None
gp.axis.visible=False
gp.grid.grid_line_color = None

gp.title.text_font_size = '18pt'
gp.title.text_color = "maroon"
gp.title.text_font = "times"
gp.title.text_font_style = "italic"

show(gp)
```

It seems that Kaggle is heavily used in India, over 25% survey users are of India !!!, USA has 2nd highest users. It would be interesting to see the gender distribution in different countries.
