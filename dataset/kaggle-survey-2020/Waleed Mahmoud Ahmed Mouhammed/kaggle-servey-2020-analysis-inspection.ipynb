# Programming Languages 
- Q7 What programming languages do you use on a regular basis ?
-  **Python** The Top of Most Language using With Programmers.

```python
#plotly HBar
lang = df.filter(regex='Q7').copy()
col_lang = ['Python','R','SQL','C','C++','Java',
                'Javascript','Julia','Swift','Bash',
                'MATLAB','None','Other']
lang.set_axis(col_lang,
                axis=1,inplace=True)
lang.fillna(0,inplace=True)
lang = lang.apply(lambda x: [1 if n else 0 for n in x ])
sorted_lang = lang.sum().sort_values(ascending=True)
fig = px.bar(sorted_lang, text='value', labels={'value':'Use on a regular',
                                                'index':'programming languages'},orientation='h',)
fig.show()
```

# Current Role (Career)& Code Write Experince
- How many years title have been writing code/programing? relation with current Role(Career)

```python
job = pd.DataFrame(df['Q5'].value_counts()).reset_index()
job.columns = ['Q5', 'count']


q=df[['Q5','Q6']]

x= pd.DataFrame(q.groupby('Q5')['Q6'].value_counts())
x.columns = ['n']
x.reset_index(inplace=True)

x = x.merge(job, how='left', on='Q5')
x['count'] = ((x['n']/x['count'])*100).round(decimals=1)

piv = x.pivot('Q5', 'Q6', 'count')

plt.figure(figsize=(10,8))
sns.heatmap(piv, annot=True, linewidths=.2, cmap ='Oranges', cbar_kws={'shrink':0.7, 'label':'% responses for Current Role and Code experiences'})
plt.title('Current Role & code experince written')
plt.show()
```


# Countries in details
