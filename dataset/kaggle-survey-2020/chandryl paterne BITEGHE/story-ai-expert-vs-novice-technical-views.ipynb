```python
def plotting_pie(quiz):
    dic = {}
    for i in data['For how many years have you been writing code and/or programming?'].unique():

        df = data[data['For how many years have you been writing code and/or programming?'] == i]

        tab = []
        for j in df.columns:
            if quiz.lower() in j.lower():

                tab.append([j.split('e - ')[1], np.shape(df)[0] - df[j].isnull().sum()])

        Tab = pd.DataFrame(tab, columns=['label','value'])
        Tab = Tab[Tab['label'].isin(['None','No / None']) ==False]

        dic[i] = Tab

    #general structure
    code = []
    for i in data.columns:
      if quiz.lower() in i.lower():
        code.append([i.split('e - ')[1], np.shape(data)[0] - data[i].isnull().sum()])

    code = pd.DataFrame(code, columns=['label', 'value'])
    code = code[code['label'].isin(['None','No / None']) == False]

    dic['general'] = code

    #figure constructtion
    fig = make_subplots(
        rows=3, cols=3,
        specs=[[{"colspan":3, 'type':'domain'}, None,None],
               [{'type':'domain'},{'type':'domain'},{'type':'domain'}],
               [{'type':'domain'}, {'type':'domain'}, {'type':'domain'}]]    )

    fig.add_trace(go.Pie(values=dic.get('general')['value'].values,
                         labels=dic.get('general')['label'].values,
                         title="Global view"
                        ),
                  row=1, col=1)


    fig.add_trace(go.Pie(labels=dic.get('< 1 years')['label'].values,
                         values=dic.get('< 1 years')['value'].values,
                         title="Less than 1 year"
                        ),
                  row=2, col=1)

    fig.add_trace(go.Pie(labels=dic.get('1-2 years')['label'],
                         values=dic.get('1-2 years')['value'],
                         title="Between 1 and 2 years"),
                  row=2, col=2)

    fig.add_trace(go.Pie(labels=dic.get('3-5 years')['label'],
                         values=dic.get('3-5 years')['value'],
                         title="Between 3 and 5 years"),
                  row=2, col=3)


    fig.add_trace(go.Pie(labels=dic.get('5-10 years')['label'],
                         values=dic.get('5-10 years')['value'],
                         title="Between 5 and 10 years"),
                  row=3, col=1)


    fig.add_trace(go.Pie(labels=dic.get('10-20 years')['label'],
                         values=dic.get('10-20 years')['value'],
                         title="Between 10 and 20 years"),
                  row=3, col=2)

    fig.add_trace(go.Pie(labels=dic.get('20+ years')['label'],
                         values=dic.get('20+ years')['value'],
                         title="More than 20 years"),
                  row=3, col=3)

    fig.update_layout(height=1100, width=1100, title_text=quiz)
    return fig.show()
```

```python
def plotting_bar(quiz):
    dic = {}
    for i in data['For how many years have you been writing code and/or programming?'].unique():

        df = data[data['For how many years have you been writing code and/or programming?'] == i]

        tab = []
        for j in df.columns:
            if quiz.lower() in j.lower():

                tab.append([j.split('e - ')[1], np.shape(df)[0] - df[j].isnull().sum()])

        Tab = pd.DataFrame(tab, columns=['label','value'])
        Tab = Tab[Tab['label'].isin(['None','No / None']) == False]

        dic[i] = Tab

    #general structure
    code = []
    for i in data.columns:
      if quiz.lower() in i.lower():
        code.append([i.split('e - ')[1], np.shape(data)[0] - data[i].isnull().sum()])

    code = pd.DataFrame(code, columns=['label', 'value'])
    code = code[code['label'].isin(['None','No / None']) == False]

    dic['general'] = code

    #figure constructtion
    fig = make_subplots(
        rows=3, cols=3,
        specs=[[{"colspan":3, "type": "xy"}, None,None],
               [{"type": "xy"},{"type": "xy"},{"type": "xy"}],
               [{"type": "xy"}, {"type": "xy"}, {"type": "xy"}]]    )

    fig.add_trace(go.Bar(y=dic.get('general')['value'].values,
                         x=dic.get('general')['label'].values,
                         name="Global view"
                        ),
                  row=1, col=1)


    fig.add_trace(go.Bar(x=dic.get('< 1 years')['label'].values,
                         y=dic.get('< 1 years')['value'].values,
                         name="Less than 1 year"
                        ),
                  row=2, col=1)

    fig.add_trace(go.Bar(x=dic.get('1-2 years')['label'],
                         y=dic.get('1-2 years')['value'],
                         name="Between 1 and 2 years"),
                  row=2, col=2)

    fig.add_trace(go.Bar(x=dic.get('3-5 years')['label'],
                         y=dic.get('3-5 years')['value'],
                         name="Between 3 and 5 years"),
                  row=2, col=3)


    fig.add_trace(go.Bar(x=dic.get('5-10 years')['label'],
                         y=dic.get('5-10 years')['value'],
                         name="Between 5 and 10 years"),
                  row=3, col=1)


    fig.add_trace(go.Bar(x=dic.get('10-20 years')['label'],
                         y=dic.get('10-20 years')['value'],
                         name="Between 10 and 20 years"),
                  row=3, col=2)

    fig.add_trace(go.Bar(x=dic.get('20+ years')['label'],
                         y=dic.get('20+ years')['value'],
                         name="More than 20 years"),
                  row=3, col=3)

    fig.update_layout(height=1000, width=1000, title_text=quiz)
    return fig.show()
```
