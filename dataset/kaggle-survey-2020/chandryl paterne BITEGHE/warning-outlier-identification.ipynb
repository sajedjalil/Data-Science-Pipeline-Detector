# Age and Diploma üéì

**INTRODUCTION:**

Depending on the country education system, the duration for obtaining a diploma can

In Europe, most of country respect [Bologna process](https://www.unipage.net/en/degrees_academic) which establishes the bachelor education at 3 years, master 2 years and 3 years for the doctorat.

Others countries may respect others duration rule for graduation

**ANALYSIS:**

In the figure Below we saw some profil get graduated by a doctorat at 18-21 years which means that they got an hight school diploma at 1O-13 years old.

AltoughThese situations are not possible but that remain a worldwide exception. 

List of outlier in this situation:


*   18-21 = Doctorat
*   22-24 = Doctorat
*   18-21 = Master ( This situation is at the boundary to be a outlier but i will not considere this situation because it is possible with a 2 years of projection in the future )


```python
table = dataset.loc[:,['What is your age (# years)?',
                       'What is the highest level of formal education that you have attained or plan to attain within the next 2 years?']]

outlier =pd.pivot_table(table, index=['What is your age (# years)?'],
               columns=['What is the highest level of formal education that you have attained or plan to attain within the next 2 years?'],
               aggfunc='size')

fig = ff.create_annotated_heatmap(outlier.to_numpy(),x=list(outlier.columns), y=list(outlier.index),colorscale='Viridis')
fig.show()
```

```python
outlier = dataset[dataset['What is your age (# years)?'].isin(['18-21', '22-24'])]
outlier = outlier[outlier['What is the highest level of formal education that you have attained or plan to attain within the next 2 years?'].isin(['Doctoral degree'])]

outlier_country = outlier.groupby('In which country do you currently reside?').agg({'In which country do you currently reside?':'count'})
fig = px.bar(outlier_country, x=outlier_country.index,
             y='In which country do you currently reside?',
             title="Country containing outlier",
            height=800)
fig.show()
```

**CONCLUSION**

Some of the countries found in the outlier respect Bologna system, which confirm that they are outlier

More over, with some discussion with my collegues, getting a Doctorat at 18 years old is also impossible in UK, USA, Morroco, Brazil and India.

# Age and Programing Experience üíª

Other Outlier are identify by analysis the crosstable between the Age and experience in programming.

Some profil have age between 18-21 years old and more than 20 years of programing experience.That mean means they start computing even before they were born üòÇüòÇüòÇ.

This is an example of outlier that can be found in the figure below.

List of outlier:


*   18-21 = 20+ years 
*   18-21 = 10-20 years
*   18-21 = 5-10 years
*   22-24 = 20+ years 
*   22-24 = 10-20 years
*   22-24 = 5-10 years ( as This situation is possible in the 2 extrem case 24-5 starting at 19, i will not considere it)
*   25-29 = 20+ years
*   25-29 = 10-20 in ( as this situation is possible in the √© extrem case 29-10 starting at 19, i will not considere it but outlier can be find in )

```python
table = dataset.loc[:,['What is your age (# years)?',
                       'For how many years have you been writing code and/or programming?']]

outlier =pd.pivot_table(table, index=['What is your age (# years)?'],
               columns=['For how many years have you been writing code and/or programming?'],
               aggfunc='size')

outlier
fig = ff.create_annotated_heatmap(outlier.to_numpy(),x=list(outlier.columns), y=list(outlier.index),colorscale='Viridis')
fig.show()
```

```python
sub_outlier1 = dataset[dataset['What is your age (# years)?'].isin(['18-21', '22-24'])][dataset['For how many years have you been writing code and/or programming?'].isin(['10-20 years', '20+ years'])]
sub_outlier2 = dataset[dataset['What is your age (# years)?'].isin(['18-21'])][dataset['For how many years have you been writing code and/or programming?'].isin(['5-10 years'])]
sub_outlier3 = dataset[dataset['What is your age (# years)?'].isin(['25-29'])][dataset['For how many years have you been writing code and/or programming?'].isin(['20+ years'])]

outlier = pd.concat([sub_outlier1, sub_outlier2, sub_outlier3])
```

```python
outlier_country = outlier.groupby('In which country do you currently reside?').agg({'In which country do you currently reside?':'count'})
fig = px.bar(outlier_country, x=outlier_country.index,
             y='In which country do you currently reside?',
             title="Country containing outlier",
            height=800)
fig.show()
```
