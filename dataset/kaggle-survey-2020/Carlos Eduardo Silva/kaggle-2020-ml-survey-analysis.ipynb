<h2 style='background:#4974a5; border:50; color:white'>Creating a function for plotting interactive charts &#128200;<h2>

```python
def plotBar(x, y, title, xaxis, yaxis, orientation = 'v',text=True):
    
    #41e1b7 #f68ca5
    if text:
        data = go.Bar(
                x=x, y=y,
                text=y,
                textposition='auto',
                orientation = orientation, 
                opacity = 0.70
            )
    else:
        data = go.Bar(
                x=x, y=y,
                textposition='auto',
                orientation = orientation, 
                opacity = 0.60, 
                marker_color = '#41e1b7',
            )
    

    layout = go.Layout(
        title       = title,
        yaxis       =  {'title': yaxis},
        xaxis       =  {'title': xaxis},

    
        hoverlabel=dict(
        bgcolor="#f5f5f5",
        font_size=14,
    ),
)  
    fig = go.Figure(data = data, layout = layout, )
    
    
    fig.show()
```

<h2 style='background:#a4b9d2; border:50; color:black'>Distribuition Gender by Age<h2>


```python
dataQ1 = df['Q1'].value_counts().sort_index()
filterGender = dfFilterGender.groupby('Q2')['Q1'].value_counts().unstack().sort_index()
man = filterGender.loc['Man']
woman = filterGender.loc['Woman']
pman = np.round(man.values/man.sum()*100,decimals=2)
pwoman = np.round(woman.values/woman.sum()*100,decimals=2)

fig = go.Figure(data=[
    go.Bar(x=dataQ1.index, y=woman, name='Woman',text=pwoman, textposition="auto"),
    go.Bar(x=dataQ1.index, y = man, name='Man',text=pman, textposition="auto")
])

fig.update_traces(marker_line_color='rgb(8,48,107)',
                  texttemplate="%{value}<br>%{text:.2f}%", 
                  marker_line_width=1, opacity=0.80)

fig.update_layout(barmode='relative',
                 title_text='Distribuition Gender by Age',
                 hoverlabel=dict(
                 bgcolor="#f5f5f5",
                 font_size=14,
    ),
)

fig.show()
```

```python
dfQ24 = df[df["Q2"].isin(["Man","Woman"])].loc[:, ["Q2", "Q3", "Q5", "Q24"]].dropna()

dfQ24["Q24"] = (dfQ24["Q24"].apply(lambda x: x.replace(",", "")).apply(lambda x: x.replace("$", "")))

dfQ24["Q24"] = dfQ24["Q24"].apply(lambda x: 500000 if ">" in x else 0.5 * (int(x.split("-")[0]) + int(x.split("-")[1])))

df_salary = dfQ24.groupby(by=["Q2","Q5"])["Q24"].agg(['mean','size']).reset_index()

fig = go.Figure()

fig.add_trace(
    go.Scatter(
        x=df_salary.loc[df_salary['Q2']=='Man','mean'],
        y=df_salary.loc[df_salary['Q2']=='Man','Q5'],
        name="Man",
        mode="markers",
        marker=dict(
            sizemode="area",
            size=df_salary.loc[df_salary['Q2']=='Man','size'],
            color="#87ceeb",
        ),
    )
)
fig.add_trace(
    go.Scatter(
        x=df_salary.loc[df_salary['Q2']=='Woman','mean'],
        y=df_salary.loc[df_salary['Q2']=='Woman','Q5'],
        name="Woman",
        mode="markers",
        marker=dict(
            sizemode="area",
            size=df_salary.loc[df_salary['Q2']=='Woman','size'],
            color="#f68ca5",
        ),
    )
)


fig.update_traces(hovertemplate="%{y}<br>Salary:%{x:$,.0f}<br>Size:%{marker.size}")

fig.update_layout(title="Average Salary by Gender and Current Role", height = 700,
        xaxis=dict(
        showgrid=True,
        showline=True,
        tickfont=dict(size=14),
        showticklabels=True,
        title=" Average Annual Salary in USD"),  
        yaxis=dict(
        tickfont=dict(size=14) 
                )
        )


fig.show()
```

<h3>The current job that has the <mark>highest average salary</mark> is <mark> product/project manager</mark> and <mark> Data Scientist</mark>. We can also observe that there is a <mark>difference</mark> in the <mark>average salary</mark> of <mark>men</mark> for <mark>women</mark>, that is, with the same function, women receive much less than men.</h3>
