# OOP Phase 1: Object building 1/2

```python


""" Development of Class 1: datasetToProcess(object) """

class datasetToProcess(object):  
     """ External data is input in datasetToProcess to output a pivot table. """
     """ Processes are insertion and pivot-table building"""
    
     def __init__(self):
        """create empty sets of feature labels, pivot features, features to drop, and instance index"""
        self.featureLabels = []
        self.pivotFeatures = []
        self.featuresToDrop = []
        self.instanceIndex = []
       
        """create empty dataframes of dataset, connections, policies, and pivot table"""
        import pandas as pd
        d = {}
        df = pd.DataFrame(data=d)
       
        self.dataset = df
        self.connections = df
        self.policies = df
        self.pivotTable = df
        self.pivotTableConverted = df
        self.pivotTable_Description = df
        
       
        """create policySelected, policy and setOfPolcies as empty lists"""
       
        self.polcySelected = []
        self.policy = []
        self.setOfPolicies = []
      
    
     def insert(self,dataset, connections, policies, policySelected, setOfPolicies, featuresToDrop, pivotFeatures):
        """Insert dataframes: (1) dataset; (2) connections; (3) policies"""
        """ Insert lists; (1) policySelected, setOfPolicies, featuresToDrop, pivotFeatures"""
        """Set policy as a main-driver of a data story"""
        self.dataset = dataset
        self.connections = connections
        self.policies = policies
        self.policy = policySelected
        self.setOfPolicies = setOfPolicies
        self.featuresToDrop = featuresToDrop
        self.pivotFeatures = pivotFeatures
       
       
     def getPivotTable(self, dataset, featuresToDrop):
        """Get and return pivot table given a selected policy"""
        import pandas as pd
        
        # Droping features and description row (Row 0)
        self.pivotTable = self.dataset.drop(columns=self.featuresToDrop)
        self.pivotTable_Description = self.pivotTable
        self.pivotTable = self.pivotTable.drop([0])
        
        
        self.featureLabels = self.pivotTable.columns[1:]
        self.instanceIndex = self.pivotTable.index
        
        # Features typification: (1) Numbers (just 'Time from Start to Finish (seconds)'); (2) Categories
        
        ## Categories
        self.pivotTableConverted = self.pivotTable.convert_dtypes(convert_string=True, convert_integer=True)
        
        ## Number
        self.pivotTableConverted = self.pivotTableConverted.astype({'Time from Start to Finish (seconds)': 'int32'})
        
        return self.pivotTable, self.pivotTableConverted, self.pivotTable_Description, self.policy
    
    
     def __repr__(self):
        """Returns class report"""
      
        import pandas as pd
      
        # Set Time
        from datetime import date
        reportDate = date.today()
      
        # Dataset report
        if(self.dataset.size != 0):
          dataset_report = 'inserted'
        else:
          dataset_report = 'not inserted'
      
        # Connections report
        if(self.connections.size != 0):
          connections_report = 'inserted'
        else:
          connections_report = 'not inserted'
          
        # Policies report
        if(self.policies.size != 0):
          policies_report = 'inserted'
        else:
          policies_report = 'not inserted'
    
        # Pivot Table report
        if(self.pivotTable.size != 0):
          pivotTable_report = 'fetched'
        else:
          pivotTable_report = 'not fetched'
    
        # Policy report
        if(len(self.setOfPolicies) != 0):
          setOfPolicies_report = 'inserted'
        else:
          setOfPolicies_report = 'not inserted'
          
        # Pivot Features report
        if(len(self.pivotFeatures) != 0):
          pivotFeatures_report = 'inserted'
        else:
          pivotFeatures_report = 'not inserted'
          
        # Pivot Features report
        if(len(self.featureLabels) != 0):
          featureLabels_report = 'fetched'
        else:
          featureLabels_report = 'not fetched'
          
        # Instances report
        if(len(self.instanceIndex) != 0):
          instanceIndex_report = 'fetched'
        else:
          instanceIndex_report = 'not fetched'
    
      # Get report statements
        reportStatementList = []
        for i in range (0,24):
            reportStatementItem = i
            reportStatementList = reportStatementList + [reportStatementItem]
    
        reportStatementList[0] = "\n" + "\n========================================="
        reportStatementList[1] = "\nOOP Phase 1 | Report | Date:" + str(reportDate.year) + str('/') + str(reportDate.month) + str('/') + str(reportDate.day)
        reportStatementList[2] = "\n========================================="
        reportStatementList[3] = "\nObject: datasetToProcess(object)"
        reportStatementList[4] = "\n========================================="
        reportStatementList[5] = "\n(1) Inputs"
        reportStatementList[6] = "\n========================================="
        reportStatementList[7] = "\n-Dataset : " + str(dataset_report) + str('| ') + str(type(self.dataset))
        reportStatementList[8] = "\n-Connections: " + str(connections_report) + str('| ') + str(type(self.connections))
        reportStatementList[9] = "\n-Policies: " + str(policies_report) + str('| ') + str(type(self.policies))
        reportStatementList[10] = "\n-Selected Policy: " + str(self.policy)
        reportStatementList[11] = "\n-Available Policies: " + str(self.setOfPolicies)
        reportStatementList[12] = "\n-# of features to drop: " + str(len(self.featuresToDrop))
        reportStatementList[13] = "\n-# Pivot Features: " + str(self.pivotFeatures)
        reportStatementList[14] = "\n========================================="
        reportStatementList[15] = "\n(2) Process"
        reportStatementList[16] = "\n========================================="
        reportStatementList[17] = "\n-Pivot Table: " + str(pivotTable_report) + str('| ') + str(type(self.pivotTable))
        reportStatementList[18] = "\n========================================="
        reportStatementList[19] = "\n(3) Outputs"
        reportStatementList[20] = "\n========================================="
        reportStatementList[21] = "\n-instances selected: " + str(instanceIndex_report) + str('| ') + str(len(self.instanceIndex))
        reportStatementList[22] = "\n-features selected: " + str(featureLabels_report) + str('| ') + str(len(self.featureLabels))
        reportStatementList[23] = "\n========================================="
        
        reportStatement = []
       
        for i in range (0,24):
            if i == 0:
                reportStatement = reportStatementList[i]
            else:
                reportStatement = reportStatement + reportStatementList[i]
       
        return super().__repr__() + reportStatement
        
    
""" Test of Class 1: datasetToProcess(object) """
    
if testModePhase1 == True:

    # Get external data
    reportMode = reportModePhase0
    datasetExternal, connectionsExternal, policiesExternal = GetInstances(reportMode)
    policySelectedExternal, setOfPoliciesExternal, featuresToDropExternal, pivotFeaturesExternal = setPolicy(policiesExternal,reportMode)
    
    # Run class
    classObject1 = datasetToProcess()
    
    # Call method insert
    classObject1.insert(datasetExternal, connectionsExternal, policiesExternal, policySelectedExternal, setOfPoliciesExternal, featuresToDropExternal, pivotFeaturesExternal)
    
    # Call method getPivotTable
    classObject1.getPivotTable(datasetExternal,featuresToDropExternal)


if reportModePhase1 == True:
    ## Call method __repr__
    print(classObject1)

""" Ouput of Class 1: output_datasetToProcess"""

if commissioningModePhase1 == True:

    # Get external data
    reportMode = reportModePhase0
    datasetExternal, connectionsExternal, policiesExternal = GetInstances(reportMode)
    policySelectedExternal, setOfPoliciesExternal, featuresToDropExternal, pivotFeaturesExternal = setPolicy(policiesExternal,reportMode)
    
    # Run class
    classObject1 = datasetToProcess()
    
    # Call method insert
    classObject1.insert(datasetExternal, connectionsExternal, policiesExternal, policySelectedExternal, setOfPoliciesExternal, featuresToDropExternal, pivotFeaturesExternal)
    
    # Call method getPivotTable
    output_datasetToProcess1, output_datasetToProcess2, output_datasetToProcess3, output_datasetToProcess4 = classObject1.getPivotTable(datasetExternal,featuresToDropExternal)
    
    
    if reportModePhase1 == True:
        ## Call method __repr__
        print(classObject1)
    

```

```python
"""Development of Class 2: processedDataset(object)"""

"""
OOP Phase 2: Commissioning
"""

class processedDataset(object):
    
    """ A pivot table is input to output a analytics. """
    """ Processes are """
    
    def __init__(self):
                    
          # Create blank variables
          self.featureQuestion1 = 0
          self.featureQuestion2 = 0
          
          self.mostWantedCompensation = 0
          self.mostWantedExpenditures = 0
          
          # Create empty lists
          self.policySelected = []
          self.expendituresMeanValues = []
          self.compensationMeanValues = []
          
          self.plot1 = []
          self.plot2 = []
          self.plot3 = []
          
          self.expendituresFindingsText = []
          self.compensationFindingsText = []
          
          # Create empty dataframes
          import pandas as pd
          d = {}
          df = pd.DataFrame(data=d)
          
          self.df = df
          self.pivotTable_Description = df
          self.pivotTableConverted = df
          self.pivotTable = df
          
          self.ExpendituresApproachFindings = df
          self.CompensationApproachFindings = df
          
          self.CompensationApproach = df
          self.ExpendituresApproach = df
          
          self.ExpendituresOutput = df
          self.CompensationOutput = df
          
    
    def insert(self,pivotTable, pivotTableConverted, pivotTable_Description, policySelected):
        """ Insert dataframe: Pivot Table """
        self.pivotTable = pivotTable # No typification
        self.pivotTableConverted = pivotTableConverted # Typification
        self.pivotTable_Description = pivotTable_Description # No typification with description
        self.policySelected = policySelected
                
    
    def analyticsCore(self):
        """ Counting grouped by pivot features: 'Q24, 'Q25' """
        
        import pandas as pd
        import numpy as np
        import matplotlib.pyplot as plt
        
        # Pair option list
        pairOptions = pivotTable_Description[pivotTable_Description.index==0].T
        listLenght = len(pivotTable_Description[pivotTable_Description.index==0].columns)
        
        pairOptions_indexList = []
        for i in range(0,listLenght):pairOptions_indexList = pairOptions_indexList + [i]

        pairOptions = pairOptions.set_index([pd.Index(pairOptions_indexList)], append=True)
        
        # Input Pair
        
        print("=========================================")
        print("OOP Phase 1: Objet building | Data story by feature-pair selection")
        print("=========================================")
        print(pairOptions)
        print("=========================================")
        print("Number range: 0 - ",listLenght-1)
        
        #featureQuestion1 = input("Select pair-1: ")
        featureQuestion1 = 6
        featureQuestion1 = int(featureQuestion1)
        featureQuestion1 = pairOptions.index[featureQuestion1][0]
        
        #featureQuestion2 = input("Select pair-2: ")
        featureQuestion2 = 27
        featureQuestion2 = int(featureQuestion2)
        featureQuestion2 = pairOptions.index[featureQuestion2][0]
                
        print("\n=========================================")
        
        # Put self variables
        self.featureQuestion1 = featureQuestion1
        self.featureQuestion2 = featureQuestion2
        
        # Findings Dataframes 
        self.CompensationApproachFindings = self.pivotTableConverted.groupby(by=['Q24', featureQuestion1, featureQuestion2]).count() # Compensation
        self.ExpendituresApproachFindings = self.pivotTableConverted.groupby(by=['Q25', featureQuestion1, featureQuestion2]).count() # Expenditures
        
        # DataFrames using pivot variables
        self.CompensationApproach = self.pivotTableConverted.groupby('Q24').count() # Compensation
        self.ExpendituresApproach = self.pivotTableConverted.groupby('Q25').count() # Expenditures
        
        # DataFrames using pivot variables and picked variable
        self.ExpendituresOutput = self.ExpendituresApproach.loc[:,[self.featureQuestion1,self.featureQuestion2]]
        self.CompensationOutput = self.CompensationApproach.loc[:,[self.featureQuestion1,self.featureQuestion2]]
        
        return self.CompensationApproach, self.ExpendituresApproach, self.CompensationApproachFindings, self.ExpendituresApproachFindings, self.ExpendituresOutput, self.CompensationOutput, self.featureQuestion1, self.featureQuestion2
        
    
    def plottingPandas(self):
        """ Numeric plotting: (1) Numeric"""
        import pandas as pd
        
        # Pivot columns
        #pivotColumns = ['Q24','Q25']
        pivotColumns = self.pivotTableConverted.columns[0]
        
        # Box plot
        self.plot1 = self.pivotTableConverted.boxplot(column = pivotColumns)
        
        # Bootstrap
        s = self.pivotTableConverted['Time from Start to Finish (seconds)']
        self.plot2 = pd.plotting.bootstrap_plot(s)
        
        # Andrews Curves
        self.plot3 = self.pivotTableConverted.plot.hist(bins=12, alpha=0.5)
        
        return self.plot1, self.plot2, self.plot3
    
    def plottingMatplotlib(self): #Raw state
        """ Bar chart on polar axis """
        import pandas as pd
        import numpy as np
        import matplotlib.pyplot as plt
                
        # Plotting
        for i in range(0,2):
        
            ## Get figure and a set of subplots
            if i == 0:
                N = len(self.ExpendituresApproach.values) # Q25: Expenditure
                approach = 'Expenditures'
                    
                featureToPlot1 = list(self.ExpendituresApproach[self.featureQuestion1]) 
                featureToPlot2 = list(self.ExpendituresApproach[self.featureQuestion2])

                tickLabel = list(self.ExpendituresApproach.index)
                
                
            if i == 1:
                N = len(self.CompensationApproach.values) # Q24: Compensation
                approach = 'Compensation'
                featureToPlot1 = list(self.CompensationApproach[self.featureQuestion1]) 
                featureToPlot2 = list(self.CompensationApproach[self.featureQuestion2])
                tickLabel = list(self.CompensationApproach.index)
            
            ## Compute pie slices
            
            theta = np.linspace(0.0, 2 * np.pi, N, endpoint=False)
            radii = np.array(featureToPlot1)
            width = np.pi / 4000 * np.array(featureToPlot2)
            #width = np.pi / 20
            
            radiiColors = 10 * np.random.rand(N)
            colors = plt.cm.viridis(radiiColors / 10.)
    
            ## Plot
            plt.figure(figsize = (12,12))
            axs = plt.subplot(projection='polar')
            axs.bar(theta, radii, width=width, bottom=0.0, color=colors, alpha=0.5, tick_label=tickLabel)
            titleToPlot = 'Pair: ' + self.featureQuestion1 + ' - ' + self.featureQuestion2 + ' | ' + approach
            axs.set_title(titleToPlot)
    
            plt.show()
        
        return 
    
    def getMeanValue(self, case, description):
        """ Return mean value given the description of a monetary range """

        self.compensationMeanValues = [500, 1500, 12500, 112500, 137500, 17500, 175000, 2500, 22500, 225000, 27500, 275000, 3500, 35000, 400000, 4500, 45000, 6250, 55000, 65000, 8250, 75000, 85000, 95000, 500000]
        compensationDescriptionList = self.CompensationApproach.index        

        self.expendituresMeanValues = [0, 50, 50000, 100000, 500, 5000] 
        expendituresDescriptionList = self.ExpendituresApproach.index

        meanValue = 0

        if (case == 'compensation'):
            
            for i in range(0,len(compensationDescriptionList)):
                itemToEvluate = compensationDescriptionList[i]
                if(description == itemToEvluate): meanValue = self.compensationMeanValues[i]
        
        
        if (case == 'expenditures'):
            
            for i in range(0,len(expendituresDescriptionList)):
                itemToEvluate = expendituresDescriptionList[i]
                if(description == itemToEvluate): meanValue = self.expendituresMeanValues[i]
        
        return meanValue
            
            
    def getFindings(self):
        """ Most wanted | Compensation approach: most significant compensation """
        """ Most wanted | Expenditure approach: most significant expenditure """
        
        # Expenditure findings
        operationsNumber_Expenditure = len(self.ExpendituresApproachFindings)
        
        columnLabel = self.ExpendituresApproachFindings.columns[0]
            
        expenditureFrequency = list(self.ExpendituresApproachFindings[columnLabel])
        
        mostWantedExpenditures = 0
        
        for i in range(0,operationsNumber_Expenditure):
            description = self.ExpendituresApproachFindings.index[i][0]
            itemToInclude = expenditureFrequency[i] * int(processedDataset.getMeanValue(self,'expenditures', description))
            
            if (mostWantedExpenditures <= itemToInclude):
                mostWantedExpenditures = itemToInclude
                self.mostWantedExpenditures_Position = i
                itemToInclude = 0
            
        self.mostWantedExpenditures = mostWantedExpenditures
        
            
        # Compensation findings
        operationsNumber_Compensation = len(self.CompensationApproachFindings)
        
        columnLabel = self.CompensationApproachFindings.columns[0]
            
        compensationFrequency = list(self.CompensationApproachFindings[columnLabel])
        
        mostWantedCompensation = 0
        
        for i in range(0,operationsNumber_Compensation):
            description = self.CompensationApproachFindings.index[i][0]
            itemToInclude = compensationFrequency[i] * int(processedDataset.getMeanValue(self,'compensation', description))
            
            if (mostWantedCompensation <= itemToInclude):
                mostWantedCompensation = itemToInclude
                self.mostWantedCompensation_Position = i
                itemToInclude = 0
        
        self.mostWantedCompensation = mostWantedCompensation
        
        return self.mostWantedExpenditures_Position, self.mostWantedCompensation_Position, self.mostWantedExpenditures, self.mostWantedCompensation
    
    
    def FindingsInText(self):
        
        """ Returns findings text"""
        # Thousands Separator in Python [duplicate]
        import locale
        locale.setlocale(locale.LC_NUMERIC, '')
        
        # Get policy name from self.policySelected
        
        if(self.policySelected == 'TEC'):
            policySelected = 'Technological'
    
        if(self.policySelected == 'CU'):
            policySelected = "Cultural"
    
        if(self.policySelected == 'SE'):
            policySelected = 'Socioeconomic'
        
        # Display text from Expenditures Findings
        
        findingExpenditureText_Title = 'Data story: Most wanted data scientists featured by significant expenditures'
        findingExpenditureText_Assumption0 = "Base Assumption (A0): Most wanted data scientists are supposed to be the collective that spent more monetary resources on machine learning and/or cloud computing services at home (or at work) in the past 5 years"
        findingExpenditureText_Assumption1 = "Assumption 1 (A1): Given the survey question 25, it is said that data scientists inject significant monetary resources around the world."
        findingExpenditureText_Assumption2 = "Assumption 2 (A2): It is assumed that the data scientist profile is fetched under the following policy: " + str(policySelected)
        findingExpenditureText_Grouping = "A target collective can be identified grouping expenditures by the following questions: " + "(1) " + str(self.featureQuestion1) + "; " + "(2) " + str(self.featureQuestion2)
        
        featureExpenditures1 = self.ExpendituresApproachFindings.index[self.mostWantedExpenditures_Position]
        featureExpenditures2 = self.ExpendituresApproachFindings['Time from Start to Finish (seconds)'][self.mostWantedExpenditures_Position]
        featureExpenditures3 = featureExpenditures2
        
        featureExpenditures4 = self.mostWantedExpenditures
        featureExpenditures4 = locale.format_string("%d",featureExpenditures4,grouping=True)
        
        findingExpenditureText_groupResult = "Most wanted datas cientists (target collective) are featured as follows: "
        feature1_ExpendituresText = "(1) Data scientist persona: " + str(featureExpenditures1)
        feature2_ExpendituresText = "(2) Collective featured by " + str(self.featureQuestion1) + ' and ' + str(self.featureQuestion2) + " : " + str(featureExpenditures2) + ' individuals'
        feature3_ExpendituresText = "(3) Collective expenditures: USD " + str(featureExpenditures4)
        
        expendituresFindingsText = [findingExpenditureText_Title] + [findingExpenditureText_Assumption0] + [findingExpenditureText_Assumption1] + [findingExpenditureText_Assumption2] + [findingExpenditureText_Grouping] + [findingExpenditureText_groupResult] + [feature1_ExpendituresText] + [feature2_ExpendituresText] + [feature3_ExpendituresText]
        
        self.expendituresFindingsText = expendituresFindingsText
        
        # Display text from Compensation Findings
        
        findingCompensationText_Title = 'Data story: Most wanted data scientists featured by significant compensation'
        findingCompensationText_Assumption0 = "Base Assumption (A0): Most wanted data scientists are supposed to be the collective that is best compensated."
        findingCompensationText_Assumption1 = "Assumption 1: Given the survey question 24, it is said that data scientists capture significant monetary resources around the world."
        findingCompensationText_Assumption2 = "Assumption 2: It is assumed that the data scientist profile is fetched under the following policy: " + str(policySelected)
        findingCompensationText_Grouping = "A target collective can be identified grouping compensation by the following questions: " + "(1) " + str(self.featureQuestion1) + "; " + "(2) " + str(self.featureQuestion2)
        
        featureCompensation1 = self.CompensationApproachFindings.index[self.mostWantedCompensation_Position]
        featureCompensation2 = self.CompensationApproachFindings['Time from Start to Finish (seconds)'][self.mostWantedCompensation_Position]
        featureCompensation3 = featureCompensation2
        
        featureCompensation4 = self.mostWantedCompensation
        featureCompensation4 = locale.format_string("%d",featureCompensation4,grouping=True)
        
        findingCompensationText_groupResult = "Most wanted data scientists (target collective) are featured as follows: "
        feature1_CompensationText = "(1) Data scientist persona: " + str(featureCompensation1)
        feature2_CompensationText = "(2) Collective featured by " + str(self.featureQuestion1) + ' and ' + str(self.featureQuestion2) + " : " + str(featureCompensation2) + ' individuals'
        feature3_CompensationText = "(3) Collective compensation: USD " + str(featureCompensation4)
        
        compensationFindingsText = [findingCompensationText_Title] + [findingCompensationText_Assumption0] + [findingCompensationText_Assumption1] + [findingCompensationText_Assumption2] + [findingCompensationText_Grouping] + [findingCompensationText_groupResult] + [feature1_CompensationText] + [feature2_CompensationText] + [feature3_CompensationText]
        
        self.compensationFindingsText = compensationFindingsText
        
        return self.expendituresFindingsText, self.compensationFindingsText
        

    def __repr__(self):
        """Returns findings and class report"""
        
        # Set Time
        from datetime import date
        reportDate = date.today()
      
        # Input report
        inputReport = []
        inputReport_output = []
        
        if (self.pivotTable.size != 0):
            inputReport = inputReport + [True]
        else:
            inputReport = inputReport + [False]
        
        if (self.pivotTableConverted.size != 0):
            inputReport = inputReport + [True]
        else:
            inputReport = inputReport + [False]
            
        if (self.pivotTable_Description.size != 0):
            inputReport = inputReport + [True]
        else:
            inputReport = inputReport + [False]
            
        if (self.policySelected != 0):
            inputReport = inputReport + [True]
        else:
            inputReport = inputReport + [False]
            
        for i in inputReport:
            if (i==True):
                inputReport_output = inputReport_output + ['Fetched']
            else:
                inputReport_output = inputReport_output + ['Not Fetched']
        
        # Process report
        processReport = []
        processReport_output = []
        
        if (self.CompensationApproach.size != 0):
            processReport = processReport + [True]
        else:
            processReport = processReport + [False]
        
        if (self.ExpendituresApproach.size != 0):
            processReport = processReport + [True]
        else:
            processReport = processReport + [False]
        
        if self.plot1 != []:
            processReport = processReport + [True]
        else:
            processReport = processReport + [False]
        
        if self.plot2 != []:
            processReport = processReport + [True]
        else:
            processReport = processReport + [False]
        
        if self.plot3 != []:
            processReport = processReport + [True]
        else:
            processReport = processReport + [False]
        
        for i in processReport:
            if (i==True):
                processReport_output = processReport_output + ['Fetched']
            else:
                processReport_output = processReport_output + ['Not Fetched']

        # Output report
        outputReport = []
        outputReport_output = []
                
        if (self.ExpendituresOutput.size != 0):
            outputReport = outputReport + [True]
        else:
            outputReport = outputReport + [False]
            
        if (self.CompensationOutput.size != 0):
            outputReport = outputReport + [True]
        else:
            outputReport = outputReport + [False]
        
        if (self.mostWantedExpenditures_Position != 0):
            outputReport = outputReport + [True]
        else:
            outputReport = outputReport + [False]    
        
        if (self.mostWantedCompensation_Position != 0):
            outputReport = outputReport + [True]
        else:
            outputReport = outputReport + [False]          
        
        for i in outputReport:
            if (i==True):
                outputReport_output = outputReport_output + ['Fetched']
            else:
                outputReport_output = outputReport_output + ['Not Fetched']

      # Get report statements
        reportStatementList = []
        for i in range (0,26):
            reportStatementItem = i
            reportStatementList = reportStatementList + [reportStatementItem]
    
        reportStatementList[0] = "\n" + "\n========================================="
        reportStatementList[1] = "\nOOP Phase 2 | Report | Date:" + str(reportDate.year) + str('/') + str(reportDate.month) + str('/') + str(reportDate.day)
        reportStatementList[2] = "\n========================================="
        reportStatementList[3] = "\nObject: processedDataset(object)"
        reportStatementList[4] = "\n========================================="
        reportStatementList[5] = "\n(1) Inputs"
        reportStatementList[6] = "\n========================================="
        reportStatementList[7] = "\n-Pivot table (raw) : " + str(inputReport_output[0]) + str('| ') + str(type(self.pivotTable))
        reportStatementList[8] = "\n-Pivot table (converted): " + str(inputReport_output[1]) + str('| ') + str(type(self.pivotTableConverted))
        reportStatementList[9] = "\n-Pivot table (description): " + str(inputReport_output[2]) + str('| ') + str(type(self.pivotTable_Description))
        reportStatementList[10] = "\n-Policy Selected: " + str(inputReport_output[3]) + str('| ') + str(self.policySelected)
        reportStatementList[11] = "\n========================================="
        reportStatementList[12] = "\n(2) Process"
        reportStatementList[13] = "\n========================================="
        reportStatementList[14] = "\n-Analytics core | Compensation data: " + str(processReport_output[0]) + str('| ') + str(type(self.pivotTable))
        reportStatementList[15] = "\n-Analytics core | Expenditure data: " + str(processReport_output[1]) + str('| ') + str(type(self.pivotTable))
        reportStatementList[16] = "\n-Visualization | Pandas: " + str(processReport_output[2]) + str('| ') + str(type(self.pivotTable))
        reportStatementList[17] = "\n-Visualization | Matplotlib: " + str(processReport_output[3]) + str('| ') + str(type(self.pivotTable))
        reportStatementList[18] = "\n========================================="
        reportStatementList[19] = "\n(3) Outputs"
        reportStatementList[20] = "\n========================================="
        reportStatementList[21] = "\n-Expenditures instances: " + str(outputReport_output[0]) + str('| ') + str(len(self.ExpendituresOutput))
        reportStatementList[22] = "\n-Compensation instances: " + str(outputReport_output[1]) + str('| ') + str(len(self.CompensationOutput))
        reportStatementList[23] = "\n-Expenditures index: " + str(outputReport_output[2]) + str('| ') + str(self.mostWantedExpenditures_Position)
        reportStatementList[24] = "\n-Compensation index: " + str(outputReport_output[3]) + str('| ') + str(self.mostWantedCompensation_Position)
        reportStatementList[25] = "\n========================================="
        
        reportStatement = []
       
        for i in range (0,26):
            if i == 0:
                reportStatement = reportStatementList[i]
            else:
                reportStatement = reportStatement + reportStatementList[i]
       
        return super().__repr__() + reportStatement
        
        

""" Test of Class 2: processedDataset(object) """

if testModePhase2 == True:
    
    # Get output from class 1
    pivotTable, pivotTableConverted, pivotTable_Description, policySelected = output_datasetToProcess1, output_datasetToProcess2, output_datasetToProcess3, output_datasetToProcess4
    
    # Run class
    classObject2 = processedDataset()
    
    # Call method insert
    classObject2.insert(pivotTable, pivotTableConverted, pivotTable_Description, policySelected)
    
    # Call method analyticsCore
    output_processedDataset1, output_processedDataset2, output_processedDataset3, output_processedDataset4, output_processedDataset5, output_processedDataset6, output_processedDataset7, output_processedDataset8 = classObject2.analyticsCore()
    
    # call method plotting
    plot1, plot2, plot3  = classObject2.plottingPandas()
    
    # call method plottingMatplotlib
    classObject2.plottingMatplotlib()
    
    # call method getMeanValue
    case = 'expenditures'
    description = '$0 ($USD)'
    classObject2.getMeanValue(case, description)
    
    # Call method getFindings
    output_processedDataset9, output_processedDataset10, output_processedDataset11, output_processedDataset12 = classObject2.getFindings()
    
    # Call method FindingsInText
    output_processedDataset13, output_processedDataset14 = classObject2.FindingsInText()

    if reportModePhase2 == True:
        
        ## Call method __repr__
        
        print(classObject2)

""" Ouput of Class 2: output_processedDataset"""

if commissioningModePhase2 == True:
    
# Get output from class 1
    pivotTable, pivotTableConverted, pivotTable_Description, policySelected = output_datasetToProcess1, output_datasetToProcess2, output_datasetToProcess3, output_datasetToProcess4
    
    # Run class
    classObject2 = processedDataset()
    
    # Call method insert
    classObject2.insert(pivotTable, pivotTableConverted, pivotTable_Description, policySelected)
    
    # Call method analyticsCore
    output_processedDataset1, output_processedDataset2, output_processedDataset3, output_processedDataset4, output_processedDataset5, output_processedDataset6, output_processedDataset7, output_processedDataset8 = classObject2.analyticsCore()
    
    # call method plotting
    plot1, plot2, plot3  = classObject2.plottingPandas()
    
    # call method plottingMatplotlib
    classObject2.plottingMatplotlib()
    
    # Call method getFindings
    output_processedDataset9, output_processedDataset10, output_processedDataset11, output_processedDataset12 = classObject2.getFindings()
    
    # Call method FindingsInText
    output_processedDataset13, output_processedDataset14 = classObject2.FindingsInText()
    
    print("=========================================")
    print("Data story proposals")
    print("=========================================")
    print(output_processedDataset13)
    print("=========================================")
    print(output_processedDataset14)
    print("=========================================")

    if reportModePhase2 == True:
        
        ## Call method __repr__
        
        print(classObject2)
```
