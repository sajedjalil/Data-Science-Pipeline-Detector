```python
def createSankey2020(df , questionNumber, title, linkColorAayush = [], linkColorVaibhav = [],topN = None, excludeVal = []):
    
    
    # Removing Values from the Dataset
    if(excludeVal):
        df = df[~(df[questionNumber].isin(excludeVal))]
    if(topN is not None):
        topN = list(df[df['isEmployed']][questionNumber].value_counts()[:topN].sort_index().index)
        df = df[df[questionNumber].isin(topN)]

    label = ["Survey 2020"]
    label.extend(['Employed', 'Unemployed'])
    label.extend(list(df[questionNumber].unique()))

    # Creating Source and Target Mappings

    source = [0,0]
    target = [1,2]
    
    value = [
              sum(df['isEmployed']), 
              len(df) -   sum(df['isEmployed'])
          ]

    labelMapping = {lbl : idx for idx, lbl in enumerate(label)}
    employedVal = df[df['isEmployed'] == True][questionNumber].value_counts().to_dict()
    for k, v in employedVal.items():
        source.append(1)
        target.append(labelMapping[k])
        value.append(v)

    unemployedVal = df[df['isEmployed'] == False][questionNumber].value_counts().to_dict()
    for k, v in unemployedVal.items():
        source.append(2)
        target.append(labelMapping[k])
        value.append(v)

    linkColor = [LINK_COLOR_BASE] * len(source)
    
    for idx in linkColorAayush:
        linkColor[idx] = LINK_COLOR_AAYUSH
    
    for idx in linkColorVaibhav:
        linkColor[idx] = LINK_COLOR_VAIBHAV
        
    # Plotting the Sankey Diagram
    
    fig = go.Figure(data=[go.Sankey(
        arrangement = "snap",
        node = dict(
          pad = 10,
          thickness = 20,
          line = dict(color = "black", width = 0.5),
          label = label,
          color = NODE_COLOR
        ),
        link = dict(
          source = source,
          target = target,
          value = value,
          color = linkColor
      ))])

    fig.update_layout(margin=dict(l=80, r=80, t=50, b=50),
                  title = { 'text' : f'<b>{title}</b>', 'x' : 0.5},
                 font_family = 'Fira Code', plot_bgcolor= BACKGROUND_COLOR ,paper_bgcolor= BACKGROUND_COLOR)
    fig.show()
```

```python
def create3StepSankey2020(df , questionNumber, title, linkColorAayush = [], linkColorVaibhav = [],topN = None, excludeVal = []):
    
    
    # Removing Values from the Dataset
    if(excludeVal):
        df = df[~(df[questionNumber].isin(excludeVal))]
    if(topN is not None):
        topN = list(df[df['isEmployed']][questionNumber].value_counts()[:topN].sort_index().index)
        df = df[df[questionNumber].isin(topN)]

    # Creating Source and Target Mappings
    label = ["Survey 2020"]
    label.extend(['Employed', 'Unemployed'])
    label.extend(list(df['Q5'].unique()))
    label.extend(list(df[questionNumber].unique()))

    # Creating Source and Target Mappings

    source = [0,0]
    target = [1,2]

    value = [
          sum(df['isEmployed']), 
          len(df) -   sum(df['isEmployed'])
      ]

    labelMapping = {lbl : idx for idx, lbl in enumerate(label)}
    employedVal = df[df['isEmployed'] == True]['Q5'].value_counts().to_dict()
    for k, v in employedVal.items():
        source.append(1)
        target.append(labelMapping[k])
        value.append(v)

    unemployedVal = df[df['isEmployed'] == False]['Q5'].value_counts().to_dict()
    for k, v in unemployedVal.items():
        source.append(2)
        target.append(labelMapping[k])
        value.append(v)

    for role in ['Student','Currently not employed', 'Data Scientist', 'Data Analyst', 'Machine Learning Engineer']:
        temp = df[df['Q5'] == role][questionNumber].value_counts().to_dict()
        for k, v in temp.items():
            source.append(labelMapping[role])
            target.append(labelMapping[k])
            value.append(v)

    linkColor = [LINK_COLOR_BASE] * len(source)
    
    for idx in linkColorAayush:
        linkColor[idx] = LINK_COLOR_AAYUSH
    
    for idx in linkColorVaibhav:
        linkColor[idx] = LINK_COLOR_VAIBHAV
        
    # Plotting the Sankey Diagram
    
    fig = go.Figure(data=[go.Sankey(
        arrangement = "snap",
        node = dict(
          pad = 10,
          thickness = 20,
          line = dict(color = "black", width = 0.5),
          label = label,
          color = NODE_COLOR
        ),
        link = dict(
          source = source,
          target = target,
          value = value,
          color = linkColor
      ))])

    fig.update_layout(title = { 'text' : f'<b>{title}</b>', 'x' : 0.5}, plot_bgcolor= BACKGROUND_COLOR ,paper_bgcolor= BACKGROUND_COLOR)
    fig.show()
```

<h2 style="text-align:center; border:2px solid Violet;">Tools in your Backpack</h2>

```python
def multiChoice3stepSankey(df, questionNumber, questionParts, title, xaxisTitle, linkColorAayush = [], linkColorVaibhav = [], excludeVal = [],
                          nodeGroups = {}, text = ''):
    
    label = ["Survey 2020"]
    label.extend(['Employed', 'Unemployed'])
    label.extend(list(df['Q5'].unique()))
    for i in range(1,questionParts+1):
        colName = f'{questionNumber}_Part_{i}'
        colVal = list(df[colName].dropna().unique())[0]
        colVal = re.sub(r'\(.*\)','',colVal)
        colVal = colVal.strip()
        if(colVal in excludeVal):
            continue
        label.append(colVal)

    # Creating Source and Target Mappings

    source = [0,0]
    target = [1,2]

    value = [
          sum(df['isEmployed']), 
          len(df) -   sum(df['isEmployed'])
      ]

    labelMapping = {lbl : idx for idx, lbl in enumerate(label)}

    employedVal = df[df['isEmployed'] == True]['Q5'].value_counts().to_dict()
    for k, v in employedVal.items():
        source.append(1)
        target.append(labelMapping[k])
        value.append(v)

    unemployedVal = df[df['isEmployed'] == False]['Q5'].value_counts().to_dict()
    for k, v in unemployedVal.items():
        source.append(2)
        target.append(labelMapping[k])
        value.append(v)

    for role in ['Student','Currently not employed', 'Data Scientist', 'Data Analyst', 'Machine Learning Engineer']:
        for i in range(1,questionParts+1):
            colName = f'{questionNumber}_Part_{i}'
            colVal = list(df[colName].dropna().unique())[0]
            if(colVal in excludeVal):
                continue
            temp = df[df['Q5'] == role][colName].value_counts().to_dict()
            if(len(temp) == 0):
                continue
            value.append(temp[colVal])
            colVal = re.sub(r'\(.*\)','',colVal)
            source.append(labelMapping[role])
            target.append(labelMapping[colVal.strip()])
            
    linkColor = [LINK_COLOR_BASE] * len(source)
    
    for idx in linkColorAayush:
        linkColor[idx] = LINK_COLOR_AAYUSH
    
    for idx in linkColorVaibhav:
        linkColor[idx] = LINK_COLOR_VAIBHAV
        
    nodeColor = [NODE_COLOR] * len(label)
    for n, (group, groupVals) in enumerate(nodeGroups.items()):
        for val in groupVals:
            valIndex = label.index(val)
            nodeColor[valIndex] = NODE_NEIGHTBOUR_PALETTE[n]
            
    """
    This entire annotated heatmap code snippet as well as idea is modified and borrowed from 
    https://www.kaggle.com/spitfire2nd/enthusiast-to-data-professional-what-changes
    Not smart enough to think this own my own XD
    """
        
    heatmapColorscale = []
    for i in range(len(nodeGroups)):
        heatmapColorscale.append([i*(1/len(nodeGroups)), NODE_NEIGHTBOUR_PALETTE[i]])
        heatmapColorscale.append([(i+1)*(1/len(nodeGroups)), NODE_NEIGHTBOUR_PALETTE[i]])
        
    colorscale = ff.create_annotated_heatmap(
        z=[list(range(len(nodeGroups)))],
            annotation_text =[[f"<span style='font-size:12px'>{text.replace(' ','<br>')}</span>" for text in nodeGroups.keys()]],
        colorscale=heatmapColorscale,
        font_colors = ["white"] * len(nodeGroups),
        text = text,
        hoverinfo = "text",
        xgap = 0,
        showscale = False
    )

    fig = make_subplots(
        rows=3, 
        cols=1, 
        specs = [[{"type":"heatmap",'rowspan' : 1}],[{"type" : "sankey","rowspan" : 2}],[{}]],
        shared_yaxes=True, 
        shared_xaxes=False, 
        horizontal_spacing = 0.02, 
        vertical_spacing = 0.01,
    )

    # Plotting the Sankey Diagram
    
    sankeyTrace = go.Sankey(
        arrangement = "snap",
        node = dict(
          pad = 10,
          thickness = 20,
          line = dict(color = "black", width = 0.5),
          label = label,
          color = nodeColor
        ),
        link = dict(
          source = source,
          target = target,
          value = value,
          color = linkColor
      ))

    fig.append_trace(colorscale.data[0],1,1); 
    fig.add_trace(sankeyTrace, 2,1)
    annot1 = list(colorscale.layout.annotations)
    for k in range(len(annot1)):
        annot1[k]['xref'] = 'x'
        annot1[k]['yref'] = 'y'
    fig.update_layout(annotations=annot1)
    
    fig.update_layout(margin=dict(l=80, r=80, t=80, b=50),
                  title = { 'text' : f'<b>{title}</b>', 'x' : 0.5},
                 font_family = 'Fira Code', plot_bgcolor= BACKGROUND_COLOR ,paper_bgcolor= BACKGROUND_COLOR,yaxis = {'domain' : [0.85,1],'showticklabels' : False},
                      xaxis = {'showticklabels' : False, 'title' : f'<b>{xaxisTitle}</b>'})
    
    fig.show()
```
