```python
import seaborn as sns
import plotly.express as px
import plotly.graph_objects as go
data1 = pd.read_csv('/kaggle/input/kaggle-survey-2020/kaggle_survey_2020_responses.csv')
arg_kag = data1[data1['Q3']=='Argentina']
```

```python
year_prog = pd.DataFrame(arg_kag['Q6'].value_counts())
year_prog = year_prog.sort_values(by='Q6')
fig = px.bar(y=year_prog.index,
             x=year_prog.Q6,
             text=round(year_prog.Q6/sum(year_prog.Q6)*100,2),
            title='Years of Coding Experience')
fig.add_shape(type='rect',
             x0=0,
             y0=5.5,
             x1=28.2,
             y1=6.5,)
fig.add_shape(type='rect',
             x0=0,
             y0=0.5,
             x1=14.2,
             y1=1.5,)
fig.add_shape(type='line',
             x0=27,
             y0=3.5,
             x1=27,
             y1=5.5,
             line={'dash':'dot'})
fig.add_shape(type='line',
             x0=14.2,
             y0=1,
             x1=27,
             y1=1,
             line={'dash':'dot'})
fig.add_shape(type='line',
             x0=27,
             y0=1,
             x1=27,
             y1=2,
             line={'dash':'dot'})
fig.add_annotation(
        x=27,
        y=2.8,
        xref="x",
        yref="y",
        text='Over 32% have less than 1 year of coding experience',
showarrow=False,
font_size=9)
fig.update_traces(marker_color='#4d8bd6'
                 )
fig.update_layout(
    title_font_size=15,
    font_color='Black',
    plot_bgcolor='rgba(0,0,0,0)',
    xaxis_title = '% of Valid Answers',
    yaxis_title = '',
    width=1000,
    height=400)

fig.show()
```

More than a third of the respondents are beginners when it comes to coding.

```python
Profs = arg_kag[arg_kag['Q5']!=('Student' and 'Other')]
temp_5 = pd.DataFrame(Profs['Q8'].value_counts())
temp_5=temp_5.sort_values(by='Q8',ascending=True)
#temp_5_per={%:.0}(temp_5.Q8/sum(temp_5.Q8))*100
fig = px.bar(y=temp_5.index,x=temp_5.Q8,
            title='What do professionals recommend?',
            text=round(temp_5.Q8/sum(temp_5.Q8)*100,2),)
fig.update_traces(marker_color='#4d8bd6'
                 )
fig.update_layout(
    title_font_size=22,
    font_color='Black',
    plot_bgcolor='rgba(0,0,0,0)',
    xaxis_title = '% of Valid Answers',
    yaxis_title = '')
fig.add_annotation(
        x=40,
        y=4,
        xref="x",
        yref="y",
        text='Over 90% of all answers recommend either Python or R as the language to learn for beginners',
        ax=22,
        ay=-2,
font_size=9)
fig.add_shape(type="rect",
    x0=0, y0=4.5, x1=83, y1=6.44,
             line_color='rgba(0,0,0,0)',
             fillcolor='beige',
             opacity=0.3,
             )
fig.show()
```

Python and R continue to be the most recommened languages to learn for beginners. Even by those who are already working in the industry.

```python
year_prog = pd.DataFrame(arg_kag['Q24'].value_counts())
temp_series = np.array(year_prog.index.values)
temp_series = pd.Series(temp_series)
year_prog['NumericValue'] = np.array(temp_series.str.split('-',expand=True)[1])
year_prog['NumericValue'] = year_prog['NumericValue'].str.replace(',','').astype(int)
year_prog = year_prog.sort_values(by='NumericValue',ascending=True)
fig = px.bar(y=year_prog.Q24,x=year_prog.index,
            title='Annual Income')
#24000 es el ingreso mediano a Julio/2020
#24000*12 = ingreso mediano anual, como esto incluye cuentapropista y trabajos en negro no se incluye sac y queda por 12.
#288000 seria el mediano anual
#TC Oficial: 87 => 3300
#TC Oficial+impuestos: 144.38 => 1994
fig.add_vline(x=2.8,line_dash='dash',line_color='lightgray')
fig.add_vline(x=1.5,line_dash='dash',line_color='lightgray')
fig.add_annotation(x=3,
                   y=14.5,
                   text=r'Argentina Median Income at Official Exchange Rate (~3300 USD)',
                  showarrow=False,
                  textangle=90,
                  font_size=8,)
fig.add_annotation(x=1.7,
                   y=14.5,
                   text=r'Argentina Median Income at Official Exchange Rate + Tax (~2000 USD)',
                  showarrow=False,
                  textangle=90,
                  font_size=8)
fig.update_xaxes(title_text='Income Range (USD)')
fig.update_yaxes(title_text='Answer Counts ')
fig.update_layout(plot_bgcolor='rgba(0,0,0,0)',
                 font_color='Black')
fig.update_traces(marker_color='steelblue')
fig.show()
```

```python
not_arg_kag = data1[data1['Q3']!='Argentina']
```

```python
Gender_RW=not_arg_kag[['Q2']].iloc[1:,:].value_counts()
Gender_RW=Gender_RW/len(not_arg_kag[['Q2']].iloc[1:,:])
Gender_RW
```

```python
Gender_Arg=arg_kag[['Q2']].iloc[1:,:].value_counts()
Gender_Arg=Gender_Arg/len(arg_kag[['Q2']].iloc[1:,:])
Gender_Arg
```

```python
Gender=pd.DataFrame([Gender_RW,Gender_Arg])
Gender=Gender.T
Gender.columns=['Rest of World','Argentina']
Gender=pd.DataFrame(Gender.to_records())
Gender=Gender.fillna(0)
Gender['Rest of World']=Gender['Rest of World'].astype('float32')
Gender['Argentina']=Gender['Argentina'].astype('float32')
Gender=Gender.sort_values('Rest of World')
```

```python
Gender['Rest of World %']=(Gender['Rest of World']*100).round(4).astype('str') + '%'
Gender['Argentina %']=(Gender['Argentina'].round(4)*100).astype('str') + '%'
fig = go.Figure()
fig.add_trace(go.Bar(y=Gender['Q2'], x=Gender['Rest of World'],orientation='h',
                    text=Gender['Rest of World %'],
                    textposition='outside',
                    name='Rest of World'))
fig.add_trace(go.Bar(y=Gender['Q2'], x=Gender['Argentina']*-1,orientation='h',
                    text=Gender['Argentina %'],
                    textposition='outside',
                    name='Argentina',
                    marker_color='steelblue'))
fig.update_xaxes(showticklabels=False,
                title='% of All Answers')
fig.update_layout(plot_bgcolor='rgba(0,0,0,0)',
                 font_color='Black',
                 title_text='Gender by Answers',
                 width=1000,
                 height=500,
                 font_size=9)
fig.update_layout(
    legend_x=0.5,
    legend_y=-0.3,
    legend_orientation='h'
)
fig.show()
```
