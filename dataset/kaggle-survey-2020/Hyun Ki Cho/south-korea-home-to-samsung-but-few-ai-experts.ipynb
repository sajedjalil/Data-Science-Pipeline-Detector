```python
#Import Kaggle survey data from 2017-2020

resp_2020 = pd.read_csv('/kaggle/input/kaggle-survey-2020/kaggle_survey_2020_responses.csv')
resp_2019 = pd.read_csv('/kaggle/input/kaggle-survey-2019/multiple_choice_responses.csv')
resp_2018 = pd.read_csv('/kaggle/input/kaggle-survey-2018/multipleChoiceResponses.csv')
resp_2017 = pd.read_csv('/kaggle/input/kaggle-survey-2017/multipleChoiceResponses.csv', encoding='ISO-8859-1')


#Function returns the users(and their information) of a country and the number of users by year
def users_of_country(df_2020, df_2019, df_2018, df_2017, country):
    
    if country == 'World':
        ctry_2020 = df_2020
        ctry_2019 = df_2019
        ctry_2018 = df_2018
        ctry_2017 = df_2017
        
        tot_2020 = ctry_2020.shape[0]
        tot_2019 = ctry_2019.shape[0]
        tot_2018 = ctry_2018.shape[0]
        tot_2017 = ctry_2017.shape[0]
        
        tot = {'Country': [country, country, country, country],
        'Kaggle Users': [tot_2020, tot_2019, tot_2018, tot_2017],
         'Year': [2020, 2019, 2018, 2017]}
        tot_df = pd.DataFrame(data = tot)
        
        return ctry_2020, ctry_2019, ctry_2018, ctry_2017, tot_2020, tot_2019, tot_2018, tot_2017, tot, tot_df
    
    ctry_2020 = df_2020[df_2020['Q3'].str.contains(country)]
    ctry_2019 = df_2019[df_2019['Q3'].str.contains(country)]
    ctry_2018 = df_2018[df_2018['Q3'].str.contains(country)]
    ctry_2017 = df_2017[df_2017['Country'].str.contains(country, na=False)]
    
    if country == 'United States of America':
        ctry_2017 = df_2017[df_2017['Country'].str.contains('United States', na=False)]
    
    tot_2020 = ctry_2020.shape[0]
    tot_2019 = ctry_2019.shape[0]
    tot_2018 = ctry_2018.shape[0]
    tot_2017 = ctry_2017.shape[0]
    
    tot = {'Country': [country, country, country, country],
        'Kaggle Users': [tot_2020, tot_2019, tot_2018, tot_2017],
         'Year': [2020, 2019, 2018, 2017]}
    tot_df = pd.DataFrame(data = tot)
    
    return ctry_2020, ctry_2019, ctry_2018, ctry_2017, tot_2020, tot_2019, tot_2018, tot_2017, tot, tot_df
```

# Factor 1 - An Archaic Education System
South Korea is obsessed with education. This isn't such a big surprise if you consider two things. One, South Korea's culture is heavily influenced by Confucianism, which places a high importance on education. Two, South Korea went from one of the poorest countries in the world in the 1950s, to the world's 12th largest economy in 2018, in no small part due to extensive government investment in education[2]. Due to this, and other historical factors, Koreans see education as a means to economic prosperity. 

This trend is well and alive today. In 2019, "Elementary, middle and high school students spent 20.99 trillion won (17.6 billion USD) on private education[3]." And some studies show that many South Korean students spend 16 hours or more studying at school and attending private prep schools called *hagwons*[2]. This is because **students' scores on *Suneung*, a nationally conducted standardized test for college admissions, is a deciding factor in what colleges they will attend, future income, employment prospects, and even future relationships[4].** Education is not just a stressful topic for South Koreans. It is the root of mental health problems that hurt countless South Korean students, even if they're still in elementary school.[5].

But if Korean students are so educated, they must be well-prepared to study topics such as AI and data science, right? Wrong. The South Korean education system is especially problematic because it is primarily based on rote memorization and test-taking[6]. This means students have trouble figuring out answers to new problems they haven't seen before[7], which is exactly what new and complex fields like AI require. Furthermore, studies show that Korean universities provide less programs to study AI than other nations do[8].

In this section, I'll show why the Korean education system, particularly universities, are ill-equipped to producing an AI workforce, and how that shows in the Kaggle survey data.

# Inexperience in Machine Learning

The lack of university resources may be to blame for Korea's inexperience in machine learning.

The National Information Society Agency, a government agency in Korea, released a report showing that **as of 2018, 0 universities in Korea offered undergraduate, graduate, or doctorate programs for learning AI**. Furthermore, the report stated that **Korea placed last amongst 7 countries (Korea, United States, China, Japan, Europe, India, Israel) in the number of scientific papers published in AI[8]**. Even if Korean university students want to learn machine learning techniques, they have little opportunity to do so in classes. If we compare Korea to the US, a current world leader in AI, we can see that the US has numerous graduate, doctorate, and even undergraduate programs focused on AI[31].

Although there is no evidence for causation, we can carefully venture to say that Korean Kagglers have less opportunity to study machine learning techniques in university compared to US Kagglers, which may be part of the reason why Korean Kagglers are inexperienced in machine learning. For a fairer comparison, I used the percentage of Kagglers per experience level rather than raw numbers because due to a larger population, the US is inevitably going to have more people in each category. I chose to compare Korea to the US because the US is a world leader in AI, which is what Korea should strive to be.

**Korean Kagglers have, on maximum, 5-10 years of experience utilizing machine learning (ML) methods, while US Kagglers have users with 10-20 years of experience, and in some cases, 20 or more years of experience. Looking at the distribution, 5.7% of Korean Kagglers have 5-10 years of ML experience, while 15.4% of US Kagglers have 5-10 years of ML experience or more.** This showcases a large gap in experience, which may be due to the lack of educational resources available.

# Getting the Education System Up to Speed 

Completely fixing the Korean education system is beyond the scope of this notebook. What can be done is touching on suggestions for improving the education system's ability to foster AI experts. First, Korea's love affair with standardized tests needs to change. It has built up a system of public and private education that kills the creativity that students will need to survive in today's economy. To do this, more holistic methods of evaluating a person's capabilities must be developed. Furthermore, universities need to create programs that allow students to study AI, and teach students skills that employers find valuable. Korea is already implementing some of these suggestions. 

Korea is already planning to change the education system to bolster the AI sector, as a part of the National Strategy for Artificial Intelligence laid out by the Moon administration shortly before COVID-19 hit. **In 2021, Korea is planning to introduce AI courses as electives for South Korean high schools**. They will be able to take introductory courses in AI, or AI mathematics, and this increased exposure could cause more students to study AI in university[9]. Furthermore, starting in 2019, AI education began to be introduced to universities. **In 2019, 4 select universities were chosen to create graduate and doctorate programs in AI, and in 2020, 3 more universities did the same**. At the undergraduate levels as well, some universities began to create curriculums specialized for AI, in cooperation with corporations to make sure the curriculum could prepare students for a job in the AI sector[10].

# Factor 2 - A Gender Chasm
Women are dissuaded from entering the workforce, and find it harder to land a job due to gender biases. This is true everywhere in the world, but compared to Western nations, Korean women have more barriers to break through. Choi Mi-Jin, president of the Women Labor Law Support Center, put it this way: **"Our society is dominated by stereotypes about gender roles. Men are the breadwinners, and women are responsible for child-rearing and housework. This is why gender discrimination in hiring is widely accepted and continues to be practiced."[11]**

It is in South Korea's best interests to bridge this gap. Korea's working age population is expected to decline very soon, which makes better integration of women into the workforce a necessity to maintain economic growth. In 2019, the IMF found that if Korea could raise female participation in the workforce to be the same as male participation by 2035, it would be able to enjoy a **7% GDP growth**. However, if it failed to do so, economic competitiveness would decline due to aging women dropping out of the workforce and there not being enough new women to replace them[12].

The declining population and gender discrimination affects the AI sector as well. Bridging the gender gap is not only a matter of achieving greater equality, but is also necessary to increase Korea's AI workforce. 

# A Comparison of Gender Ratios

The following graphs show percentage data on how likely women are to be X or Y and how likely men are to be X or Y. # Women Outnumbered in STEM

Data from the UNESCO Institute for Statistics shows that **for every 100 STEM graduates in South Korea, only 26 are women[14].** Since backgrounds in STEM are favorable for learning AI, starting from university, women seeking to pursue careers in AI are disadvantaged compared to men who decide to do the same. 

The gap is present in the survey data from 2018 as well. **While 70% of Korean female Kagglers majored in a STEM field, 82% of Korean male Kagglers majored in STEM.** Worldwide, the gap in STEM majors was 5% and in the US, the gap was 7%. Considering that Kaggle is a website for learning about data analysis and machine learning, we could presume that the 12% gap in STEM majors of female and male Korean Kagglers is even wider within Korean society. 

Backgrounds in STEM are much more favorable for learning AI than backgrounds in other majors, making it imperative that Korean women are encouraged to study in STEM fields.

```python
#Function for getting stem majors and non stem majors from 2018 data
def stem_major(df, gender):
    #Takes in df of all major data and outputs combined df that's ready to plot gender distribution in STEM majors
    
    #Excluded people with 'I never declared a major' because they don't add to the data
    df = df[df['index'] != 'I never declared a major']
    
    #Create stem dataframe containing STEM majors and create non_stem dataframe containing non-STEM majors
    non_stem_majors = ['Fine arts or performing arts', 'Social sciences (anthropology, psychology, sociology, etc.)',
           'Humanities (history, literature, philosophy, etc.)', 'Other', 'A business discipline (accounting, economics, finance, etc.)']
    pat = '|'.join(map(re.escape, non_stem_majors))
    stem = df[~df['index'].str.contains(pat)]
    non_stem = df[df['index'].str.contains(pat)]
    
    #Assign label based on gender
    if gender == 'Women':
        stem = stem.assign(Gender='Women')
        non_stem = non_stem.assign(Gender='Women')
    if gender == 'Men':
        stem = stem.assign(Gender='Men')
        non_stem = non_stem.assign(Gender='Men')
    
    #Rename columns for clarity
    stem = stem.rename(columns = {'Q5': 'Percentage of Majors'})
    non_stem = non_stem.rename(columns = {'Q5': 'Percentage of Majors'})
    
    #Count number of people present in STEM, non-STEM, and the total count
    stem_count = stem['Percentage of Majors'].sum()
    non_stem_count = non_stem['Percentage of Majors'].sum()
    total_count = stem_count + non_stem_count
    
    #Create dataframe for counting STEM and non-STEM majors
    total = pd.DataFrame({'Majors': ['STEM', 'Non-STEM'], 'Percentage of Majors': [(stem_count/total_count) * 100, (non_stem_count/total_count) * 100],
                             'Gender': [gender, gender]})
    
    return stem_count, non_stem_count, total

#Gather data and create dataframe for plotting for Korea
kr_men_maj = kr_2018[kr_2018['Q1']=='Male']['Q5'].value_counts().to_frame().reset_index()
kr_men_stem , kr_men_n_stem, kr_men_maj_perc = stem_major(kr_men_maj, 'Men')
kr_wom_maj = kr_2018[kr_2018['Q1']=='Female']['Q5'].value_counts().to_frame().reset_index()
kr_wom_stem , kr_wom_n_stem, kr_wom_maj_perc = stem_major(kr_wom_maj, 'Women')

#Combine data for men and women in Korea
kr_maj_perc = pd.concat([kr_wom_maj_perc, kr_men_maj_perc])

#Gather data and create dataframe for plotting for Globe
glob_men_maj = resp_2018[resp_2018['Q1']=='Male']['Q5'].value_counts().to_frame().reset_index()
glob_men_stem , glob_men_n_stem, glob_men_maj_perc = stem_major(glob_men_maj, 'Men')
glob_wom_maj = resp_2018[resp_2018['Q1']=='Female']['Q5'].value_counts().to_frame().reset_index()
glob_wom_stem , glob_wom_n_stemm, glob_wom_maj_perc = stem_major(glob_wom_maj, 'Women')

#Combine data for men and women in Globe
glob_maj_perc = pd.concat([glob_wom_maj_perc, glob_men_maj_perc])


#Gather data and create dataframe for plotting for US
us_men_maj = us_2018[us_2018['Q1']=='Male']['Q5'].value_counts().to_frame().reset_index()
us_men_stem , us_men_n_stem, us_men_maj_perc = stem_major(us_men_maj, 'Men')
us_wom_maj = us_2018[us_2018['Q1']=='Female']['Q5'].value_counts().to_frame().reset_index()
us_wom_stem , us_wom_n_stem, us_wom_maj_perc = stem_major(us_wom_maj, 'Women')

#Combine data for men and women in US
us_maj_perc = pd.concat([us_wom_maj_perc, us_men_maj_perc])

#Create figure
maj_comp_fig, maj_comp_ax = plt.subplots(1,3)
maj_comp_fig.set_size_inches(21, 6)
maj_comp_fig.suptitle('Gender Distribution in Undergraduate Majors by Region', y=1.10)


#Plot data for Korea
kr_maj_barplot = sns.barplot(data = kr_maj_perc,
                            x='Majors', y='Percentage of Majors', hue='Gender',
                            ax = maj_comp_ax[0], palette = ['#EA744A', '#4AC0EA'])
kr_maj_barplot.set_xticklabels(kr_maj_barplot.get_xticklabels(), rotation = 0)
kr_maj_barplot.set_title('South Korea')


#Plot data for Globe
glob_maj_barplot = sns.barplot(data = glob_maj_perc,
                            x='Majors', y='Percentage of Majors', hue='Gender',
                            ax = maj_comp_ax[1], palette = ['#EA744A', '#4AC0EA'])
glob_maj_barplot.set_xticklabels(glob_maj_barplot.get_xticklabels(), rotation = 0)
glob_maj_barplot.set_title('Global')


#Plot data for US
us_maj_barplot = sns.barplot(data = us_maj_perc,
                            x='Majors', y='Percentage of Majors', hue='Gender',
                            ax = maj_comp_ax[2], palette = ['#EA744A', '#4AC0EA'])
us_maj_barplot.set_xticklabels(us_maj_barplot.get_xticklabels(), rotation = 0)
us_maj_barplot.set_title('United States');

#Standardize y axises
kr_maj_barplot.set_yticks(ticks = [100, 90, 80, 70, 60, 50, 40, 30, 20, 10]);
glob_maj_barplot.set_yticks(ticks = [100, 90, 80, 70, 60, 50, 40, 30, 20, 10]);
us_maj_barplot.set_yticks(ticks = [100, 90, 80, 70, 60, 50, 40, 30, 20, 10]);
sns.despine(left='True')
```

# Women in Engineering

Women are present in the sciences, but much less so in engineering. A study by Korea's Center for Women in Science, Engineering, and Technology (WISET), showed that in 2018, **although 53.3% of undergraduates in natural sciences were female, just 21.8% of undergraduates in engineering were female**. Even in the natural sciences, the equality drops off at the doctorate level, where only 35.9% of doctorate graduates were female. These factors contribute to gender inequality in STEM fields[15].

At first glance, it looks like the difference between the percentage of STEM majors for men and women isn't that large, but if we look at some selected STEM majors, we can get more insight. **Korean male Kagglers were 13% more likely to major in Computer Science/Engineering than Korean female Kagglers were**, which is the major that arguably sets up students best for a career in AI. The gap in Korea is relatively larger than that of the world or the US. 

An interesting finding in all three regions is that women are more likely to, or are almost just as likely as men to pursue Math/Statistics, Life Sciences/Medical, Environmental Sciences/Geology as an undergraduate major. This could be due to more women preferring to major in the natural sciences instead of engineering. 

Disclaimer: Percentages in the graph do not add to 100 because not all of the majors from the STEM major category are included. 

```python
#Function for creating dataframe to plot gender distribution within STEM majors
def spec_major(df, gender):
    #takes in df of all major data and outputs percentage data for selected STEM majors 
    
    #excluded people with 'I never declared a major'
    df = df[df['index'] != 'I never declared a major']
    
    #Take in desired majors
    spec = ['Computer science (software engineering, etc.)', 'Mathematics or statistics',
            'Medical or life sciences (biology, chemistry, medicine, etc.)',
           'Environmental science or geology']
    pat = '|'.join(map(re.escape, spec))
    spec_maj = df[df['index'].str.contains(pat)]
    other_maj = df[~df['index'].str.contains(pat)]
    
    #Rename majors for readability
    conv_dict = {'Computer science (software engineering, etc.)':'Computer Science/Engineering', 
                'Mathematics or statistics':'Math/Statistics', 
                'Medical or life sciences (biology, chemistry, medicine, etc.)': 'Life Sciences/Medical',
                'Environmental science or geology': 'Environ. Science/Geology'
                }
    spec_maj['index'] = spec_maj['index'].map(conv_dict)
    
    #Assign labels for plotting
    if gender == 'Women':
        spec_maj = spec_maj.assign(Gender='Women')
    if gender == 'Men':
        spec_maj = spec_maj.assign(Gender='Men')
    
    #Rename for clarity
    spec_maj = spec_maj.rename(columns = {'Q5': 'Percentage of Majors'})
    other_maj = other_maj.rename(columns = {'Q5': 'Percentage of Majors'})
    spec_maj = spec_maj.rename(columns = {'index': 'Majors'})
    other_maj = other_maj.rename(columns = {'index': 'Majors'})
    
    #Count sums of majors in each category then calculates total sum
    spec_count = spec_maj['Percentage of Majors'].sum()
    other_count = other_maj['Percentage of Majors'].sum()
    total_count = spec_count + other_count
    
    #Turn raw major counts into percentages
    spec_maj['Percentage of Majors'] = spec_maj['Percentage of Majors'].apply(lambda x: (x/total_count) * 100)
    
    return spec_maj

#Create dataframe for Korea
kr_spec_men = spec_major(kr_men_maj, 'Men')
kr_spec_wom = spec_major(kr_wom_maj, 'Women')
kr_spec_wom = kr_spec_wom.reset_index(drop=True).reindex([0, 5, 6, 1, 2, 4, 3])
kr_spec_perc = pd.concat([kr_spec_wom, kr_spec_men])

#Create dataframe for Globe
glob_spec_men = spec_major(glob_men_maj, 'Men')
glob_spec_wom = spec_major(glob_wom_maj, 'Women')
glob_spec_wom = glob_spec_wom.reindex([0, 2, 8, 1, 4, 10, 5])
glob_spec_perc = pd.concat([glob_spec_wom, glob_spec_men])

#Create dataframe for US
us_spec_men = spec_major(us_men_maj, 'Men')
us_spec_wom = spec_major(us_wom_maj, 'Women')
us_spec_wom = us_spec_wom.reindex([0, 3, 9, 1, 4, 10, 7])
us_spec_perc = pd.concat([us_spec_wom, us_spec_men])

#Create figure
spec_comp_fig, spec_comp_ax = plt.subplots(1,3)
spec_comp_fig.set_size_inches(27, 6)
spec_comp_fig.suptitle('Gender Distribution Within STEM Majors by Region', y=1.10)

#Plot data for Korea
kr_spec_barplot = sns.barplot(data = kr_spec_perc,
                            x='Majors', y='Percentage of Majors', hue='Gender',
                            ax = spec_comp_ax[0], palette = ['#EA744A', '#4AC0EA'])
kr_spec_barplot.set_xticklabels(kr_spec_barplot.get_xticklabels(), rotation = 80)
kr_spec_barplot.set_title('South Korea')


#Plot data for Globe
glob_spec_barplot = sns.barplot(data = glob_spec_perc,
                            x='Majors', y='Percentage of Majors', hue='Gender',
                            ax = spec_comp_ax[1], palette = ['#EA744A', '#4AC0EA'])
glob_spec_barplot.set_xticklabels(glob_spec_barplot.get_xticklabels(), rotation = 80)
glob_spec_barplot.set_title('Global')


#Plot data for US
us_spec_barplot = sns.barplot(data = us_spec_perc,
                            x='Majors', y='Percentage of Majors', hue='Gender',
                            ax = spec_comp_ax[2], palette = ['#EA744A', '#4AC0EA'])
us_spec_barplot.set_xticklabels(us_spec_barplot.get_xticklabels(), rotation = 80)
us_spec_barplot.set_title('United States');

#Standardize y axises
kr_spec_barplot.set_yticks(ticks = [50, 40, 30, 20, 10]);
glob_spec_barplot.set_yticks(ticks = [50, 40, 30, 20, 10]);
us_spec_barplot.set_yticks(ticks = [50, 40, 30, 20, 10]);
sns.despine(left='True')
```

In case you're unfamiliar with violin plots, violin plots are practically identical to box and whisker plots except for the fact that the width of the plot shows how concentrated datapoints are for that particular y value. For example, since the violin plot is the thickest/stretched out for Korean women at the 45,000 USD mark, we know that of the Korean women surveyed, the highest proportion of them answered that they earn 45,000 USD per year. Within the data, I excluded any unrealistic data below about 16,000 USD, based on Korea's yearly minimum wage. I also excluded wage data above 300,000 USD because they made the graphs unreadable. An interesting point I found when excluding outliers above 300,000 USD was that globally and in the US, both men and women had these high earners. **In Korea, women did not have high outlier salaries above 300,000 USD, while men did.**

```python
#Function converting data labels to numbers
def cat_to_num(df):
    #for responses in 2020
    #converts from strings to numerical data
    #I put inputted the value of salary that was right in between the range of salaries in order for a fair estimate of salaries
    conv_dict = {'$0-999':500, '10,000-14,999':12500, '1,000-1,999':1500, '100,000-124,999': 112500,
                '40,000-49,999':45000, '30,000-39,999':35000, '50,000-59,999': 55000, '5,000-7,499':6250,
                '15,000-19,999':17500, '60,000-69,999':65000, '20,000-24,999': 22500, '70,000-79,999':75000,
                '7,500-9,999':8750, '150,000-199,999':175000, '2,000-2,999':2500, '125,000-149,999':137500,
                '25,000-29,999':27500, '90,000-99,999':95000, '4,000-4,999':4500, '80,000-89,999':85000,
                '3,000-3,999':3500, '200,000-249,999':225000, '300,000-500,000':400000, '> $500,000':500000,
                '250,000-299,999':275000}
    df = df['Q24'].map(conv_dict)
    return df

#min yearly salary was calculated by hourly wage * 40(hr workweek) * 52(weeks in a year)
#16259.152 USD--> korea's lowest possible yearly salary
#300000 --> took out the biggest salary outliers so the graph looked a little bit better

#Get male and female salary dataframes for korea
kr_sal_wom = cat_to_num(kr_2020[kr_2020['Q2']=='Woman']['Q24'].to_frame()).to_frame()
kr_sal_wom = kr_sal_wom.drop(kr_sal_wom[kr_sal_wom['Q24'] < 16259.152].index) #take out outliers
kr_sal_men = cat_to_num(kr_2020[kr_2020['Q2']=='Man']['Q24'].to_frame()).to_frame()
kr_sal_men = kr_sal_men.drop(kr_sal_men[(kr_sal_men['Q24'] < 16259.152) | (kr_sal_men['Q24'] > 300000)].index) #take out outliers
kr_sal_men = kr_sal_men.rename(columns = {'Q24': 'Salary in USD'})
kr_sal_wom = kr_sal_wom.rename(columns = {'Q24': 'Salary in USD'})
kr_sal_men = kr_sal_men.assign(Gender='Men')
kr_sal_wom = kr_sal_wom.assign(Gender='Women')

#Combine male and female dataframes for korea
kr_womenvsmen = pd.concat([kr_sal_wom, kr_sal_men])

#Get male and female salary dataframes for US
us_sal_wom = cat_to_num(us_2020[us_2020['Q2']=='Woman']['Q24'].to_frame()).to_frame()
us_sal_wom = us_sal_wom.drop(us_sal_wom[(us_sal_wom['Q24'] < 15080.0) | (us_sal_wom['Q24'] > 300000)].index) #take out outliers
us_sal_men = cat_to_num(us_2020[us_2020['Q2']=='Man']['Q24'].to_frame()).to_frame()
us_sal_men = us_sal_men.drop(us_sal_men[(us_sal_men['Q24'] < 15080.0) | (us_sal_men['Q24'] > 300000)].index) #take out outliers
us_sal_men = us_sal_men.rename(columns = {'Q24': 'Salary in USD'})
us_sal_wom = us_sal_wom.rename(columns = {'Q24': 'Salary in USD'})
us_sal_men = us_sal_men.assign(Gender='Men')
us_sal_wom = us_sal_wom.assign(Gender='Women')

#Combine male and female dataframes for US
us_womenvsmen = pd.concat([us_sal_wom, us_sal_men]) #df for boxplot


#Get male and female salary dataframes for Globe
glob_sal_wom = cat_to_num(resp_2020[resp_2020['Q2']=='Woman']['Q24'].to_frame()).to_frame()
glob_sal_wom = glob_sal_wom.drop(glob_sal_wom[(glob_sal_wom['Q24'] < 15080.0) | (glob_sal_wom['Q24'] > 300000)].index) #take out outliers
glob_sal_men = cat_to_num(resp_2020[resp_2020['Q2']=='Man']['Q24'].to_frame()).to_frame()
glob_sal_men = glob_sal_men.drop(glob_sal_men[(glob_sal_men['Q24'] < 15080.0) | (glob_sal_men['Q24'] > 300000)].index) #take out outliers
glob_sal_men = glob_sal_men.rename(columns = {'Q24': 'Salary in USD'})
glob_sal_wom = glob_sal_wom.rename(columns = {'Q24': 'Salary in USD'})
glob_sal_men = glob_sal_men.assign(Gender='Men')
glob_sal_wom = glob_sal_wom.assign(Gender='Women')

#Combine male and female dataframes for Globe
glob_womenvsmen = pd.concat([glob_sal_wom, glob_sal_men]) #df for boxplot

#Create figure and use gridspec to create plots
sal_comp_fig = plt.figure(constrained_layout = True)
sal_comp_fig.suptitle('Male and Female Wage Comparisons by Region', y=1.10)
sal_comp_fig.set_size_inches(20, 12)
sal_comp_gs = gridspec.GridSpec(ncols=2, nrows=2, figure=sal_comp_fig)
sal_comp_ax1 = plt.subplot(sal_comp_gs[0, :])
sal_comp_ax2 = plt.subplot(sal_comp_gs[1, 0])
sal_comp_ax3 = plt.subplot(sal_comp_gs[1, 1])

    
#Plot data for Korea
kr_womvsmen_boxplot = sns.violinplot(data = kr_womenvsmen, x='Gender', y='Salary in USD', ax=sal_comp_ax1,
                                    palette = ['#EA744A', '#4AC0EA'])
kr_womvsmen_boxplot.set_title('South Korea')
kr_sal_medians = kr_womenvsmen.groupby(['Gender'])['Salary in USD'].median() #added median salary text
kr_sal_v_offset = kr_womenvsmen['Salary in USD'].median()*0.02
for xtick in kr_womvsmen_boxplot.get_xticks():
    kr_womvsmen_boxplot.text(xtick,kr_sal_medians[xtick-1]+ kr_sal_v_offset,kr_sal_medians[xtick-1], 
            horizontalalignment='center',color='white',weight='semibold', size = 20) #took out size='small'


#Plot data for Globe
glob_womvsmen_boxplot = sns.violinplot(data = glob_womenvsmen, x='Gender', y='Salary in USD', ax = sal_comp_ax2,
                                      palette = ['#EA744A', '#4AC0EA'])
glob_womvsmen_boxplot.set_title('Global')
glob_sal_medians = glob_womenvsmen.groupby(['Gender'])['Salary in USD'].median() #added median salary text
glob_sal_v_offset = glob_womenvsmen['Salary in USD'].median()*0.02
for xtick in glob_womvsmen_boxplot.get_xticks():
    glob_womvsmen_boxplot.text(xtick,glob_sal_medians[xtick-1]+ glob_sal_v_offset,glob_sal_medians[xtick-1], 
            horizontalalignment='center',color='white',weight='semibold', size = 20) #took out size='small'

#Plot data for US
us_womvsmen_boxplot = sns.violinplot(data = us_womenvsmen, x='Gender', y='Salary in USD', ax = sal_comp_ax3,
                                    palette = ['#EA744A', '#4AC0EA'])
us_womvsmen_boxplot.set_title('United States')
us_sal_medians = us_womenvsmen.groupby(['Gender'])['Salary in USD'].median() #added median salary text
us_sal_v_offset = us_womenvsmen['Salary in USD'].median()*0.02 
for xtick in us_womvsmen_boxplot.get_xticks():
    us_womvsmen_boxplot.text(xtick,us_sal_medians[xtick-1]+ us_sal_v_offset,us_sal_medians[xtick-1], 
            horizontalalignment='center',color='white',weight='semibold', size = 20) #took out size='small'
sns.despine(left='True')
```

Without women present in all parts of the field, it becomes more likely that gender biases and stereotypes will make their way into technologies, further cementing a dominant ideology as part of our lives[17]. The only way to make sure that AI doesn't contain biases against women is to make sure that women contribute to the creation of such AI.

```python
#Function for counting how many elements within el_list exist in one row of dataframe
def count_ele(df, el_list):
    #returns count of elements that are within el_list in one row of dataframe 
    count = df[df['Job Type'].str.contains('|'.join(el_list))]
    count = count['Percentage of Jobs'].sum()
    
    return count

#Function for returning job distribution data for men and women
def job_dist(df):

    #Split jobs into 4 categories - scientist, analyst, leader, engineer
    job_excl = ['Currently not employed', 'Student', 'Other']
    sci = ['Data Scientist', 'Research Scientist', 'Statistician']
    eng = ['Software Engineer', 'Data Engineer', 'DBA/Database Engineer', 'Machine Learning Engineer']
    lead = ['Product/Project Manager']
    anal = ['Data Analyst', 'Business Analyst']

    #Take out null values
    w_out_null = df[~df['Q5'].isnull()]
    job_titles = w_out_null[~w_out_null['Q5'].str.contains('|'.join(job_excl))]

    #Rename and reorder dataframe
    job_titles = job_titles[['Q2', 'Q5']].value_counts().to_frame().reset_index()
    job_titles = job_titles.rename(columns = {'Q2': 'Gender', 'Q5': 'Job Type', 0: 'Percentage of Jobs'})

    #Count number of men in each category
    man_job_titles = job_titles[job_titles['Gender']=='Man']
    man_sci_count = count_ele(man_job_titles, sci)
    man_eng_count = count_ele(man_job_titles, eng)
    man_lead_count = count_ele(man_job_titles, lead)
    man_anal_count = count_ele(man_job_titles, anal)

    #Create dataframe for plotting for men
    total_man = pd.DataFrame({'Gender': ['Men', 'Men', 'Men', 'Men'],
                              'Job Type': ['Engineers', 'Leaders', 'Scientists', 'Analysts'],
                             'Percentage of Jobs': [man_eng_count, man_lead_count, man_sci_count, man_anal_count]})
    man_sum = man_job_titles['Percentage of Jobs'].sum()
    total_man['Percentage of Jobs'] = total_man['Percentage of Jobs'].apply(lambda x: (x/man_sum) * 100)

    #Count number of men in each category
    wom_job_titles = job_titles[job_titles['Gender']=='Woman']
    wom_sci_count = count_ele(wom_job_titles, sci)
    wom_eng_count = count_ele(wom_job_titles, eng)
    wom_lead_count = count_ele(wom_job_titles, lead)
    wom_anal_count = count_ele(wom_job_titles, anal)

    #Create dataframe for plotting for women
    total_wom = pd.DataFrame({'Gender': ['Women', 'Women', 'Women', 'Women'],
                              'Job Type': ['Engineers', 'Leaders', 'Scientists', 'Analysts'],
                             'Percentage of Jobs': [wom_eng_count, wom_lead_count, wom_sci_count, wom_anal_count]})
    wom_sum = wom_job_titles['Percentage of Jobs'].sum()
    total_wom['Percentage of Jobs'] = total_wom['Percentage of Jobs'].apply(lambda x: (x/wom_sum) * 100)
    
    #Combine male and female dataframes
    df_for_graph = pd.concat([total_wom, total_man])

    return df_for_graph

#Create figure
job_comp_fig, job_comp_ax = plt.subplots(1,3)
job_comp_fig.set_size_inches(27, 6)
job_comp_fig.suptitle('Gender Distribution in Job Types by Region', y=1.10)

#Plot data for Korea
kr_job_comp = job_dist(kr_2020)
kr_job_comp_barplot = sns.barplot(data = kr_job_comp,
                                 x='Job Type', y='Percentage of Jobs', hue='Gender',
                                 ax=job_comp_ax[0], palette = ['#EA744A', '#4AC0EA'])
kr_job_comp_barplot.set_title('South Korea')

#Plot data for Globe
glob_job_comp = job_dist(resp_2020)
glob_job_comp_barplot = sns.barplot(data = glob_job_comp,
                                 x='Job Type', y='Percentage of Jobs', hue='Gender',
                                 ax=job_comp_ax[1], palette = ['#EA744A', '#4AC0EA'])
glob_job_comp_barplot.set_title('World')

#Plot data for US
us_job_comp = job_dist(us_2020)
us_job_comp_barplot = sns.barplot(data = us_job_comp,
                                 x='Job Type', y='Percentage of Jobs', hue='Gender',
                                 ax=job_comp_ax[2], palette = ['#EA744A', '#4AC0EA'])
us_job_comp_barplot.set_title('United States');

#Standardize y axises
kr_job_comp_barplot.set_yticks(ticks = list(range(0, 110, 20)));
glob_job_comp_barplot.set_yticks(ticks = list(range(0, 110, 20)));
us_job_comp_barplot.set_yticks(ticks = list(range(0, 110, 20)));
sns.despine(left='True')
```

# Korean Companies Slow To Utilize Machine Learning (ML)

**In 2020, large Korean tech firms such as Samsung Electronics and Kakao are joined together in "artificial intelligence alliances" to compete with global tech firms such as Apple and Google [24].** However, what must be considered is that, in places like the US, firms have started earlier and have been quick to implement ML and AI into their services and products. **Korea's Ministry of Science and ICT (MSICT) predicted that Korean AI technology is 1.8 years behind the US[25]**. 

Another reason AI experts look to the US for employment comes from the fact that US firms may have far more job opportunities for these AI experts because they have more firms that are actively using ML. Comparison of ML Usage in Korean and US firms show that a much larger proportion of US firms use ML in some capacity(ML Used for Insight, Models Usage <2 Years, Models Usage +2 Years) compared to Korea. **While only 59 of 122 Korean firms actively utilize ML, 850 of 1,325 US firms utilize ML.** The data suggests that when Korean AI experts look abroad for opportunities, they find a greater variety and number of opportunities. **One optimistic datapoint is that approximately a third of Korean firms are "Exploring ML," meaning that a large proportion of Korean firms are getting ready to incorporate ML into their business operations. **

```python
#Function for getting ML usage dataframe that is plottable
def ml_usage(df):
    
    #Create dictionary to rename options for better readability
    conv_dict = {'We have well established ML methods (i.e., models in production for more than 2 years)': 'Model Usage 2+ Years',
                'We are exploring ML methods (and may one day put a model into production)': 'Exploring ML(NM)',
                'We recently started using ML methods (i.e., models in production for less than 2 years)': 'Model Usage <2 Years',
                'No (we do not use ML methods)': 'No ML Usage',
                'We use ML methods for generating insights (but do not put working models into production)': 'ML Used for Insight(NM)'}
    
    #Create new dataframe
    ml_usage = df['Q22'].value_counts().to_frame().reset_index()
    ml_usage = ml_usage.rename(columns = {'index': 'ML Usage for Firms', 'Q22': 'Number of Firms'})
    ml_usage = ml_usage[ml_usage['ML Usage for Firms'] != 'I do not know'] #Removed these because they weren't useful for analysis
    ml_usage['ML Usage for Firms'] = ml_usage['ML Usage for Firms'].map(conv_dict)

    return ml_usage

#Gather and reorder data for Korea
kr_ml_usage = ml_usage(kr_2020)
kr_ml_usage = kr_ml_usage.reindex([3, 0, 2, 1, 5])

#Gather and reorder data for US
us_ml_usage_index = [3, 1, 4, 2, 0]
us_ml_usage = ml_usage(us_2020)
us_ml_usage = us_ml_usage.reindex(us_ml_usage_index)

#Create figure
ml_usage_comp_fig, ml_usage_comp_ax = plt.subplots(1,1)
ml_usage_comp_fig.set_size_inches(20, 6)
ml_usage_comp_fig.suptitle('Machine Learning(ML) Method Usage in Firms by Region (NM = No Model)', y=1.10)

#Combine dataframes amongst regions after naming them
kr_ml_usage = kr_ml_usage.assign(Country = 'South Korea')
us_ml_usage = us_ml_usage.assign(Country = 'United States')
tot_ml_usage = pd.concat([kr_ml_usage, us_ml_usage])

#Plot data of all regions
tot_ml_usage_barplot = sns.barplot(data = tot_ml_usage,
                                  x='ML Usage for Firms', y='Number of Firms', hue = 'Country',
                                  palette = [
                                            '#FA62A1', '#62FABC', '#FA62A1', '#62FABC', '#FA62A1', '#62FABC'])
sns.despine(left='True')
```

# Better and More Research/Work Environments Lie Outside of Korea

South Korean work environments can be toxic, and encourage employees to put in long hours - **in a 2017 OECD survey, Korea ranked third in OECD nations for weekly working hours with 47.8 hours per week for men and 45.2 for women[26]**. The workforce is primarily male-dominated, and the culture is infused with an almost militaristic hierarchy(often based on age), not to mention that hiring practices are corrupt and gender discriminating[27], resulting in an un-inclusive, undesirable workplace. AI researchers in particular want to leave Korea because its research infrastructure is lacking, and because legal systems in Korea prevent researchers from collecting large datasets to protect the privacy of individuals[19]. 

Researching AI can be expensive, as it may involve working with large cloud databases and specialized software. A snapshot of what looks like Korea's weak research infrastructure reveals itself in the small number of research scientists/research teams that get access to expensive machine learning and cloud computing services. On the other hand, **The US has 3 times more researchers/teams that are able to spend 10,000-99,999 USD per year and 21 times more researchers/teams that are able to spend 100,000 USD or more per year compared to Korea**. Not only does the US have more groups of research scientists that are funded, but they are also generously funded. This makes the US a much more desirable country to research AI at than Korea. 


Note: I excluded datapoints that used amounts under 100 USD because I wanted to see individuals/teams that were doing large-scale data related projects. Although I found in my research that China was heavily investing into AI research, I excluded China because research scientists did not provide substantial information on how much they/their teams were spending. 

```python
#Function creates df for plotting money used for cloud services
def res_funds(df, country):
    #Takes in df of country and country name in string
    
    #Remove 0 USD, 1-99, because I want to look at researchers who need funding for big projects not just personal ones
    conv_dict = {'$0 ($USD)': '0', '$1000-$9,999': '1000-9,999' , '$100-$999': '100-999',
            '$1-$99': '1-99', '$10,000-$99,999': '10,000-99,999', '$100,000 or more ($USD)': '100,000 or more'}
    
    #Create dataframe for raw numbers of people using cloud services
    fund_df = df[df['Q5'] == 'Research Scientist']['Q25'].dropna()
    fund_df = fund_df.map(conv_dict)
    fund_df = fund_df.value_counts().to_frame().reset_index()
    fund_df = fund_df.rename(columns = {'index': 'Cost of ML and Cloud Computing Services(USD)', 'Q25': 'Number of People/Teams Using Services'})
    fund_df = fund_df[(fund_df['Cost of ML and Cloud Computing Services(USD)'] != '0') & (fund_df['Cost of ML and Cloud Computing Services(USD)'] != '1-99')]
    fund_df = fund_df.assign(Country = country)
    
    #Create dataframe for percentage of people using cloud services(Unused)
    perc_df = fund_df
    perc_df = perc_df.rename(columns = {'Cost of ML and Cloud Computing Services(USD)': 'Cost of ML and Cloud Computing Services(USD)', 
                                        'Number of People/Teams Using Services': 'Percentage of People/Teams Using Services'})
    count = perc_df['Percentage of People/Teams Using Services'].sum()
    perc_df['Percentage of People/Teams Using Services'] = perc_df['Percentage of People/Teams Using Services'].apply(lambda x: x/count)
    
    return fund_df, perc_df

#Create dataframe for Korea
kr_res_funds, perc_kr_res_f = res_funds(kr_2020, 'South Korea')
kr_res_funds = kr_res_funds.reindex([2,3,1,4])
perc_kr_res_f = perc_kr_res_f.reindex([0, 3, 1, 5])

#Create dataframe for Globe
glob_res_funds, perc_glob_res_f = res_funds(resp_2020, 'World')
glob_res_funds = glob_res_funds.reindex([2, 1, 4, 5])
perc_glob_res_f = perc_glob_res_f.reindex([2, 1, 4, 5])

#Create dataframe for US
us_res_funds, perc_us_res_f = res_funds(us_2020, 'United States')
us_res_funds = us_res_funds.reindex([4, 1, 3, 2])
perc_us_res_f = perc_us_res_f.reindex([4,1,3,2])

#Combine dataframes amongst regions
tot_res_funds = pd.concat([kr_res_funds, us_res_funds])
tot_perc_res_funds = pd.concat([perc_kr_res_f, perc_glob_res_f, perc_us_res_f])

#Create figure
res_funds_comp_fig, res_funds_comp_ax = plt.subplots(1,1)
res_funds_comp_fig.set_size_inches(10, 6)

#Plot data for all 3 regions
tot_res_funds_barplot = sns.barplot(data = tot_res_funds,
                                   x='Cost of ML and Cloud Computing Services(USD)', y = 'Number of People/Teams Using Services',
                                   hue='Country',
                                   palette = ['#FA62A1', '#62FABC', '#FA62A1', '#62FABC', '#FA62A1', '#62FABC'])
tot_res_funds_barplot.set_title('Researcher/Team Usage of ML/Cloud Computing Services by Region', y=1.10);
res_funds_comp_ax.set_ylabel('Number Using Services');
tot_res_funds_barplot.set_xticklabels(tot_res_funds_barplot.get_xticklabels(), rotation = 30);
sns.despine(left='True')
```
