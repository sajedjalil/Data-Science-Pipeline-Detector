## Bridging the Gender Gap!!

Data Science; and programming in general was assumed to be a Man's  world. But did you know, the first program was actually written by a woman! As per Wikipedia, **Ada Lovelace** has been called the world's first computer programmer. What she did was write the world's first machine algorithm for an early computing machine that existed only on paper. Of course, someone had to be the first, but Lovelace was a woman, and this was in the 1840s.  Let's try to see what is the distribution  by gender in the data that we have. There is steady increase in numbers from 16.71% in 2017 to current 19.36% in 2020 as evident from an analysis of fellow [kagglers](https://www.kaggle.com/aagghh/animated-plots-and-kaggle-surveys#section-three) in an animated visual "Respondents' sex change over time". Irrespective of steady  increase it is still a long way and there are factors which keep Women away from Data science and representation of other genders is very low. An interesting read about [what keeps women away](https://www.bcg.com/publications/2020/what-keeps-women-out-data-science) and current gender gap in 2020  analysis by fellow [Kaggler](https://www.kaggle.com/lucabasa/gender-gap-in-the-kaggle-community)

```python
def func(x):
    if x =='Man':
        return "Male"
    elif x =='Woman':
        return "Female"
    else:
        return "Others"
data['gender']=data['Q2'].apply(func)
genderlist=data['gender'].unique()

# Creating plot 
fig = plt.figure(figsize =(6, 6)) 
explode = (0.1, 0.1, 0.1)  

#my_colors = ['lightblue', 'lightpink','lightgreen']
colors = ['#ffcc99','#66b3ff','#99ff99']
wedges, texts, autotexts=plt.pie(data['gender'].value_counts(),autopct='%1.1f%%',explode=explode,colors=colors, startangle=15, shadow = True,textprops=dict(color="w")) 

for text in texts:
    text.set_color('black')
for autotext in autotexts:
    autotext.set_color('black')
    
plt.title('Gender Ratio')
plt.axis('equal')
plt.legend(wedges, genderlist,
          title="Gender",
          loc="center left",
          bbox_to_anchor=(1, 0, 0.5, 1))
# show plot 
plt.show() 
```

# Coding Language Preference among the entire group 

Python has emerged as language of choice irrespective of age, profession, gender any variable. Python’s huge ecosystem of rich libraries and implementation for various purposes makes it a genuinely more productive programming environment and multi-faceted option. It is simplicity and scalibility which makes python popular along with other benefits as explained in this [article](https://towardsdatascience.com/top-10-reasons-why-you-need-to-learn-python-as-a-data-scientist-e3d26539ec00). python is gradually becoming more popular than R and a [2017 article](https://www.datacamp.com/community/tutorials/r-or-python-for-data-analysis) clearly shows how percentage of users have increased year over year and inclined towards python 


```python

# group age 
def func(x):
    if x =='18-21' or x =='22-24':
        return "<25"
    elif x =='25-29' or x =='30-34' or x =='35-39':
        return "25-39"
    elif  x =='40-44' or x =='45-49' or x =='50-54' :
        return "40-54"
    else:
        return "55+"
data['age_grouped']=data['Q1'].apply(func)

#group profession
def func(x):
    if x in ["Data Engineer","Data Scientist","Data Analyst","Statistician","Machine Learning Engineer","Business Analyst" ,"DBA/Database Engineer"]:
        return("Data Related profession")
    else:
        return x
data['profession_grouped']=data['Q5'].apply(func)


df_prof=data[["id","profession_grouped","Q7_Part_1", "Q7_Part_2"]]
df_prof_unpivoted = df_prof.melt(id_vars=['id','profession_grouped'], var_name='Option', value_name='Language')
#df_lang_unpivoted.sort_values(by=['id'])



df_lang=data[["id","age_grouped","Q7_Part_1", "Q7_Part_2"]]
df_lang_unpivoted = df_lang.melt(id_vars=['id','age_grouped'], var_name='Option', value_name='Language')
#df_lang_unpivoted.sort_values(by=['id'])

language_list=["id","profession_grouped", "Q4", "Q6", "age_grouped", "Q24"]
for col in data.columns:
    if 'Q7' in col:
          language_list.append(col)
            
#metric_cols=["id","profession_grouped", "Q4"]  

df_lang_all=data[language_list]
      
df_lang_unpivoted_all = df_lang_all.melt(id_vars=["id","profession_grouped", "Q4", "Q6", "age_grouped", "Q24"], var_name='Option', value_name='Language')
#df_lang_unpivoted.sort_values(by=['id'])

      
plt.figure(figsize=(15,5))
c=sns.countplot(x="Language",data=df_lang_unpivoted_all,alpha=0.8, hue="Language", dodge=False, palette="muted",order=df_lang_unpivoted_all.Language.value_counts().iloc[:].index)
#plt.xticks(rotation=20)
plt.title('Language Distribution Overall',fontsize=15 )
plt.ylabel('Number of People', fontsize=12)
plt.xlabel('Language', fontsize=12)
for p in c.patches:
             c.annotate("%.0f" % p.get_height(), (p.get_x() + p.get_width() / 2., p.get_height()),
                 ha='center', va='center', fontsize=10, color='black', xytext=(0, 5),
                 textcoords='offset points')

plt.show()

  
```

# How well are the companies reaping the benefits of ML??

Most of ML usage seems to be in "Nascent Phases" or "not yet using". But according to [Gartner](https://www.gartner.com/en/newsroom/press-releases/2019-07-15-gartner-survey-reveals-leading-organizations-expect-t) companies are expected to double number of AI projects  


![](https://thedatascientist.com/wp-content/uploads/2019/06/what-is-data-science.jpg)

```python
# size of the company vs people involved in data science
def company_mp_exp(x):
        if x in ["We have well established ML methods (i.e., models in production for more than 2 years)"]:
                return("using since >2 years")
        if x in ["No (we do not use ML methods)", "We use ML methods for generating insights (but do not put working models into production)","I do not know"]:
                return("Not yet using")
        if x in ["We are exploring ML methods (and may one day put a model into production)", "We recently started using ML methods (i.e., models in production for less than 2 years)"]:
                return("Nascent  Phases")
        
data["ML_expertise_company"]=data["Q22"].apply(company_mp_exp)

plt.figure(figsize=(20,5))
size_c=["0-49 employees", "50-249 employees","250-999 employees","1000-9,999 employees", "10,000 or more employees"]
c=sns.countplot(x="Q20", data=data,alpha=0.7,hue="ML_expertise_company" , dodge=False, palette="muted", order=size_c)
plt.title('Company Size vs how well are they using ML', fontsize=15)
#plt.xticks(rotation=10)
plt.xlabel('Size of Company', fontsize=15)
plt.ylabel('Number of People', fontsize=15)				 
leg = c.axes.get_legend()
new_title = 'ML Usage in years '
leg.set_title(new_title)

#plt.legend([],[], frameon=False)
plt.show()
```
