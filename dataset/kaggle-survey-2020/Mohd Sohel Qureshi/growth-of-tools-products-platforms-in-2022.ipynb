### Before going into the EDA of "Growth of Products, Platforms & Tools in 2022.", 
### we will going to see the usual routine of Respondent's gender, country of residence and professional role/status they hold.
## i. Gender Diversity in this survey

```python
import matplotlib.pyplot as plt


survey_gender=[ 
                survey_raw[survey_raw['Q2']=='Man']['Q2'].count(),
                survey_raw[survey_raw['Q2']=='Woman']['Q2'].count(),
                survey_raw[survey_raw['Q2']=='Nonbinary']['Q2'].count(),
                survey_raw[survey_raw['Q2']=='Prefer not to say']['Q2'].count(),
                survey_raw[survey_raw['Q2']=='Prefer to self-describe']['Q2'].count()
             ]



genders=['Man','Woman', 'Nonbinary', 'Prefer not to say', 'Prefer to self-describe']
color=['RebeccaPurple','steelblue', 'DarkOrange','red','green']

explode =(0.2,0,0.5,0,0.4)

fig1,ax1= plt.subplots(figsize = (15,8))

plt.pie(survey_gender, explode=explode, labels=genders, colors=color, autopct='%1.2f%%')

ax1.legend(genders, loc = 'upper right')

plt.tight_layout()

plt.savefig('gender_diversity.png')
```

* From the piechart it can be seen that Male Gender is the major portion of respondent in this survey constituting 78.80%.

* Women as second large, constituting upto 19.36%.

## Let Us see EDA of "Growth of Products, Platforms & Tools in 2022"
## 1. Compensation and Spending 




**We are going to analyse the yearly compensation (approximate $USD) of the survey respondent according to their Professional Role/Title.**


```python
import plotly.express as px


fig3 = px.violin(survey, x='Q24', y='Q5',orientation="h").update_traces(side="positive", width=3)


fig3.update_layout(template="plotly_dark",
                    autosize=False,
                   width=750,height=600,
                xaxis={'title':'current yearly compensation (approximate $USD)'},
                  yaxis={'categoryorder':'total ascending','title':'Respodent of respective Role'},
                  titlefont=dict(size=16),
                  title_text="Respondents with their Yearly Compensation (approximate $USD)"
                 )

fig3.update_xaxes(ticktext=(pd.unique(survey['Q24'].dropna())),tickangle=45)

fig3.show()
```

* Most of the Professional Respondent observe to have USD 0-999 USD yearly compensation.

* Around 40% of Data Scientist and Data Engineer have yearly compensation 125K-149K$.

* And around 45% of Product Manger have yearly compensation 125K-149K$.


```python
xs=survey["Q25"]


xs.replace({"$100,000 or more ($USD)": "100K or more",
                       "$0 ($USD)": "0",
                        "$1-$99":"1-100",
                       "$10,000-$99,999": "10K-100K",
                       "$1000-$9,999": "1k-10K",
                       "$100-$999": "100-1K",}, inplace=True)


fig4 = px.violin(survey, x=xs, y='Q5',orientation="h").update_traces(side="positive", width=3)


fig4.update_layout(template="plotly_dark",
                    autosize=False,
                   width=740,height=600,
                xaxis={'title':'Spending on machine learning and/or cloud computing services($USD)'},
                  yaxis={'categoryorder':'total ascending','title':'Respodent of respective Role'},
                  titlefont=dict(size=16),
                  title_text="Respondents spending on machine learning and/or cloud computing services"
                 )

fig4.update_xaxes(ticktext=xs,tickangle=45)

fig4.show()
```

* Most of the Most of the Professional Respondent observe to have $0 spending on machine learning and/or cloud computing services.

* Around 47% of Data Scientist respondent have 1k-10k$ spending on machine learning and/or cloud computing services.

* And around 26% of Data Scientist respondent have 100k$-more spending on machine learning and/or cloud computing services.


## 2. Cloud Computing Platforms & Products


```python
import plotly.graph_objects as go
from plotly.subplots import make_subplots


#Platform
platform20=survey.iloc[:,120:130].count()

platform22=survey.iloc[:,256:266].count()

labelplatorm=list(survey.iloc[:,120:130].mode().iloc[0])

#Product
product20=survey.iloc[:,133:140].count()

product22=survey.iloc[:,268:278].count()

labelproduct=list(survey.iloc[:,133:140].mode().iloc[0])

#figure

fig5 = go.Figure(data=[
                    go.Bar(legendgroup="group",
                           name='Platforms Regularly Used',
                           x= platform20,
                           y= labelplatorm,
                           marker_color ='steelblue',
                           orientation='h'),
    
                    go.Bar(legendgroup="group",
                           name='Platforms Getting Familiar in 2022',
                           x= platform22,
                           y= labelplatorm,
                           marker_color ='firebrick',
                           orientation='h'),
    
                    go.Bar(legendgroup="group2",
                           name='Products Regularly Used',
                           x= product20,
                           y= labelproduct,
                           marker_color ='RebeccaPurple',
                           orientation='h',
                           xaxis='x2', yaxis='y2'), 

                    go.Bar(legendgroup="group2",
                           name='Products Getting Familiar in 2022',
                           x= product22,
                           y= labelproduct,
                           marker_color ='Orange',
                           orientation='h',
                           xaxis='x2', yaxis='y2'),



                    ],
                 layout = go.Layout(
                   xaxis = dict(domain = [0.0, 0.4]),
                   xaxis2 = dict(domain = [0.65, 1]),
                   yaxis2 = dict(anchor = 'free',position = 0.65)
                )
                )

fig5.update_layout(template="plotly_dark",
                    autosize=False,
                   width=780,height=600 
                  ,yaxis={'categoryorder':'total ascending'},
                  titlefont=dict(size=16),
                  title_text="Cloud Computing Platforms & Products",
                 font=dict(size=9,color="white"),
                  legend=dict(orientation="h", yanchor="bottom", y=1.00, xanchor="right",x=0.95)

                 )
                


fig5.show()
```

*Cloud Computing Platform*

* *Amazon Web Services (AWS)* is in regular use followed by *Google Cloud Platform (GCP)* and *Microsoft Azure*.

* It is observed that the large number of respondent are going to get familiar ,in year 2022, with *Amazon Web Services (AWS)* but also *Google Cloud Platform (GCP)* is nearly gaining on AWS in getting familiar in next 2 years.


*Cloud Computing Product*

* *Google Cloud Compute Engine* is in regular use with *AWS Lambda* at second followed by *Azure Cloud Services*.

* In the "getting familiar in next 2 years", *Google Cloud Functions* is most favoured Cloud Computing Product followed by *Microsoft Azure Container*.


## 3. Machine Learning Products


```python
import plotly.graph_objects as go

#Regular

mal20=survey.iloc[:,144:153].count()
mllabel=list(survey.iloc[:,144:153].mode().iloc[0])

#Familiar

mal22=survey.iloc[:,280:289].count()
mllabel2=list(survey.iloc[:,280:289].mode().iloc[0])

#Plot

fig6 = go.Figure(data=[
                    go.Bar(name='Regular Used',x= mllabel,y= mal20,marker_color ='RebeccaPurple'),
                    go.Scatter(name='Getting Familiar in 2022',x= mllabel2,y= mal22,marker_color ='Orange')
                    ],layout_title_text="Machine Learning Products"
                )

fig6.update_layout(
                  template="plotly_dark",
                   autosize=True,
                  width=780,height=700,
                  yaxis={'categoryorder':'total ascending'},
                  titlefont=dict(size=20),
                  font=dict(size=9, color="white"),
                 legend=dict(orientation="h", yanchor="bottom", y=1.02, xanchor="right",x=1)
                )

fig6.show();
```

**Looks like Respondent are likely to going to opt for Machine Learning products till 2022 in comparison to to their regualr use, which means growth of Machine Product is promising**

* In "regular use" *Google Cloud AI Platform/Google Cloud ML Engine* is top choice.

* And with "getting familiar till 2022", *Google Cloud AI Platform/Google Cloud ML Engine* followed by *Goggle Cloud Vision AI * and *Google Cloud NAtural Language*. Making Google Cloud Machine learning product top choice in next two year.


## 4. Big Data Product


```python
#Regular
bdmost=(survey['Q30'].value_counts())
bdlabelmost=list(pd.unique(survey['Q30'].dropna()))


#Most
bdregular20=survey.iloc[:,155:173].count()
bdlabel20=list(survey.iloc[:,155:173].mode().iloc[0])
bdlabel20

#Familiar
bdregular22=survey.iloc[:,291:309].count()
bdlabel22=list(survey.iloc[:,291:309].mode().iloc[0])
bdlabel22

#plot
fig7 = go.Figure(data=[
                        go.Bar(name='Used Most often',x=bdmost, y=bdlabelmost,marker_color ='orange',orientation='h'),
                        go.Bar(name='Regularly Used',x= bdregular20,y= bdlabel20,marker_color ='steelblue',orientation='h'), 
                        go.Bar(name='Getting Familiar in 2022',x= bdregular22,y= bdlabel22,marker_color ='RebeccaPurple',orientation='h') 
                    ]
                )

fig7.update_layout(
                   template="plotly_dark",
                   autosize=True,
                   barmode='stack',
                  width=750,height=550
                 ,yaxis={'categoryorder':'total ascending'},
                  titlefont=dict(size=20),
                  title_text="Big Data Products",
                  font=dict(size=10,color="white"),
                legend=dict(orientation="h", yanchor="bottom",y=1.02,xanchor="right",x=1)
)

fig7.show()
```
