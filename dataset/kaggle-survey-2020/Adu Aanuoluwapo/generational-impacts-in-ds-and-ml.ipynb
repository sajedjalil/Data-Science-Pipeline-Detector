This notebook seeks to explore **age** from the generational perspective.

In 2017, Millennials were aged between 21-36 and Generation Z were aged between 5-20. Generations before Millennials have been merged to be called 'Older generations'. Based on the survey data, the following are the three groups of generations to be considered in this notebook.

### **For 2020 survey data:**

**Millennials** = 25-39 years,
**Generation Z** = 18-24 years,
**Older generations** = 40+ years


### **For 2017 survey data:**

**Millennials** = 21-36 years,
**Generation Z** = 18-20 years,
**Older generations** = 36+ years

## **What generation is responsible for population growth in the ML and DS community?**

```python
#create bins and additional generation column for the three generation categories 
bins = 18, 21, 37, 40
labels = ['generation Z','millenial','Older Generations']

kaggle2017['generation'] = pd.cut(kaggle2017['Age'],bins,labels=labels, include_lowest = True)
```

```python
#remove row 0
kaggle2020 = kaggle2020.drop(index = 0)

#define the generational categories
gen_z = kaggle2020[kaggle2020['Q1'].isin(['18-21','22-24'])]
gen_z.insert(0,'generation','generation Z')

millenial = kaggle2020[kaggle2020['Q1'].isin(['25-29','30-34','35-39'])]
millenial.insert(0,'generation','millenial')

olderGen = kaggle2020.loc[~kaggle2020['Q1'].isin (['18-21','22-24','25-29','30-34','35-39'])]
olderGen.insert(0,'generation','Older Generations')

kaggle2020 = pd.concat([gen_z,millenial,olderGen])
```

```python
groupby = kaggle2017.groupby(['generation'],as_index = False)['Age'].count()
groupby2 = kaggle2020.groupby(['generation'],as_index = False)['Q1'].count()

sns.set(rc={'figure.figsize':(8,8)})
fig = pd.concat({
    'Kaggle 2017': groupby.set_index('generation'), 'Kaggle 2020': groupby2.set_index('generation')
}, axis=1).plot.bar()
fig.set(Title = 'Count of Kagglers by generation')  

for p in fig.patches:
    width = p.get_width()
    height = p.get_height()
    x, y = p.get_xy() 
    fig.annotate(f'{height}', (x + width/2, y + height*1.02), ha='center')
    

sns.set(rc={'figure.figsize':(8,8)})
#calculate percentage
groupby['Age'] = (groupby['Age']/groupby['Age'].sum()*100)
groupby2['Q1'] = (groupby2['Q1']/groupby2['Q1'].sum()*100)

fig1 = pd.concat({
    'Kaggle 2017': groupby.set_index('generation'), 'Kaggle 2020': groupby2.set_index('generation')
}, axis=1).plot.bar()
fig1.set_ylabel('Percentage')
fig1.set(Title = 'Generational percentage per year') 

for p in fig1.patches:
    width = p.get_width()
    height = p.get_height()
    x, y = p.get_xy() 
    fig1.annotate(f'{height}', (x + width/2, y + height*1.02), ha='center')
```

Above chart shows a reduction in the total number of Millennials in the Data Science and Machine Learning community over the years. In the 2017 survey, over 11,000 Millennial Kagglers took part in the survey making them 83% of the total participants. However, in 2020, there is a reduction in the number of Millennials within the community. Although, Millennials still constitute the highest number of kagglers, they now make up about 43% percent of the entire community and they are not responsible for the growth in terms of numbers within the community. 

Generation Z on the other hand have increased by almost 400% since 2017 also Older Generations have increased by over 200%.

Yes,the Data Science community has a larger percentage of Millennials but these generation is not responsible for the increase in population.

## **Who earns what?**

```python
groupby15 = kaggle2020.groupby('Q24')['Q24'].count()
groupby15
```

```python
groupby5 = kaggle2020.groupby(['Q24','generation'],as_index = False)['Q6'].count()

groupby5['Q6']= (groupby5['Q6']/groupby5['Q6'].sum()*100)

plt.figure(figsize=(12,10))
sns.set(style = "white")
fig4 = sns.barplot(x='Q24',y='Q6', hue= 'generation', data = groupby5)
fig4.set_title('Who earns what?')
fig4.set_ylabel('Percentage')
plt.xticks(rotation='vertical')
```

```python
groupby6 = gen_z.groupby(['Q24','generation'],as_index = False)['Q6'].count()
groupby6 = groupby6.sort_values(by = 'Q6', ascending = False)

groupby7 = millenial.groupby(['Q24','generation'],as_index = False)['Q6'].count()
groupby7 = groupby7.sort_values(by = 'Q6', ascending = False)

groupby8 = olderGen.groupby(['Q24','generation'],as_index = False)['Q6'].count()
groupby8 = groupby8.sort_values(by = 'Q6', ascending = False)


plt.figure(figsize=(8,5))
sns.set(style = "white")
fig5 = sns.barplot(x='Q24',y='Q6', data = groupby6, color='blue')
fig5.set_title('What Generation Z earns')
fig5.set_ylabel('Count')
plt.xticks(rotation='vertical')
fig5.set(yscale='linear')
plt.show()

plt.figure(figsize=(8,5))
sns.set(style = "white")
fig6 = sns.barplot(x='Q24',y='Q6', data = groupby7, color='green')
fig6.set_title('What Millennials earn')
fig6.set_ylabel('Count')
plt.xticks(rotation='vertical')
fig6.set(yscale='linear')
plt.show()

plt.figure(figsize=(8,5))
sns.set(style = "white")
fig7 = sns.barplot(x='Q24',y='Q6', data = groupby8, color='red')
fig7.set_title('What older generations before the millennials earn')
fig7.set_ylabel('Count')
plt.xticks(rotation='vertical')
fig7.set(yscale='linear')
plt.show()

```

## What job titles are common per generation?

```python
groupby11 = kaggle2020.groupby(['Q5','generation'],as_index = False)['Q6'].count()

groupby11 = groupby11.groupby(['generation','Q5']).agg({'Q6': 'sum'})
groupby12 = groupby11.groupby(level=0).apply(lambda x:100 * x/float(x.sum()))
groupby12
```

```python
groupby11['Percent'] = 100 * groupby11['Q6']/groupby11.groupby('Q5')['Q6'].transform('sum')
print(groupby11.sort_values(['Q5','generation']).reset_index(drop=False))
```

For the Older Generations, the highest percentage have other job titles not included in the survey followed by Data Scientist, Software Engineer and Research Scientist respectively

Millennials mostly work as  Data Scientist(18%), then Student(12%) and Software Engineer(11%)

58% of Generation Z are students followed by 8% who work as Data Scientists. 8% Software Engineer and 7% are currently unemployed

Software Engineer and Data Scientist job titles are consistent are consistent across all 3 generations. Percentages of Research Scientist job title are also low for Millennials and Generation Z participants.

## Where do kagglers live?

```python
groupby13 = kaggle2020.groupby(['generation','Q3'],as_index = False)['Q6'].count()
groupby13['Percent']= (groupby13['Q6']/groupby13['Q6'].sum()*100)
groupby13.sort_values(by = 'Percent',ascending = False)
```

```python
groupby13 = pd.pivot_table(groupby13, values= 'Percent', index=['Q3'], columns='generation')
plt.figure(figsize=(8,12))
sns.heatmap(groupby13, cmap="RdBu")
```
