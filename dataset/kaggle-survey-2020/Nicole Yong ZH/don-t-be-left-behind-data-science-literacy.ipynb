# Is data science a trend, or is it here to stay?

Much has already been said about the benefits of data-driven decisions. There has been continuous hype surrounding data sciences that being a data scientist was popularly described as the sexiest job of the 21st century by [Harvard Business Review back in 2012](http://hbr.org/2012/10/data-scientist-the-sexiest-job-of-the-21st-century).

# Part 1: Kaggle Survey 2020: Exploring the data scientists subset

```python
# arrange df for Q5
df5 = pd.DataFrame(df['Q5'].value_counts())
df5 = df5[:-1]

# plot
plt.figure(figsize=(9, 6))
ax = sns.barplot(x="Q5", y=df5.index, data=df5)
sns.despine()
plt.title("Respondents' Profile") 
plt.xlabel("No. of Respondents")

warnings.filterwarnings("ignore")
```

# Part 2: Who is the data scientist

<h2>Part 2.1: What companies are data scientists working in</h2>

```python
# arrange df for Q20
q20 = pd.DataFrame(dsdf['Q20'].value_counts().reindex(['0-49 employees',
                                          '50-249 employees',
                                          '250-999 employees',
                                          '1000-9,999 employees',
                                          '10,000 or more employees']))


# plot
plt.figure(figsize=(9, 6))
ax = sns.barplot(x="Q20", y=q20.index, data=q20)
sns.despine()
plt.title("Size of the Company, by No. of Employees") 
plt.xlabel("No. of Respondents")
plt.ylabel("No. of Employees in Company")

warnings.filterwarnings("ignore")
```

<h3>üîç Results</h3>

Most data scientists work in smaller companies with fewer employees (49 and below), and this is followed by working in very large companies (10,000 or more employees).

We can speculate various reasons for this result:

For the **smaller companies**:
* They could be start ups, open to new technologies and aggressively adopting data-based solutions to problems
* With lesser levels of hierarchy, it enables the company to make a quick decision to adopt data-based solutions
* They see the largest benefit from the skill sets of data scientists, where a small investment in resources could yield results from data they always had but had no idea what to make sense of

For the **larger companies**:
* They are aware of the benefits of data-driven decisions, and have both the resources and reputation to attract data scientist to join them
* Being in short supply, data scientists can take their pick and choose to work in larger companies where compensation, benefits, and progression routes are better

This is not to say data scientists only work in very small or very large companies. It is good to note that there is a healthy spread of data scientists working across companies of different sizes. This signals that companies, regardless of sizes or perhaps industry type (perhaps we can include this in next year's Kaggle survey!), are hiring data scientists, embracing data front and center.

<h3>üì£ What It Means</h3>

Clearly, more opportunities for data scientists in the future as data science become a must-have in all companies.

*Reference: Q20 - What is the size of the company where you are employed?*

<h2>Part 2.2: How many data scientists among us</h2>

```python
# arrange df for Q20
q21 = pd.DataFrame(dsdf['Q21'].value_counts().reindex(['0', '1-2', '3-4', '5-9', '10-14', '15-19', '20+']))

# plot
plt.figure(figsize=(9, 6))
ax = sns.barplot(x="Q21", y=q21.index, data=q21)
sns.despine()
plt.title("No. of individuals are working on data science in the company") 
plt.xlabel("No. of Respondents")
plt.ylabel("No. of Inviduals Responsible for Data Science Workloads")

warnings.filterwarnings("ignore")
```

<h3>üîç Results</h3>

You will notice that many survey respondents' companies have more than 20+ individuals or 1-2 individuals responsible for data science workloads.

With more than 20 employees working on data science workloads, we can safely assume there is an established data science team / function / department. This also tells us that data science value-adds to these companies such that they are willing to spend resources to hire and build a team. Such work environments will provide more job stability and career progression for the data scientist.

For companies with 1-2 individuals working on data science, this could suggest perhaps a one-man show or distributed data science duties / capabilities / multi-hatting in addition to other roles employees may be responsible for.

<h3>üì£ What It Means</h3>

There is a wide range in terms of numbers of employees working on data science workloads in the companies. However, this also suggests that perhaps many companies have yet to or are in the process of building up a data science team, thus providing many opportunities to suitable candidates. 

*Reference: Q21 - Approximately how many individuals are responsible for data science workloads at your place of business?*

<h2>Part 2.3: What is the data scientist up to at work</h2>

```python
# arrange df for q23
q23 = pd.DataFrame(dsdf.loc[:,dsdf.columns.str.startswith("Q23")])
q23e = q23.apply(pd.Series.value_counts)
q23e['Q23_count'] = q23e.loc[:,q23e.columns.str.startswith("Q23")].max(axis=1)
q23e = q23e.sort_values(['Q23_count'], ascending=False)

# plot
plt.figure(figsize=(9, 6))
ax = sns.barplot(x="Q23_count", y=q23e.index, data=q23e)
sns.despine()
plt.title("Activities at work") 
plt.xlabel("No. of Respondents")

warnings.filterwarnings("ignore")
```

<h3>üîç Results</h3>

The largest chunk of work that occupies a data scientist at work is analyzing and understanding data to influence product or business decisions. This is expected and also demonstrates that data science can provide a tangible and measurable benefit to the business operations.

As a basis of all data science projects, analyzing and understanding data is critical and wrongly interpreting data could result in unpleasant outcomes.

<h3>üì£ What It Means</h3>

Since companies are in possession of various data of different sorts depending on the industry, one can assume that having specific industry / product knowledge will also help a data scientist in interpreting data and providing solutions / making recommendations.

Moving forward, there will be a demand for data scientists with not only data science skills, but also specific industry knowledge. Of course, such a candidate will be hard to find, but a rare and valuable one! ü¶Ñ

<h2>Part 2.4: Is the reward worthy enough to pursue</h2>

```python
# arrange df for q24
q24 = pd.DataFrame(dsdf['Q24'].value_counts())

# plot
plt.figure(figsize=(9, 6))
ax = sns.barplot(x="Q24", y=q24.index, data=q24)
sns.despine()
plt.title("How much does a data scientist make") 
plt.xlabel("No. of Respondents")

warnings.filterwarnings("ignore")
```

<h3>üîç Results</h3>

From the survey results, it seems that most data scientists make less than USD 999 on an annual basis, which at first glance does not seem to make the endeavor a worthy pursuit. However, we bear in mind that perhaps this reflects on data scientists beginning on this path, those working part-time, pursuing data science as an interest, or even accounting for wage differences in different countries.

Encouragingly, the second most common response to the yearly compensation question is $100,000 to 124,999 - which is a pretty decent wage! üí∞

<h3>üì£ What It Means</h3>

According to [PayScale](http://www.payscale.com/research/SG/Job=Data_Scientist/Salary/54970bde/Singapore), the average salary of a data scientist in Singapore is around USD 54,000. This is around 31 percent higher than the median wage in Singapore, which is around USD 41,000 in [2020](http://stats.mom.gov.sg/Pages/Income-Summary-Table.aspx) - a reflection of the demand for data science skills.

Given the well-publicized shortage of data scientists, we can expect compensation to increase to compensate for the low supply and high demand.

*Reference: Q24 - What is your current yearly compensation ( approximate $USD )?*

# Part 4: Keeping Up-To-Date

<h2>Part 4.1: Keeping the tools sharp</h2>

```python
# arrange df for q37
q37 = pd.DataFrame(dsdf.loc[:,dsdf.columns.str.startswith("Q37")])
q37e = q37.apply(pd.Series.value_counts)
q37e['Q37_count'] = q37e.loc[:,q37e.columns.str.startswith("Q37")].max(axis=1)
q37e = q37e.sort_values(['Q37_count'], ascending=False)

# plot
plt.figure(figsize=(9, 6))
ax = sns.barplot(x="Q37_count", y=q37e.index, data=q37e)
sns.set_style("whitegrid")
plt.title("Which platforms are preferred for a data science course") 
plt.xlabel("No. of Respondents")

warnings.filterwarnings("ignore")
```

<h3>üîç Results</h3>

[Coursera](http://www.coursera.org/) is a clear winner from the survey results, with a significant lead over the second choice - [Udemy](http://www.udemy.com/). It is telling of the times we live in given the top 2 choices are online learning platforms. 

Interesting to note, third in place is university courses resulting in a university degree. This could be due to the increasing availability of university courses specializing in data sciences, since demand for data science would naturally lead to a increase in supply in the form of future workers currently studying in university.

<h3>üì£ What It Means</h3>

It is no longer the case that one has to go a traditional school to acquire knowledge. Knowledge is widely available on various platforms now, and on-demand knowledge allows for flexibility both in time and in location. It is also usually cheaper to attend an online course compared to a traditional school.

However, given that online learning platforms are still facing issues of legitimacy and market over-saturation, perhaps employers are still keen to hire those with formal qualifications from an established university.

Having said that, I am optimistic that online certifications will be even more widely-accepted in the future, perhaps being on par with traditional school certificates. Perhaps, employers will move from assessing candidates based on qualifications to assessing candidates based on the skill sets they possess.
