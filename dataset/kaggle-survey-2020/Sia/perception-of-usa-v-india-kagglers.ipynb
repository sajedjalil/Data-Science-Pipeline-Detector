```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

import plotly.express as px
from plotly.offline import init_notebook_mode, iplot, plot
import plotly as py
import plotly.graph_objs as go
from plotly import subplots
import re
from pywaffle import Waffle
```

```python
survey=pd.read_csv("../input/kaggle-survey-2020/kaggle_survey_2020_responses.csv")
```

```python
check=survey.iloc[0:1,:].values.tolist()# used to  check the title of columns.
title=[]
for i in check[0]:
    title.append(i)
```

```python
def cc(df,column_name):#used to count the answers to a unique question.
    count=0
    for column in df.columns:
        for line in re.findall(column_name,column):
            count=count+1
    return count
```

```python
def col_num(df,column_name):#to count  survey results.
    k=cc(df,column_name + ".+")
    data=[]
    count=[]
    per=[]
    for i in range(1,k):
        data_count=(df[column_name + "Part_"+ "{}".format(i)].value_counts())
        data.append(data_count.index[0])
        count.append(data_count.values[0])
    other=df[column_name+"OTHER"].value_counts()
    data.append(other.index[0])
    count.append(other.values[0])
    for i in count:
        k=(np.round_((i/sum(count))*100,2))
        per.append(k)
    return data,count,per
```

The chart illustrates the Age group distribution of USA and INDIA kaggle-survey-2020 respondents in percentages. 

Overall,The Indian kaggle survey respondents are more compared to USA kaggle survey respondents. The age group distribution of Indian respondents is right skewed with around 35% of respondents are at 18-21 years and 0.22% of respondents belong to 60-69 years age group. The age group distribution of american respondents doesn't follow any standard distribution.

Around 18% of respondents belong to 25-29 years age group in both countries. In USA, the percentage of respondents that belong to 30-34 are double than that of Indian responents which was 9%.Approximately 80% of respondents are between 18-29 years age group in INDIA.
