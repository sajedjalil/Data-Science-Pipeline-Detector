### Top highlights from the insights found from the data collected through Kaggle Survey 2000

 **1. From 30 years age onwards, with every year (or age group) it is evident people tend to move into senior/managerial positions with less programming exposure (and/or not much time spent on ML/DL models)** 

 **2. Overwhelming number of participants from one country (India) in top 1 position , we can see that last 4 in top 5 cumulatively is less than 50% of top 1 position.**

 **3. We see more active female participation in ML going ahead which is evident from almost 30% of females are practicing ML who are currently studying.**

 **4. More and more entrants into IT are into practicing ML. It is clearly evident that AI/ML is going to be part of each and everyone going ahead.**

 **5. We see more active female participation in ML going ahead which is evident from almost 30% of females are practicing ML among the recent entrants into IT.**

 **6. Python is the most recommended language for data scientist aspirants,  by more than 90% participants.**

 **7. Jupyter (Notebooks) is the mostly used IDE.  Jupyter is opensource and is almost possibly installed by each ML practitioner on their own systems/laptops.**

 **8. Azure and Google Colab notebooks are the mostly used Cloud notebooks by some distance when compared to others.**  

 **9. Matplotlib and Seaborn are the top most used Visualization libraries.**

 **10. sklearn and tensorflow (+keras) are the top most used ML frameworks followed by PyTorch.**

 **11. RandomForest (+Decision Trees) and Gradient Boosting are the top most regularly used ML algorithms followed by GAN and Bayesian model. It also indicates the acceptance of Bagging and Boosting in order to overcome overfitting and help in increasing generalization of models.**

 **12. Currently top most used Cloud Computing Platform is AWS, but there is clear interest towards trying Google followed by Azure.**

 **13. There is more interest towards using cloud computing products, cloud big data products, cloud BI products as well as cloud ML products for all different ML tasks going ahead.**
