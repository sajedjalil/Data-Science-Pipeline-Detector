# 1. Preprocessing

```python
responses = pd.read_csv('/kaggle/input/kaggle-survey-2020/kaggle_survey_2020_responses.csv')
responses.head(1)
```

```python
#This data frame will be enriched as we proceed with the data.
response_data = pd.DataFrame() 
response_data['time_taken'] = responses['Time from Start to Finish (seconds)'][1:]
response_data['age'] = responses['Q1'][1:]
response_data['gender'] = responses['Q2'][1:]
response_data['country'] = responses['Q3'][1:]
response_data['highest_education'] = responses['Q4'][1:]
response_data['job_title'] = responses['Q5'][1:]
response_data['experience'] = responses['Q6'][1:]
response_data['recommended_language'] = responses['Q8'][1:]
response_data['computing_platform'] = responses['Q11'][1:]
response_data['use_of_tpu'] = responses['Q13'][1:]
response_data['ml_exp_years'] = responses['Q15'][1:]
response_data['size_of_company'] = responses['Q20'][1:]
response_data['emp_in_ds_dept'] = responses['Q21'][1:]
response_data['ml_in_company'] = responses['Q22'][1:]
response_data['salary'] = responses['Q24'][1:]
response_data['ml_cc_spend'] = responses['Q25'][1:]
response_data['big_data_products'] = responses['Q30'][1:]
response_data['bi_tools'] = responses['Q32'][1:]
response_data['analysis_tools'] = responses['Q38'][1:]
```

```python
#Shortening the names of the country.
response_data['country'] = response_data['country'].replace({'United States of America':'USA', 'United Kingdom of Great Britain and Northern Ireland':'UK', 'Republic of Korea': 'Korea', 'Iran, Islamic Republic of...':'Iran'})
```

```python
def isNan(num):
    return num != num

def get_avg_X(val):
    val = str(val)
    val = val.replace(',','')
    val_list = re.findall(r'\d+', val)
    val_list = list(map(int, val_list))
    try:
        return (sum(val_list)/len(val_list))
    except:
        return 0
    
def get_max_in_range(val):
    val = str(val)
    val = val.replace(',','')
    val_list = re.findall(r'\d+', val)
    val_list = list(map(int, val_list))
    try:
        return max(val_list)
    except:
        return 0
    
def get_avg_exp(exp):
    exp = str(exp)
    exp_list = re.findall(r'\d+', exp)
    exp_list = list(map(int, exp_list))
    try:
        return (sum(exp_list)/len(exp_list))
    except:
        return 0
```

```python
response_data['salary_avg'] = response_data['salary'].apply(lambda x: get_avg_X(x))
response_data['ml_cc_spend_avg'] = response_data['ml_cc_spend'].apply(lambda x: get_avg_X(x))
```

```python
response_data['max_size_of_company'] = response_data['size_of_company'].apply(lambda x: get_max_in_range(x))
response_data['max_emp_in_ds_dept'] = response_data['emp_in_ds_dept'].apply(lambda emp: get_max_in_range(emp))
```

```python
ds = ['Data Scientist', 'Data Analyst', 'Data Engineer', 'Machine Learning Engineer']
not_ds_df = response_data[~(response_data['job_title'].isin(ds))]
not_ds_df['job_title'].value_counts()
```

```python
response_data['time_taken'] = response_data['time_taken'].astype(int)
response_data['time_taken'].max()
print(f"Maximum time taken to complete the survey: {response_data['time_taken'].max()} seconds\nMinimum time taken to complete the survey: {response_data['time_taken'].min()} seconds")
```

>## 2.2 Gender Distribution accross countries that are in data science community.

```python
n=10
top_10_countries = tuple(top_n_countries(n))
col=int(n/2)

fig = make_subplots(rows=  2, cols = col,
                  specs = [[{'type':'pie'}]*col,
                          [{'type':'pie'}]*col],
                  subplot_titles = top_10_countries)

count=0
for i in range(1,3):
    for j in range(1,col+1):
        gd = response_data[response_data['country']==top_10_countries[count]]['gender'].value_counts()
        fig.add_trace(go.Pie(labels=gd.index , values=gd.values) , row=i , col=j)
        count+=1
fig.show()
```

As it is not possible to represent this distribution for all countries let's see the top countries from the data set who top's in other minority developers.

```python
def country_with_highest_pecentage_of(gender, n):
    '''
    This method will return a dataframe grouped and sorted by the input parameter gender.
    '''
    count_genders = response_data.groupby(['country', 'gender']).count()['time_taken']
    count_gender_percentage = count_genders.groupby(level=0).apply(lambda x: 100 * x / float(x.sum()))
    count_gender_percentage_sorted = count_gender_percentage.unstack(level=1).sort_values(gender, ascending=True)
    return count_gender_percentage_sorted[:n]


minority = tuple(np.delete(response_data['gender'].unique(), 0))

fig = make_subplots(rows=  4, cols = 1,
                    specs = [[{'type':'bar'}]]*len(minority),
                    subplot_titles = minority)

count = 0
for j in range(1, 5):
    fig.add_trace(go.Bar(orientation='h', y=country_with_highest_pecentage_of(minority[count], 10).index, x=country_with_highest_pecentage_of(minority[count], 10)[minority[count]]), j, 1)
    count+=1     
fig.update_layout(title='Highest percentage countries for minority genders', showlegend=False, boxgap=0.9, height=1000)
fig.show()
```
