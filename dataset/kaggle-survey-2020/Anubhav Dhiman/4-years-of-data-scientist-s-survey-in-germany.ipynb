## üë®üë© Men still dominate the field but the ratio in Women is steadily increasing

```python
col = "gender"
question = "What's your gender?"

gr = pd.DataFrame(df.groupby([col, "year"]).size()).reset_index()
gr["size"] = gr.groupby("year")[0].transform(np.sum)
gr.columns = [col, "year", "percent", "size"]

gr["percent"] = np.round(100*gr["percent"]/gr["size"])
gr = gr[gr.percent > 0]

ax = sns.catplot(
    x='percent', y=col, data=gr, kind="bar", hue="year", height=4,
    order = ["Man", "Woman"],
    palette=np.array(['#f0f0f0', '#bdbdbd', '#636363', '#2DA5EA']),
    legend=None,
)

ax.set(xlabel='Percentage', ylabel='', title=question)
plt.legend(bbox_to_anchor=(1.01, 1), loc=2, borderaxespad=0.)
sns.despine()
plt.tight_layout()
plt.show()
```
