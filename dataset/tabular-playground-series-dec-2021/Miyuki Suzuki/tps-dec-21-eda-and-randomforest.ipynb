# 3.Let's find an importance feature

> I try a simple feature this time.(no categorical features)üëÄ but..It is not the best score.
> So I revert it back to the original feature.

```python

tc = train[['Elevation','Cover_Type']]
te = test[['Elevation']]

idx = train[train["Cover_Type"] == 5].index
train.drop(idx, axis=0, inplace=True)

cols = ["Soil_Type7", "Soil_Type15"]
train.drop(cols, axis=1, inplace=True)
test.drop(cols, axis=1, inplace=True)

# labelE=encoder
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
train["Cover_Type"] = encoder.fit_transform(train["Cover_Type"])


X = train.drop('Cover_Type', axis = 1).copy()
y = train['Cover_Type'].copy()

X_test = test.copy()

del train
del test
```

```python

# scaler
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X = pd.DataFrame(scaler.fit_transform(X))
X_test = pd.DataFrame(scaler.transform(X_test))
```

n_estimators = 100 > 150,200, random_state = 0 > 42, 2021 üëç



```python
from sklearn.metrics import accuracy_score
from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier(n_estimators = 100, random_state = 0)
model.fit(X,y)

pred_valid = model.predict(X)
ascore = accuracy_score(y, pred_valid)
print('accuracy_score : ', ascore)

test_preds = model.predict(X_test)

labels = X.columns.values
importances = model.feature_importances_
idx = np.argsort(importances)
print(idx)
```

```python
imp = pd.DataFrame()
imp['importance'] = importances
imp = imp['importance']
```

```python
plt.figure(figsize = (10,10))
plt.title('feature importances')
imp.nlargest(25).plot(kind='barh')
plt.show()
```

```python

prediction = encoder.inverse_transform(test_preds)

sub['Cover_Type'] = prediction

sub.to_csv("submission.csv",index=False)
sub[0:10]
```
