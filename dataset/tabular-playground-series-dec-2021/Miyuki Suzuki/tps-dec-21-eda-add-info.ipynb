# Count ( Soil and Wildness Area )

```python
soil_ = [x for x in train.columns if x.startswith("Soil_Type")]
wilderness_ = [x for x in train.columns if x.startswith("Wilderness_Area")]

train["Soil_Count"] = train[soil_].apply(sum, axis=1)
test["Soil_Count"] = test[soil_].apply(sum, axis=1)
train["Wilderness_Area_Count"] = train[wilderness_].apply(sum, axis=1)
test["Wilderness_Area_Count"] = test[wilderness_].apply(sum, axis=1)
```
