```python
def read_data(df):
    print(f"Total rows {df.shape[0]}")
    print(f"Columns {df.shape[1]}")
    print(df.columns)
    print(df.head(3))
```

Reading Data

We see a target column to be predicted, cat0 - cat18 are categorical, and the feature columns cont0 - cont10 are continuous.

```python
train_df = pd.read_csv("/kaggle/input/tabular-playground-series-mar-2021/train.csv")
test_df = pd.read_csv("/kaggle/input/tabular-playground-series-mar-2021/test.csv")
print("Train Data")
read_data(train_df)
print("Test Data")
read_data(test_df)
```

```python
feature_cols = train_df.drop(['id', 'target'], axis=1).columns
# Seprating numerical and categprical columns
numerical_columns = train_df[feature_cols].select_dtypes(include='float64').columns
categorical_columns = train_df[feature_cols].select_dtypes(include='object').columns


# combinning both train and test df
df = pd.concat([train_df, test_df], axis=0).reset_index(drop=True)
targets = df['target']
df = df.drop(['id', 'target'], axis=1)
print(df.shape)
```

```python
# Lets check some numerical columns
df_numerical = df[numerical_columns]
df_numerical.head(5)
```

```python
# Correlation matrix
corr = df.corr()
corr.style.background_gradient(cmap='coolwarm')
```

Observations:

* Seems like cont6 is not correlated to any of the other columns.
* cont1 and cont2 seem to be highly correlated (0.86193) same could be said for cont0 & cont10, cont0 & cont7.

```python
# Lets check some categorical columns
df_categorical = df[categorical_columns]
df_categorical.head(5)
```

Distribution of unique values

```python
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style="ticks", context="talk")

def plot_unique(columns, df):
    unique_count = []
    for column in columns:
        # print(f'{column}: {df[column].unique().size}')
        unique_count.append(df[column].unique().size)
    plt.xticks(rotation='vertical')    
    plt.bar(categorical_columns, unique_count)
    plt.xlabel('Column name')
    plt.ylabel('Count of unique values')
    print(f"Min count: {min(unique_count)}")
    print(f"Max count: {max(unique_count)}")
plot_unique(categorical_columns, df_categorical)        
```

Observations:

* Minimum number fo unique values present in the categorical columns is 2 and max is 309.
* For other columns the value ranges from 2-100 unique values.
* cat5, cat8 & cat10 have 2 letter alphabets whereas other contain only single alphabet value.

Lets have a look at the target column

```python
import matplotlib.pyplot as plt
import seaborn as sns

target_counts = targets.value_counts()
print(target_counts.index)
sns.barplot(x=target_counts.index, y=target_counts.values)
plt.xlabel('Targets')
plt.ylabel('Count')
plt.show()

print("Count of targets:")
print(targets.value_counts())
```

Seems like the targets are skewed with significantly more number of 0's than 1's
