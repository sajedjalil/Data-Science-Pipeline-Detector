```python
from fastai.tabular.all import *
```

```python
df = pd.read_csv('../input/tabular-playground-series-mar-2021/train.csv')
test_df = pd.read_csv('../input/tabular-playground-series-mar-2021/test.csv')
```

```python
len(df.columns)
```

```python
df.head()
```

```python
cat_vars = ['cat0', 'cat1', 'cat2', 'cat3', 'cat4', 'cat5', 'cat6', 'cat7', 'cat8', 'cat9', 'cat10', 'cat11', 'cat12', 'cat13', 'cat14', 'cat15', 'cat16', 'cat17', 'cat18']
cont_vars = ['cont0', 'cont1', 'cont2', 'cont3', 'cont4', 'cont5', 'cont6', 'cont7', 'cont8', 'cont9', 'cont10']
dep_var = ['target']
```

```python
df.isna().sum()
```

```python
procs = [Categorify, Normalize]
splits = RandomSplitter()(df)
to = TabularPandas(df, procs, cat_names=cat_vars, cont_names=cont_vars, y_names=dep_var, y_block=CategoryBlock, splits=splits)
dls = to.dataloaders(bs=512)
test_dl = dls.test_dl(test_df)
```

```python
dls.show_batch()
```

```python
learn = tabular_learner(dls, metrics=accuracy)
learn.lr_find()
```

```python
learn.fit_one_cycle(6, 6e-3)
```

```python
learn.recorder.plot_loss()
```

```python
preds,targs = learn.get_preds(dl=test_dl)
res = preds[:,1]
final_df = pd.DataFrame({'id': test_df['id'], 'target': res})
final_df.to_csv('./submission.csv', header=True, index=False)
```
