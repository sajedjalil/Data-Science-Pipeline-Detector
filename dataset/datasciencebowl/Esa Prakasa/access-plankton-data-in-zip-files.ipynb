This initial step aims to acces zip data of plankton images. The code can list all files in the input folder. The code can also select and display a single or several plankton images

```python
import os

for dirname, _, filenames in os.walk('/kaggle/input'):
    for filename in filenames:
        print(os.path.join(dirname, filename))
```

Zip file of train dataset is held by a variable namely *files_in_zip*.

```python
from zipfile import ZipFile
from  matplotlib import pyplot as plt

datasetPath = '/kaggle/input/datasciencebowl'

zip_path = '/kaggle/input/datasciencebowl/train.zip'
with ZipFile(zip_path) as myzip:
    files_in_zip = myzip.namelist()
    
files_in_zip[:5]

NMax = len(files_in_zip)

print("Total number of the train images is %d"%(NMax))

#files_in_zip[:5]
```

```python
# Display a single image of plankton

import random

fileIdx = random.randint(0, NMax)
if (fileIdx<3):
    fileIdx = fileIdx + 5

from PIL import Image
with ZipFile(zip_path) as myzip:
    with myzip.open(files_in_zip[fileIdx]) as myfile:
        img = Image.open(myfile)

#print(img.size)
M = img.size[0]
N = img.size[1]

Mf = int((M/M)*5)
Nf = int((N/M)*5)

img.show()
plt.rcParams["figure.figsize"] = (Mf,Nf)

plt.imshow(img, cmap="gray")
plt.title("File idx: "+str(fileIdx) +"  "+myfile.name)
plt.show()
```

```python

idxF1 = 5000
idxF2 = idxF1 + 10

plt.rcParams["figure.figsize"] = (20,10)
idx = 0

with ZipFile(zip_path) as myzip:
    for i in range(idxF1,idxF2,1):
    
        myfile = myzip.open(files_in_zip[i])
        #print(str(idx)+" "+myfile.name)
        idx = idx + 1
        img = Image.open(myfile)       
        
        plt.subplot(2,5,idx)
        plt.imshow(img, cmap="gray")
        plt.title(myfile.name)
```
