<br><b style="text-decoration: underline; font-family: Verdana; text-transform: uppercase;">GENERAL EVALUATION INFORMATION</b>


This competition is evaluated on the [**F2 Score**](https://en.wikipedia.org/wiki/F-score) at different intersection over union (IoU) thresholds
* <mark><b>The F2 metric weights [recall](https://www.wikiwand.com/en/Precision_and_recall) more heavily than [precision](https://www.wikiwand.com/en/Precision_and_recall), as in this case it makes sense to tolerate some false positives in order to ensure very few starfish are missed.</b></mark>
* The metric sweeps over IoU thresholds in the range of 0.3 to 0.8 with a step size of 0.05 calculating an F2 score at each threshold. 
    * For example, at a threshold of 0.5, a predicted object is considered a "hit" if its IoU with a ground truth object is at least 0.5
* A true positive is <b>the first (in confidence order, see details below) submission box in a sample with an IoU greater than the threshold against an unmatched solution box</b>.
* Once all submission boxes have been evaluated, any unmatched submission boxes are false positives
    * Any remaining unmatched solution boxes are false negatives
* The final F2 Score is calculated as the mean of the F2 scores at each IoU threshold. 
* Within each IoU threshold the competition metric uses <b><mark>micro averaging</b></mark>; 
    * every true positive, false positive, and false negative has equal weight compared to each other true positive, false positive, and false negative.
    
<br><b style="text-decoration: underline; font-family: Verdana; text-transform: uppercase;">SUBMISSION FORMAT</b>

The submission format requires a space delimited set of bounding boxes. 
* For example:
    * `abc,0.5 0 0 100 100` indicates that image abc has a bounding box with a confidence of 0.5, at x == 0 and y == 0, with a width and height of 100.
    * wheras `def,0.3 0 0 50 50 0.5 10 10 30 30` indicates an image with two bounding boxes
* Each prediction row needs to include all bounding boxes for the image.

<h3 style="font-family: Verdana; font-size: 20px; font-style: normal; font-weight: normal; text-decoration: none; text-transform: none; letter-spacing: 2px; color: #40E0D0; background-color: #ffffff;">2.3 LEVERAGING XLA OPTIMIZATIONS</h3>

---


**XLA** (Accelerated Linear Algebra) is a domain-specific compiler for linear algebra that can accelerate TensorFlow models with potentially no source code changes. **The results are improvements in speed and memory usage**.

<br>

When a TensorFlow program is run, all of the operations are executed individually by the TensorFlow executor. Each TensorFlow operation has a precompiled GPU/TPU kernel implementation that the executor dispatches to.

XLA provides us with an alternative mode of running models: it compiles the TensorFlow graph into a sequence of computation kernels generated specifically for the given model. Because these kernels are unique to the model, they can exploit model-specific information for optimization.<br><br>

<div class="alert alert-block alert-danger" style="margin: 2em; line-height: 1.7em; font-family: Verdana;">
    <b style="font-size: 16px;">üõë &nbsp; WARNING:</b><br><br>- XLA can not currently compile functions where dimensions are not inferrable: that is, if it's not possible to infer the dimensions of all tensors without running the entire computation<br>

````python
def flatten_l_o_l(nested_list):
    """ Flatten a list of lists """
    return [item for sublist in nested_list for item in sublist]

def tf_load_img(img_path, reshape_to=None):
    if reshape_to is None:
        return tf.image.decode_image(tf.io.read_file(img_path), channels=3)
    else:
        return tf.image.resize(tf.image.decode_image(tf.io.read_file(img_path), channels=3), reshape_to)

# ORIGINAL CREDIT 
#      --> https://www.kaggle.com/diegoalejogm/great-barrier-reefs-eda-with-animations
def create_animation(img_path_list, annotations=None, fps=12):
    """ Create an animtation from a list of image paths """
    fig = plt.figure(figsize=(20, 11))
    plt.axis('off')
    
    img = plt.imshow(cv2.imread(img_path_list[0])[..., ::-1].astype(np.int32))
    
    def animate_func(i):
        tmp_img = cv2.imread(img_path_list[i])[..., ::-1].astype(np.int32)
        if annotations is not None:
            if len(annotations[i])>0:
                for bbox in annotations[i]:
                    tl_box, br_box = get_tl_br(bbox)
                    tmp_img = cv2.rectangle(tmp_img, tl_box, br_box, (255,0,0), 8)
        
        img.set_array(tmp_img)
        return [img]
    plt.close()
    return animation.FuncAnimation(fig, animate_func, frames=len(img_path_list), interval=1000//fps)


def get_tl_br(bbox):
    """ Return the top-left and bottom-right bounding box """
    return (bbox['x'], bbox['y']), (bbox['x']+bbox["width"], bbox['y']+bbox["height"])

def plot_image(img_path, annotations=None, **kwargs):
    """ Plot an image and bounding boxes """
    img = np.array(tf_load_img(img_path))
    
    if annotations:
        plt.figure(figsize=(20,10))
        for i, bbox in enumerate(annotations):
            tl_box, br_box = get_tl_br(bbox)
            img = cv2.rectangle(img, tl_box, br_box, (255-2*i,14*i,0), 4)
        plt.imshow(img)
        plt.axis(False)
        plt.title(f"Bounding Boxes Shown in Red ({len(annotations)})", fontweight="bold")
    else:
        plt.figure(figsize=(20,10))
        plt.imshow(img)
        plt.axis(False)
        plt.title("No Bounding Boxes", fontweight="bold")
    plt.tight_layout()
    plt.show()
````

````python
def compute_intersection(box_1, box_2):
    """Computes pairwise IOU matrix for given two sets of boxes
    
    REFERENCE: https://keras.io/examples/vision/retinanet/
    
    Args:
        box_1 (tf.Tensor): An array with shape `(4,)` representing bounding boxes
            where each box is of the format `[xmin, ymin, xmax, ymax]`
        box_2: An array with shape `(4,)` representing bounding boxes
            where each box is of the format `[xmin, ymin, xmax, ymax]`

    Returns:
        intersection area between two boxes
    """
    
    # determine the (x, y)-coordinates of the intersection rectangle
    tl_x_max = max(box_1[0], box_2[0])
    tl_y_max = max(box_1[1], box_2[1])
    br_x_min = min(box_1[2], box_2[2])
    br_y_min = min(box_1[3], box_2[3])
    
    inter_area = max(0, br_x_min-tl_x_max+1) * max(0, br_y_min-tl_y_max+1)
    inter_width, inter_height = br_x_min-tl_x_max, br_y_min-tl_y_max
    return inter_area, inter_width, inter_height

def get_grid_images(img, annotation, img_dim=(720,1280,3), grid_dim=(160,160,3), stride=(80,80), min_inter_area=250, min_inter_width=16, min_inter_height=16, return_as_ordered_dict=False):
    
    x_points = np.linspace(0, img_dim[1], img_dim[1]//stride[1]+1, dtype=np.int32)[:-grid_dim[1]//stride[1]]
    y_points = np.linspace(0, img_dim[0], img_dim[0]//stride[0]+1, dtype=np.int32)[:-grid_dim[0]//stride[0]]

    if return_as_ordered_dict:
        # top left corners
        ordered_grid_image_map = {i:{j:{"annotation":[], "image":None, "label":0} for j, y_pt in enumerate(y_points)} for i, x_pt in enumerate(x_points)}
        for ii, i in enumerate(x_points):
            for jj, j in enumerate(y_points):
                grid_img=img[j:j+grid_dim[1], i:i+grid_dim[1], :]
                ordered_grid_image_map[ii][jj]["image"]=grid_img
                ordered_grid_image_map[ii][jj]["tl"]=(i,j)
                ordered_grid_image_map[ii][jj]["br"]=(i+grid_dim[0],j+grid_dim[1])
                
                is_bg=True
                for box in annotation:
                    inter_area, inter_width, inter_height = compute_intersection(
                        box_1=(box["x"], box["y"], box["x"]+box["width"], box["y"]+box["height"]),
                        box_2=(i, j, i+grid_dim[0], j+grid_dim[1]),
                    )
                    if inter_area>=min_inter_area and inter_width>=min_inter_width and inter_height>=min_inter_height:
                        is_bg=False
                        ordered_grid_image_map[ii][jj]["annotation"].append(box)
                        ordered_grid_image_map[ii][jj]["label"]=1                
        return ordered_grid_image_map
                    
    else:
        fg_grid_images = []
        bg_grid_images = []
        for i in x_points:
            for j in y_points:
                grid_img=img[j:j+grid_dim[1], i:i+grid_dim[1], :]
                is_bg=True
                for box in annotation:
                    inter_area, inter_width, inter_height = compute_intersection(
                        box_1=(box["x"], box["y"], box["x"]+box["width"], box["y"]+box["height"]),
                        box_2=(i, j, i+grid_dim[0], j+grid_dim[1]),
                    )
                    if inter_area>=min_inter_area and inter_width>=min_inter_width and inter_height>=min_inter_height:
                        is_bg=False
                        fg_grid_images.append(grid_img)
                        break
                if is_bg:
                    bg_grid_images.append(grid_img)

        return fg_grid_images, bg_grid_images


def draw_od_grid(img, annotation=None, img_dim=(720,1280,3), grid_dim=(160,160,3), stride=(80,80), ):
    
    x_points = np.linspace(0, img_dim[1], img_dim[1]//stride[1]+1, dtype=np.int32)
    y_points = np.linspace(0, img_dim[0], img_dim[0]//stride[0]+1, dtype=np.int32)
    
    for x_pt in x_points[1:-1:2]:
        img = cv2.line(img, (x_pt, 0), (x_pt, img_dim[0]), (255,200,0), 2)
        
        for y_pt in y_points[1:-1:2]:
            img = cv2.line(img, (0, y_pt), (img_dim[1], y_pt), (255,200,0), 2)
            
    for x_pt in x_points[:-1:2]:
        img = cv2.line(img, (x_pt, 0), (x_pt, img_dim[0]), (0,200,255), 2)
        
        for y_pt in y_points[:-1:2]:
            img = cv2.line(img, (0, y_pt), (img_dim[1], y_pt), (0,200,255), 2)

    if annotation is not None:
        for box in annotation:
            img = cv2.rectangle(img, (box["x"], box["y"]), (box["x"]+box["width"], box["y"]+box["height"]), (255, 64, 64), 2)
            
    return img
````

I was doing this with raw images... but now I am creating tfrecords and saving them as a dataset so that I can leverage TPU in the future.

````python
def _bytes_feature(value, is_list=False):
    """Returns a bytes_list from a string / byte."""
    if isinstance(value, type(tf.constant(0))):
        value = value.numpy() # BytesList won't unpack a string from an EagerTensor.
    
    if not is_list:
        value = [value]
    
    return tf.train.Feature(bytes_list=tf.train.BytesList(value=value))

def _float_feature(value, is_list=False):
    """Returns a float_list from a float / double."""
        
    if not is_list:
        value = [value]
        
    return tf.train.Feature(float_list=tf.train.FloatList(value=value))

def _int64_feature(value, is_list=False):
    """Returns an int64_list from a bool / enum / int / uint."""
        
    if not is_list:
        value = [value]
        
    return tf.train.Feature(int64_list=tf.train.Int64List(value=value))

def serialize_raw(img, lbl):
    """
    Creates a tf.Example message ready to be written to a file from 4 features.
    """
    
    img = tf.io.encode_png(tf.constant(img, dtype=tf.uint8))
    
    # Create a dictionary mapping the feature name to the 
    # tf.Example-compatible data type.
    feature_dict = {
        'img': _bytes_feature(img),
        'lbl': _int64_feature(lbl),
    }
       
    # Create a Features message using tf.train.Example.
    example_proto = tf.train.Example(features=tf.train.Features(feature=feature_dict))
    return example_proto.SerializeToString()

def img_lbl_generator(subsample_neg_at=15):
    for idx, row in obj_only_df.sample(len(obj_only_df)).iterrows():
        fg_images, bg_images = get_grid_images(cv2.imread(row.img_path)[..., ::-1], row.annotations,
                                               img_dim=IMG_SHAPE, grid_dim=GRID_IMG_SHAPE, stride=STRIDE, 
                                               min_inter_area=300, min_inter_width=16, min_inter_height=16,)
        for image in fg_images:
            yield image, 1
        
        for image in bg_images[::subsample_neg_at]:
            yield image, 0


def write_tfrecords(n_ex, n_ex_per_rec=250, serialize_fn=serialize_raw, out_dir="/kaggle/working/tfrecords", ds_type="all"):
    """ Function to create our tfrecords 
    
    Args:
        TBD
    
    Returns:
        TBD
    """
    
    n_recs = int(np.ceil(n_ex/n_ex_per_rec))
    
    # Make dataframe iterable
    iter_il_gen = img_lbl_generator()
        
    out_dir = os.path.join(out_dir, ds_type)
    
    # Create folder
    if not os.path.isdir(out_dir):
        os.makedirs(out_dir, exist_ok=True)
        
    # Create tfrecords
    for i in tqdm(range(n_recs), total=n_recs):
        print(f"\n... Writing {ds_type.title()} TFRecord {i+1} of {n_recs} ...\n")
        tfrec_path = os.path.join(out_dir, f"{ds_type}__{(i+1):02}_{n_recs:02}.tfrec")
        
        # This makes the tfrecord
        with tf.io.TFRecordWriter(tfrec_path) as writer:
            for ex in tqdm(range(n_ex_per_rec), total=n_ex_per_rec):
                try:
                    example = serialize_fn(*next(iter_il_gen))
                    writer.write(example)
                except:
                    break
                    
TFRECORD_FILES = [x for x in tf.io.gfile.glob(os.path.join(DATASET_DIR, "**/**/*.tfrec")) if tf.io.gfile.GFile(x).size()>0]
if len(TFRECORD_FILES)>10:
    print("\n... LOADING THE TFRECORD FILES ...\n")
else:
    print("\n... WRITE THE TFRECORD FILES ...\n")
    write_tfrecords(n_ex=360000, n_ex_per_rec=1800, serialize_fn=serialize_raw, out_dir="/kaggle/working/tfrecords")
````

For <b>EfficientNetV2</b>, by default input preprocessing is included as a part of the model (as a Rescaling layer), and thus <code style="color: darkred; font-weight: bold;">tf.keras.applications.efficientnet_v2.preprocess_input</code> is actually a pass-through function. At the same time, preprocessing as a part of the model (i.e. Rescaling layer) can be disabled by setting <code style="color: darkred; font-weight: bold;">include_preprocessing</code> argument to <code style="color: darkred; font-weight: bold;">False</code>. With preprocessing disabled <b>EfficientNetV2</b> models expect their inputs to be float tensors of pixels with values in the <code style="color: darkred; font-weight: bold;">[-1, 1]</code> range<br>

````python
def add_model_head(_model, fc_units=512, dropouts=[0.45, 0.45], freeze_backbone=True, freeze_up_to="block6h"):
    
    # Set trainability of model
    if freeze_backbone:
        for _l in _model.layers:
            if _l.name.startswith(freeze_up_to): 
                break
            _l.trainable=False
    
    x = tf.keras.layers.Dropout(dropouts[0])(_model.output)
    x = tf.keras.layers.Dense(fc_units, activation="relu")(x)
    x = tf.keras.layers.Dropout(dropouts[1])(x)
    
    _outputs = tf.keras.layers.Dense(1, activation="sigmoid")(x)
    
    return tf.keras.Model(inputs=_model.inputs, outputs=_outputs)

LOAD_MODEL_PATH=""
with strategy.scope():
    if LOAD_MODEL_PATH!="":
        model = tf.keras.models.load_model(LOAD_MODEL_PATH, 
                                           options=load_locally, compile=False)
    else:
        # tf.keras.applications.EfficientNetB0
        model = tf.keras.applications.EfficientNetV2B0(
            include_top=False, weights='imagenet', 
            input_shape=GRID_IMG_SHAPE, pooling="avg",
        )

        # RANGER Optimizer With Default HyperParameters
        radam = tfa.optimizers.RectifiedAdam()
        ranger = tfa.optimizers.Lookahead(radam, sync_period=6, slow_step_size=0.5)

        model = add_model_head(model, fc_units=1024)
        model.compile(optimizer=ranger, loss="binary_crossentropy", metrics=["acc", 
                                                                             tf.keras.metrics.TrueNegatives(name="tn"), 
                                                                             tf.keras.metrics.TruePositives(name="tp"),
                                                                             tf.keras.metrics.FalseNegatives(name="fn"), 
                                                                             tf.keras.metrics.FalsePositives(name="fp"),
                                                                             tfa.metrics.FBetaScore(
                                                                                 num_classes=1, average="micro", beta=2.0,
                                                                             )])
        model.summary()
````

````python
def decode_image(image_data, n_channels=3, reshape_to=GRID_IMG_SHAPE, cast_to=tf.float32):
    image = tf.image.decode_png(image_data, channels=n_channels)    
    image = tf.reshape(image, reshape_to) 
    return tf.cast(image, cast_to)

def decode(serialized_example, n_channels=3, reshape_to=GRID_IMG_SHAPE):
    """ Parses a set of features and label from the given `serialized_example`.
        
        It is used as a map function for `dataset.map`

    Args:
        serialized_example (tf.Example): A serialized example
        
    Returns:
        A decoded tf.data.Dataset object representing the tfrecord dataset
    """
    
    # Defaults are not specified since both keys are required.
    feature_dict = {
        'img': tf.io.FixedLenFeature(shape=(), dtype=tf.string),
        'lbl': tf.io.FixedLenFeature(shape=(), dtype=tf.int64),
    }
    
    # Define a parser
    features = tf.io.parse_single_example(serialized_example, features=feature_dict)   
    img = decode_image(features['img'], n_channels, reshape_to)
    lbl = tf.cast(features["lbl"], tf.uint8)
    
    return img, lbl

def tf_load_grid_img(img_path, _shape=GRID_IMG_SHAPE):
    return tf.cast(tf.reshape(tf.image.decode_image(tf.io.read_file(img_path)), _shape), tf.float32)

def augment_fn(img_batch, lbl_batch):
    """ super simple augmentation fn (for now) """
    
    img_batch = tf.image.random_flip_left_right(img_batch)
    img_batch = tf.image.random_flip_up_down(img_batch)
    
    return img_batch, lbl_batch
````

````python
MODEL_CKPT_DIR = "/kaggle/working/model_wts"
if not os.path.isdir(MODEL_CKPT_DIR): os.makedirs(MODEL_CKPT_DIR, exist_ok=True)

callback_list = []
with strategy.scope():    
    callback_list.append(tf.keras.callbacks.ModelCheckpoint(
        filepath=os.path.join(MODEL_CKPT_DIR, "epoch_{epoch:02d}__fbeta_{val_fbeta_score:.3f}"),
        save_weights_only=False, monitor='val_fbeta_score', mode='max', save_best_only=True,
        options=save_locally,
    ))
    callback_list.append(tf.keras.callbacks.ReduceLROnPlateau(
        monitor='val_fbeta_score', factor=0.9, patience=2, verbose=1, mode='max',
    ))


N_EPOCHS = 100
BATCH_SIZE = 192*N_REPLICAS
SHUFFLE_BUFFER = 10*BATCH_SIZE
    
TRAIN_TFREC_PATHS = TFRECORD_FILES[5:]
VAL_TFREC_PATHS = TFRECORD_FILES[:5]

train_ds = tf.data.TFRecordDataset(TRAIN_TFREC_PATHS, num_parallel_reads=tf.data.AUTOTUNE)
val_ds = tf.data.TFRecordDataset(VAL_TFREC_PATHS, num_parallel_reads=tf.data.AUTOTUNE)

train_ds = train_ds.map(decode)
val_ds = val_ds.map(decode)

print("\n\n\n... DATASET STATISTICS ...\n")
class_weighting = {0:0.625, 1:1.125}
print("WEIGHT MAPPING --> ", class_weighting)
    
train_ds = train_ds.shuffle(SHUFFLE_BUFFER).batch(BATCH_SIZE, drop_remainder=True)
train_ds = train_ds.map(augment_fn, num_parallel_calls=tf.data.AUTOTUNE)
train_ds = train_ds.prefetch(tf.data.AUTOTUNE)

val_ds = val_ds.shuffle(SHUFFLE_BUFFER).batch(BATCH_SIZE, drop_remainder=True)
val_ds = val_ds.prefetch(tf.data.AUTOTUNE)

print("\nTRAIN : ", train_ds)
print("\nVAL   : ", val_ds)
````
