# Lets define our model
Model trained with SemEval 2017 corpus (around ~40k tweets). Base model is BerTweet, a RoBERTa model trained on English tweets.
https://huggingface.co/finiteautomata/bertweet-base-sentiment-analysis

We are going to remove the last layer and append softmax dense layer for our output.

**Lets fine tune only our last layer by freezing the bert layers to maximize our classifier layer training**. 
This means we need to make sure the model layers are all frozen

We have created a basic model with 2 Dense layer in between the BERT and the 

```python
MAXLEN = training_encodings['input_ids'].shape[1]

with strategy.scope():    
    input_ids = tf.keras.Input(shape =(MAXLEN,), dtype=tf.int32, name='input_ids') 
    attention_mask = tf.keras.Input(shape=(MAXLEN,),dtype=tf.int32, name='attention_mask')  
    
    tf_model = TFAutoModel.from_pretrained('vinai/bertweet-base', from_pt=True)
    for layers in tf_model.layers:
        layers.trainable = False
        
    tf_model_output = tf_model([input_ids, attention_mask])[1]

    tf_model_output = tf.keras.layers.Dense(256, activation = 'relu')(tf_model_output)
    tf_model_output = tf.keras.layers.Dropout(0.2)(tf_model_output)
    tf_model_output = tf.keras.layers.Dense(64, activation = 'relu')(tf_model_output)
    output = tf.keras.layers.Dense(1,activation='sigmoid')(tf_model_output)

    model = tf.keras.Model(inputs=[input_ids, attention_mask], outputs=output)

    lr = 1e-3
    weight_decay = 5e-4
    model.compile(optimizer = tfa.optimizers.AdamW(weight_decay = weight_decay, lr=lr), 
                          loss='binary_crossentropy', 
                          metrics=['accuracy'])
    model.summary()
```

I am only creating a early stoppping callbacks with patience =2. You can have multi callbacks like saving the checkpoints file along the training.

# Lets start the training 

```python
early_stop = tf.keras.callbacks.EarlyStopping(monitor='val_accuracy',patience=4, restore_best_weights=True)

model.fit(train_tf_dataset, 
        validation_data=eval_tf_dataset,
        callbacks = [early_stop],
        epochs=25,
        verbose=1)
```
