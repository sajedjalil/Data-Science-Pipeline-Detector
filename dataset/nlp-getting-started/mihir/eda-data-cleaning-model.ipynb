```python
#Count of Words
train['Number_of_words'] = train['text'].apply(lambda x: len(word_tokenize(x)))
test['Number_of_words'] = test['text'].apply(lambda x: len(word_tokenize(x)))

#Count of Sentences
train['Number_of_Sentences'] = train['text'].apply(lambda x: len(sent_tokenize(x)))
test['Number_of_Sentences'] = test['text'].apply(lambda x: len(sent_tokenize(x)))

#Count of Unique Words
train['Number_of_Unique_Words'] = train['text'].apply(lambda x: len(set(x.split())))
test['Number_of_Unique_Words'] = test['text'].apply(lambda x: len(set(x.split())))

#Count of Stop words like the, this, is , a, ..
#############
def Number_of_Stop_Words(tweet):
  word_tokens = word_tokenize(tweet) #splitta i pezzi

  stopwords_x = [w for w in word_tokens if w in STOPWORDS]

  return len(stopwords_x)
train['Number_of_Stop_Words'] = train['text'].apply(lambda x: Number_of_Stop_Words(x))
test['Number_of_Stop_Words'] = test['text'].apply(lambda x: Number_of_Stop_Words(x))
############

#Count of '#'
train['Number_of_Hashtage'] = train['text'].apply(lambda x: x.count('#'))
test['Number_of_Hashtage'] = test['text'].apply(lambda x: x.count('#'))

#Count of '@'
train['Number_of_Mentions'] = train['text'].apply(lambda x: x.count('@'))
test['Number_of_Mentions'] = test['text'].apply(lambda x: x.count('@'))

#Count of Punctuations
train['Punctuation_Count'] = train['text'].apply(lambda x: len([c for c in str(x) if c in string.punctuation]))
test['Punctuation_Count'] = test['text'].apply(lambda x: len([c for c in str(x) if c in string.punctuation]))


```

```python
def generate_ngrams(text,n_gram=1):
    token = [token for token in text.lower().split(' ') if token != '' if token not in STOPWORDS]
     
    ngrams = zip(*[token[i:] for i in range(n_gram)])
    
    return [' '.join(ngram) for ngram in ngrams]
    
```

```python
#Unigrams

fake_unigrams = dict()
true_unigrams = dict()

for tweet in train[train['target']==0]['text']:
    for w in generate_ngrams(tweet):
        try:
            fake_unigrams[w] +=1
        except:
            fake_unigrams[w] = 1
        
        
for tweet in train[train['target']==1]['text']:
    for w in generate_ngrams(tweet):
        try:
            true_unigrams[w] +=1  
        except:
            true_unigrams[w] = 1
        
```

```python
true_unigrams = dict(sorted(true_unigrams.items(), key=lambda x: x[1],reverse=True))
fake_unigrams = dict(sorted(fake_unigrams.items(), key=lambda x: x[1],reverse=True))
```

```python
def bert_encode(texts, tokenizer, max_len=512):
    all_tokens = []
    all_masks = []
    all_segments = []
    
    for text in texts:
        text = tokenizer.tokenize(text)
            
        text = text[:max_len-2]
        input_sequence = ["[CLS]"] + text + ["[SEP]"]
        pad_len = max_len - len(input_sequence)
        
        tokens = tokenizer.convert_tokens_to_ids(input_sequence)
        tokens += [0] * pad_len
        pad_masks = [1] * len(input_sequence) + [0] * pad_len
        segment_ids = [0] * max_len
        
        all_tokens.append(tokens)
        all_masks.append(pad_masks)
        all_segments.append(segment_ids)
    
    return np.array(all_tokens), np.array(all_masks), np.array(all_segments)
```

```python
def build_model(bert_layer, max_len=512):
    input_word_ids = Input(shape=(max_len,), dtype=tf.int32, name="input_word_ids")
    input_mask = Input(shape=(max_len,), dtype=tf.int32, name="input_mask")
    segment_ids = Input(shape=(max_len,), dtype=tf.int32, name="segment_ids")

    _, sequence_output = bert_layer([input_word_ids, input_mask, segment_ids])
    clf_output = sequence_output[:, 0, :]
    out = Dense(1, activation='sigmoid')(clf_output)
    
    model = Model(inputs=[input_word_ids, input_mask, segment_ids], outputs=out)
    model.compile(Adam(lr=2e-6), loss='binary_crossentropy', metrics=['accuracy'])
    
    return model
```

```python
module_url = "https://tfhub.dev/tensorflow/bert_en_uncased_L-24_H-1024_A-16/1"
bert_layer = hub.KerasLayer(module_url, trainable=True)
```

```python
vocab_file = bert_layer.resolved_object.vocab_file.asset_path.numpy()
do_lower_case = bert_layer.resolved_object.do_lower_case.numpy()
tokenizer = tokenization.FullTokenizer(vocab_file, do_lower_case)
```
