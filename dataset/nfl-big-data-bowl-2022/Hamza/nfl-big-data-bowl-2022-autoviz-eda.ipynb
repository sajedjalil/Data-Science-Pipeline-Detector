# Usage

```python
from autoviz.AutoViz_Class import AutoViz_Class
AV = AutoViz_Class()
```

# 1.Players

```python
players
```

```python
filename = "/kaggle/input/nfl-big-data-bowl-2022/players.csv"
sep = ","
dft = AV.AutoViz(
    filename,
    sep=",",
    depVar="",
    dfte=None,
    header=0,
    verbose=0,
    lowess=False,
    chart_format="svg"
)
```

```python
fig, ax = plt.subplots(figsize=(16, 6))
sns.histplot(players['Position'],color="blue",ax=ax)
```
