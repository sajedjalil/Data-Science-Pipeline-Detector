```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import os
for dirname, _, filenames in os.walk('/kaggle/input'):
    for filename in filenames:
        print(os.path.join(dirname, filename))
```

```python
train = pd.read_csv('/kaggle/input/tabular-playground-series-feb-2022/train.csv')
test = pd.read_csv('/kaggle/input/tabular-playground-series-feb-2022/test.csv')
sub = pd.read_csv('../input/early-ensemble/submission.csv')
```

## 1. Drop duplicates

```python
# drop duplicate rows
cols = [e for e in test.columns if e not in ('row_id')]
train.drop_duplicates(subset=cols, keep='first',inplace = True)
```

```python
print(train.shape)
```

## 2. Check the intersection between train & test dataset

```python
s1 = pd.merge(train, test, how='inner', on=cols)
```

```python
s1
```

```python
s1.row_id_y.nunique()
```

* <h4> We have 486 samples from the test dataset that already exist in the train dataset ðŸ‚ðŸ» => we already know the target of these samples...

## 3. Change the value of these samples in the submission file

```python
dic = {}
for i in range(len(s1)):
    dic[s1.loc[i]['row_id_y']] = s1.loc[i]['row_id_x']
```

```python
len(dic)
```

```python
for e in dic.items():
    print(e)
```

* The keys of dic : row ids of these samples in the test dataset
* The values of dic : row ids of these samples in the train dataset

```python
for e in dic:
    sub.loc[sub[sub['row_id']==e].index.to_list(),'target'] = train.loc[train[train['row_id']==dic[e]].index.tolist()[0],'target']
```

```python
sub.to_csv("submission.csv", index=False)
```
