**<br><font size=5>Background</font><br>**

Coronavirus disease 2019 (COVID-19) pandemic is highly infectious cause by severe acute respiratory syndrome coronavirus 2 (SARS-CoV-2), which can lead to fatal comorbidities especially acute respiratory distress syndrome (ARDS). Although the pathogenesis of SARS-CoV-2 infection remains unclear, the previous researcher has found several risk factors increase susceptibility to infection based on the epidemiologic characteristics observed so far in globe wide including age, smoking status, Diabetes, cardio- and cerebrovascular disease, cancer, chronic respiratory diseases and so on<sup>1-5</sup>.factor can help to better understand the disease, thereby accelerating the research of precision treatment. 

Available evidence indicates that the innate and adaptive immune system of COVID-19 patients both occurred changes. In particular, a state of a variety of inflammatory cytokines releasing out-of-control, "cytokine storm", a key hallmark of the overreaction of the body's immune system as well as the major responsibility for the occurrence of ARDS6<sup>6</sup>.
General considered that autoimmune disease patients facing increasing infectious risk compared to the general population whose immune system has an overall impairment and long-term using the corticosteroids and immunosuppressive drugs, such as cytokine blockers<sup>7-9</sup>. However, some anti-rheumatic drugs have been used as potential treatment options in the clinic, for instance, glucocorticoids, IL-6 antagonist, JAK inhibitors and choloroquine/hydrocholoroquine<sup>10</sup>.

Understanding the association between autoimmune disease and COVID-19 will contribute to drugs investigation and treatment therapy development. Here, we hypothesis that autoimmune disease is a risk factor of COVID-19 which affair the immune system of patients.

**<br><font size=5>Study strategy</font><br>**

1.	Collected candidate target gene sets in terms of cytokine storm and several autoimmune diseases from literature. The autoimmune diseases involved this study including inflammatory bowel disease(IBD), rheumatoid arthritis(RA), multiple sclerosis(MS), systemic lupus erythematosus(SLE).

2.	Perform meta-analysis of Single-cell RNA sequencing data of varying severity of COVID-19 and from healthy people to select the overlapping genes between COVID-19 and cytokine storm as well as several autoimmune diseases, which may priority as COVID-19 drug target genes.

3.	Compare the gene expression of these overlapping genes in different COVID-19 disease states and health states.

4.	Investigate which enrichment pathway are these overlapping genes involved.

5.	Using Scanpy API to read large-scale single-cell h5 matrix and Keras API to build an artificial neural network model to conduct a COVID-19 patient state forecast model.

**<br><font size=5>Materials</font><br>**


**<br><font size=3>Data Sources</font><br>**

Single-cell RNA sequencing data of bronchoalveolar lavage fluid immune cells from patients with varying severity of COVID-19 and from healthy people were obtained from the research by Mengfeng Liao et al which contained three moderate cases (M1–M3), one severe case (S1) and five critical cases (S2–S6), three healthy controls(HC1 to HC3). Data download from the Gene Expression Omnibus (GEO https://www.ncbi.nlm.nih.gov/) in terms of GSE145926.

**<br><font size=3>Gene sets collection</font><br>**

**<font size=3>Gene set-1 : Cytokine storm related genes</font>**

We collected the gene set base on the recently published literature.*ACE2* and *TMPRSS2* have been identified as SARS-CoV-2 host cell factors<sup>11,12</sup>. *TNF* is critical in nearly all acute inflammatory reactions and the anti-tumor necrosis factor (anti-TNF) antibodies are publicly using in severe cases of autoimmune inflammatory disease such as RA, IBD, or AS for around 20 years. Now trials of anti-tumour necrosis factor therapy for COVID-19 are vigorously promoted13. A single-cell sequencing of Immune cell proﬁling of COVID-19 by Wen, W.et al have identify IL-1β and *M-CSF*(or *CSF1*) as novel candidate target genes for inﬂammatory storm and some cytokine including *TNFSF13, IL-18, IL-2*, and *IL-4* maybe contribute to the recovery of COVID-19 patients<sup>14</sup>.Besides, *FCN1* and *SPP1* were highly expressed by severe/critical COVID-19 patients while *FABP4* was preferentially expressed by patients with moderate infection and healthy control<sup>15</sup>.

Overall, the gene-set including 13 genes, they are *ACE2, TNF, TNFSF13, TNFSF13B, TMPRSS2, FCN1, SPP1, FABP4, IL-18, IL-2, IL-4, IL-1β*(or *IL1B*), *M-CSF*(or *CSF1*). While IL-2 is absent in the **GSE145926** data. So only 12 genes used in the further study. Details information of the genes please see Table1. 

**<br><font size=3>Gene set-2 : 4 major kind of autoimmune disease HPO associate genes</font><br>**

These 4 major kind of autoimmune disease are inflammatory bowel disease(IBD), rheumatoid arthritis(RA), multiple sclerosis(MS), systemic lupus erythematosus(SLE).
We download the HPO associate genes of these 4 major kind of autoimmune disease from the Human Phenotype Ontology(HPO) database on 14, June, 2020.Stored in the Table 2-5, which contains 50, 18, 3, 6 genes respectively.

```python
# making input data

import numpy as np
import pandas as pd

import scanpy as sc
import h5py
    
M1_141 = "/path/GSE145926_RAW/GSM4339769_C141_filtered_feature_bc_matrix.h5"
M2_142 = "/path/GSE145926_RAW/GSM4339770_C142_filtered_feature_bc_matrix.h5"
M3_144 = "/path/GSE145926_RAW/GSM4339772_C144_filtered_feature_bc_matrix.h5"

S1_145 = "/path/GSE145926_RAW/GSM4339773_C145_filtered_feature_bc_matrix.h5"

S2_143 = "/path/GSE145926_RAW/GSM4339771_C143_filtered_feature_bc_matrix.h5"
S3_146 = "/path/GSE145926_RAW/GSM4339774_C146_filtered_feature_bc_matrix.h5"
S4_148 = "/path/GSE145926_RAW/GSM4475051_C148_filtered_feature_bc_matrix.h5"
S5_149 = "/path/GSE145926_RAW/GSM4475052_C149_filtered_feature_bc_matrix.h5"
S6_152 = "/path/GSE145926_RAW/GSM4475053_C152_filtered_feature_bc_matrix.h5"

HC1_51 = "/path/GSE145926_RAW/GSM4475048_C51_filtered_feature_bc_matrix.h5"
HC2_52 = "/path/GSE145926_RAW/GSM4475049_C52_filtered_feature_bc_matrix.h5"
HC3_100 = "/path/GSE145926_RAW/GSM4475050_C100_filtered_feature_bc_matrix.h5"

sc.settings.verbosity = 3             # verbosity: errors (0), warnings (1), info (2), hints (3)
sc.logging.print_versions()
sc.settings.set_figure_params(dpi=80, facecolor='white')

files = [M1_141, M2_142, M3_144, S1_145, S2_143, S3_146, S4_148, S5_149, S6_152, HC1_51, HC2_52, HC3_100]
marker_genes = ['ACE2', 'TNF', 'TNFSF13', 'TNFSF13B', 'TMPRSS2', 'FCN1', 'SPP1', 'FABP4',
                'IL10', 'SLC22A4', 'CASP10', 'PRKCD', 'TNFAIP3', 'FOXP3', 'RET', 'NOD2', 'PLCG2', 'TCF4', 'RTEL1',
                'HLA-DQB1', 'CIITA', 'HLA-DRB1', 'FCGR2A', 'FCGR2B', 'CTLA4', 'TREX1', 'IL18', 'IL1B','CSF1','IL4']

samples = ['M1', 'M2', 'M3', 'S1', 'S2', 'S3', 'S4', 'S5', 'S6', 'HC1', 'HC2', 'HC3']

def preprocess(file):
    adata = sc.read_10x_h5(file, genome=None, gex_only=True)
    adata.var_names_make_unique()
    adata.var['mt'] = adata.var_names.str.startswith('MT-')  # annotate the group of mitochondrial genes as 'mt'
    sc.pp.calculate_qc_metrics(adata, qc_vars=['mt'], percent_top=None, log1p=False, inplace=True)
    sc.pp.normalize_total(adata, target_sum=1e4)
    sc.pp.log1p(adata)
    sc.pp.highly_variable_genes(adata, min_mean=0.0125, max_mean=3, min_disp=0.5) 
    return adata

def inputs(files):
    dfs = []
    for file in files:
        adata = preprocess(file)
        
        col_ACE2 = adata.var.loc['ACE2','mean_counts']
        col_TNF = adata.var.loc['TNF','mean_counts']
        col_TNFSF13 = adata.var.loc['TNFSF13','mean_counts']
        col_TNFSF13B = adata.var.loc['TNFSF13B','mean_counts']
        col_TMPRSS2 = adata.var.loc['TMPRSS2','mean_counts']
        col_FCN1 = adata.var.loc['FCN1','mean_counts']
        col_SPP1 = adata.var.loc['SPP1','mean_counts']
        col_FABP4 = adata.var.loc['FABP4','mean_counts']
        col_IL10 = adata.var.loc['IL10','mean_counts']
        col_SLC22A4 = adata.var.loc['SLC22A4','mean_counts']
        col_CASP10 = adata.var.loc['CASP10','mean_counts']
        col_PRKCD = adata.var.loc['PRKCD','mean_counts']
        col_TNFAIP3 = adata.var.loc['TNFAIP3','mean_counts']
        col_FOXP3 = adata.var.loc['FOXP3','mean_counts']
        col_RET = adata.var.loc['RET','mean_counts']
        col_NOD2 = adata.var.loc['NOD2','mean_counts']
        col_PLCG2 = adata.var.loc['PLCG2','mean_counts']
        col_TCF4 = adata.var.loc['TCF4','mean_counts']
        col_RTEL1 = adata.var.loc['RTEL1','mean_counts']
        col_HLADQB1 = adata.var.loc['HLA-DQB1','mean_counts']
        col_CIITA = adata.var.loc['CIITA','mean_counts']
        col_HLADRB1 = adata.var.loc['HLA-DRB1','mean_counts']
        col_FCGR2A = adata.var.loc['FCGR2A','mean_counts']
        col_FCGR2B = adata.var.loc['FCGR2B','mean_counts']
        col_CTLA4 = adata.var.loc['CTLA4','mean_counts']
        col_TREX1 = adata.var.loc['TREX1','mean_counts']
        col_IL18 = adata.var.loc['IL18','mean_counts']
        col_IL1B = adata.var.loc['IL1B','mean_counts']
        col_CSF1 = adata.var.loc['CSF1','mean_counts']
        col_IL4 = adata.var.loc['IL4','mean_counts']

        column_names= [col_ACE2, col_TNF, col_TNFSF13, col_TNFSF13B, col_TMPRSS2, col_FCN1, col_SPP1, col_FABP4,
                       col_IL10, col_SLC22A4, col_CASP10, col_PRKCD, col_TNFAIP3, col_FOXP3, col_RET, col_NOD2, 
                       col_PLCG2, col_TCF4, col_RTEL1, col_HLADQB1, col_CIITA, col_HLADRB1, col_FCGR2A, col_FCGR2B, 
                       col_CTLA4, col_TREX1, col_IL18, col_IL1B, col_CSF1, col_IL4]
        df = pd.DataFrame(np.column_stack(column_names), columns=marker_genes)
        dfs.append(df)
        inputs = pd.concat(dfs, axis=0)
    return inputs

model_inputs = inputs(files).to_numpy()
```

```python
# building Keras mode
# training
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import LabelBinarizer
from sklearn.model_selection import train_test_split,KFold
from keras.models import Sequential
from keras.layers import Dense, Activation, Dropout
from keras import optimizers
from keras.optimizers import RMSprop
from keras import metrics
from keras import regularizers
from keras.callbacks import EarlyStopping, ModelCheckpoint, ReduceLROnPlateau
import tensorflow as tf
import keras.backend.tensorflow_backend as KTF
from matplotlib import pyplot as plt
import numpy as np
    
num_input = 30 #column
time_step = 1 #row
n_hidden = 1 
num_class = 3
batch_size = 1
n_epoch = 500

model = Sequential()
model.add(Dense(32, input_shape=(time_step,num_input),
                batch_size=batch_size,
                activation='sigmoid'))
model.add(Dropout(0.5))
model.add(Dense(3, activation='sigmoid'))
##model.add(Dense(32,activation='relu'))
#model.add(Dense(16, activation='sigmoid'))
rmsprop = optimizers.RMSprop(lr=0.0005,clipnorm=1.0)
model.compile(optimizer=rmsprop, loss='binary_crossentropy',
              metrics=['accuracy'])

def get_callbacks(patience_lr):
##    mcp_save = ModelCheckpoint(name_weights, save_best_only=True, monitor='val_loss', mode='min')
    early_stop = EarlyStopping(monitor='loss',min_delta=0,patience=10,mode='min')
##    reduce_lr_loss = ReduceLROnPlateau(monitor='loss',factor=0.1,patience=patience_lr,verbose=2,min_delta=1e-4,mode='min')
    return [early_stop]

labels = ['M', 'M', 'M', 'S', 'S', 'S', 'S', 'S', 'S', 'HC', 'HC', 'HC']
encoder = LabelBinarizer()
_labels = encoder.fit_transform(labels)
#print(OneHot_labels)
xs = model_inputs.reshape(12,1,30) #row 1, col 30
ys = _labels.reshape(12,1,3)

# enc = OneHotEncoder()
# enc.fit(_labels)
# ys = enc.transform(_labels).toarray().reshape(12,1,3)

folds = list(KFold(n_splits=6,shuffle=True,random_state=1).split(xs,ys))

## Training
score = []

train_loss = []
train_acc = []
val_loss = []
val_acc = []

for j,(train_idx, val_idx) in enumerate(folds):
    X_train_cv = xs[train_idx]
    y_train_cv = ys[train_idx]
    X_valid_cv = xs[val_idx]
    y_valid_cv = ys[val_idx]
    
    name_weights = 'final_model_fold' + str(j) + 'weights.h5'
    callbacks=get_callbacks(patience_lr=10)
    history = model.fit(X_train_cv, y_train_cv, batch_size = batch_size,epochs = n_epoch,
                        verbose = 2, shuffle=True, validation_data=(X_valid_cv, y_valid_cv),
                        callbacks=callbacks)
    metrics = model.evaluate(X_valid_cv,y_valid_cv,batch_size=batch_size)
    print(metrics)
    score.append(metrics[1])
    
##    plt.plot(history.history['loss'], label='train loss')
##    plt.plot(history.history['accuracy'], label='train accuracy')
##    plt.plot(history.history['val_loss'], label='test loss')
##    plt.plot(history.history['val_accuracy'], label='test accuracy')
##    plt.title('Feedforward with BP Classification')
##    plt.xlabel('Epochs')
##    plt.ylabel('Performance')
##    plt.legend(loc='upper right')
##    plt.show()
    
    train_loss.extend(history.history['loss'])
    train_acc.extend(history.history['accuracy'])
    val_loss.extend(history.history['val_loss'])
    val_acc.extend(history.history['val_accuracy'])

print(score)
print(np.mean(score))

fig, ax = plt.subplots(figsize = (8,7), dpi=100)
fig.tight_layout()
plt.plot(train_loss, label='train loss')
plt.plot(val_loss, label='val loss')
plt.plot(train_acc, label= 'train accuracy')
plt.plot(val_acc, label='val accuracy')
plt.xlabel('Epochs')
plt.legend(loc='upper left')
plt.show()

```
