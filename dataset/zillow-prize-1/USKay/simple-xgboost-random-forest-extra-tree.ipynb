```python
import pandas as pd
import numpy as np
from sklearn.cross_validation import train_test_split
np.set_printoptions(threshold=np.inf)
```

```python
# Data Import
train = pd.read_csv("../input/train_2016_v2.csv")   
props = pd.read_csv("../input/properties_2016.csv") 
samp = pd.read_csv("../input/sample_submission.csv") 
```

```python
train = train.drop(["transactiondate"], axis=1)
```

```python
DF = pd.merge(train, props, how='left', on='parcelid')
```

```python
X = DF.drop(['logerror'],axis=1,inplace=False)
X = X.select_dtypes(exclude=[object]) 
y = DF['logerror']
```

```python
X = X.fillna(0)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.2)
```

```python
## xgboost
from xgboost import XGBRegressor
model = XGBRegressor()
```

```python
model.fit(X, y)
```

```python
model.score(X_test, y_test)
```

```python
## random forest
from sklearn.ensemble import RandomForestRegressor
```

```python
forest = RandomForestRegressor()
```

```python
forest.fit(X, y)
```

```python
forest.score(X_test, y_test)
```

```python
## ExtraTreesRegressor
from sklearn.ensemble import ExtraTreesRegressor
```

```python
ext = ExtraTreesRegressor()
```

```python
ext.fit(X, y)
```

```python
ext.score(X_test, y_test)
```
