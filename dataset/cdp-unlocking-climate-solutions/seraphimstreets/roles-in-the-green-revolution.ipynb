# How does environmental protection create inequality? An introduction


Here, we will explore the main arguments for environmental regulation exacerbating inequity, which are largely economical in nature. 



First, is the opportunity cost associated with implementing these green policies. In other words, public funds spent on green projects such as renewable energy investments  or reforestation etc. could  have been used for schemes that more directly benefit the poor, such as food subsidies and public education. This may result in a welfare loss for society if the legislation does not bring aboout a commensurate positive impact that covers the costs. 

For instance, a common way of measuring the economic efficacy of environmental policies is via a marginal carbon abatement curve (MACC), which is an estimation of the volumes and costs of opportunities to reduce emissions in a given year. This allows policymakers to easily compare the relative costs and GHG emissions reductions of each course of action. Here, we see that the most effective environmental policies are also the most expensive, resulting in these funds being unable to be used for poverty-alleviating measures instead. 

![Carbon Abatement Curve](https://i.imgur.com/CKD8XJk.png)

Next, is that negative impacts of environmental regulations fall disproportionately on the poor, in terms of employment and direct cost.

For employment, increased carbon taxation and a shift towards cleaner energies will cause a long-term decline in certain industries which largely support blue-collar workers. This is especially egregious if the workers do not have the requisite skills or opportunities to transition to new jobs in the renewable energy sector.

For instance, a hot-button topic in US politics involves the decline in manufacturing and energy jobs in the Rust Belt, supposedly spurred by "job-killing" environmental regulations.  

While the decline in manufacturing jobs over the decades has more to do with cheaper labor costs overseas and US trade agreements, it is true that the energy sectors have been greatly pressured by the regulations. For instance, it has been generally agreed that the decline of the coal industry was induced by a shift towards renewable energy.

# Finally, some KPIs for Environment-Equity Alignment

And now, we have finally reached the part where we discuss some possible KPIs for measuring the relationship between reducing enviromental degradation and inequity. However, the issues we discussed above were not for naught; they will help play a key role in the formulation of these KPIs. Essentially, we want to find an alignment between environmental and equity goals; policies that help alleviate inequality more than exacerbate it. To do so, we will need to factor in all the concerns that we have listed above. 

We start by considering the most intuitive choice for measuring equity, which is the commmonly used Gini index. The Gini is a single-value measure of the inequality (usually income) in a population. It is defined mathematically as the ratio of the area between the Lorenz curve and the line of equality (perfectly equal scenario) over the total area under the line of equality. Thus, its value ranges from 0 (perfect equality) to 1 (perfect inequality). It is currently the de facto choice for measuring inequality in a population, largely due to its ability to capture complex income distributions in a simple number.  

![GINI](https://i.imgur.com/OPJyGDI.jpg)


A naive approach would be to plot the Gini index of a population with a proxy of enviromental performance (such as pollution readings) over time. However, this would be grossly unrepresentative, because in reality income inequality in a nation is dependent on a host of macroeconomic factors, and is in fact usually hardly impacted by environemntal policy, due to its relatively small contribution to overall GDP.

Therefore, we need a more nuanced approach that attempts to isolate the impact of environmental policy on equity as much as possible. So we propose an enviroment-based Ginis instead, based mainly on the factors that we had discussed above. So, the objective is for these KPIS to be able to measure inequality in **environmental health** as well as **employment created/destroyed by environmental policy**. 


## Pollution Gini

Let's begin by trying to measure inequality in environmental outcomes across a nation. There are many ways to measure environmental health of a community: through level of pollution, water security,climate vulnerability etc. For this part, however, we will be mainly exploring pollution as quantified by annual PM 2.5 levels, since this is the only information that is widely available and easily understood.  

PM 2.5 Levels are usually measured across a large area (eg. city/state level. Therefore, we will utilize area-based Gini coefficient (AR-Gini), where each area is treated as an a uniform group in the Gini tabulation. 

Since we are using a discretized version of Gini, the formula of Gini coefficient is defined as follows:

$$ G = 1 - \sum\limits_{k=1}^{k=n-1} {(x_k - x_{k+1})(y_k + y_{k+1}) } $$

where y is the cumulative sum of the dependent variable (eg. PM 2.5 level) in the kth state, x is the cumulative sum of the population in the kth state, k is the number of states.

(Note: this is equivalent to the geometric definition we saw before, where G = A/(A+B). Since A+B = 0.5, we have G = 2A = 1 - 2B. B is the Lorenz curve and can be estimated by trapezoids defined by the x/y values, which is represented in the second term in our equation above.)

Let's illustrate the concept of Environmental Gini with the USA.  We would like for our Gini to summarize the change in inequality of pollution across its population. We could achieve this in several ways. First, we could treat each state as a separate entity and calculate the Gini as if there were only that number of people. Here, we only have PM 2.5 data for 47 states, so we will be calculating it as if the US is a 47 person population. Doing so for the year 2003 yields this graph:

```python



def get_cum_values_states_year(states, year, criteria="state", power=1):
    usa_ppp_pollution = ppp_pollutions_region[ppp_pollutions_region['Region'].isin(states)]
    usa_ppp_pollution_year = usa_ppp_pollution[usa_ppp_pollution.Year_left == year]
    if power != 1:
        usa_ppp_pollution_year['nPM2.5 Level (µg/m³)'] =  usa_ppp_pollution_year['PM2.5 Level (µg/m³)'].apply(lambda x: max(0, x-2.4))
    else:
        usa_ppp_pollution_year['nPM2.5 Level (µg/m³)'] = usa_ppp_pollution_year['PM2.5 Level (µg/m³)']
    usa_ppp_pollution_year['pow_PM2.5 Level (µg/m³)'] = np.power(usa_ppp_pollution_year['nPM2.5 Level (µg/m³)'], power)
    
   
    total_pop = usa_ppp_pollution_year.Population.sum()
    total_pm25 =  usa_ppp_pollution_year["pow_PM2.5 Level (µg/m³)"].sum()
    usa_ppp_pollution_year.sort_values('PM2.5 Level (µg/m³)',inplace=True)
    
    
    
    if criteria == "state":
        usa_ppp_pollution_year['cumsum_pop'] = np.linspace(0, 1, usa_ppp_pollution_year.shape[0])
        usa_ppp_pollution_year['cumsum_pm25'] = usa_ppp_pollution_year["pow_PM2.5 Level (µg/m³)"].cumsum()/total_pm25
    else:
        usa_ppp_pollution_year['total_2.5'] =   usa_ppp_pollution_year["pow_PM2.5 Level (µg/m³)"] * usa_ppp_pollution_year[criteria] 

      
        total_crit = usa_ppp_pollution_year[criteria].sum()
        total_pm25 =  usa_ppp_pollution_year["total_2.5"].sum()
        usa_ppp_pollution_year['cumsum_pop'] = usa_ppp_pollution_year[criteria].cumsum()/total_crit
        usa_ppp_pollution_year['cumsum_pm25'] = usa_ppp_pollution_year["total_2.5"].cumsum()/total_pm25        
   
   
    
    return usa_ppp_pollution_year

def calculate_env_gini(y_vals, x_vals):
    trap_abs = []
    nick = y_vals.values
    nick = np.concatenate([np.array([0]), nick])
    for i in range(len(nick) - 1):
        trap_abs.append(nick[i] + nick[i+1])

    nick = x_vals.values
    nick = np.concatenate([np.array([0]), nick])
    trap_hs = []
    for i in range(len(nick) - 1):
        trap_hs.append(nick[i+1] - nick[i])

    curve_area = 0
    for i in range(len(trap_abs)):

        
        curve_area += (trap_abs[i] * trap_hs[i])

    return 1 - curve_area


def ginis_pollution_yearly(states, code, criteria="state", power=1):
    cou_years = sorted(ppp_pollutions_region[ppp_pollutions_region['Region'].isin(states)].Year_left.unique())
    pollution_wealth_ginis = []
    for year in cou_years:
        cou_ppp_pollution_year = get_cum_values_states_year(states, year, criteria, power=power) 
        k = calculate_env_gini(cou_ppp_pollution_year['cumsum_pm25'], cou_ppp_pollution_year['cumsum_pop'])
        pollution_wealth_ginis.append(k)

    total_cou_pollution = pollution_region[pollution_region.REG_ID == code]
    total_cou_pollution = total_cou_pollution[total_cou_pollution.VAR == 'AIR_POL']
    total_cou_pollution = total_cou_pollution[total_cou_pollution.Year.isin(cou_years)]
    total_cou_pollution =  total_cou_pollution.sort_values('Year').Value
    return cou_years, pollution_wealth_ginis, total_cou_pollution


def plot_pollution_wealth_gini(states, country_code, country_name, criteria="state", power=1):
    
    years_to_use, pollution_wealth_ginis, total_pollution = ginis_pollution_yearly(states,  country_code, criteria, power=power)
   
    fig = make_subplots(specs=[[{"secondary_y": True}]])

    # Add traces
    fig.add_trace(
        go.Scatter(x=years_to_use, y=pollution_wealth_ginis, name="Pollution Gini"),
        secondary_y=True,
    )

    fig.add_trace(
        go.Scatter(x=years_to_use, y=total_pollution, name="Air Pollution in {} (µg/m³)".format(country_name)),
        secondary_y=False,
    )

    # Add figure title
    fig.update_layout(
        title_text= country_name +" Pollution Gini and PM2.5 levels"
    )

    # Set x-axis title
    fig.update_xaxes(title_text="Year")

    # Set y-axes titles
    fig.update_yaxes(title_text="Pollution Gini", secondary_y=True)
    fig.update_yaxes(title_text="Air Pollution in PM2.5 (µg/m³)", secondary_y=False)


    fig.show()
```

However, some may take issue with this interpretation of pollution levels, and for good reason. Here, we are treating PM 2.5 levels as a uniform measure with every unit having the same (negative) value. So, with our current version, we would view 2 states with 5µg/m³ annually as being as bad as 1 state with 105µg/m³ annually. However, this is clearly at odds with reality. A measurement of 5µg/m³ is usually considered 'a clear and non-hazy day', [while EPA researchers have associated PM 2.5 levels of 10µg/m³ with higher incidence of pollution-related death in black communities](https://www.pri.org/stories/2020-01-07/air-pollution-silent-killer-even-when-it-meets-epa-standards).  To align our metric with our expections, we propose scaling these levels exponentially, as a way of better approximating the actual risk levels associated with higher PM 2.5 levels.

We use the baseline of 2.4µg/m as a baseline "safe" value based on [this study](https://www.sciencedirect.com/science/article/abs/pii/S004896972032355X), while having higher values rise by a power of 1.5. Of course, the choice of power function is arbitrary here, and I don't really have any justification of one value over another. Still, I think this presents at least a more accurate picture of reality compared to our linear function previous.

## Net Energy Change Gini and Fossil Fuel Reliance as a proxy for employment

From available litreature, we also know that fossil fuel jobs would be lost and renewable energy jobs created from environmental policy, especially in a push towards clean energies. So, it would be prudent for us to calculate an employment-wealth Gini from this as well. However, this is where we run into a roadblock, because unfortunately there is no publicly available dataset containing information on renewable jobs AND energy jobs created on the state/country level. (believe me, I've looked) 

So we have to make do with a proxy. Fortunately, we can use an important finding confirmed independently by many pieces of peer-reviewed research, which is that renewable energy tends to create more jobs than it removes per unit energy produced compared to fossil fuels - up to 12 times, some say. So, in our analyis, we can safely make a worst-case assumption that a unit of energy produced for renewables is equivalent in job production/loss to a unit produced for fossil fuels. 

Fortunately for us, data for renewable and fossil fuel energy generation is available for US states. So we can subtract the difference in fossil fuel generation over years from renewable generation to get the net energy change for each state. So we take the change in these energy generation levels as a very rough proxy for employment in these areas. (An approach with many limitations, but we makek do with what we have). Here, we calculate Ginis for renewable energy generation, fossil fuel generation, and net energy change = renewable energy change - fossil fuel energy change. Since the Net Energy Change column has negative values which is incompatible with Gini, we normalize by adding the minimum value in the year to all values. 


```python

def get_cum_values_states_year(year, crit):
    
    usa_ppp_energy_year = energy_state_table[energy_state_table.YEAR == year]
    
    if crit == 'Net Energy Change' or 'diff' in crit: 
    
        min_crit = usa_ppp_energy_year[crit].min()
        usa_ppp_energy_year[crit] = usa_ppp_energy_year[crit].apply(lambda x: x - min_crit)
    
    
    
    
    
    total_pop =  usa_ppp_energy_year["Population"].sum()
    total_crit = usa_ppp_energy_year[crit].sum()
    usa_ppp_energy_year['crit_per_person'] = usa_ppp_energy_year[crit]/usa_ppp_energy_year["Population"]
    usa_ppp_energy_year.sort_values('crit_per_person',inplace=True)
   
    
    usa_ppp_energy_year['cumsum_pop'] = usa_ppp_energy_year['Population'].cumsum()/total_pop 
    
    usa_ppp_energy_year['cumsum_crit'] = usa_ppp_energy_year[crit].cumsum()/total_crit
    
    return usa_ppp_energy_year

def ginis_energy_yearly(crit):
    cou_years = sorted(us_ppp_energy_region.Year_left.unique())
    energy_wealth_ginis = []
    for year in cou_years:
        if 'diff' in crit and year == 2003:
            continue
        
        cou_ppp_pollution_year = get_cum_values_states_year(year, crit) 
        
        k = calculate_env_gini(cou_ppp_pollution_year['cumsum_crit'], cou_ppp_pollution_year['cumsum_pop'])
        
        energy_wealth_ginis.append(k)
    
    total_cou_energy = []
    for year in cou_years:
        
        kr = energy_state_table[energy_state_table.YEAR == year]
        renew_sum = kr['Renewables'].sum()
        ff_sum = kr['Fossil Fuels'].sum()
    
        total_cou_energy.append(ff_sum/(ff_sum + renew_sum))
    
  
    return cou_years, energy_wealth_ginis, total_cou_energy


def plot_energy_wealth_gini(crit):
    
    years_to_use, energy_wealth_ginis, total_energy = ginis_energy_yearly(crit)
   
    fig = make_subplots(specs=[[{"secondary_y": True}]])

    # Add traces
    fig.add_trace(
        go.Scatter(x=years_to_use, y=energy_wealth_ginis, name= crit + " Gini"),
        secondary_y=True,
    )

    fig.add_trace(
        go.Scatter(x=years_to_use, y=total_energy, name="Renewable Energy Generation (Megawatthours)"),
        secondary_y=False,
    )
    
    

    # Add figure title
    fig.update_layout(
        title_text= crit + " Gini and Renewable Energy Generation"
    )

    # Set x-axis title
    fig.update_xaxes(title_text="Year")

    # Set y-axes titles
    fig.update_yaxes(title_text= crit+ " Gini", secondary_y=True)
    fig.update_yaxes(title_text="Renewable Energy (% of Total US Generation)", secondary_y=False)


    fig.show()
    

```

# Intersection between corporate and environmental concerns 

Privately-owned businesses and enterprises are estimated to release 70% of all carbon emissions, and governments often bend to the interests of these massively powerful corporations. (to varying degrees) Therefore, they are undoubtedly an important stakeholder to consider, some might say even more than policymakers. After all, environmental regulations may help curtail some of the actions of these polluting industries, but at the expense of society's welfare since energy is a necessity\. Instead, if there was an incentive for private businesses to naturally adopt more sustainable practices instead, then we could let the hand of the free market guide coroporations to a cleaner, more equitable world. 

## Renewables

In the past, renewable energy had been earmarked by investors as being too expensive and inefficient to be financially feasible in  any manner. However, with the advancement of green technologies and a shift in global perceptions towards climate change over the past decade, they have been growing in prominence and attractiveness, and are looking poised to overtake traditional fossil fuels as the main energy source over the next decade or two. 

However, how viable is renewable energy really compared to fossil fuels? As an business owner/investor, would it really be in my best interest to invest into clean energy? We invesetigate several key issues surrounding the adoption of renewable energies, and therefore find the intersection between corporate and environmental concerns. 

### 1. Price of Renewables

Costs in producing energy can be broadly broken into two categories. First, is the capital costs involved to set up the infrastructure and machinery needed to generate the energy, whether that comes in the form a power plant, or wind farms. The second are the costs for the production of the energy itself, the "running costs of the operation". 

We start with running costs. Running costs can be quite handily summarized with the levelized cost of energy (LROE)  metric, which takes the sum of costs over a plant's lifetime and divides it by the sum of electricity produced over its lifetime. Taken over many plants, this provides a good estimate for us to compare the running costs of different energy types. The data below was collected by IRENA, and shows declining LCOEs for renewables over time. We use Lazard's estimation of LCOE for gas combined cycle power plant and coal for 2020.

```python
fig = go.Figure()
for et in renewable_energy_stats['Energy Type'].unique():
    et_data = renewable_energy_stats[renewable_energy_stats['Energy Type'] == et]
    et_lcoe = et_data["LCOE (2019 USD/kWh) (Weighted Average)"]
    fig.add_trace(
        go.Scatter(x=et_data.Year, y=et_lcoe, name= et ),

    )

fig.add_trace(go.Scatter(x=[2010, 2019], y=[0.112, 0.112],  name='Coal (Lazard)', line=dict(color='Red', dash='dot')))
fig.add_trace(go.Scatter(x=[2010, 2019], y=[0.059, 0.059],  name='Gas/oil combined cycle power plant (Lazard)', line=dict(color='Purple', dash='dot')))

fig.update_layout(
    title_text= "Levellised cost of energy for different energy sources"
)

fig.update_xaxes(title_text="Year")
fig.update_yaxes(title_text="Levelized cost of energy (2019 USD/kWh)") 
fig.show()
```

Of course, the estimates for these costs vary, and Lazard's own estimation for the average cost of solar PV puts it below the combined cycle power plant at 0.036. Whichever the case though, this shows that the most common renewables, solar photovoltaic and onshore wind, are price-competitive with fossil fuels in terms of running costs.  

Then, we move on to the initial capital costs, the barrier to entry for all energy aspirants. This is typically measured using overnight cost per watt. As of 2019, the estimated average capital costs per kilowatt for coal is 3500, while for gas/oil combined cycle it is 1000. Most renewables have fallen below the installed price of coal as of 2019, but only solar PV is able to compete with the combined cycle power plant.

```python
fig = go.Figure()
for et in renewable_energy_stats['Energy Type'].unique():
    et_data = renewable_energy_stats[renewable_energy_stats['Energy Type'] == et]
    et_lcoe = et_data["Total Installed Costs (2019 USD/kW) (Weighted Average)"]
    fig.add_trace(
        go.Scatter(x=et_data.Year, y=et_lcoe, name= et ),

    )
    
fig.add_trace(go.Scatter(x=[2010, 2019], y=[3500, 3500],  name='Coal (2019)', line=dict(color='Red', dash='dot')))
fig.add_trace(go.Scatter(x=[2010, 2019], y=[1000, 1000],  name='Gas/oil combined cycle power plant (2019)', line=dict(color='Purple', dash='dot')))

fig.update_layout(
    title_text= "Total Installed Costs for different energy sources"
)

fig.update_xaxes(title_text="Year")
fig.update_yaxes(title_text="Total Installed Costs (2019 USD/kW) (Average)") 
fig.show()
```

In conclusion, it seems that costs should not be a deterrent for businesses to switch from fossil fuels to the most cost efficient renewables. In fact, one could make substantial cost savings by moving from coal to solar PV/onshore wind power in the long run. Further government subsidization and schemes should only serve to reinforce this.

### 3. Investment potential of renewables 

Investor sentiment is an indispensable factor when it comes to business feasibility. It can be said that even the CEO of a company is just another employee in the eyes of its shareholders. So how do we convince investors to make the switch from traditional to renewable, assuming players that are driven not only by alturism and expect some kind of commensurate return? 

A [study](https://www.imperial.ac.uk/business-school/faculty-research/research-centres/centre-climate-finance-investment/research/energy-investing-exploring-risk-and-return-the-capital-markets/) by the Imperial College London and the International Energy Agency this year sought to answer this question. They first identified some "pure play" companies in renewable energy and fossil fuels in  Germany/France,the UK and the US, using them to build two separate portfolios respectively. They then analyzed the activity of these two portfolios over 5 and 10 year period. They found that renewables yielded a 178.2% return over -20.7% for fossil fuel investments over a 5 year period for France+Germany, 75.4% compared to 8.8% returns for the UK, and 200.3% against 97.2% for the US.

![Returns](https://i.imgur.com/UUjY4oY.png)

If renewables are such a lucrative market, why aren't more instituational investors jumping on? Well, the authors themselves proffered an explanation,  “It is easier to allocate a meaningful percentage of their assets under management to renewables if the market is deep and liquid...Currently, that is not the case.” Currently, the renewable market is still too nascent and its waters too untested for big-name firms to shake off their hesitancy. Still, even on this front, this may not be the case for long. Investment in renewable energy has been steadily rising for the past decade and governments have already pledged a collective US$1 trillion on clean energy in the next decade, so it seems only a matter of time until the market becomes deep and tested enough to inspire the confidence of shareholders. Maybe then, it will reach the crucial inflexion point it needs to really take off.

### So what's stopping people?

I have not shown all sides of this argument, naturally. It would be impossible to capture all political, economic, (and administrative?) contexts unique to each region that might possibly act as a roadblock for providers of renewable energy. However, the fact is that renewables have seen widespread success in many regions across the world. In Europe, especially, there have been many success stories. I have already mentioned Germany, but the other EU states are perhaps even more impressive. The UK and Denmark have achieved 50% renewable usage, and Scotland, incredibly, is on track to power 100% of its domestic electricity consumption with clean energy by the end of this year. 

All this was achieved wfrom a collaboration of public and private sectors, with governments generously supplying subsidies and infrastructures and the business owners managing the production and distribution. The Aberdeen Renewable Energy Group (AREG), for instance,  is a public-private partnership created in Scotland to identify and promote renewable energy opportunities for businesses in the country. 

The existence of such exemplars, together the aforementioned factors of costs, reliability, and investment potential, suggest that the inability of other developed nations to transition to renewables stems from factors besides economic. For rapidly industrializing economies, the situation is more fraught, but that doesn't mean that they cannot contribute in small ways. I[ndia has committed to reducing their greenhouse gas emissions intensity by 33% from 2005 to 2030](https://www.globalenergyinstitute.org/indias-conditional-unconditional-climate-pledge), and [China has pledged to reach "peak carbon" by 2030 and zero emissions by 2060](https://www.theguardian.com/environment/2020/sep/27/china-carbon-pledge-put-energy-system-reverse-wind-solar). Although it still remains to be seen whether they can deliver on these promises, the conditions seem favorable for a smooth transition, and perhaps a day may come when it becomes more economically vialbe to adopt renewables as the primary energy source. For the whole world, it only seems a matter of time. 



## What about energy-efficient companies? 

Not all companies need to be in the energy sector to make a difference to climate change, though. We examine companies from all different sectors provided in the CDP dataset. Since the question posed was whether that could be any intersection between corporate and environmental interests, we will focus on the possible profitability derived from selling of environmentally friendly products. 

```python
fig = go.Figure()
all_corps = [corp_18, corp_19, corp_20]
print("Percentage of companies who sell low-carbon products/help lower GHG emissions for third-parties:")
years = []
yes = []
no = []
for i, corp in enumerate(all_corps):
   
    r = corp[corp["question_number"]== "C4.5"].response_value.value_counts().values
    yes_r = r[0]/(r[0] + r[1])
    years.append(2018+i) 
    yes.append(r[0])
    no.append(r[1])
    print("{}: {:.1f}%".format(2018+i, yes_r*100))
    
fig.add_trace(go.Bar(x=years, y=yes, name='Yes'))
fig.add_trace(go.Bar(x=years, y=no, name='No'))
fig.update_layout(
    title_text= "Do you classify any of your existing goods and/or services as low-carbon products or do they enable a third party to avoid GHG emissions?"
)
fig.show()

```

```python
y1 = []
y2 = []
y3 = []
for corp in all_corps:
    qin = corp[corp["column_name"].str.contains("C4.5a_C3")]
    r = qin[qin.response_value.notna()].response_value.value_counts().values
   
    y1.append(r[0])
    y2.append(r[1])
    y3.append(r[2])
    


years = ['2018', '2019', '2020']

fig = go.Figure()
fig.add_trace(go.Bar(
    x=years,
    y=y1,
    name='Low-carbon product',
    marker_color='indianred'
))
fig.add_trace(go.Bar(
    x=years,
    y=y2,
    name='Avoided Emissions',
    marker_color='lightsalmon'
))

fig.add_trace(go.Bar(
    x=years,
    y=y3,
    name='Both',
    marker_color='blue'
))


# Here we modify the tickangle of the xaxis, resulting in rotated labels.
fig.update_yaxes(title_text="No. of Companies")
fig.update_layout(barmode='group', xaxis_tickangle=-45)
fig.update_layout(
    title_text= "Are these low-carbon product(s) or do they enable avoided emissions?"
)
fig.show()
```

```python
fig = go.Figure()
for i, corp in enumerate(all_corps):
    qin = corp[corp["column_name"].str.contains("C4.5a_C5")]
    r = qin[qin.response_value.notna()]
    fig.add_trace(
        go.Box(y=r.response_value.values,  boxpoints='all', jitter=0.3,  name=str(i + 2018))
    )
fig.update_layout(
    title_text= "% revenue from low-carbon product(s) in the reporting year"
)    
fig.show()
```

While the median company still makes only about 2% of their revenue from low-carbon products in 2020, this is a rise from the  previous years of 1%, and if we extend it to the 75th percentile, we find a marked increase from 25% to 29.5%, which is promising. There are also a group of companies that make a 100% of their revenue from low-carbon products, but this is likely from the nature of the product they are selling.
