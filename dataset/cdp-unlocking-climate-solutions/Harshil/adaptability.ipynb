```python
def standardize_rank(arr, direction=1):
    rank = (stats.mstats.rankdata(np.ma.masked_invalid(arr)))
    rank[rank == 0] = np.nan
    if(direction == 1):
        return rank/(np.nanmax(rank)) 
    else:
        return 1-(rank/(np.nanmax(rank)))
```

```python
def score_adaptation_actions(acct):
    df = fetch_answer_individual(acct, '3.0')
    rows = np.unique(df["Row Number"])
    nactions = len(rows)
    alls = []
    for i in range(len(rows)):
        sub = df[(df["Row Number"] == rows[i])]
        try:
            scst = score_status[sub[sub["Column Number"] == 4.0]["Response Answer"].iloc[0]]
        except:
            scst = 0
        try:
            scf = score_finance[sub[sub["Column Number"] == 9.0]["Response Answer"].iloc[0]]
        except:
            scf = 0.05
        cobenefits = len(sub[sub["Column Number"] == 6.0])
        alls.append(scst*scf*cobenefits)
    return np.sum(alls)

def cost_adaptation(acct):
    df = fetch_answer_individual(acct, '3.0')
    rows = np.unique(df["Row Number"])
    nactions = len(rows)
    alls = []
    for i in range(len(rows)):
        sub = df[(df["Row Number"] == rows[i])]
        try:
            cost = float(sub[sub["Column Number"] == 11.0]["Response Answer"].iloc[0])
        except:
            cost = 0
        alls.append(cost)
    return np.nansum(alls)
```

```python
from nltk.sentiment.vader import SentimentIntensityAnalyzer
from nltk import tokenize

sid = SentimentIntensityAnalyzer()

def polarity_score_city(question, column, sent_thresh=1):
    subset = cities_2020[cities_2020['Question Number'] == question]
    subset_city_answers = subset[subset["Column Number"] == column]
    
    compound_score = []
    for i in range(len(acct_number)):
        ss = list(subset_city_answers[subset_city_answers["Account Number"] == acct_number[i]]["Response Answer"])
        ss = ' '.join(map(str, ss))
        sents = tokenize.sent_tokenize(ss)
        if(len(sents) > sent_thresh):
            cc = 0
            for s in sents:
                r = sid.polarity_scores(s)
                cc += r['compound']
            avg_cc = cc/len(sents)
            compound_score.append(avg_cc)
        else:
            compound_score.append(np.nan)
    return (compound_score)
```

```python
bric_2010 = pd.read_excel("../input/resiliencescoresbric/bric/BRIC2010_FIPS2019_nov 20.xlsx")
bric_2015 = pd.read_excel("../input/resiliencescoresbric/bric/BRIC2015_FIPS2019_nov 20.xlsx")

def resilience_bric(type_resilience, year):
    if(year == 2015):
        bric = bric_2015
    else:
        bric = bric_2010
        
    res = []
    for i in range(len(match_fips)):
        if(not flag_multi_counties[i]):
            fips = int(match_fips[i])
            cdf = bric[bric["GEOID"] == fips]
            try:
                res.append(float(cdf[type_resilience]))
            except:
                res.append(np.nan)
        else:
            fips = match_fips[i]
            temp = []
            for j in range(len(fips)):
                cdf = bric[bric["GEOID"] == fips[j]]
                temp.append(np.sum(np.array(cdf[type_resilience])))
            ar = np.array(temp)
            weighted_avg = np.average(ar, weights=np.array(ast.literal_eval(counties_weights[i])))
            res.append(weighted_avg)
    return np.array(res)
```

```python
social_resilience = resilience_bric("SOCIAL", 2015)
econ_resilience =  resilience_bric("ECONOM", 2015)
infra_resilience = resilience_bric("HOUSING/INFRA", 2015)
comm_resilience = resilience_bric("COMM CAPITL", 2015)
inst_resilience = resilience_bric("INSTITUTIONAL", 2015)
env_resilience = resilience_bric("ENVIRONMNT", 2015)
tot_resilience = resilience_bric("TOT RESIL2015", 2015)

social_resilience_ranked = standardize_rank(social_resilience, 1)
econ_resilience_ranked =  standardize_rank(econ_resilience, 1)
infra_resilience_ranked = standardize_rank(infra_resilience, 1)
comm_resilience_ranked = standardize_rank(comm_resilience, 1)
inst_resilience_ranked = standardize_rank(inst_resilience, 1)
env_resilience_ranked = standardize_rank(env_resilience, 1)
tot_resilience_ranked = standardize_rank(tot_resilience, 1)
```

```python
delta_social_resilience = (social_resilience - resilience_bric("SOCIAL", 2010))/resilience_bric("SOCIAL", 2010)
delta_econ_resilience = (econ_resilience - resilience_bric("ECONOM", 2010))/resilience_bric("ECONOM", 2010)
delta_infra_resilience = (infra_resilience - resilience_bric("HOUSING/INFRA", 2010))/resilience_bric("HOUSING/INFRA", 2010)
delta_comm_resilience = (comm_resilience - resilience_bric("COMM CAPITL", 2010))/resilience_bric("COMM CAPITL", 2010)
delta_inst_resilience = (inst_resilience - resilience_bric("INSTITUTIONAL", 2010))/resilience_bric("INSTITUTIONAL", 2010)
delta_env_resilience = (env_resilience - resilience_bric("ENVIRONMNT", 2010))/resilience_bric("ENVIRONMNT", 2010)
delta_tot_resilience = (tot_resilience - resilience_bric("TOT RESIL2010", 2010))/resilience_bric("TOT RESIL2010", 2010)

delta_social_resilience_ranked = standardize_rank(delta_social_resilience, 1)
delta_econ_resilience_ranked =  standardize_rank(delta_econ_resilience, 1)
delta_infra_resilience_ranked = standardize_rank(delta_infra_resilience, 1)
delta_comm_resilience_ranked = standardize_rank(delta_comm_resilience, 1)
delta_inst_resilience_ranked = standardize_rank(delta_inst_resilience, 1)
delta_env_resilience_ranked = standardize_rank(delta_env_resilience, 1)
delta_tot_resilience_ranked = standardize_rank(delta_tot_resilience, 1)
```
