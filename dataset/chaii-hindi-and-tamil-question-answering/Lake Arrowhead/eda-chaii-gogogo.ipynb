```python
language_name = train['language'].value_counts().index.tolist()
language_val = train['language'].value_counts().tolist()

fig = px.bar(
    x=language_name,
    y=language_val,
    title="Training Samples by Language",
    labels={
        'x': 'Language',
        'y': 'Sample count'
    },
    color = language_name
)
fig.show()
```

```python
labels = list(train['language'].value_counts().index)
values = train['language'].value_counts()
colors = ['#FF0000', '#E066FF']

trace = [go.Pie(labels = labels, 
                values = values,
                rotation = 30,
                opacity = 1,
                showlegend = True,
                hoverinfo = 'label+percent', 
                textinfo = 'label+value',#textinfo = 'value',
                insidetextorientation='radial',
                textfont = dict(size = 30, color = 'white'),
                marker = dict(colors = colors, 
                   line = dict(color = '#000000',width=2)),
               title = 'Pie Chart - Training Samples by Language')]
fig = go.Figure(data = trace)

fig.show()
```

# 4„ÄÅWordCloud 

```python
# get the Hindi and Tamil text 
Hindi_text = ''.join(train[train['language'] =='hindi']['question'])
Tamil_text = ''.join(train[train['language'] =='tamil']['question'])

# Get the tokens and frequencies for hindi and tamil language

hindi_nlp = Hindi()
hindi_doc = hindi_nlp(Hindi_text)
hindi_tokens = set([token.text for token in hindi_doc])
hindi_tokens_counter = Counter(hindi_tokens)

tamil_nlp = Tamil()
tamil_doc = tamil_nlp(Tamil_text)
tamil_tokens = set([token.text for token in tamil_doc])
tamil_tokens_counter = Counter(tamil_tokens)
```

```python
wordcloud = WordCloud(font_path="Devanagari/Lohit-Devanagari.ttf",
                      width=600,
                      height=600,
                      background_color='black',
                      stopwords=hindi_stopwords).generate(str(hindi_tokens_counter))
plt.figure(figsize=(20, 15))
plt.title('Sample WordCloud of Hindi',fontsize=25,color='r')
plt.imshow(wordcloud)
plt.axis("off")
plt.show()
```

```python
wordcloud = WordCloud(font_path="Tamil/Lohit-Tamil.ttf",
                      width=600,
                      height=600,
                      background_color='white',
                      stopwords=hindi_stopwords).generate(str(tamil_tokens_counter))
plt.figure(figsize=(20, 15))
plt.title('Sample WordCloud of Tamil',fontsize=25,color='r')
plt.imshow(wordcloud)
plt.axis("off")
plt.show()
```
