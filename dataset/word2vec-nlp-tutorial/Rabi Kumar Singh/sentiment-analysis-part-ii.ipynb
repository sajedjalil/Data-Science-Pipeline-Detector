# Text cleanings
* Lower case
* Remove Puncuation
* Remove Frequents Words
* Remove Rare Words
* Stemmer
* Lemmatization
* Removing stop words
* Remove URLS
* Expanding the deontracted words 

```python
def clean_text(df, col):
    # lowering the case 
    def lower_case(df,col):
        df[col]=df[col].str.lower()
    # remove punctuation 
    
    punc_to_remove=string.punctuation
    def remove_punctuation(text):
        return text.translate(str.maketrans('','',punc_to_remove))
    df[col]=df[col].apply(lambda text: remove_punctuation(text))
    
    #remove_frequent_words
    FREQWORDS = set([w for (w, wc) in cnt.most_common(10)])
    def remove_freqwords(text):
        """custom function to remove the frequent words"""
        return " ".join([word for word in str(text).split() if word not in FREQWORDS])
    df[col]=df[col].apply(lambda text: remove_freqwords(text))
    
    # remove_rare words
    n_rare_word=10
    RAREWORDS=set([w for (w ,wc) in cnt.most_common()[:-n_rare_word-1:-1]])

    def remove_rareword(text):
        return " ".join([word for word in str(text).split() if word not in RAREWORDS])
    df[col]=df[col].apply(lambda text: remove_rareword(text))
    
    #Porter Stemming
    stemmer=PorterStemmer()

    def stem_porter(text):
        return " ".join([stemmer.stem(word) for word in text.split()])
    df[col]=df[col].apply(lambda text: stem_porter(text))
    
    # Lemmatization
    
    lematizer=WordNetLemmatizer()
    def lemmatizer_words(text):
        return " ".join([lematizer.lemmatize(word) for word in text.split()])
    df[col]=df[col].apply(lambda text: lemmatizer_words(text))
    
    lemmmatizer=WordNetLemmatizer()

    wordnet_map={"N":wordnet.NOUN, "V":wordnet.VERB, "J":wordnet.ADJ, "R":wordnet.ADV}
    def lemmatized_words(text):
        pos_tagged_text = nltk.pos_tag(text.split())
        return " ".join([lematizer.lemmatize(word , wordnet_map.get(pos[0], wordnet.NOUN)) for word, pos in pos_tagged_text])
    df[col]=df[col].apply(lambda text: lemmatized_words(text))
    
    # Remove HTML tag from the text
    
    TAG_RE = re.compile(r'<[^>]+>')

    def remove_tags(text):
        return TAG_RE.sub('', text)
    df[col]=df[col].apply(lambda text: remove_tags(text))
    
    # remove urls
    
    def remove_urld(text):
        url_pattern=re.compile(r'https?://\S+|www\.\S+')
        return url_pattern.sub(r'', text)

    df[col]=df[col].apply(lambda text: remove_urld(text))
    
    # expamnds the decontracted words
    
    c_re = re.compile('(%s)' % '|'.join(cList.keys()))
    def expandContractions(text, c_re=c_re):
        def replace(match):
            return cList[match.group(0)]
        return c_re.sub(replace, text)
    df[col]=df[col].apply(lambda text: expandContractions(text, c_re=c_re))
    
    
```

# counting the most common words

```python
cnt=Counter()

for text in train['review'].values:
    for word in text.split():
        cnt[word]+=1
        
cnt.most_common(50)
```

```python
most_common_word=pd.DataFrame(cnt.most_common(50), columns=['words','count'])
most_common_word.head()
```

# Visualize the Most Common words in review

```python
fig, ax = plt.subplots(figsize=(8, 8))

# Plot horizontal bar graph
most_common_word.sort_values(by='count').plot.barh(x='words',
                      y='count',
                      ax=ax,
                      color="purple", figsize=(20,20))

ax.set_title("Common Words Found in Review (Clearned Words)")
plt.grid(b=True)
plt.show()
```
