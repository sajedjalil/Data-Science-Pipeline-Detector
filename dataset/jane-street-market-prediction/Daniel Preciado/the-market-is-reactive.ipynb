```python
import janestreet
import numpy as np
import pandas as pd
import lightgbm as lgb
from sklearn import *

train = pd.read_csv('../input/jane-street-market-prediction/train.csv')
train = train.query('date > 85').reset_index(drop = True) 
train = train[train['weight'] != 0]
features = [c for c in train.columns if 'feature' in c]
for c in features:
    train[c] = train[c].fillna(train[c].mean())
#features = pd.read_csv('../input/jane-street-market-prediction/features.csv')
```

```python
%%time
train['action'] = (train['resp'].values > 0).astype('int')
col = [c for c in train.columns if c not in ['resp_1', 'resp_2', 'resp_3', 'resp_4', 'resp', 'ts_id', 'date', 'action']]

params = {'objective':'binary', 'boosting': 'gbdt', 'learning_rate': 0.4, 'max_depth': -1, 'random_state': 20}
x1, x2, y1, y2 = model_selection.train_test_split(train, train['action'], test_size=0.3, random_state=20)
del train
model = lgb.train(params, lgb.Dataset(x1[col], y1), 250,  lgb.Dataset(x2[col], y2), verbose_eval=100, early_stopping_rounds=100)
```

```python
env = janestreet.make_env()
test = env.iter_test()

for (t, sub) in test:
    sub.action = (model.predict(t[col], num_iteration=model.best_iteration)  > 0.503).astype('int')
    env.predict(sub)
```

```python
lgb.plot_importance(model, importance_type='gain', max_num_features=20)
```
