### (Almost) dummy prediction for Jane Street

After studying the data, it looks like `-feature_96` is a measure of net volume traded over the last 30 minutes. 
* If the value is positive, more people bought than sold during the last period. 
* If the value is negative, more people sold. 


Together with the fact that `feature_0` is the side of the trade, this notebook implements a simple rule: I always follow whatever other people did in the last 30 mins. 

```python
import numpy  as np
import pandas as pd

import janestreet
env = janestreet.make_env()
iter_test = env.iter_test()
for (test_df, sample_prediction_df) in iter_test:
    side        = test_df['feature_0'].item()
    net_volume  = -test_df['feature_96'].item()
    action      = 1 * np.logical_not(np.logical_xor(side > 0, net_volume > 0)) * (net_volume == net_volume)
    sample_prediction_df.action = action
    env.predict(sample_prediction_df)
```
