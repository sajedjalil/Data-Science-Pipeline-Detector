```python
import numpy as np 
import pandas as pd
```

```python
submission_1_df = pd.read_csv('../input/jane-street-submissions/submission_1.csv')
submission_2_df = pd.read_csv('../input/jane-street-submissions/submission_2.csv')
submission_3_df = pd.read_csv('../input/jane-street-submissions/submission_3.csv')
```

```python
submission_1_df.head()
```

```python
submission_overall_df = submission_1_df.merge(submission_2_df, on = 'ts_id').merge(submission_3_df, on = 'ts_id')
```

```python
submission_overall_df.columns = ['ts_id', 'action_1', 'action_2', 'action_3']
```

```python
submission_overall_df.head()
```

```python
# Here we will use voting mechanism. If vote count is greater than 1, then they are said to be majority and hence the action value is assigned as 1.
submission_overall_df['vote_count'] = submission_overall_df['action_1'] + submission_overall_df['action_2'] + submission_overall_df['action_3']
```

```python
# Now create final action column
submission_overall_df['action'] = np.where(submission_overall_df['vote_count'] > 1,1, 0)
```

```python
submission_overall_df.head()
```

```python
submission_overall_df[['ts_id', 'action']].to_csv("submission.csv", index = False)
```

```python
import janestreet
env = janestreet.make_env()
env_iter = env.iter_test()

i = 0

for (test_df, pred_df) in env_iter:
#     print(test_df.columns)
#     print(pred_df.head())
    pred_df.action = submission_overall_df[submission_overall_df['ts_id'] == i]['action'].values[0]
    i = i + 1
    env.predict(pred_df)
```
