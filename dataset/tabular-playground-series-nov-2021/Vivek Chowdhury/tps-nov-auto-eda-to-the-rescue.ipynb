```python
# Importing all the necessary modules
import gc
import numpy as np
import pandas as pd

# Pandas Profiling
from pandas_profiling import ProfileReport

# SweetViz
import sweetviz as sv

# AutoViz
from autoviz.AutoViz_Class import AutoViz_Class

# D-Tale
import dtale
import dtale.app as dtale_app
dtale_app.USE_NGROK = True

# Data-Prep
from dataprep.eda import plot, plot_correlation, create_report, plot_missing, plot_diff

# Import Warnings
import warnings
warnings.filterwarnings("ignore")
```

```python
# Looking at the shapes of both the train and test dataset

train = pd.read_csv('../input/tabular-playground-series-nov-2021/train.csv')
test = pd.read_csv('../input/tabular-playground-series-nov-2021/test.csv')

print('Number of rows and columns in train data:', train.shape)
print('Number of rows and columns in test data:', test.shape)
```

```python
# creating the y column (target)
y = train.pop('target')

# dropping the id columns from both the dataframe
train = train.iloc[:,1:]
test = test.iloc[:,1:]

# Combining the train and test dataset
master_df = pd.concat([train,test], axis=0)
master_df.shape
```

```python
gc.collect() # For memory optimization
```
