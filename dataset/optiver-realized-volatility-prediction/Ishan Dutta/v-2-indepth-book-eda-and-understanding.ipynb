<div class="list-group" id="list-tab" role="tablist">
<h2 class="list-group-item list-group-item-action active" data-toggle="list" style='background:orange; border:0; color:white' role="tab" aria-controls="home"><center>Distribution Plots</center></h2>


```python
book = book.head(100000)
```

```python
def plot_distribution(x, title):

    fig = px.histogram(
    book, 
    x = x,
    width = 800,
    height = 500,
    title = title
    )
    
    fig.show()
```

```python
plot_distribution(x = 'time_id', title = 'Time Id Distribution')
```

```python
plot_distribution(x = 'seconds_in_bucket', title = 'Seconds in Bucket Distribution')
```

```python
plot_distribution(x = 'stock_id', title = 'Stock Id Distribution')
```

```python
plot_distribution(x = 'bid_price1', title = 'Bid Price 1 Distribution')
```

```python
plot_distribution(x = 'ask_price1', title = 'Ask Price 1 Distribution')
```

```python
plot_distribution(x = 'bid_price2', title = 'Bid Price 2 Distribution')
```

```python
plot_distribution(x = 'ask_price2', title = 'Ask Price 2 Distribution')
```

```python
plot_distribution(x = 'bid_size1', title = 'Bid Size 1 Distribution')
```

```python
plot_distribution(x = 'ask_size1', title = 'Ask Size 1 Distribution')
```
